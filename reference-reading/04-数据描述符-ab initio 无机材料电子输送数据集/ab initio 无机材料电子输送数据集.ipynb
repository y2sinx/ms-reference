玻尔兹曼方程的积分需要对带结构进行分析描述. BoltzTraP代码使用基于带能量的傅立叶展开的插值方法来提供这种功能, 该方法通过使用星函数来维持空间群对称性. 
这种技术的基本思想是使用比带能量(band-energies)更多的星形函数, 但是限制了拟合带 $\tilde{\varepsilon}$ 的数量(使其等于能带数 $\varepsilon$), 并使用额外的自由度来最小化粗糙函数 $\rho$. 该方法由Shankland 引入, 由Koelling和Wood 验证和测试, 并由Pickett 等人修改. 过去十年, BoltzTraP代码在不同应用(体系)中已被大范围的测试, 从超导体至热电材料, 并且在一些案例中与实验值保持一致.
从实际的角度来看, BoltzTraP代码将先前通过DFT代码(或其他方法)计算的不同k点的电子能量作为输入, 对能带进行插值处理, 并计算不同温度和费米能级的费米积分. 最后, 它返回所有传输系数, 以及其他数据, 例如插值函数的系数.

最后, 我们还想提一下BoltzWann代码: 最近尝试使用Wannier函数插值能带. 虽然它为插值带结构提供了更高的精度, (例如, 更好地处理带交叉), 但这种方法还没有像BoltzTraP那样广泛测试, 并且由于Wannier函数的自动构建仍处于早期阶段, 因此很难在HT(高通量)框架内利用它.

## 计算参数

运行BoltzTraP所需的输入数据是晶体结构和均匀网格上的电子带结构. 这两个输入都使用标准的高通量密度函数理论来计算, HT-DFT配方来自MP. 使用维也纳从头计算仿真程序包(VASP)执行DFT计算, 使用PBE-GGA, 和PAW方法. 对于具有局部d轨道的过渡金属氧化物, 采用GGA+U方法设定MP标准的Hubbard校正. 包含在MP数据库中大部分结构来自ICSD. 其他项目来自先前的高通量项目(例如, 锂离子电池筛选项目 51)以及其他数据库(例如, 开放量子材料数据库)). 使用两步程序使所有结构完全松弛(原胞和原子位置), 直到能量差低于0.0005eV /原子. 所有的弛豫都是在自旋极化的基础上进行的, 并在高自旋铁磁体中初始化磁离子. 对于后续计算, 仅当弛豫结果表明原子`具有非零投射磁矩时`才保留自旋极化. 取决于标准原胞的带结构计算根据Setyawan和Curtarolo 的惯例进行. **首先**, 执行静态自洽计算是为了电荷密度的收敛, 为大的间隙体系(≥0.5EV)使用中等k点密度(90/埃^3)来采样布里渊区, 对于那些有小间隙(<0.5电子伏特))体系采用450点/埃^3. 在大多数情况下, 使用四面体方法计算k空间上的能带结构积分. 每当此方法失败, 使用高斯smearing. **然后**, 执行两个非自洽计算以评估带结构: 第一个沿着ref9中的对称线, 第二个采用均匀k点网格(大带隙1000个k点,小带隙1500个k点. 在当前的研究中没有考虑自旋轨道耦合, 但可以作为改进数据库的下一步来实现.

掺杂(即引入额外的载流子如空穴或电子)对电子传输性质具有巨大影响. 掺杂将重设费米能级(μ)并直接影响输送性质的值. **第一个数据集**提供了对n型和p型掺杂的所有输送量, 固定掺杂水平下, 范围从 $10^{16} - 10^{20}/cm^{-3}$(间隔为一个数量级). 第二个更精细的数据集提供了各种费米能级(在费米能级附近-1.5到1.5 eV, 能量增量为0.005eV)和温度(范围从100-1300K, 增量为100 K)的电子传输特性. 表1和表2列出了两个数据集中可访问的输送物理量. 我们应该注意, 对掺杂水平值不在 $10^{16}-10^{20}cm^{-3}$ 的固定掺杂范围内的用户掺杂水平, 可以使用更精细的数据集来计算更精确的掺杂(参见使用说明).

## 局限性

在这里,我们想讨论在这项工作中做出的主要近似:恒定的弛豫时间.观察电导率张量,弛豫时间 $\tau_{i,\mathbf{k}}$ 以一般形式写为张量,取决于能量和方向. 所有散射事件都可以影响电子传导,诸如杂质散射,声子散射等,都包含在这个参数. 因此, 将该项视为常数意味着它被建模为各向同性并且在  $\kappa_{b}T$ 的能量尺度上没有强烈变化. 这是一个很牵强的近似值, 已知它与几种材料的实验值相差甚远. 为了考虑到不同的散射过程, 提出并测试了许多模型(无论经验还是第一原理). 然而, 超越恒弛豫时间的此类模型更复杂, 并且引入了对诸如电声相互作用,变形势,弹性常数和介电常数等的其他材料性质的依赖性. 因此,它们在数千种材料的高通量规模上更难以使用. 我们应该强调的是, 虽然有更精确的方法存在, 特别是对于单一材料的详细研究, 恒弛豫时间近似是第一性原理筛选是非常有用的, 并且具有越来越普遍的趋势(但是用户应该铭记它的局限性).

由于热导率和电导率在我们的恒定弛豫时间框架内与弛豫时间成正比关系, 再此, 我们使用单位弛豫时间内的物理量. 因此, 用户可以简单地将这些值乘以恒定的弛豫时间(通常为 $10^{-14}-10^{-15}s$, 以获得最终的传输属性. 塞贝克系数不依赖于恒定弛豫时间近似内的弛豫时间. 我们提醒, 在这种近似中塞贝克系数的符号对于某些金属来说是错误的.

另一个问题与k点网格有关. 其密度对于通过插值计算的传输特性的精度非常重要. 傅立叶插值的一个已知问题是在带交叉附近不能正确地确定能带导数. 在参考文献中已经分析了这个问题, 表明如果带交叉不太接近费米能级, 则带的导数和曲率不会受到太大影响. Uehara 等人提出了一种可能的解决方案.另外, 如参考文献中所述, 这个问题仅局限于高对称线. 密集的k点网格通常会解决这个问题, 并且由于本文中的性质是相对于k点和能带的平均, 因此它们的准确性不会受到显着影响.当考虑有限数量的材料时,通常使用非常致密的k网格. 例如, Madsen 在完整的布里渊区域中建议 $64\cdot10^6/V$ k点. 由于我们处理的是数千种材料, 因此该项目中使用的k点网格较粗糙. 它代表了计算时间和准确度之间的折衷. 但是,我们在此强调,我们使用验证方法(参见验证部分), 该方法测试带结构插值的质量, 并评估k点网格是否足够密集以避免插值方案的任何大的失败.

最后, 我们已知, 在这项工作中使用的标准密度泛函(GGA)会低估带隙. 我们由其发现, 我们预测带隙偏差小于约 $10\kappa_BT$, 但真正的间隙若高于此值可能会使预测性能出现较大误差.

## 属性

表1和表2中报告了两个数据集中包含的属性.每个属性都存储在字典中,除了有效质量外,还计算了各种掺杂类型,温度,掺杂水平和数据类型.所有这些情况都可以通过表5中报告的子键访问.

在第一个数据集中,存储以下属性:塞贝克系数,电子电导率(除以τ)和不同掺杂(类型和水平)和温度下的电子热导率(除以τ); 不同温度下载流子和霍尔载流子浓度(作为费米能级的函数)(mu_steps); 掺杂浓度为 $ 10^{18} cm^{ -3}$, 温度为300 K时的有效质量, **其中n和p型分别表示电子和空穴质量**; 费米能级值.

在第二个数据集(包含供专家用户使用的附加信息)中,存储以下属性:塞贝克系数,电子电导率(除以τ)和不同温度下的电子热导率(除以τ)作为费米能
级的函数(能量步骤包含为mu_steps键的值); 300 K时不同掺杂水平(n型和p型)的有效质量; 对应于每种掺杂水平(n-和p-型)的化学势的值; 费米能级值.

在两个数据集中,对于塞贝克系数,电子电导率(除以τ)和电子热导率(除以τ))存储了其全张量及其特征值(按升序排序).对于有效质量,仅存储全张量的排序(按升序)特征值.关于霍尔载流子浓度,仅存储完整霍尔张量的平均迹(对角线的和). 我们提供特征值, 因为它们是轴选择的不变量. 因此它们对查询非常有用.例如,搜索高塞贝克材料将涉及对塞贝克特征值的查询.为了便于查询,特征值按升序排序(第一特征值是最小的特征值).可以通过最后和第一特征值之间的差异直接评估属性的各向异性.我们强调所提供的特征值按升序排序,并且不包含有关相应主要方向的任何信息.为了获得晶体学方向和特征值之间的对应关系,我们建议研究全张量(和晶体结构信息)并应用寻找特征值和特征向量的算法(参见使用说明).我们还提醒说,有效质量仅用于半导体材料(金属没有有效质量?),即GGA或GGA + U中带隙大于零的化合物.

## 技术验证

### 验证插值精度

鉴于带结构的初始均匀k点网格可能不足以进行所有带结构的良好插值,我们在存储数据之前执行了后处理检查.将插值给出的沿对称线的带结构与在材料项目中报告的用更密集的k点网格明确计算的带结构进行比较.这种比较已经在pymatgen中实现.

比较分为两个过程. 首先,我们评估两个能带之间的相关距离(如在scipy.spatial.distance.correlation类中定义的;基本上1-ρ,其中ρ是皮尔森系数), 以确定它们是否表现类似. 
其次,我们通过平均每个段中k点数的绝对差值之和来评估每个高对称路径段的能量距离: $pass$. 该检查的输出存储在表4中描述的字典中.它主要包含相关距离和能量距离(对于每个区段和整个区域),对于非金属的最后(第一)四个价(传导)带,或者对于金属在费米能级以上和四个以下的四个带.对于快速筛选,它还包含一个警告标志(参见表4中的`acc_err`键),相关距离和能量距离(对于整个能带),设置为True当他们在八个频段的平均值高于0.03时. 根据该阈值,大约2.5%的GGA / GGA + U带结构对相关性有警告,4%对能量距离有警告.应谨慎使用带插值警告的数据.

# 使用说明

我们的论文提供了在恒定弛豫时间近似内从DFT(GGA/GGA+U)带结构和玻尔兹曼输运计算得到的约48,000种材料的传输性质数据集. 这类数据已被用于深入了解电子或热电材料的基本材料特性. 虽然我们警告用户始终小心使用此数据集(请记住我们的方法的限制), 但此数据库构建了有关材料传输属性的材料搜索和数据挖掘的强大基础.

需要澄清BolzTraP提供并在我们的数据集中使用的掺杂的含义. 掺杂水平不是载流子的总量. (等式(10))指出掺杂浓度是未掺杂材料中存在的每体积电子数与给定费米能级下每体积电子数之间的差.为了更好地理解,我们可以将掺杂浓度定义为与给定费米能级的自由电子数相比的过量空穴数.现在更清楚的是,掺杂浓度对于p型掺杂是正的,其中存在比自由电子更多的空穴,并且对于n型掺杂是负的,其中相反的情况是正确的.我们注意到本质上产生的移动载流子,导致相同数量的空穴和自由电子,不被视为掺杂浓度的一部分.例如,金属和小间隙材料可能包括一个固有的载流子浓度,与本工作中报告的掺杂水平不同.对于这样的材料,可以使用例如霍尔载流子浓度直接获得总载流子浓度.我们还提醒用户要记住,霍尔载流子浓度不一定与掺杂浓度相同.当半导体高度退化时,这种相等性仅对于抛物线带是精确的. 当比较实验和理论结果时,应该记住,**实验提供的绝大多数情况下载流子浓度是霍尔载流子浓度**. 此外,载流子浓度的这种定义影响由等式(11)给出的电导率有效质量的评估.因此,我们仅报告GGA或GGA + U中能隙高于零的材料的有效质量,我们建议用户小心使用能隙小于0.1 eV的材料的有效质量.  

如上所述,我们在第一个数据集中提供固定掺杂水平下的所有传输属性.如果需要不同掺杂水平的特定属性的值,则用户可以使用提供费米能级函数的第二数据集.当设定目标掺杂时,用户可以找到费米能级将在所需温度下提供该掺杂水平并使用与该费米能级和给定温度相对应的性质. 在两个数据集中,我们存储了塞贝克系数的全张量及其排序(按升序)特征值,电子电导率(除以τ)和电子热导率(除以τ).还提供了有效质量的特征值(也按升序排序).如果需要沿着晶体的特定方向的特性值,则必须使用全张量和结构.同样重要的是要注意,当需要派生属性(例如,功率因数小号2个σ),对特征值进行操作是错误的(因为它们可能不参考相应的方向).因此,我们强烈建议改为对完整张量执行操作.可以通过在得到的全张量上运行适当的算法来获得特征值.
