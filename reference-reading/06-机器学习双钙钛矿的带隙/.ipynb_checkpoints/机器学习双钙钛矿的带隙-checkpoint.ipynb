**摘要**

对双钙钛矿的带隙进行快速准确预测的能力对于一系列应用具有很大的实际意义. 因为高保真带隙的量子力学计算是计算时间密集的,因此在高通量研究中是不切实际的,但基于信息学的统计学习方法可能是一种很有前景的选择.在这里,我们展示了一种系统的特征工程方法和一个强大的学习框架,用于高效准确地预测双钙钛矿的电子带隙.在评估了一组超过120万个特征之后,我们将最低占据的Kohn-Sham能级和构成原子物种的元素电负性确定为最关键和相关的预测因子.

# 介绍

在最近的过去,巨大的化学空间的高吞吐量的勘探已显著的辅助材料的合理设计和发现过程. 用于计算/预测材料的性质(包括电子结构,热力学和结构特性)的大量开放访问数据库现在亦是可用的5,6,7. 材料科学家目前正在研究从材料大数据中提取知识和挖掘趋势的有效方法8. 因此,完善的机器学习统计技术(ML)正在逐步进入材料科学9. 这些数据科学和信息理论方法已经在化学信息学10,博弈论11,模式识别12,人工智能13,事件预测14 等领域取得了巨大的成功,现在正在为材料信息学定制,以帮助识别下一步代材料突破和工艺优化15,16.

鉴于过去的知识 - 一组有限的材料候选者的(感兴趣的)属性的高质量数据 (在一个明确定义的化学空间内)- 基于信息学的统计学习方法提供了一个有效的途径对其中的新化合物进行高保真度预测(在目标化学空间). 最近的一些使用信息学性能预测材料的实例包括: 分子预测, 周期表属性,过渡态23,电位24,25,结构分类26,27,28,介电性能,量子力学的自洽的解决方案 29和带隙的预测.

在这个贡献中,我们的目标是建立一个经过验证的统计学习模型,用于特定类别的复合氧化物,即双钙钛矿.如图1a所示的双钙钛矿结构由化学式 $AA'BB'O_6$ 表示; 其中A和A'阳离子通常具有较大的半径并且具有12倍的配位(八个氧八面体的中心),而相对较小的B和B'金属离子占据氧八面体中的六倍配位位置(氧八面体的中心). A位离子通常具有+1,+ 2或+3标称电荷状态,而B位阳离子的电荷状态由系统的总电荷中性控制. 我们认为双钙钛矿氧化物是一种感兴趣的材料类,这是由于钙钛矿骨架在适应广谱原子取代时所具有的化学柔韧性,以及由双钙钛矿32跨越的组成和构型空间的广阔性.

快速且准确地预测双钙钛矿带隙的能力对于对带隙具有预定约束的材料的一系列应用非常感兴趣,例如,闪烁33,光伏34和催化剂35等等.在密度泛函理论中使用的局部和半局部函数 - 当前用于电子结构计算的主力 - 具有严重低估带隙的众所周知的缺陷. 更先进的方法,例如GW方法36和混合泛函37计算时间密集,因此在高通量研究中是不切实际的(旨在用有针对性的带隙筛选有希望的候选者). 这是寻求开发统计(或机器)学习模型的最重要原因之一,其中人们可以使用材料的易于访问的属性(也称为特征)以有效且准确的方式直接预测其带隙. 我们的主要目的是开发一种有效的和预测模型,其建立的数学关系(或映射)可以预测在化学空间中的材料(i,数据点)带隙, 一个Ω维的特征矢量 $f_i$. 这里,Ω-D向量 f_i(也称为描述符)由Ω不同的特征组成,并且唯一地描述了材料 i. 还希望具有既简单(即,具有足够小的Ω)又相当准确的模型. 在这里,我们报告一个特征选择(即,如何找到最佳Ω-D特征向量)和学习框架(即,如何建立带隙和特征向量之间的映射),以便有效和准确地预测双电子带隙钙钛矿.

根据阳离子的性质, 可以在双钙钛矿中产生各种类型的阳离子顺序. 对于双重取代的$AA'BB'O_6$ 类钙钛矿,有三种常见的方式(cations
at each of the two sublattices can order,两个子晶格中的每一个的阳离子都可以排序),导致总共九种不同的有序排列. 具体而言,A和A'(B和B')阳离子可以形成层状,柱状或岩盐排列. B位点亚晶格最常见的排序类型是阳离子在所有三个维度上交替的类型,模仿**岩盐型**有序亚晶格以有效地适应由于两种阳离子的尺寸不匹配而产生的任何局部应变. 较不常见的是,B位阳离子可以形成分层顺序,其中它们仅在一个方向上交替并且在另外两个正常笛卡尔轴中形成连续层.然而,很少会发生柱状秩序,其中两种不同的阳离子在两个正交方向上交替,但沿第三个方向形成一个连续的列. CMR数据库报告所有双钙钛矿的都是阳离子岩石排序.

报告的带隙(图1c,d)使用密度泛函理论(DFT)计算,如GPAW代码42中实施的那样(使用GLLB势), 由 Kuisma 等人进一步针对固体进行了优化. GLLB泛函具有用于评估导数不连续性的内置处方,其被添加到Kohn-Sham带隙以校正常规DFT内的带隙低估. 实际上,已发现几种单金属氧化物的GLLB-SC带隙与相应的实验值非常一致(补充资料). 此外,GLLB-SC功能最近针对更先进和要求更高的特征值自洽GW方法进行了测试,并且已经证明可以很好地与20个随机选择的系统的带隙相一致,形成一组非常规的三元和四元化合物(来自材料项目数据库). 最后,我们还要注意,尽管与GW方法相比计算成本显着低,但GLLB-SC功能与使用本地或半本地功能的传统DFT计算相比,时间花费大约是其两倍.

任何针对学习预定材料属性的ML方法都依赖于两个主要成分:学习算法本身和学习(或训练)数据集中材料的数字表示(以描述符的形式).为给定的预测任务确定适当且最合适的指纹是当前由社区积极追求的主要挑战之一.这个数值表示的具体选择是完全取决于应用以及在高级别功能方面的一些提议(例如,d波段中心,元素电负势和电离势,原子价轨道半径),拓扑特征49,原子径向分布函数,组成,构型和端基指纹, 已经作出.无论所追求的具体选择如何,期望表示满足某些基本要求,例如关于材料变换的不变性,例如相似元素的平移,旋转和置换.另外,还希望指纹识别在化学上是直观的并且在物理上是有意义的.

本研究采用的整体工作流程,将有效的特征搜索策略与最先进的统计学习方法相结合,如图2所示.我们的系统方法首先确定形成双钙钛矿结构的每种阳离子物质的七个原子(或元素)特征. 这些**元素特征(孤立的中性原子的鲍林的电负性(χ),电离势(I),最高占据原子能级(h),最低未占据原子能级(l)和s- ,p- 和d- 价轨道半径r_s,r_p和r_d)**是易于获得的, 并且是A位和B位原子成分的物理直观属性. χ,I,h和l自然与能量尺度的形成有关(与带隙的预测相关), 但是原子价轨道半径的选择(为什么选择这几个参数)是基于它们在AB二元固体的所显示出的优异性能. 对四个原子中的每一个取这七个元素特征,占据A或B位置,形成我们的二十八个元素特征的起始集合. 有关功能集的更多详细信息,请参见"方法"部分.

此时值得注意的是,研究中的双钙钛矿结构在两个A位阳离子以及两个B位阳离子(即 AA'BB'O 6,A')的交换方面是不变的. ABB'O 6和AA'B'BO 6都是相同的系统.为了将这种结构对称性结合到模型中,我们将上述28个元素特征对称化,使得它们能够反映双钙钛矿晶体结构A位和B位子晶格的基础对称性. 这是通过获取每对元素的绝对和 $f_A+f_{A'}$ 和绝对差 $f_A-f_{A'}$ 来实现的. B位阳离子的特征也以类似的方式转变. 在特征级别建立这种对称性确保了所考虑的ML模型对对称独特系统能预测出相同的带隙,而不管两个A位和两个B位原子的任何特定标记. 由此实现的这组28个对称特征在下文中称为**主要特征**.

此时,我们采用双重(两步)路线进行特征选择. 虽然主要特征可以直接用于统计学习模型,但我们还考虑以受控方式构建的大量**联合或复合特征,以允许特征级别的非线性**. 复合特征集按以下方式构建:6个原型函数,即 $x,x^{ 1/2},x^2,x^3,ln(1+ x), e^x$, x是28个主要特征之一. 这会立即生成168个特征(28\*6). 简单地将单个函数的这些特征一次乘以两个或三个,分别导致额外的16,464和1,229,312个特征. 因此,这种方法为我们提供了1,245,944个复合特征,每个特征都涉及多达3个主要特征. 最后,采用基于最小绝对收缩和选择算子(LASSO)的模型选择来下选一组40个复合特征,这些特征被认为与带隙的预测最相关. 我们在此注意到,这种使用LASSO创建大量初始化合物特征和向最相关特征下调的策略最近已成功用于识别新的晶体结构分类器51. 基于LASSO的形式主义也被用于识别合金簇膨胀的低维表示52.

主要描述符从3-d到4-d可以发现在模型预测的性能方面显著改善. 例如,测试集上的平均R^2从0.69增加到0.90,并且平均rms误差从~0.87eV减小到~0.50 eV. 然而,超越4-D描述符只会导致边际改进. 例如,对于16-D描述符(包含所有主要特征),获得的平均测试集R 2是~0.94,仅略好于4-D描述符的平均测试集. 图4b-d使用代表性的训练/测试集拆分,将4-D描述符的KRR预测性能与16-D主要描述符和16-D复合描述符在单独的奇偶校验图中进行比较. 可以看出,虽然具有较高维度描述符的KRR模型中的训练集性能明显更好,但是这些模型的测试集性能可以被认为与4-D描述符的性能相当(或仅略微更好).

# 讨论

我们现在检查组合成**4-D主描述符**的各个主要特征. 这些特征包括: $l_A^+, l_A^-, \chi_B^+, \chi_B^-$, 分别表示两个A位元素最低占据能级的绝对和与绝对差, 两个B位原子电负性的绝对值和绝对差. 我们还注意到这些元素特征的两个共轭对一起出现, 并且没有出现具有价轨道半径的主要特征. 此外,该描述符相对于A位原子和B位原子所所参与的特征非常平衡. 除了化学上直观,优雅和对称之外,所识别的描述符也简单且易于访问. 总是希望具有基于更简单(即,低维)和直观描述符的ML预测模型,因为对于高维复杂描述符,总是存在过度拟合的危险,导致较差的模型概括性.因此, 优先选择4-D主描述符而不是在16-D描述符, 我们正在处理一些模型精度,以简化模型和更好的模型概括性.

为了进一步测试模型的可预测性,我们使用经过随机选择的90%双钙钛矿数据集训练的交叉验证KRR学习模型来预测原始53个单亲钙钛矿的带隙. 我们注意到,单钙钛矿,由于制约条件A = A'和B = B',只有四个当中的两个特征存在(对于所有的单钙钛矿,$l_A^-=0, \chi_B^-=0$ ). 图5比较了模型预测的带隙与使用DFT和GLLB-SC功能计算的带隙. 鉴于该模型从未在单个钙钛矿上进行过训练,并且四个主要特征中只有两个在单个钙钛矿中有效存活, 但是这种预测的性能相当显着.

最后,我们评论ML模型的适用性的局限性和领域. 所提出的模型适用于所考虑的化学空间(即上述A位和B位阳离子的选择)和非磁性AA'BB'O 6型钙钛矿,其可分成两个电荷中性ABO3和A'B'O3单钙钛矿. 发现不能以这种方式分解的双钙钛矿化合物的测试性能差,这并不奇怪,因为学习模型从未接受过这种化合物的训练. 将ML模型扩展到这些化合物并考虑其他可能的A-和B-位阳离子排序仍有待在未来的研究中进行. 通过使用它来预测其他材料类别的带隙来检查所识别描述符的一般适用性也是有趣的,这与钙钛矿和相关化学物质完全不同.

总之,我们提出了一个强大的ML模型以及一个简单的元素描述集,用于有效预测双钙钛矿的电子带隙. 通过搜索大部分特征空间来识别所提出的最优描述符集,该特征空间涉及超过约120万个描述符,这些描述符通过组合简单的元素特征(例如电负性,电离势,电子能级和组成原子物种的价轨道半径)而形成. 这里开发的基于KRR的统计学习模型在数据库上进行训练和测试,该数据库由在密度泛函理论框架内使用GLLB-SC功能计算的~1300双钙钛矿的精确带隙组成. 从采用的学习框架中得出的最重要的化学见解之一是,通过A位元素的最低占据能级和B位元素的电负性,主要控制带隙(因此可以有效地学习). 平均测试集RMS发现仅具有四个主要特征(即,4-D主要描述符)的交叉验证模型的误差为0.5eV,其利用主要(复合)16-D描述符进一步减小到~0.37eV(0.36eV). 通过其预测几种单一钙钛矿的带隙的能力进一步证明了训练模型的样本外预测性能. 最后,我们已经表明,通过构造几个二维特征-属性等值线图,**可以在视觉上合理化模型的预测性能**. 我们相信这里提出的ML方法是通用的,可以应用于具有给定晶体结构的受限化学空间中的任何材料类,以有效地预测带隙.

# 方法

## 特征集的详细信息
对于特征集积累,我们从双钙钛矿结构中每个金属原子A的7个原子特征开始.这些主要原子特征是鲍林的电负性(χ),电离势(I),最高占据原子Kohn-Sham能级(h),最低未占据原子Kohn-Sham能级(l)和s - ,p - 和d - Zunger's价轨道半径r_s,r_p和r_d的孤立中性原子. 电离势和鲍林的电负性数据来自文献, 分离原子物种的最高占用最低未占用Kohn-Sham能级使用GGA-PBE交换相关函数计算.

## 机器学习模型
在当前基于相似性的KRR学习模型中,测试集中系统的带隙由整个训练集上的加权高斯的总和给出.作为模型训练过程的一部分,通过最小化表达来执行学习: $\sum^N_{i=1}( E^i_{g,KRR} - E^i_{g,DFT})^2 + \lambda \sum^N_{i=1} \omega_i^2$, 其中 $E^i_{g,KRR}$ 是KRR估计的带隙值, $E^i_{g,DFT}$ 是DFT计算的带隙值, $\lambda$ 是正则化参数. 这个最小化问题的显式解(explicit solution)是 $\omega = (\kappa +\lambda I)^{-1} E_{g,DFT}$ ,其中 $I$ 是单位矩阵, $\kappa_{ij} = exp(-\frac{1}{a\sigma^2} \left\| f_i-f_j \right\|)$ 是训练集中所有化合物的核心矩阵元素. 参数 $\lambda,\sigma$ 是使用对数缩放的精细网格在五重交叉验证的内循环中确定.我们注意到,仅使用训练数据进行KRR训练和超参数确定,并且在训练过程期间KRR模型从未看到测试集样本.
