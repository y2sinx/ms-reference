**摘要**:  
我们介绍了C2DB(计算2D材料数据库),它组织了大约1500种二维材料的各种结构,热力学,弹性,电子,磁性和光学性质,这些材料分布在**30多种不同的晶体结构上**(mx2占多少种?). 材料特性通过最先进的密度泛函理论和多体扰动理论(以及~250材料的Bethe-Salpeter方程)系统地计算,遵循半自动化工作流程以获得最大的一致性和透明度. C2DB完全开放,可以在线浏览(http://c2db.fysik.dtu.dk)或完整下载. 在本文中,我们描述了数据库背后的工作流程,概述了当前可用的属性和材料,并探索了数据中的趋势和相关性. 此外,我们确定了大量新的可能合成的2D材料,这些材料具有针对自旋电子学,(光电子学)电子学和等离子体学应用的有趣特性. C2DB为快速扩展的2D材料系列提供了全面且易于访问的概述,并形成了新2D材料和范德华异质结构的计算建模和设计的理想平台.

# 简介

在过去的十年中,原子薄的二维(2D)材料已经作出自己的方式来的几个研究领域前沿包括电池,(电)催化,电子和光子学[ 1,2 ]. 这种发展是由原子级薄晶体的有趣且易于调节的特性所促成的,并且由于不断发现新的2D材料以及横向[ 3 ]和垂直[ 4 ] 2D异质结构的新兴概念而得到了推动,这为设计具有定制和卓越性能的材料.

到目前为止,已经合成或剥离了超过50种化合物作为单层(参见图7). 这些包括众所周知的单元素晶体(Xenes,例如石墨烯,磷芴)[ 5 ]及其配体官能化衍生物(Xanes,例如CF,GeH)[ 6 ],过渡金属二硫属化物(TMDCs,例如MoS 2,TaSe 2)[ 7] ],过渡金属碳化物和-氮化物(MXenes,例如钛2 CO 2)[ 8 ],III-V族半导体和绝缘体(例如氮化镓,BN)[ 9,10 ],过渡金属卤化物(例如值CrI 3)[ 11,12 ],后过渡金属硫族化合物(例如气体和GASE)[ 13,14 ]和有机-无机混合钙钛矿(例如铅(C 4 H ^ 9 NH 3)2我4)[ 15 ]. 然而,已知的单层膜只是更大冰山的一角. 事实上,最近的数据挖掘的研究表明,几百二维材料可以从已知的分层块状晶体[剥落16 - 19 ].在目前的工作中,我们采用基于组合晶格装饰的2D材料发现的补充方法,并识别另外几百个先前未知且可能合成的单层膜.  

在寻找具有定制特性或新颖功能的新材料时,第一原理计算正在发挥越来越重要的作用. 计算能力的不断提高以及理论方法和数值算法的显着进步已经将该领域推向了这样的程度,即第一性原理计算在精度方面与实验相当,并且在速度和成本方面大大超过了它们. 一个多世纪以来,关于结构,热学和电子特性的实验数据库一直是材料科学的基石,在过去十年中,实验数据通过从第一原理计算中获得的计算数据的爆炸性增强而得到增强. .20,21 ]. 一些较大的储存库,一起包含数百万材料条目,是材料项目[ 22 ]时,自动流量为材料发现(AFLOWLIB)[ 23 ],则开放量子材料数据库(OQMD)[ 24,25 ],和新材料发现(NOMAD)存储库[ 26 ].

计算材料数据库的优点很多. 最明显的是,它们促进了研究数据的公开共享和比较,同时减少了重复工作.此外,它们通过提供对公共参考系统的轻松访问来支持新方法的开发和基准测试[ 27 ]. 最后,数据库启用的机器学习技术的应用,以确定材料的空间深刻而复杂的相互关系,并利用它们设计量身定制的特性材料和加快新材料[发现28 - 30 ]. 计算数据库面临的挑战之一是存储数据的质量,其取决于数值精度(例如,所采用的k)点网格和基础集大小)和所用物理模型的准确性(例如交换相关功能). 大多数现有的计算数据库存储标准密度泛函理论(DFT)计算的结果. 虽然这些方法在适当地进行时对于诸如结构和热力学性质的基态性质是非常可靠的,但它们对于诸如电子能带结构和光学吸收光谱的激发态性质通常不是定量精确的.

与体相材料的数据库相比,2D材料的数据库仍然很少且不太发达.早期工作中使用DFT探索III-V族蜂窝状格子[单层的稳定性和电子结构31,32 ]和类MX的2过渡金属二硫属化物和氧化物[ 33 ].后来,通过对无机晶体结构数据库(ICSD)进行数据滤波,确定了92个实验已知的层状晶体,并在DFT水平上计算了它们的电子能带结构[ 34 ]. 另一项DFT研究也侧重于稳定性和能带结构,研究了大约100种从不同结构类别中选择的2D材料[ 35].为了克服DFT的已知限制,建立了用于50个半导体TMDC的具有多体G 0 W 0带结构的数据库[ 36 ]. 最近,该材料项目的数据挖掘和实验结晶结构的数据库中的精神[ 34 ],导致接近千个实验已知的分层晶体的鉴定从中单个层可能被剥落[ 16 - 19 ]. 这些工作还计算了DFT水平的单层(或至少选定的子集)的基本能量,结构和电子特性.

在本文中,我们介绍了开放的计算2D材料数据库(C2DB),它为超过1500种不同的2D材料组织了各种ab initio计算的属性. C2DB的主要特点是:

* 材料:数据库完全侧重于2D材料,即通过已知晶体结构原型的组合晶格装饰获得的孤立单层.
* 一致性:所有材料的所有属性都使用相同的代码和参数设置在相同的工作流程后计算,以实现最大的透明度,可重复性和数据的一致性.
* 属性:数据库包含大量不同的属性,包括结构,热力学,磁性,弹性,电子,电介质和光学性质.
* 精度:使用混合功能(HSE06)以及超越DFT多体微扰理论(G 0 W 0)来获得定量精确的能带结构,并且从随机相位近似(RPA)和Bethe-Salpeter获得光学性质方程(BSE).
* 开放性:数据库可以自由访问,可以使用简单和高级查询直接下载和在线浏览.

用于在数据库中生成结构的系统组合方法不可避免地产生许多不稳定的材料, 因此不切实际且不可能在现实中合成. 然而,这种"假设"结构在许多情况下仍然有用,例如方法开发和基准测试,机器学习算法的测试和培训,趋势识别和结构 - 属性关系等. 出于这个原因,我们绘制出来除了最不稳定(因此化学上不合理)的化合物之外的所有化合物的性质. 然而,对候选结构的稳定性和可合成性的可靠评估是一个重要问题. 使用已经以单层形式实验合成的C2DB中的55种材料作为指导,**我们设定了假设的2D材料应该满足的标准**,以使其"可能是可合成的". 在这些标准的基础上,我们介绍一个简单的稳定性量表,以量化候选材料的动态和热力学稳定性. 在最初的约1900个单层分布在32种不同的晶体结构中,**我们发现350种是最稳定的类别**. 除了55个实验合成的单层外,该组还包括来自实验已知的vdW层状体相材料的约80个单层,因此包括约200个全新的和可能合成的2D材料.

### 磁性
最近,据报道单层CrI 3表现出铁磁有序,居里温度为45 K [ 12 ].这包括显示出磁性顺序纯2D材料的第一实施例和目前还为新的二维材料激烈搜索与磁性顺序持续高于室温[ 150 - 152 ].

对于2D材料,磁性顺序仅在存在磁各向异性(MA)的情况下在有限温度下持续存在.实际上,根据Mermin-Wagner定理,除非旋转的旋转对称性被破坏,否则在2D中磁性顺序是不可能的[ 153 ].具有平面外易轴的有限MA打破了Mermin-Wagner定理背后的假设,并使得有限温度下的磁性有序.因此,2D材料中磁性有序的临界温度将强烈依赖于各向异性.

MA源自自旋轨道耦合,并且在此定义为磁矩的面内和面外取向之间的能量差,参见等式(4).根据我们的定义,负MA对应于平面外易轴.我们注意到C2DB中的大多数材料在平面内几乎是各向同性的.因此,如果易轴位于平面中,则自旋将表现出近似的平面内旋转对称性,这禁止在有限温度下的磁性有序.由于自旋轨道耦合对于重元素变大,我们通常期望对于含有较重元素的材料发现较大的MA.通常,MA的大小很小.例如,对于居里温度为45 K 的单层CrI 3 [ 12我们发现每个Cr原子的MA为-0.85 meV,与先前的计算一致[ 154 ].虽然很小,但MA对于在有限温度下出现的磁性顺序是至关重要的.

# 结论和展望

C2DB是一个开放式数据库,具有二维材料的计算属性.它目前包含超过1500种材料,分布在32种不同的晶体结构上.在采用最先进的DFT和多体方法的高通量,半自动化工作流程之后,计算各种结构,弹性,热力学,电子,磁性和光学性质.随着新结构和房产的增加,C2DB不断增长; 因此,本文提供了数据库当前状态的快照.可以使用简单和高级查询在线浏览C2DB,可以在Creative Commons许可下通过https://c2db.fysik.dtu.dk/免费下载.

C2DB中的材料包括实验已知的和未先前合成的结构.它们是通过不同2D晶格的组合装饰以系统方式生成的.完整属性工作流程仅针对发现动态稳定且具有负形成热的结构执行.我们采用自由稳定性标准,以便排除可能通过外部手段如衬底相互作用或掺杂来稳定的潜在有趣材料,即使它们在独立形式下不稳定.作为一个重要且相当独特的特性,C2DB采用超-DFT方法,例如多体GW近似,随机相位近似(RPA)和Bethe-Salpeter方程(BSE).这些方法对于获得带隙和光谱等关键特性的定量准确描述至关重要.这对于2D材料尤其重要,因为在减小的尺寸中进行弱电介质筛选,这倾向于增强多体效应.为了最大限度地提高C2DB中数据的透明度和可重复性,本文提供了所有相关参数.此外,用于生成数据的所有脚本都可以在GPL许可下免费下载.

除了明显用作查找表之外,C2DB还提供了大量记录良好的高质量计算,使其非常适合不同代码和方法的基准测试和比较.大量不同的可用属性使C2DB成为探索结构 - 属性关系以及应用和推进材料科学中机器学习方法的游乐场.此外,C2DB应该可以作为开发更复杂的二维结构(如范德华异质结构)的理论模型的垫脚石(见下文).

正如本工作所报道的,基于组合筛选方法,我们已经确定了许多具有有趣特性的新的,可能合成的2D材料,包括具有大磁各向异性的铁磁体,具有高固有载流子迁移率的半导体,以及具有可见频率的等离子体的金属范围.这些预测均基于完美结晶材料的计算属性.虽然原始晶体构成了重要的基线参考,但它仍然是任何真实材料的理想模型.将来,将数据库扩展到具有吸附物种和/或点缺陷的单层将是有趣的.这不仅可以对磁性和(光)电子特性进行更真实的评估,还可以促进e的2D材料的设计和发现.G.电池电极和(电)催化[166,167 ].
