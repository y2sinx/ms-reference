**摘要**:  
我们提出了一个全面的第一原理研究,在2H和1T六方相半导体单层过渡金属二硫化物和氧化物的电子结构方面. 具有自旋轨道耦合的准粒子(QP)能带结构在G0W0近似下被计算,并且与不同密度泛函理论描述进行比较. 并讨论了与2D材料GW计算收敛相关的缺陷的可能的解决方案. 单层带边位置(相对于真空)用于估计各种异质结界面处的带对准. 根据电子结构, 对带结构对面内晶格常数的灵敏度进行了分析和合理化. 最后,从QP能带结构获得q依赖的介电函数和有效的电子和空穴质量,并用作2D氢模型的输入以估计激子结合能. 在整篇论文中,我们关注电子结构中的趋势和相关性,而不是具体材料的详细分析. 所有计算数据都可在开放数据库中获得.

# 简介

由于其独特的光电特性, 目前正在深入研究原子级薄的二维(2D)材料,例如石墨烯,六方氮化硼和过渡金属二硫化物(TMD). 具有化学式MX2(X = S,Se,Te;M =过渡金属)的TMD代表包含半导体和金属的特别有趣的2D材料类别.(1)例如,显示MoS2,MoSe2,WS2和WSe2在其厚度减薄至单层时经历从间接带隙材料到直接带隙材料的转变. (2-6)与它们与光的强烈相互作用(7,8)和相对较高的电荷载流子迁移率(9,10)这开辟了在包括超薄场效应晶体管在内的一系列应用中使用少层TMD的可能性, (11)光电探测器,(12-15)发光二极管,(16)和太阳能电池. (17,18)此外,2H单层结构中缺乏反转对称性导致价带的自旋轨道驱动分裂,这反过来允许电荷载流子的谷选择性激发. (19-22)增加了通过应变调整电子特性的可能性,(23)介电屏蔽,(24)静电门控,(25,26)纳米结构,(27)或通过将各个2D材料组合成范德华异质结构,(28,29)很明显,单层TMD作为基础物理平台和纳米级器件应用的构建模块具有巨大潜力.  
迄今为止,单层TMD中的光电子研究主要集中在基于Mo和W的化合物上,这些化合物的(光学)带隙在1.6-2.0 eV,(2,3,5,6,30)范围内显着大于光伏和晶体管应用的理想值. (31)为了将2D材料的使用从基础研究的水平提高到实际应用,必须扩大可用的2D构建块的空间, 不仅仅是目前考虑的少数材料. 为此,不仅带隙而且绝对带边缘位置,有效质量和介电函数对于预测给定2D材料的有用性将是至关重要的.  
体相TMD的层间(interlayer)强度相似(大约20兆电子伏/A^2)(32)表明,对于许多不同的TMD的单个层的剥离应该是可行的. 实际上,已经证明了TaSe2, NbSe2, NiTe2和MoTe2的纳米片的液体剥离. (33)在这方面,值得注意的是,已有40多种TMD以体相形式出现,可以形成新2D材料的基础. (34)这种2D单层在环境条件下的稳定性是一个关键问题,但最近在MoS2中证明(在六方氮化硼中),它可以通过封装在保护层中来缓解.(35)  
在以前的工作中,Ataca等人. 在局部密度近似(LDA)中使用密度泛函理论(DFT)对88个单层TMD和TMO进行了广泛的稳定性分析,并确定了包括金属和半导体在内的52种稳定化合物. (36)虽然稳定性是他们的主要焦点,但他们还使用GW0近似计算了稳定化合物和一些选定化合物的LDA能带结构. 令人惊讶的是,他们得出结论,LDA与GW0的现有实验提供了良好的一致性显着高估了带隙. 这个错误的结论是基于光学和准粒子(QP)带隙之间的常见混淆. 前者在光学实验中探测,并且通过激子结合能低于QP间隙. 原子级薄半导体的特征之一是激子结合能非常大(大约1eV). 这导致QP和光谱之间的显着差异,这两者分别通过应用GW近似和Bethe-Salpeter方程的多体计算很好地再现.(37-43)

我们基于27种不同的金属,对2H和1T结构中各种单层TMD和TMO的电子结构进行了广泛的第一性原理研究. 作为参考,2H和1T相的原子结构示于图1一个,对应的布里渊区(BZ)与特殊K点示于图1b,在周期表中所考虑的元素突出显示在图1c中. 在216种被研究的化合物中,我们发现171是稳定的(由相对于标准状态的负形成热定义). 这些结果显示了与参考文献36基于LDA的稳定性分析的一致扩展. 在171个稳定的单分子层中,我们重点关注51种被发现为非磁性和非金属的化合物. 对于这些材料,我们使用包含自旋轨道耦合的G0W0近似计算带结构以及价带和导带边缘相对于真空的绝对位置. 绝对G0W0准粒子能量的收敛特别苛刻,因此我们将详细讨论这个问题. G0W0使用不同的交换相关函数将带隙和带边缘位置与Kohn-Sham DFT进行比较. 我们发现带隙通常由GLLB-SC功能良好地再现,而LDA提供了对**带隙中心**的令人惊讶的良好描述. 相比之下,发现从组成原子的电负性估计带边缘位置的经验公式显着偏离第一原理结果,因为**电荷从金属转移到氧或硫属元素原子**和相关的静电势(降低相对于真空的电子带能量). 我们进一步计算(静态)q所有化合物的独立介电函数,并讨论了准二维介电筛选的一些基本性质. 有效电荷载流子质量源自G0W0能带结构,并与介电函数一起用作激子结合能的有效2D模型的输入.

总的来说,我们的结果揭示了所研究材料的电子特性的很大程度的变化. 例如,材料MX2(X = S,Se,Te; M = Cr,Mo,W)在0.9-2.5eV范围内具有直接QP带隙,而所有其他化合物在0.5-7.0eV范围内具有间接间隙. 对于硒化物和碲化物,带隙中心(相对于真空)对于一些氧化物跨越-8eV至高于-5eV. 有效质量变化几乎2个数量级,q依赖介电函数也是如此.
所有计算数据包括弛豫结构,DFT和G0W0能带结构,绝对能带边缘位置,有效质量和介电函数,可在计算材料库中在线获取(http://cmr.fysik.dtu.dk/).

# 计算方法
所有计算均使用GPAW代码中实现的PAW方法进行.(44) GPAW支持三种不同类型的基组,即真实空间网格,数值原子轨道和平面波. 我们在目前的工作中使用后者,因为GPAW的激发态计算仅针对平面波实现.

## 原子结构

216个单层TMD和TMO的晶格常数(图1a)通过结构弛豫确定,使用Perdew-Burke-Ernzerhof(PBE)(45)交换相关(xc)功能与750eV平面波截止,18× 18×1 Monkhorst-Pack k点采样,周期性重复层之间20Å .对于2H和1T相,最小晶胞的晶格常数和氧或硫属元素原子的垂直位置被放松,直到所有的力都低于0.01eV/Å. 我们使用最小单位晶胞,并没有研究最近发现的某些1T金属化合物发生的对称性减少变形.(46-48) 因为这些失真是由金属-绝缘体转变(Peierls畸变)驱动的,所以我们不认为它们对于作为本工作重点的半导体材料是重要的.

## 电子结构

所有化合物的Kohn-Sham带结构使用**LDA,PBE和GLLB-SC(49) xc-泛函计算**. `Spin−orbit couplingwas included in a non-self-consistent manner by diagonalizing the total Hamiltonian consisting of the spin−orbit interaction (which is applied inside the PAW spheres) and the selfconsistently determined Kohn−Sham Hamiltonian`. 我们发现,如果我们在Kohn-Sham哈密顿量中使用G0W0能量而不是LDA能量,则对带结构的自旋轨道校正不变(小于0.02 eV). 对于10种代表性材料,我们对用GPAW获得的自旋轨道校正LDA能带结构与全电子线性化增强平面波ELK代码(50)进行了基准测试. 并且发现了极好的一致性(差异在0.02 eV以内).  

QP能带结构使用GPAW中实现的G0W0近似中计算.(24)我们使用LDA波函数(从KS哈密顿量的精确对角化得到)(600eV,30×30的K点)作为GW计算的输入. 用于构建屏蔽相互作用和自能的平面波截止值在150和500eV之间变化,并外推至无限截止能量,如下所述. 对于所有计算,用于构建屏蔽交互和GW自能的未占用轨道的数量被设置为等于平面波的数量. 频率依赖性表示在从0eV到包括在基础中的最高转变的能量的非线性网格上,其中网格间距从0.1eV开始逐渐增加并且在ω= 15eV时达到0.2eV. 频率网格通常包含300到350个网格点.在这项工作中应用的PAW潜力包括半微秒状态,即低于真空至少1 hartree的原子状态,而冻结核心中包括更深的状态.

在参考文献43中,我们证明了在2D材料的GW计算中使用截断的库仑相互作用的重要性.在本研究中,我们使用了Wigner-Seitz截断方案.(51)对于一组有代表性的材料,我们检查了当k点网格从30×30增加到45×45 时,QP带隙(更一般地说是绝对带边缘位置)的变化小于0.1eV . 我们顺便指出,2D系统的大多数先前的GW计算已经应用了显着更小的k点网格.(36,37,52) 如参考文献43中所解释的,`k点慢收敛的物理原因是强q-二维半导体的介电函数的依赖性`. 虽然3D半导体的ε(q)平滑地趋于q →0的恒定值,但对于2D系统,ε(q)= 1 + O(q)(图2). 因此,需要更密集的k点网格来捕获 q=0 附近的ε(q)的变化. 例如,当增加k时(点网格从15×15到30×30),2H-MoS 2的G0W0带隙减小0.4 eV. 因为ε(q)的强烈变化局限于周围的一个小区域q = 0,在该区域的细网格上对筛选的相互作用W(q)进行采样就足够了,而在BZ的剩余部分可以使用更粗略的采样(但是,这种非均匀采样在本文中没有使用). 我们强调这些事实仅适用于孤立的二维半导体,这在实践中意味着何时使用截断的库仑相互作用. 只有这样才是ε(q)= 1 + O(q). 相反,如果使用完整的1 / r库仑相互作用,则计算收敛得快得多,但取决于层间距离而错误的值. (43)介电函数和图2将在第3.6节中进行更深入的讨论.

最后,我们讨论了G0W0能量相对于平面波数N_G的收敛,用于表示屏蔽的相互作用和自能. 先前已经发现,体半导体和绝缘体的GW校正遵循1/N_G比例(53)这使得有可能将QP能量外推到无限基组限制. 根据我们的计算,截止能量来自150 eV,在某些情况下高达500 eV,我们观察到以下情况:(i)外推程序是必不可少的,可以校正150 eV截止值获得的QP能量高达0.5 eV.(ii)外推曲线的斜率对于不同的状态(波段和k点)可以是不同的,但是通常表示作为N_G的函数的减小.(iii)带隙趋于比绝对能带更快地收敛. 在参考文献53中,还表明PAW电位缺乏规范守恒会影响GW能量的收敛,如N_G增加了.对于更多局部化状态,特别是3d状态,其效果更大,其中违反规范守恒可能是显着的. 尽管可以构建保守的PAW潜力,但我们在目前的工作中并没有追求这一点.

我们没有提供与QP带隙的先前文献值的详细比较. 对于MoS2,在参考文献43中进行了这样的比较. 然而,大多数先前的用于2D TMD的GW计算使用相当粗糙的k点网格并且没有使用截断的库仑相互作用(两者都对计算的间隙具有显着影响(43)),旋转的重要性不总是包括轨道相互作用,以及间隙对面内晶格常数的敏感性(参见第3.4节),使得一般比较变得困难. 我们认为,目前工作的QP能带结构是最精心收敛的G0W0到目前为止,已报告了2D TMO和TMD的计算结果. 关于独立TMD单层的QP带隙,只有少数实验报告. 对于2H-MoS2,我们的2.48 eV的G0W0间隙与基于悬浮的MoS2的光电流光谱的参考15中报道的2.5 eV一致. 我们注意到,与我们之前的工作相比,这里获得的MoS2(2.48 eV)的带隙略小(43)(2.65 eV)主要是由于在目前的工作中包含了自旋轨道相互作用. 或者,可以从光学吸收或光致发光光谱推断出QP间隙,这在实验上更容易获得. 然而,这需要知道激子结合能,而激子结合能又取决于从基底的筛选.(56)

## 带的绝对位置

对于许多应用,不仅关注带和未占用带之间的距离,即带隙,而且带边缘相对于真空的绝对位置也是令人感兴趣的. 我们通过将带能量引用到层之间的真空区域中的Hartree势的渐近值来计算这些. 对于体相材料来说,这是一项艰巨的任务,因为它需要使用厚板来代表整体内部及其表面. 此外,Hartree电位取决于表面偶极子(在板的两侧),这使得该问题高度依赖于表面并且使得与实验的比较复杂化. 对于这里研究的单层膜,这些问题显然不存在,使它们成为带对准问题的基准系统的理想选择.

## 激子

原子级薄的2D半导体的最典型特征之一是激子的大结合能.(15,41,64)这样做的原因是由于尺寸较小而导致的筛选减少,从而在电子和空穴之间产生更强的吸引力(见前一节的讨论).用于从第一原理计算激子结合能的常规方法是Bethe-Salpeter方程(BSE).BSE在计算上要求很高,不适合大规模研究,如目前的工作.相反,我们使用最近开发的2D Mott-Wannier模型用于仅需要激子有效质量和准2D介电函数作为输入的激子.在现实空间中,模型需要2D薛定谔方程的形式

Mott-Wannier激子模型背后的四个基本假设是  
* 各向同性激子质量,

* 仅通过价带和导带之间的跃迁很好地描述了激子
* 价带和导带波函数均匀分布在层上,即它们沿z的分布可以通过阶梯函数近似.  
虽然发现所有材料的介电函数几乎都是各向同性的,但激子质量不是这种情况,见图11C.虽然可以修改模型以允许各向异性质量,但我们在这里仅限于具有各向同性激子质量的材料.从模型获得的激子结合能显示为图13中条形顶部的暗区(也见表5).条的总高度表示G 0 W 0计算的QP间隙.对于直接(间接)带隙材料,我们在模型中使用了直接(间接)激子质量.

# 结论

我们已经提出了51个单层过渡金属二硫化物和氧化物的详细电子结构研究.**从初始的216种化合物中选择51个单层**,这些化合物具有有限的带隙和负的形成热. 计算的性质包括平衡结构周围范围内面内晶格常数的LDA能带结构,平衡晶格常数的准粒子带结构,在G0W0近似中评估并包括自旋轨道耦合,绝对位置相对于真空的传导和价带边缘,有效电子和空穴质量,以及静态q独立的介电功能. 作为一个例子,我们展示了计算数据(在这种情况下是有效质量和介电函数)如何用于从2D Mott-Wannier模型获得最低激子束结合能. 
我们选择关注材料电子结构的一般趋势和相关性,而不是详细说明特定材料的电子结构. 但是,由于所有计算数据都可在开放数据库中使用,因此可以直接更详细地检索和分析特定材料的数据. 我们目前正在努力扩展数据库以包括其他2D材料和属性. 我们坚信,这样的数据库既可用于指导寻找新2D材料的实验工作,也可用作预测更复杂材料(例如范德华异质结构)的性质的平台.
