**Machine-Learning-Augmented** Chemisorption Model for CO2 Electroreduction Catalyst Screening  
用于$CO_2$电还原催化剂筛选的 **ML增强型** 化学吸附模型

**Abstract:**  

我们提出了一种 `ML增强型` 化学吸附模型, 能够快速准确地预测金属合金的表面反应活性.  
具体来说, 我们的工作表明: 通过拟合一系列的从头算吸附能和理想化双金属表面的电子指纹, 人工神经网络可捕捉吸附物( $*CO$ )在多金属表面上复杂, 非线性的相互作用([~0.1eV 误差]()), 在预测中优于 [两级交互模型]().  
通过比较(leveraging)相似吸附质吸附能之间的大小(scaling)关系, 这种集成方法极大地促进了催化剂的高通量筛选, 具体来说, 有望在多金属合金 {100面的端基处 提高 $CO_2$ 还原为 $C_2$ 的效率和选择性. 通过统计分析--[神经网络对输入特征扰动的响应](), 可以加强我们在金属表面上化学键合的基础. 

**Bimetallic or multimetallic materials:**  

具有原子级精确结构和成分的二元或多元金属材料, 在许多化学催化和电化学催化反应中具有广阔的前景.  
金属合金可以利用各金属物种之间的协同作用, 使其表面位点的理化性质具有优异的性能.   
但是, 通过高通量实验和量子化学计算搜索高度优化的合金非常耗时且成本高昂.   
在这方面, 化学吸附的**d带理论**, 可以将活性位点的电子特性与关键中间体的吸附能(通常称为描述符)联系起来, 为探索最佳催化剂(optimal catalyst)的特性提供了物理上的洞察力(insight), 成为催化剂筛选的理论基础.    
由于d带理论中紧束缚近似的约束, 只有相对于主合金材料具有微小扰动的元素可以直接被探索(在一个合理的精度内), 这限制了它的应用.  
为了解决这个问题, 我们在这里提出了ML增强化学吸附模型, 以捕捉吸附质和基底之间复杂且非线性的相互作用, 从而能够在大规模合金材料空间中应用反应描述符. 

**In this work:**  

在这项工作中, 将以二氧化碳在金属电极上的电化学还原为例进行说明.  
在过渡金属和后过渡金属中, $Cu$ 显示出对 $CO_2$ 电还原具有明显的活性, 归因于其均衡的表面活性(balanced surface reactivity), 既可以强烈的吸附$CO_2$ 和表面中间体(以打破所需的化学键), 又对还原产物(如碳氢化合物或氧化物)是弱吸附的.  
众所周知, 该反应与表面金属原子的几何排列有关. 例如: $Cu(111)$ 可以将 $CO_2$ 还原为  $CH_4$, 法拉第效率约为50％, 尽管要求非常高的过电势(~1.0V).   
有趣的是, $Cu(100)$ 对 $C_2$ 物种具有特殊的选择性如$C_2H_4, C_2H_5OH$, 可以在较低的过电位(~0.8V)下进行.  

**Some recent studies:**  

最近关于 $CO/CO_2@Cu(100)$ 上电还原的一些研究表明: $*CO$ 的二聚化(顺序电子和质子转移)控制着 $C_2$ 路径的起始电位, 而到 $*CO$ 的, 协调的质子-电子转移 是 $C_1$ 途径中的关键步骤.   
基于此机制, 我们的目标是识别反应描述符(表征有效合金电催化剂的选择性)(以将 $CO_2$ 转化为 $C_2$ 物种). 要正确描述在 $Cu(100)$上 具有竞争性的 $C_1$和$C_2$ 途径的能量, 明确考虑现实的电化学还原条件是重要的, 在现实的化学环境中, $*CO$ 的覆盖率很高.  
图1显示了沿 $C_1$和$C_2$ 途径相关步骤的自由能, 在 0V和-0.7V(相对于RHE)下, 通过DFT计算得到, $*CO$的覆盖率为3/8 ML.  
在支持信息的图S1中(Supporting Information Figure S1), 显示了 $*CO$ 覆盖率为 1/8 ML 时的自由能途径.

**To address this challenge:**  

为了应对这一挑战, 已经做出了重多努力开发预测模型, 将金属位点的表面反应与其电子性质联系起来, 然后通过标准表或数据库即可预测其电子性质.   
在这方面, 最初由 Hammer和Nørskov开发的d-带理论, 已广泛用于了解各种吸附物的在纯过渡金属及其合金表面吸附能的变化.  
d带中心, $\epsilon_d$, 即电子的d态投影到一个表面金属原子之上的加权平均能量, 是该理论框架中是最重要的参数.   
如果 d带中心在能量上向上移动, 那么吸附质-金属之间的反键态也会发生上移, 可以减少对反键状态的占用, 提高键合.   
这中简单的在金属表面上的化学键合图景, 已经被光谱技术验证.   
许多研究表明, d带模型对过渡金属和和一小部分合金材料的表面反应进行了相当好的预测.
