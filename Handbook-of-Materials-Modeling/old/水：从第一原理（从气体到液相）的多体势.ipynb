1简介
水的重要性怎么强调都不过分。这种情况主要是因为独特的物理特性和广泛的，它存在于液相条件的结合使水分必不可少的生活，因为我们知道它（马雷夏尔2007年）。经过数十年的研究，现在几乎没有问题，水的独特行为不仅仅来自三原子，10电子分子，而是来自个别水分子之间多体相互作用的微妙相互作用，这是进一步的由温度和压力调节（Eisenberg和Kauzmann，1969）。

鉴于水作为生命矩阵的关键作用（Ball 2008），许多理论和计算研究试图得出水从气体到凝聚相的分子级图像并不奇怪。自从第一次蒙特卡罗（MC）和分子动力学（MD）模拟液态水分别由Barker和Watts（1969）以及Rahman和Stillinger（1971）完成），基于分子力学（通常定义为“力场”）或量子力学（通常定义为“ab initio”模型）开发了无数模型。多年来已经提出了具有不同经验主义的各种力场，从粗粒度表示没有原子细节到经典参数化（点电荷和刚性键），以及更复杂的模型来解释分子灵活性，电子极化和电荷转移（Guillot 2002 ; Vega和Abascal 2011 ; Cisneros等2016 ; Shvab和Sadus 2016）。尽管取得了很大进展，但已经表明即使是最成功和最受欢迎的水力场仍然表现出有限的准确性并且实际上缺乏预测能力（Cisneros等人，2016）。

另一方面，ab initio模型基于波函数理论（WFT）或密度泛函理论（DFT）。尽管相关的WFT方法，例如具有单，双和微扰三重激发的耦合簇，通常缩写为CCSD（T），可以实现具有化学准确度的水的分子水平研究，通常定义为1 kcal / mol的偏差。 “精确”的结果（雷扎克和Hobza 2013 ; Simova等人2013），相关联的计算成本限制了其应用到含有比分子的少数没有更多的系统（甘露等人2017年）。这意味着DFT仍然是唯一可行的ab initio方法，用于模拟不同阶段的水。然而，现有的DFT模型受到若干缺点的影响，这些缺点严重限制了它们描述水从气相中的小团簇到液相和冰的性质的预测能力（Gillan等人，2016）。

20世纪50年代Frank和Wen已经认识到水体中多体效应的重要性，他们介绍了由“闪烁的氢键分子簇”组成的液态水的分子图，强调了氢键的合作性质（Frank和Wen）1957年）。第一尝试开发多体（MB）的势能函数（PEFS），用于水仅使用从头数据通过克莱门蒂和同事（Matsuoka等制成。1976 ; Evans等人。1987 ; Niesar等人1990年）。这些开创性的研究表明，Eq。（1）快速收敛水，表明3B效应可以贡献多达15-20％的循环水结构的相互作用能量，4B效应平均表示相互作用能的1％。

随着用于相关WFT方法的更快硬件和有效算法的不断发展，现在常规地执行小分子系统的精确相互作用能量的计算，其然后可以作为用于推导等式的各个项的分析表示的参考。（1）。特别是，利用MBE对水的快速收敛，“分层近似多体方法”（SAMBA）已被证明通过应用逐步降低水平的电子结构方法来为水簇提供高度精确的相互作用能量在MBE（戈拉等人的随后更高体条款。2011）。与硬件和算法开发并行，在构建能够准确表示具有大量自由度的全球势能面的多维数学函数方面取得了巨大进展（Braams和Bowman，2009 ;Bartók等，2010 ; Behler，2016）。这一进展伴随着在多体分子系统中开发有效的经典方案来处理静电相互作用（Thole 1981 ; van Duijnen和Swart 1998 ; Cisneros等人2006 ; Piquemal等人2006 ; Torheyden和Jansen 2006 ; Cisneros等人，2007年; Handley等人。2009年 ; 高等人。2014）。因此，在过去几年中已经提出了几种MB PEF，其中最着名的是CC-pol（Bukowski等人2007），WHBB（Wang等人2011）HBB2-pol（Babin等人2012）和MB-pol（Babin等人，2013年，2014年 ; Medders等人，2014年）。

之间存在多体PEFS，MB-POL已显示正确预测的跨不同阶段水的性质（Reddy等人。2016），再现水二聚体的振动旋转隧道谱（的Babin等人2013），小团簇的能量学，量子平衡和红外光谱（Babin等人2014 ; Brown等人2017），液态水的结构，热力学和动力学特性（Medders等人，2014），冰相的能量学（Pham等人，2017），液态水的红外（IR）和拉曼光谱（Medders和Paesani，2015 ; Reddy等，2017）），空气/水界面的振动和频生成谱（Medders和Paesani 2016），以及冰的IR和拉曼光谱I h（Moberg等人2017）。

本章回顾了使用MB-pol PEF进行计算机模拟的不同阶段水的特性建模的最新进展。在简要概述了MB-pol和相关计算算法的理论形式之后，提出了水的MBE的系统分析，并与力场和ab initio模型进行了比较。然后评估MB-pol在表示液相和冰中水的多维Born-Oppenheimer势能面的准确度。最后，讨论了MB-pol预测结构，热力学和动力学性质以及从气相中的小团簇到液相和冰的水的振动光谱的能力。

MB-pol的1B术语对应于Partridge和Schwenke开发的单体PEF，它精确地再现了气相中分离的水分子的振动谱（Partridge和Schwenke 1997）。与1B能量表达式相关联的是非线性偶极矩表面，其由两个氢原子上的几何相关（正）点电荷表示，其由置于M位点上的负点电荷平衡（图1）。分离的水分子的分子极化张量由位于三个物理原子上的各向同性原子极化率表示。

MB-pol 2B术语通过结合永久和诱导静电的经典表达式和具有多维项的分散能来表达，该多维项纠正了与短程范围内分子间相互作用的纯经典表示相关的缺陷（Stone 1997）。该附加术语，其平滑地切换到零作为二聚体内的两个水分子的氧原子之间的间隔接近6.5埃，有效地表示从单体电子密度的重叠所产生的量子力学相互作用（的Babin等人2013）如电荷转移和渗透和泡利斥力（Khaliullin等人。2007年，2008年）。2B永久静电和诱导均由来自TTM4-F模型的改进的Thole型表达式表示（Burnham等人，2008），而分散能量通过严格拟合至最初引入的渐近ab initio参考能量来表示。由CC-pol PEF（Bukowski等人，2008）。遵循相同的理论方案，MB-pol通过将经典的3B诱导项与多维短程函数相结合来表示3B相互作用，当三聚体内所有水分子对之间的氧 - 氧分离接近4.5Å时，它平稳地切换到零。 （Babin等，2014）。鉴于其功能形式，MB-pol可以被视为基于基线可极化模型的多体PEF，辅以短距离2B和3B贡献，有效地代表纯粹经典表达式无法描述的量子力学相互作用（Medders）和Paesani 2015）。

4从Born-Oppenheimer势能面到实验可观测量
考虑到MB-pol预测水从气体到凝聚相的能量的准确性，有理由怀疑这是否足以同样准确地预测可以通过实验测量的结构，热力学，动力学和光谱学性质。因为通过构造，MB-pol有效地提供了水的Born-Oppenheimer势能面的分析表示，从严格的理论角度来看，MB-pol的计算机模拟应该明确地解释核量子效应（Tuckerman 2010）。对于小团簇，已经证明基于基集扩展和有限网格的方法在用MB-pol描述核量子动力学方面非常有效（Babin等人2013 ; Richardson等人。2016年 ; 布朗等人。2017）。对于具有多个自由度的系统，基于Feynman路径积分形式（Tuckerman 2010）的方法，如路径积分分子动力学（PIMD），可以对给定的Born-Oppenheimer势能进行数值精确的结构和热力学性质计算表面。迄今为止，对于具有许多自由度的系统来说，量子动力学特性（例如，扩散和振动光谱）的精确计算是无法承受的。已经提出了几种用于凝聚相系统的近似量子动力学方法，包括质心分子动力学（CMD）（Voth， 1996），其已经广泛用于使用MB-pol的液态水和冰的量子模拟中（Medders等人。2014年，Medders和Paesani 2015年， 2016年 ; Straight和Paesani 2016 ; Moberg等。2017年 ; 雷迪等人。2017）。

5总结与展望
基于Stillinger及其同事（Hankins等人1970）和Clementi及其同事的开创性研究（Matsuoka等人1976 ; Evans等人1987 ; Niesar等人1990年）），过去十年中，人们重新开始关注精确的势能函数的开发，这些函数严格地来自相互作用能量的多体膨胀。这些努力得到了持续的硬件改进以及相关电子结构方法的有效算法的开发的支持，这些算法现在能够以化学精度实现小分子系统的相互作用能的常规计算。与这些技术和算法的发展并行，使用置换不变多项式（Braams and Bowman 2009），神经网络（Behler 2016），在具有大量自由度的多维势能面的表示方面取得了巨大进步。）和高斯近似电位（Bartók等人，2010）。

考虑到水作为生命矩阵的重要性（Ball 2008），并考虑到自Barker和Watts（1969）以及Rahman和Stillinger（Rahman和Stillinger，1971）首次研究以来水在计算机模拟历史中发挥的关键作用，最近已经开发了几种水体的多体势能函数，例如CC-pol（Bukowski等人，2007），WHBB（Wang等人，2011）HBB2-pol（Babin等人，2012），和MB-pol（Babin等人，2013年，2014年 ; Medders等人，2014年）。当用于允许明确处理核量子效应的模拟时，已经证明这些PEF能够正确地再现从气体到凝聚相的水的性质，从而增强计算机模拟的预测能力。

本章对MB-pol在表示从二聚体到小团簇，液态水和冰的水中的Born-Oppenheimer势能面的性能进行了批判性回顾（Paesani 2016）。结果表明，MB-pol通过2B和3B相互作用的显式短程表示以及多体的物理正确描述的组合，定量地再现了相互作用能量的多体膨胀的个体项，从而实现了高精度。所有距离的静电相互作用。

尽管向长期追求的“通用模型”迈进了一步，能够正确描述不同阶段水的结构，热力学，动力学和光谱学特性，但应始终牢记MB-pol仍然是“水的模型，有自己的近似和局限性。特别地，MB-pol是极限pK w中的水的物理模型→∞，即不允许自电离事件。虽然这是纯水的合理近似值，但是在pH值低于或高于纯水的复杂水溶液中情况急剧变化。此外，正如振动光谱分析所证明的那样，一些实验特征不能通过MB-pol的量子动力学模拟进行定量再现。尽管有人认为这些差异可能与量子动力学方法的近似性质有关，但还需要进一步研究以解决这些不足。最后，MB-pol在复杂解决方案研究中的应用将需要开发通用分子系统的多体表示。虽然沿着这个方向已经取得了一些进展（Bajaj等人，2016年）; Riera等。2017年 ; Sode和Cherry 2017 ; Wang和Bowman 2017），需要进一步的理论和计算发展。在这种情况下，考虑到MB-pol所采用的非常规功能形式，这种形式不受流行的计算机模拟软件的支持，理论/计算化学家/物理学家和计算机科学家之间的协同努力将是开发专业和更有效的软件的关键。用于多体分子动力学模拟，可以充分利用现代硬件。
