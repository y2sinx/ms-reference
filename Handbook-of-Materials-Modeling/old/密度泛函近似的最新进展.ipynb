```python

跳到主要内容跳到部分
SpringerLink数据库
搜索 
菜单
家
登录
材料建模手册 第1-14页| 引用为

密度泛函近似的最新进展
作者
作者和附属机构
李莉Kieron Burke电邮作者
1. 
2. 
生活参考工作条目
首先在线： 2018年6月12日
抽象
我们调查了密度函数计算中使用的一些标准近似值，其中大部分至少已有20年历史，并且已经开发了一些新的方法。

下载 参考工作条目PDF
表格1
按字母顺序排列的首字母缩略词表。参考文献首先出现在正文中

 	
一般条款

DFT

密度泛函理论

G3

分子能量差异的标准数据库

GGA

广义梯度近似

GKS

广义的Kohn-Sham

HF

哈特里 - 福克

KS

科恩深水

LC20

一个20个固体晶格常数的数据库

元GGA

GGA的扩展，使用KS动能

RPA

随机相位近似

S22

弱键合分子的数据库

TDDFT

时间密度泛函理论

XC

交易所相关

 	
近似函数

B3LYP

化学中最受欢迎的功能

B88

交换GGA是化学中最受欢迎的

LDA

局部密度近似

PBE

一种流行的材料科学GGA

PW86

PBE的前身

PW91

PBE简化的GGA

SCAN

最近的meta-GGA

1简介
每年，使用密度泛函理论发表至少30,000篇论文来进行电子结构计算（Pribram-Jones等人，2015）。几乎都解决了Kohn-Sham（KS）方程（Kohn和Sham 1965）的自洽，并使用一些近似的交换相关（XC）能量作为（自旋） - 浓度E XC [ n ] 的函数。本章将介绍一些比较流行的近似值。Burke（2012）涵盖了一些背景，更侧重于化学。

1.1常用的近似值
Kohn和Sham本身（Kohn和Sham 1965）提出了原始近似值，即局部密度近似（LDA），其中系统中每个点的XC能量密度被均匀电子气体的XC能量密度所取代。那时的密度。交换贡献首先由Bloch（1929）编写，现在通过均匀气体的量子蒙特卡罗模拟（Ceperley和Alder 1980）已知相关性，在简单公式中进行参数化（Vosko等人1980 ; Perdew和Wang 1992） 。LDA对于几何形状非常准确，但通常每个键超过分子约1 eV，这对热化学来说相对无用（Jones和Gunnarsson 1989））。

复杂性的下一步是广义梯度近似（GGA），它使用密度及其在每个点的梯度产生能量密度（Burke等，1997）。在Ma和Brueckner（1968）的开创性工作中已经给出了基本概念及其首次实现。这是精心提炼的Langreth的工作，导致Langreth-Mehl的功能（Langreth和Mehl的1981年，1983年）。Perdew率先使用真实空间截止来创建GGA，从而实现了PW86的功能（Perdew和Wang，1986）。这种详细结构的最高点是PW91功能（Perdew 1991 ; Perdew等1992 ; Burke等。1997年，并且在1993年，显示它们产生结合能的有用准确度，即约6-10kcal / mol（1eV = 23kcal / mol）的误差（Gill等人， 1992）。今天材料中最常用的GGA是PW91形式的简化，称为PBE（Perdew等， 1996a），而在化学中它是BLYP，具有B88交换（Becke 1988）和Lee-Yang-Parr相关性（Lee等人） 。 1988年）。这两种都是在惰性气体原子能量上训练，为这些原子产生更准确的能量（Elliott和Burke， 2009 ; Burke等， 2016）。

最后一个标准步骤是通过用Hartree-Fock计算的精确交换能量来创建GGA的混合，通过用Hartree-Fock（HF）交换替换GGA交换的分数a，如Becke首先建议的那样（1993）。这一部分在着名的B3LYP功能中占20％（其代表Becke，3参数，Lee-Yang-Parr），这是当今化学中最常用的近似值（Becke 1993 ; Lee等人1988 ; Vosko等人。1980 ; Stephens等，1994）。其类似物是PBE0（Perdew等人1996b ; Ernzerhof和Scuseria 1999），其使用25％混合。

目前正在执行的所有DFT计算中至少80％使用上述近似值之一。人们可以想到使用更多密度成分的更复杂的近似，如雅各布的阶梯（Perdew和Schmidt 2001）。在进行密度函数逼近方面取得进展的一个重要原则是确保（几乎）在任何地方，每个近似级别至少执行以及前一级别，并且还改进某些特征，使得先前级别过时（除了节省计算）时间）。应避免为不同目的使用不同的功能。

1.2超越基态DFT
以上描述的是非相对论极限和共线磁场中的通用基态DFT。超出该领域的DFT的最大单一使用是在线性响应机制中使用时间依赖性DFT（TDDFT）（Runge和Gross 1984 ; Maitra等人2004 ; Maitra 2016）来提取电子跃迁频率（Casida 1996 ; Petersilka等人。1996年）。所有DFT出版物中有多达10％包括TDDFT激励估计。这样的计算几乎都使用XC内核的绝热近似，因此通过选择基态近似来唯一地确定。大多数这样的计算是分子（Burke等2005），因为需要非局部贡献来正确地产生体半导体光谱的细节（例如激子结合能），并且上述常用功能中缺少这种贡献（Onida等人2002 ; Martin等人2016）。

其他扩展包括用于非共线场的磁DFT，相对论DFT（Engel和Dreizler 2011）甚至QED（Flick等人2015），使用Mermin定理在有限温度下的DFT（Smith等人2017），核与原子之间的耦合。电子（Requist和Gross 2016）等。同样，使用标准的基态近似，除非特定的效果需要超出这些，例如提取双激发（Maitra等人，2004）。

2最新动态
本节专门介绍过去20年的发展。

2.1 RPA型功能
着名的DFT的绝热 - 连接波动 - 耗散公式在密度 - 密度响应函数上的频率和耦合常数以及空间积分方面产生E XC（Langreth和Perdew 1975 ; Harris和Jones 1974 ; Gunnarsson和Lundqvist 1976）。这可以直接从KS响应函数中提取，该函数由占用和未占用的轨道构成，以及TDDFT的XC内核（Gross和Kohn 1985）。忽略内核产生RPA，也称为直接随机相近似（RPA）（犁沟2001 ; Langreth和Perdew 1977年，1975年 ; Chen等人2017年）或TD-Hartree。因此，该方案用于第五梯级（以雅各梯）近似，即可能是昂贵的评价，但相对负担总是减小（犁沟2008 ; Eshuis等人2010）。直接RPA过度修正系统，因为它在能量的多体扩张中仅包含“气泡”图，并且在高阶处错过了其他贡献以减少相关性。它还具有自相互作用的困难，因为它甚至仅对一个电子产生有限的相关能量。这两种效应产生分子解离能的不准确性。最近的进展包括对XC内核的各种近似，以产生改进的能量和计算成本节省。最近的一项发展，使用精确的频率相关交换核，以及更高阶的巧妙和物理动机的恢复，似乎克服了稳定性问题，甚至允许精确计算N 2的结合曲线（Erhard等人2016）; Burke 2016）（即使在耦合簇理论中也很困难，因为大分离时波函数的多参考性质）。

2.2弱相互作用
在过去二十年中，将弱范德瓦尔斯（vdW）相互作用纳入DFT计算已经取得了巨大进展。标准功能可以很好地模拟共价键，离子键和金属键，但是对于更长，更弱的键而言则失败。由于它们的半双数性质，它们不能产生作为R 6脱落的弱结合，其中R是两个原子或分子之间的分离。因此，必须将校正添加到标准功能以捕获这些效果。这样做有三种主要方案。

在显性非局部密度泛函的领域中，存在由Langreth和Lundqvist及其同事最初开发的近似序列（Dion等人，2004）。这些近似值是非惯性导出的，显着地从对均匀气体相关能量的贡献开始。原始版本来自2004年，并且在vdW2中给出了改进的参数化（Klimeš等人，2009）。这些显式密度泛函可以应用于从分子到固体的所有材料，但是假设系统具有间隙（不是金属）而设计（Berland等人2015）。

在规模的极端另一端，有Grimme和同事的参数方案（Grimme 2006），通常称为DFT-D。这些不是明确的密度泛函，而是额外的能量，包括原子间vdW能量的渐近展开中的C 6（和更高阶）系数的估计，结合阻尼因子以保持贡献有限，因为分离减少。对于给定的标准XC近似，这样的方案需要每个原子的经验参数，但是可以为小的弱结合分子产生高度精确的能量曲线（Burns等人，2017）。

中间是Tkatchenko和Scheffler开发的方案，它只需要一个给定XC功能的参数，并对DFT能量产生准确的附加校正（Tkatchenko和Scheffler 2009）。这已经扩展到包含集体静电效应，因此可以处理许多尺寸和不同长度尺度的金属和材料（Hermann等人2017）。最终方案是由Becke和Johnson开始的，它使用交换孔的偶极矩确定C 6（和更高）系数（Johnson和Becke 2006）。

当然，更昂贵的治疗，例如上面提到的RPA，自动包括对vdW力的近似。

2.3 Meta-GGA
Jacob梯子的第三个梯级是meta-GGA，它增加了一种新的成分，超过密度和梯度。这通常被选择为KS轨道的（正）动能密度。良好的meta-GGA的目标是在没有精确交换贡献的计算开销的情况下瞄准混合的准确性。在使用以原子为中心的基函数的分子计算中，精确交换的成本是相对可控的，但在使用周期性边界条件和平面波基组时通常会令人望而却步。运行范围分离的混合（参见下一节）有时比材料计算中的典型GGA长1000倍。

Perdew和同事和其他许多人花了几十年的发展荟萃GGA的，许多有缺陷的尝试（Sun等2015年）。但最近的努力，称为强约束和适当规范的半密度函数（SCAN），已通过许多标准测试，并且看起来非常有希望加入常用功能的万神殿。G3数据集（Curtiss等人，2005）是LDA过量约3eV（通常约1eV /键）的一组标准化学键。PBE将其降低至约1eV，而SCAN将其降低至约1/4 eV。SCAN也有比PBE的S22数据集（Jurecka等小2-3倍的错误。2006年）弱结合系统。同时，SCAN将LC20数据上的晶格参数误差（Sun等人2011）设置为从约0.05ÅPBE到约0.01Å。SCAN也比PBE产生更好的水性质（Perdew等，2017）。另一方面，PBE对化学屏障高度的低估仅略微提高了约30％（而杂种通常是2-3倍）。因此，对于许多属性，SCAN产生类似于混合功能的精度，但是计算成本的一小部分（对于材料代码）。

2.4范围分离的混合动力车
范围分离背后的理论是精确的，由Andreas Savin首先开发（Toulouse等人2009 ; Savin 1996 ; Leininger等人1997）。人们简单地将库仑排斥写为短程贡献的总和（比分离的倒数更快地衰减）和远程贡献，其在零分离时没有库仑奇点。然后，可以在KS方程的一些推广中包括一个贡献作为相互作用，并且重新定义的XC贡献考虑另一个。这一切都是正式的，并且对于这样的方案存在精确的XC功能（尽管它们与它们的常规KS对应物不同）。

但现在有太多的选择等待着。第一个是执行范围分离的长度标度，通常表示为1 / ω。这是一个连续的参数，由于将对相应的E XC进行近似，因此结果对它敏感。就像全球混合动力车中精确交换的一小部分一样，总是有一种调整它的诱惑。

此外，对于交换贡献（最常见），相关性或两者，可以分离成长期和短期贡献。最后，人们可能希望用近似函数来处理短距离贡献，或者在其他情况下，远距离贡献！例如，TDDFT中的一个棘手的问题是准确地计算良好分离的供体 - 受体复合物的电荷转移激发。包括长程福克交换（Stein等恰好都非常好，这个问题2009年）。在另一方面，非常成功HSE06功能（Heyd等。2003）是一种范围分离的混合物，其中远程交换用近似功能处理，但是短距离交换被精确处理，以模拟在延伸绝缘体中的筛选（Janesko等人2009）。这种特殊的混合通常可以为中等间隙半导体和绝缘体提供精确的间隙（Perdew等，2017）。

2.5固体间隙
标准近似的一个关键性失败是它们无法预测半导体和绝缘体的间隙。LDA低估了体硅的间隙2倍，使Ge成为金属，而GGA的表现稍好一些。从很早开始，GW方法（Aryasetiawan和Gunnarsson 1998）的强大功能就是它能够提供准确可靠的间隙。

为了理解这个问题，重要的是首先要注意周期性固体的KS间隙与固体的基本间隙不匹配（Perdew等，1982 ; Perdew，1985）。所有迹象表明，事实上LDA和GGA产生相当准确的KS间隙（即，接近于具有精确基态功能的KS间隙）（Grüning等人，2006）。但是，与LDA和GGA不同，精确的功能允许进入基本间隙，这就是I  -  A，系统的电离能和电子亲和力之间的差异。考虑一个非常大但有限的材料集群。然后可以添加和减去电子以找到I和A.并推断出确切的差距。实际上，存在用于定期计算的现代方法（Stadele等人，1999）。但是在LDA或GGA中，增加的电子或空穴在整个系统上离域，并且由于它们缺乏衍生物不连续性，I  -  A崩溃到HOMO-LUMO KS能量差异，即KS间隙。

在过去二十年中取得的巨大成功是使用混合功能（如HSE06）准确计算中等间隙（Heyd等人，2003年）。这可以通过一个广义的KS方案来实现（Seidl等，1996），其中功能的轨道依赖部分被处理为HF理论，而不是纯KS理论（需要用优化的有效势处理它）（ OEP）方法（Kümmel和Kronik 2008））。通过具有轨道相关电位，可以证明这种计算的广义Kohn-Sham（GKS）间隙与I  -  A相匹配并且由于轨道依赖性，包括近似的导数不连续性。因此，这种计算中的GKS间隙是近似的基本间隙，并且比相应的KS间隙宽。这就是混合功能和meta-GGA如何产生比GGA更广泛且更准确的间隙（Perdew等人2017）。

3挑战和希望
在这里，我们回顾一下当前近似的一些令人沮丧的失败。

3.1强相关
DFT的标准逼近强关联系统的故障，因为已经知道它的发明前，如HF具有用于h问题2当其被拉伸（海特勒和伦敦1927）。当整数（或半整数）电子数量定位在不同位置时，可以分析问题并将其与标准近似的定位/离域误差相关联（Cohen等人，2008）。这通常被称为量子化学中的静态相关，并且涉及两个状态之间的KS间隙变得非常小并且精确的多体波函数变成两个Slater行列式的（几乎）相等的混合。

重要的是要注意，困难只在于在这些情况下的近似失败，而不是KS方案本身。使用双站点Hubbard模型可以很好地说明这一点，即使在强相关时，构建精确的KS系统也是如此（Carrascal和Ferrer 2012）。即使在实际情况下（尽管在1d中），如果使用精确的XC功能，则强相关系统的KS方程总是可以收敛到精确的地面能量和密度（Wagner等人2013）。但当然，KS系统的许多功能与这种情况下的物理系统不同（Carrascal等2015））。这一点经常被多体理论的实践者所迷惑。KS响应函数与强相关系统的多体模拟之间的差异不是密度函数逼近不能为这样的系统产生精确能量的信号。

但固体中的强相关性甚至比分子中的静态相关性更难。简单地看一下，考虑均匀间隔的H原子链。随着间距的增加，电子在每个位置定位。对于H 2，真波函数结合了两个Slater行列式。但是对于H 4，存在四个这样的决定因素，并且用LDA计算的DFT将将对称性分解为四种不同的解，其中一种将具有最低的能量（Wagner等人，2014）。）。随着链中原子数的增加，几乎简并的溶液的数量也会增加，通过非常小的能量的自旋激发分离。在热力学极限中，这些变得无限，并且单个Slater行列式的通常量子化学起点变得无望（Qiu等人2017）。因为这是一个非常重要的问题，大量的研究已经开始研究这个限制，特别是由Weitao Yang（Zheng等人，2011），还有Scuseria（Motta等人，2017）和Becke（Johnson和Becke 2017）。

3.2经验主义的作用
构造密度泛函的最实用的系统方法得到了Perdew的支持：将与给定的近似水平相关的精确条件与适当的规范（例如均匀气体或氢原子）相结合，以创建近似的巨大一般性（梅德韦杰夫等人，2017年）。这种方法的一个关键方面是，当一个人在雅各布的阶梯上升起时，在几乎所有情况下（并且它通常不是非常有用的情况），每个逐次逼近比前一个更好（或者至少不会更糟）。 。因此，对于给定的计算成本，存在单个（或至少非常相似）明显的选择，其很少比使用较低的梯级更差。使用适当的规范可能看起来是经验性的，但可以理解为在近似中选择参数以捕获尚未完全导出的限制，但原则上可以是。非常成功的B88交换功能（Becke 1988 ; Elliott和Burke 2009）可以被视为包含适当的规范（原子的交换能量）。

另一方面，大量廉价计算资源导致许多数据库具有实验结果或高级计算化学结果，可以测试新的近似值。这也导致了许多参数密度泛函近似的经验拟合，通过Truhlar和同事（赵和Truhlar为倡导2008年，2006年 ;赵等人。2006年）。这种方法通常产生比Perdew等人更准确的近似值。适用于所安装的系统和特性以及相关系统和特性。但它并没有产生单一的通用近似值，这种近似值通常比梯子上的先前步骤有所改进。当应用超出其适用范围时，这种近似可能会严重失败。

功能近似的一种全新方法是使用机器学习来从准确的数据中学习（Snyder等人，2012 ; Rupp等人，2012 ;Bartók等人，2010）。这与先前的方法不同，因为它自动包括高度非局部贡献，例如，在内核中通过整个系统上的积分测量密度分离（Snyder等人2012 ; Li等人2016b ; Vu等人。2015 ; Snyder等，2013）。最近的进展包括首次采用ML-DFT近似的KS-MD仿真来绕过求解KS方程（Brockherde等2017）和完全相互作用功能（包括XC）的精确近似（Li等人2016a），即使对于强相关甚至扩展系统也是如此（但到目前为止，仅在1d内，因为产生准确数据的成本）。

3.3新视野
在过去十年左右，DFT的一个全新领域已经大大增加了温暖致密物质的应用（Smith等人2017），电子秤上的温度很高（约100,000 K），但不如Thomas-Fermi高。理论（或经典行为）占主导地位。应用范围从建模行星内部到惯性约束聚变。这个领域是如此“热”，甚至热LDA的输入，即均匀气体的XC能量随温度的变化，现在才以高精度计算（Groth等人2017）。

参考
Aryasetiawan F，Gunnarsson O（1998）gw方法。Rep Prog Phys 61（3）：237。http://stacks.iop.org/0034-4885/61/i=3/a=002
ADS CrossRef Google学术搜索
BartókAP，Payne MC，Kondor R，CsányiG（2010）高斯近似势：量子力学的准确性，没有电子。Phys Rev Lett 104：136403。http://link.aps.org/doi/10.1103/PhysRevLett.104.136403
ADS CrossRef Google学术搜索
Becke AD（1988）具有正确渐近行为的密度函数交换能量近似。Phys Rev A 38（6）：3098-3100。https://doi.org/10.1103/PhysRevA.38.3098
ADS CrossRef Google学术搜索
Becke AD（1993）密度 - 功能热化学。III。精确交换的作用。J Chem Phys 98（7）：5648-5652。https://doi.org/10.1063/1.464913，http://link.aip.org/link/?JCP/98/5648/1
ADS 交叉引用谷歌学术
Berland K，Cooper VR，Lee K，SchröderE，Thonhauser T，Hyldgaard P，Lundqvist BI（2015）密度泛函理论中的Van der waals力：对vdW-DF方法的回顾。Rep Prog Phys 78（6）：066501
ADS CrossRef Google学术搜索
Bloch F（1929）Überdiequantenmechanik der elektronen in kristallgittern。ZeitschriftfürPhysikA Hadrons and Nuclei 52（7）：555-600
zbMATH Google学术搜索
Brockherde F，Vogt L，Li L，Tuckerman ME，Burke K，Mller KR（2017）绕过机器学习的Kohn-Sham方程。Nature Commun https://doi.org/10.1038/s41467-017-00839-3
Burke K（2012）透视密度泛函理论。J Chem Phys 136. http://link.aip.org/link/doi/10.1063/1.4704546
Burke K（2016）观点：改进电子结构计算。物理9（108）。https://physics.aps.org/articles/v9/108
Burke K，Perdew JP，Wang Y（1997）推导广义梯度近似：PW91密度泛函，Plenum，第81页.http：//link.springer.com/chapter/10.10077
Burke K，Werschnik J，Gross EKU（2005）时间密度泛函理论：过去，现在和未来。J Chem Phys 123（6）：062206。https://doi.org/10.1063/1.1904586，http://link.aip.org/link/?JCP/123/062206/1
ADS 交叉引用谷歌学术
Burke K，Cancio A，Gould T，Pittalis S（2016）密度泛函理论中的相关性局部。J Chem Phys 145（5）：054112。https://doi.org/10.1063/1.4959126，http://scitation.aip.org/content/aip/journal/jcp/145/5/10.1063/1.4959126
ADS 交叉引用谷歌学术
Burns LA，Faver JC，Zheng Z，Marshall MS，Smith DG，Vanommeslaeghe K，MacKerell AD Jr，Merz KM Jr，Sherrill CD（2017）生物片段数据库（BFDb）：用于非共价相互作用的计算化学分析的开放数据平台。J Chem Phys 147（16）：161727
ADS CrossRef Google学术搜索
Carrascal DJ，Ferrer J（2012）Exact Kohn-Sham本征态与强相关电子的简单模型中的准粒子。Phys Rev B 85：045110。http://link.aps.org/doi/10.1103/PhysRevB.85.045110
ADS CrossRef Google学术搜索
Carrascal DJ，Ferrer J，Smith JC，Burke K（2015）Hubbard dimer：一个关于多体问题的密度泛函案例研究。J Phys Condens Matter 27（39）：393001。http://stacks.iop.org/0953-8984/27/i=39/a=393001
CrossRef Google学术搜索
Casida ME（1996）分子系统的时间密度函数响应理论：理论，计算方法和功能。在：Seminario JM（ed）密度泛函理论的最新发展和应用。Elsevier，阿姆斯特丹
谷歌学者
Ceperley DM，Alder BJ（1980）通过随机方法研究电子气的基态。Phys Rev Lett 45：566
ADS CrossRef Google学术搜索
Chen GP，Voora VK，Agee MM，Balasubramani SG，Furche F（2017）随机相位近似方法。Ann Rev Phys Chem 68（1）：421-445。https://doi.org/10.1146/annurev-physchem-040215-112308，PMID：28301757
ADS 交叉引用谷歌学术
Cohen AJ，Mori-SánchezP，Yang W（2008）洞察当前密度泛函理论的局限性。Science 321（5890）：792-794
ADS CrossRef Google学术搜索
Curtiss LA，Redfern PC，Raghavachari K（2005）对g3 / 05测试实验能量集的高斯-3和密度泛函理论的评估。J Chem Phys 123：124107
ADS CrossRef Google学术搜索
Dion M，Rydberg H，SchröderE，Langreth DC，Lundqvist BI（2004）Van der Waals density functional for general geometry。Phys Rev Lett 92（24）：246401。http://link.aps.org/doi/10.1103/PhysRevLett.92.246401
ADS CrossRef Google学术搜索
Elliott P，Burke K（2009）b88交换函数中参数的非经验推导。Can J Chem Ecol 87（10）：1485-1491。https://doi.org/10.1139/V09-095
CrossRef Google学术搜索
Engel E，Dreizler RM（2011）密度泛函理论：高级课程。Springer，柏林
CrossRef 谷歌学者
Erhard J，Bleiziffer P，GörlingA（2016）相关核的幂级数近似导致Kohn-Sham方法结合了准确性，计算效率和一般适用性。Phys Rev Lett 117：143002。https://link.aps.org/doi/10.1103/PhysRevLett.117.143002
ADS MathSciNet CrossRef Google学术搜索
Ernzerhof M，Scuseria GE（1999）对Perdew-Burke-Ernzerhof交换相关函数的评估。J Chem Phys 110：5029。http://link.aip.org/link/JCPSA6/v110/i11/p5029/s1
ADS CrossRef Google学术搜索
Eshuis H，Yarkony J，Furche F（2010）利用同一性和虚频率积分的分辨率快速计算分子随机相位近似相关能量。Journal of Chemical Physics 132（23）：234114。https://doi.org/10.1063/1.3442749，http://link.aip.org/link/?JCP/132/234114/1
ADS 交叉引用谷歌学术
Flick J，Ruggenthaler M，Appel H，Rubio A（2015）Kohn-Sham方法量子电动力学密度泛函理论：真实空间中精确的时间依赖有效势。Proc Natad Sci 112（50）：15285-15290
ADS CrossRef Google学术搜索
Furche F（2001）随机相位逼近交换相关能函数的分子检验。Phys Rev B 64：195120
ADS CrossRef Google学术搜索
Furche F（2008）将随机相位近似发展为实际的后Kohn-Sham相关模型。J Chem Phys 129（11）：114105
ADS CrossRef Google学术搜索
Gill PMW，Johnson BG，Pople JA，Frisch MJ（1992）对hartree-fock和密度泛函理论的混合性能的研究。Int J Quantum Chem 44（S26）：319-331。https://doi.org/10.1002/qua.560440828
CrossRef Google学术搜索
Grimme S（2006）通过长程色散校正构建的半经验GGA型密度泛函。J Comput Chem 27（15）：1787-1799
CrossRef Google学术搜索
Gross E，Kohn W（1985）频率相关线性响应的局部密度泛函理论。Phys Rev Lett 55：2850
ADS CrossRef Google学术搜索
Groth S，Dornheim T，Sjostrom T，Malone FD，Foulkes WMC，Bonitz M（2017）在温暖的致密物质条件下均匀电子气的Ab initio交换相关自由能。Phys Rev Lett 119：135001。https://link.aps.org/doi/10.1103/PhysRevLett.119.135001
ADS CrossRef Google学术搜索
GrüningM，Marini M，Rubio A（2006）来自多体扰动理论的密度函数：半导体和绝缘体的带隙。J Chem Phys 124：154108
ADS CrossRef Google学术搜索
Gunnarsson O，Lundqvist B（1976）通过自旋密度函数形式主义在原子，分子和固体中交换和相关。Phys Rev B 13：4274
ADS CrossRef Google学术搜索
Harris J，Jones R（1974）有界电子气的表面能。J Phys F 4：1170
ADS CrossRef Google学术搜索
Heitler W，London F（1927）根据量子力学，中性原子与单极结合之间的相互作用。Z Physik 44：455
ADS CrossRef Google学术搜索
Hermann J，DiStasio RA，Tkatchenko A（2017）范德瓦尔斯在分子和材料中相互作用的第一原理模型：概念，理论和应用。Chem Rev 117（6）：4714-4758。https://doi.org/10.1021/acs.chemrev.6b00446，PMID：28272886
交叉引用谷歌学术
Heyd J，Scuseria GE，Ernzerhof M（2003）基于屏蔽库仑势的混合功能。J Chem Phys 118（18）：8207-8215。https://doi.org/10.1063/1.1564060，http://link.aip.org/link/?JCP/118/8207/1
ADS 交叉引用谷歌学术
Janesko BG，Henderson TM，Scuseria GE（2009）筛选了用于固态化学和物理学的混合密度函数。Phys Chem Chem Phys（3）：443-454
CrossRef Google学术搜索
Johnson ER，Becke AD（2006）Van der waals来自交换孔偶极矩的相互作用：应用于生物有机基准系统。Chem Phys Lett 432（4-6）：600-603
ADS CrossRef Google学术搜索
Johnson ER，Becke AD（2017）沟通：Dft处理3D过渡金属硅原子的强相关性。J Chem Phys 146（21）：211105
ADS CrossRef Google学术搜索
Jones R，Gunnarsson O（1989）密度函数形式，其应用和前景。Rev Mod Phys 61：689
ADS CrossRef Google学术搜索
Jurecka P，Sponer J，Cerny J，Hobza P（2006）基准数据库的精确（MP2和CCSD（T）完全基组限制）小模型复合物，DNA碱基对和氨基酸对的相互作用能。Phys Chem Chem Phys 8：1985-1993
CrossRef Google学术搜索
KlimešJ，Bowler DR，Michaelides A（2009）化学精确度为范德华密度泛函。J Phys Condens Matter 22（2）：022201
ADS CrossRef Google学术搜索
Kohn W，Sham LJ（1965）自洽方程，包括交换和相关效应。Phys Rev 140（4A）：A1133-A1138。http://link.aps.org/doi/10.1103/PhysRev.140.A1133
ADS MathSciNet CrossRef Google学术搜索
KümmelS，Kronik L（2008）轨道相关密度泛函：理论和应用。Rev Mod Phys 80（1）：3-60。https://doi.org/10.1103/RevModPhys.80.3
ADS MathSciNet CrossRef Google学术搜索
Langreth D，Mehl M（1981）易于实现的非局部交换相关能量功能。Phys Rev Lett 47：446。https://journals.aps.org/prl/abstract/10.1103/PhysRevLett.47.446
ADS CrossRef Google学术搜索
Langreth D，Mehl M（1983）超越了基态电子特性计算中的局部密度近似。Phys Rev B 28：1809。https://journals.aps.org/prb/abstract/10.1103/PhysRevB.28.1809
ADS CrossRef Google学术搜索
Langreth D，Perdew J（1975）金属表面的交换相关能。Solid State Commun 17：1425
ADS CrossRef Google学术搜索
Langreth D，Perdew J（1977）金属表面的交换相关能量：波矢量分析。Phys Rev B 15：2884
ADS CrossRef Google学术搜索
Lee C，Yang W，Parr RG（1988）将colle-salvetti相关能量公式发展成电子密度的函数。Phys Rev B 37（2）：785-789。http://link.aps.org/doi/10.1103/PhysRevB.37.785
ADS CrossRef Google学术搜索
Leininger T，Stoll H，Werner HJ，Savin A（1997）将长程配置相互作用与短程密度泛函相结合。Chem Phys Lett 275（3-4）：151-160
ADS CrossRef Google学术搜索
Li L，Baker TE，White SR，Burke K（2016a）纯密度泛函，强相关性和机器学习的热力学极限。Phys Rev B 94：245129。http://link.aps.org/doi/10.1103/PhysRevB.94.245129
ADS CrossRef Google学术搜索
Li L，Snyder JC，Pelaschier IM，Huang J，Niranjan UN，Duncan P，Rupp M，Mller KR，Burke K（2016b）理解机器学习密度函数。Int J Quantum Chem 116（11）：819-833。https://doi.org/10.1002/qua.25040
CrossRef Google学术搜索
Ma SK，Brueckner K（1968）电子气的相关能量，具有缓慢变化的高密度。Phys Rev 165：18
ADS CrossRef Google学术搜索
Maitra NT（2016）观点：时间密度泛函理论的基本方面。J Chem Phys 144（22）：220901。http://scitation.aip.org/content/aip/journal/jcp/144/22/10.1063/1.4953039
ADS CrossRef Google学术搜索
Maitra NT，Zhang F，Cave RJ，Burke K（2004）时间依赖密度泛函理论线性响应中的双激发。J Chem Phys 120（13）：5932-5937。https://doi.org/10.1063/1.1651060，http://link.aip.org/link/?JCP/120/5932/1
ADS 交叉引用谷歌学术
Martin RM，Reining L，Ceperley DM（2016）相互作用的电子。剑桥大学出版社，剑桥
谷歌学者
Medvedev MG，Bushmarinov IS，Sun J，Perdew JP，Lyssenko KA（2017）对密度泛函理论的评论的回应正在走向确切功能的道路。Science 356（6337）：496-496
ADS CrossRef Google学术搜索
Motta M，Ceperley DM，Chan GKL，Gomez JA，Gull E，Guo S，Jiménez-Hoyos CA，Lan TN，Li J，Ma F，Millis AJ，Prokof'ev NV，Ray U，Scuseria GE，Sorella S，Stoudenmire EM，Sun Q，Tupitsyn IS，White SR，Zgid D，Zhang S（2017）对于真实材料中的多电子问题的解决方案：具有最先进多体的氢链状态方程方法。Phys Rev X 7：031059。https://link.aps.org/doi/10.1103/PhysRevX.7.031059
Google学术搜索
Onida G，Reining L，Rubio A（2002）Electronic excitations：density-functional与many-body green-function approach。Rev Mod Phys 74（2）：601-659。https://doi.org/10.1103/RevModPhys.74.601
ADS CrossRef Google学术搜索
Perdew J（1985）Kohn-Sham轨道是什么意思？原子是如何分离的？Plenum，物理学中的密度泛函方法。斯普林格，波士顿，马萨诸塞州（第265-308页）。https://link.springer.com/chapter/10.1007/978-1-4757-0818-9_10
CrossRef Google学术搜索
Perdew J（1991）Electronic structure of solids 91，Ziesche P，Eschrig H（eds）（Berlin：Akademie-verlag）p。11; Perdew JP和Wang Y 1992.Phys Rev B 45（13）：244
Google Scholar
Perdew JP，Schmidt K（2001）Jacobs ladder of density functional approximations for the exchange-correlation energy。在：AIP会议论文集，AIP，第577卷，第1-20页
Google学术搜索
Perdew J，Wang Y（1986）电子交换能量的精确且简单的密度函数：广义梯度近似。Phys Rev B 33：8800
ADS CrossRef Google学术搜索
Perdew JP，Wang Y（1992）准确而简单的电子 - 气体相关能的分析表示。Phys Rev B 45（23）：13244-13249。http://link.aps.org/doi/10.1103/PhysRevB.45.13244
ADS CrossRef Google学术搜索
Perdew JP，Parr RG，Levy M，Balduz JL（1982）Density-functional theory for fractional particle number：能量的导数不连续性。Phys Rev Lett 49：1691-1694。http://link.aps.org/doi/10.1103/PhysRevLett.49.1691
ADS CrossRef Google学术搜索
Perdew JP，Chevary JA，Vosko SH，Jackson KA，Pederson MR，Singh DJ，Fiolhais C（1992）原子，分子，固体和表面：用于交换和相关的广义梯度近似的应用。Phys Rev B 46：6671
ADS CrossRef Google学术搜索
Perdew JP，Burke K，Ernzerhof M（1996a）广义梯度近似变得简单。Phys Rev Lett 77（18）：3865-3868。https://doi.org/10.1103/PhysRevLett.77.3865，同上78：1396（E）（1997）
ADS CrossRef Google Scholar
Perdew JP，Ernzerhof M，Burke K（1996b）用于将精确交换与密度函数近似混合的基本原理。J Chem Phys 105（22）：9982-9985。https://doi.org/10.1063/1.472933，http://link.aip.org/link/?JCP/105/9982/1
ADS 交叉引用谷歌学术
Perdew JP，Yang W，Burke K，Yang Z，Gross EKU，Scheffler M，Scuseria GE，Henderson TM，Zhang IY，Ruzsinszky A，Peng H，Sun J（2017）在广义Kohn-Sham理论中理解固体的带隙。Proc Natad Sci。http://www.pnas.org/content/early/2017/02/28/1621352114.full
Petersilka M，Gossmann UJ，Gross EKU（1996）来自时间密度泛函理论的激发能。Phys Rev Lett 76：1212
ADS CrossRef Google学术搜索
Pribram-Jones A，Gross DA，Burke K（2015）DFT：一个充满洞的理论？Ann Rev Phys Chem 66（1）：283-304。http://www.annualreviews.org/doi/abs/10.1146/annurev-physchem-040214-121420
ADS CrossRef Google学术搜索
Qiu Y，Henderson TM，Zhao J，Scuseria GE（2017）投影耦合簇理论。J Chem Phys 147（6）：064111
ADS CrossRef Google学术搜索
Requist R，Gross E（2016）基于精确因子分解的电子和原子核密度泛函理论。Phys Rev Lett 117（19）：193001
ADS CrossRef Google学术搜索
Runge E，Gross EKU（1984）时间依赖系统的密度泛函理论。Phys Rev Lett 52（12）：997。http://link.aps.org/doi/10.1103/PhysRevLett.52.997
ADS CrossRef Google学术搜索
Rupp M，Tkatchenko A，MüllerKR，von Lilienfeld OA（2012）利用机器学习快速准确地建模分子雾化能量。Phys Rev Lett 108：058301。http://link.aps.org/doi/10.1103/PhysRevLett.108.058301
ADS CrossRef Google学术搜索
Savin A（1996）关于简并，近简并和密度泛函理论。技术报告，路易斯安那州立大学，Baton Rouge
CrossRef 谷歌学者
Seidl A，GörlingA，Vogl P，Majewski JA，Levy M（1996）广义Kohn-Sham方案和带隙问题。Phys Rev B 53：3764-3774。http://link.aps.org/doi/10.1103/PhysRevB.53.3764
ADS CrossRef Google学术搜索
Smith JC，Sagredo F，Burke K（2017）Warming up density functional theory。提交。https://arxiv.org/abs/1701.00873
Snyder JC，Rupp M，Hansen K，Mueller KR，Burke K（2012）利用机器学习寻找密度函数。Phys Rev Lett 108：253002
ADS CrossRef Google学术搜索
Snyder JC，Rupp M，Hansen K，Blooston L，MüllerKR，Burke K（2013）Orbital-free bond breaking via machine learning。J Chem Phys 139（22）：224104
ADS CrossRef Google学术搜索
Stadele M，Moukara M，Majewski JA，Vogl P，GörlingA（1999）Exact exchange Kohn-Sham形式主义应用于半导体。Phys Rev B 59：10031
ADS CrossRef Google学术搜索
Stein T，Kronik L，Baer R（2009）使用时间密度泛函理论可靠地预测分子复合物中的电荷转移激发。J Amer Chem Soc 131（8）：2818-2820。http://pubs.acs.org/doi/abs/10.1021/ja8087482
CrossRef Google学术搜索
Stephens PJ，Devlin FJ，Chabalowski CF，Frisch MJ（1994）J Phys Chem 98：11623
CrossRef Google Scholar
Sun J，Marsman M，Csonka GI，Ruzsinszky A，Hao P，Kim YS，Kresse G，Perdew JP（2011）投影仪 - 增强波方法中的自洽元广义梯度近似。Phys Rev B 84（3）：035117
ADS CrossRef Google学术搜索
Sun J，Ruzsinszky A，Perdew JP（2015）强烈约束和适当规范的半密度函数。Phys Rev Lett 115：036402。http://link.aps.org/doi/10.1103/PhysRevLett.115.036402
ADS CrossRef Google学术搜索
Tkatchenko A，Scheffler M（2009）精确分子范德瓦尔斯与基态电子密度和自由原子参考数据的相互作用。Phys Rev Lett 102：073005
ADS CrossRef Google学术搜索
Toulouse J，Gerber IC，Jansen G，Savin A，ÁngyánJG（2009）Adiabatic-connection fluctuation-dissipation density-functional theory based range range。Phys Rev Lett 102：096404。https://link.aps.org/doi/10.1103/PhysRevLett.102.096404
ADS CrossRef Google学术搜索
Vosko SH，Wilk L，Nusair M（1980）精确的自旋相关电子液相关能量，用于局部自旋密度计算：一个关键分析。Can J Phys 58（8）：1200-1211。http://www.nrcresearchpress.com/doi/abs/10.1139/p80-159
ADS CrossRef Google学术搜索
Vu K，Snyder JC，Li L，Rupp M，Chen BF，Khelif T，Mller KR，Burke K（2015）了解内核脊回归：从简单函数到密度泛函的常见行为。Int J Quantum Chem 115（16）：1115-1128。https://doi.org/10.1002/qua.24939
CrossRef Google学术搜索
Wagner LO，Stoudenmire EM，Burke K，White SR（2013）保证Kohn-Sham方程的收敛性。Phys Rev Lett 111：093003。http://link.aps.org/doi/10.1103/PhysRevLett.111.093003
ADS CrossRef Google学术搜索
Wagner LO，Baker TE，Stoudenmire ME，Burke K，White SR（2014）Kohn-Sham计算具有确切的功能。Phys Rev B 90：045109。http://link.aps.org/doi/10.1103/PhysRevB.90.045109
ADS CrossRef Google学术搜索
Zhao Y，Truhlar DG（2006）一种新的局部密度，主要用于主族热化学，过渡金属键合，热化学动力学和非共价相互作用。J Chem Phys 125（19）：194101。https://doi.org/10.1063/1.2370993，http://link.aip.org/link/?JCP/125/194101/1
ADS 交叉引用谷歌学术
Zhao Y，Truhlar D（2008）用于主要组热化学，热化学动力学，非共价相互作用，激发态和过渡元素的m06密度泛函套件：两个新的功能和四个m06级功能和12个其他功能的系统测试。Theor Chem Accounts 120：215-241
CrossRef Google Scholar
Zhao Y，Schultz NE，Truhlar DG（2006）通过将约束满足方法与热化学，热化学动力学和非共价相互作用的参数化相结合来设计密度泛函。J Chem Theory Comput 2（2）：364-382
CrossRef Google学术搜索
Zheng X，Cohen AJ，Mori-SánchezP，Hu X，Yang W（2011）改进密度泛函理论中的带隙预测，从分子到固体。Phys Rev Lett 107：026403。http://link.aps.org/doi/10.1103/PhysRevLett.107.026403
ADS CrossRef Google学术搜索
版权信息
©Springer International Publishing AG，Springer Nature 2018的一部分
部门编辑和附属机构
天使卢比奥1
1. 德国汉堡物质结构和动力学理论系MPI
关于此条目
CROSSMARK
引用此条目为：
Li L.，Burke K.（2018）密度泛函近似的最新进展。在：Andreoni W.，Yip S.（编辑）材料建模手册。施普林格，湛
首次在线
2018年6月12日
 DOI
https://doi.org/10.1007/978-3-319-42913-7_11-1
出版商名称
Springer，Cham
 在线ISBN
978-3-319-42913-7
电子书包
物理和天文学
重印和权限

密度泛函近似的最新进展

引用 条目
怎么引用？
.RIS
论文
参考经理
RefWorks
Zotero
.ENW
EndNote
.BIB
BibTeX
JabRef
Mendeley
操作
超过1000万份科学文档触手可及


Academic Edition
企业版
家 IMPRESSUM 法律信息 隐私声明 我们如何使用cookies 无障碍 联系我们
施普林格自然
©2018 Springer Nature Switzerland AG。部分斯普林格性质。

未登入 东吴大学图书馆（3000131270） - 5102 SpringerLink China OAC National Consortium（3000202650） - 7217 SpringerLink江苏电子书联盟2010-2012版权年（3000203264） - 10786 SLCC江苏（3000803042） - LID 4548 - SpringerLink江苏eBo联盟2006- 9版权所有（3001144869） - SpringerLink江苏电子书联盟2016-2018版权所有年份（3002351942） - 13433 SpringerLink江苏电子书联盟2013-2015版权所有（3991438083） - SLCC江苏电子期刊联盟2015-2017（3991465546） 42.244.24.216

沪ICP备15051854号-2

```
