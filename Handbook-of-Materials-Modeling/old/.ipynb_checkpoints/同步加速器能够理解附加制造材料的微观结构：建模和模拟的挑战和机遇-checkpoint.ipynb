同步加速器能够理解附加制造材料的微观结构：建模和模拟的挑战和机遇

抽象
从建模和模拟的角度来看，增材制造是一个明确的多尺度问题。无论3D打印是通过熔化，聚合还是使用粘合剂完成，该工艺的规模都是亚毫米级，这意味着在制造零件时会累积数十到数千层。几何形状的变化意味着光或电子束所遵循的路径（除了在基于整个层的照明的情况下）导致连续的重叠热输入之间的高度可变的时间间隔。特别是在熔化粉末的工艺的情况下，这可能导致偏离导致缺陷的预期热输入。将微观结构作为晶体和缺陷结构的总和，这意味着使用模拟来预测微观结构需要在多个尺度上进行计算：当部件在所有尺寸上延伸到厘米时，用亚毫米尺度模拟微观结构发展是不可行的，例如晶粒，取向和孔隙。同步辐射非常适合探测增材制造中涉及的单元过程，因此专注于亚毫米级材料加工为评估建模和仿真的需求和前景提供了有用的框架。

1简介
本章旨在向读者介绍对表征附加制造材料最有用的技术以及最近结果对建模和模拟的影响。这些技术强调结晶材料，主要是金属和陶瓷。对于添加制造的聚合物存在类似的挑战，并且由于在撰写本文时存在的快速发展，生物打印特别令人感兴趣。尽管同步辐射对这些应用的发展较少，但预计未来同样具有相同的影响力。讨论的技术包括X射线微型计算机断层扫描（xμCT），广角X射线散射（WAXS），高能衍射显微镜（HEDM），残余应力（RS）和动态X射线照相术（ DXR）。

增材制造（AM）涵盖了用于制造几乎所有材料中的原型，中间形式和最终部件的大量技术。一些书籍和评论文章是可用的，在不同程度上解释了技术是如何工作的（Gibson等2010）。简而言之，AM领域已经从主要采用聚合物原型制造的物体转向制造聚合物和金属的最终部件，只需要极少的精加工（Wohlers 2017））。设计的复杂性和减少零件数，以及快速推向市场，都有利于AM方法。专注于金属部件，小尺寸是有帮助的，因为主要粉末床机器的建造率大约为3毫米/小时，这意味着高层建筑必须运行几天。当然，提高制造速度是机器制造商的主要推动力。其他技术，例如基于电子束或激光焊接的送丝，允许更大的部件，虽然分辨率更低，后处理更显着。粘合剂喷射技术避免了其不可避免的残余应力熔化，但需要去除粘合剂和烧结。正在开发其他技术，例如精密金属液滴沉积（Murr和Johnson 2017）。

即使对于沉积阶段，这种方法的模拟也是显而易见的多尺度：例如，金属粉末床需要熔化许多米的熔体线，每个熔体线的横截面（宽度×深度）为200×100μm。与熔化通过每层到（至少）下面的层相结合的熔体线的邻接意味着每个位置的热历史需要模拟几条线和层。然而，模拟整个构建顺序也是必要的，用于计算残余应力，变形以及影响预测微观结构所需的实际热历史的热积累。因此，通常合并步骤使得整个层（或一组层）被视为单个热事件。尽管有这些挑战，2017）。此外，还在各种尺度上进行验证，证明预测的热历史是可行的（例如，Beuth和Klingbeil， 2001）。残余应力已被证明是可预测的，但热变形被认为对金属更具挑战性，因为与熔点相比，床温通常较低（例如，Szost等人 2016 ; Mukherjee等人 2016）。这与使用粉末的聚合物印刷形成对比，其中有效熔融温度相对于床温度低，使得热应力和变形不太严重。

2同步加速器功能
如上所述，基于同步加速器的X射线所特有的主要能力是微断层扫描（CT），广角X射线散射（WAXS），小角度X射线散射（SAXS），高能量衍射显微镜（HEDM），残余应力（RS）和动态X射线照相术（DXR）。我们现在简要回顾一下它们各自的能力，以告知添加剂制造材料的微观结构发

2.1 X射线微断层扫描（CT）
微米和纳米尺度的计算机断层扫描是一种广泛使用的方法，用于材料内部结构的非破坏性表征。在较粗糙的尺度上，中子（例如，Cao等人2016）和μ子（Muon Tomography 2017）也是有用的，尽管作者还没有意识到AM的应用。它对于大密度对比度最有效，因为它主要依靠比尔 - 朗伯定律来测量吸收的变化，并通过相位对比来检测密度的急剧变化（Gursoy等，2014）。AM的主要应用是测量孔隙度，其中孔隙构成结构应用中材料的重要缺陷（Hudák等人2016 ; Eylon和Strope 1979）; 斯嘉丽等人。2016a，b）。孔是拉伸载荷下的应力上升，并且在大约10μ的尺寸以上，可以作为疲劳裂纹的来源，如许多关于添加制造的材料的报告中所记载的（例如，Leuders等人，2015）。疲劳是一个典型的极值问题，因为原始缺陷可以追溯到材料中可用的最大缺陷，有大量文献（Weibull 1951）。同样，同步加速器断层摄影术可用于提供数据集，然后可用于在相关长度尺度上执行机械响应的模拟（Cunningham等人2017b ; Kantzos等人，2018）。图1图2是气体雾化Ti-6Al-4V粉末颗粒的两种视图，是一种常用于航空航天和医疗行业的材料; 平均尺寸约为60μm，这是激光或电子束粉末床机器中使用的粉末的典型尺寸。图（a）显示了每个粒子的外表面，其具有随机指定的对比色; 图（b）显示了深红色的每个空隙或孔（颗粒内）用于强调。使用这种精细粉末，可用于具有同步辐射的高Z材料的亚微米分辨率对于在整个尺寸范围内分辨颗粒特别有用。

AM的一个吸引人的特征是能够在相对于结构材料的各种密度下印刷任意连接的材料。在粉末床AM中，这种方法所暗示的广泛自由度受到对最小横截面的要求和悬垂固有的困难的限制，即在未熔化粉末的显着区域上印刷固体材料。尽管存在这些局限性，但已经在晶格结构上发表了许多结果，并采用了许多优化结构的原始方法（例如，Calignano 2014））。在格子中，格子中的各个支柱的横截面变化并且具有附着在其表面上的部分熔化的粉末颗粒。与凝固和随后的热收缩相关的收缩也导致产品相对于原始设计的扭曲（Dunbar等人2016））。众所周知，这种失真的程度取决于预热，所谓的支撑结构和许多其他因素。较高的预热会降低热收缩的程度。支撑结构既可以打印悬垂部分，也可以将部件连接到底板上。这种连接意味着在切割通过支撑结构以将其分离之前，可以在底板上对部件进行热处理以通过蠕变消除应力。同步CT还被证明可用于测量格子的质量，因为它可以直接测量整个3D结构。Bormann等人。（2013年）发表了一种优雅的方法，在这种方法中，当他们加热印在NiTi中的晶格结构时，他们不断获得X光片，用作组织向内生长的支架。作者发现支架与预期设计存在显着偏差，其中更多位置显示出多余的材料，反之亦然。正如Khairallah等人。（2016）已经指出，这并不令人惊讶，因为熔池往往会在其周边拉出颗粒，部分熔化的颗粒可能超出目标熔化体积。布伦等人。（2013）报道了对聚己内酯支架的类似研究，包括传统制造方法和3D打印材料之间的比较。

Carlton等人。（2016）使用X射线微CT和原位拉伸试验来研究3D打印不锈钢的损伤和断裂。他们发现孔隙度分布可能具有重要影响，因为具有大且不均匀孔隙分布的样品表现出从现有空隙开始的裂缝。然而，在空隙率低的试样中，断裂行为不受小孔均匀分布的影响。Sandgren等。（2016）使用micro-CT研究在粉末喂料机中印刷的Ti-6Al-4V中的疲劳裂纹扩展。

2.2广角X射线散射（WAXS）
Cakmak等。（2016）在APS上使用广角X射线散射来测量电子束印刷的IN718（Ni基高温合金）的纹理，其表现出与构建方向（BD）平行的强纤维纹理。具有<100> // BD的强纤维织构与观察到的强柱状微结构一致。虽然基于同步加速器的X射线由于其穿透力而具有纹理确定的一些优点，但是EBSD和基于实验室的X射线极图的组合对于大多数研究是有效的。

2.3小角度X射线散射（SAXS）
同步加速器的一个显着优点是具有高强度的高能X射线的可用性，其允许深度穿透并因此需要查询大量的体积。许多合金经过热处理，最常见的是提高强度或达到强度和韧性的最佳组合。添加制造的金属通常表现出与常规加工材料不同的热处理响应。例如，已知Ni基合金IN718和IN625通过γ '和Laves 等相固化，碳化物和δ相对于晶粒尺寸控制更为重要。张等人。（2017年）使用SAXS研究主要合金元素即Ni，Cr，Nb和Mo的均匀化动力学。利用特征条纹图案，他们确定大部分偏析被限制在6nm范围内的区域内。每个枝晶。连续SAXS图像之间的相关性分析提供了关于扩散控制的均质化过程的动力学的定量信息。同样，薛等人。（2016）使用SAXS在Ti-48Al和Al-Cu-Mg合金中在高冷却速率条件下检测沉淀物形成。Idell等人进一步研究了IN625中的硬化响应。（2016）和Lass等。（2017）谁用常规方法确定δ阶段从一开始就占主导地位。δ相不是在晶界处非均匀地出现，而是以盘形板的形式在整个材料中沉淀，与基质具有明确的取向关系。Idell等人。（2015）对IN718进行了类似的研究，并指出基于同步加速器的SAXS再次用于测量沉淀过程的进展，因为X射线的通量大和能量高。

2.4高能衍射显微镜（HEDM）
有几种基于衍射的技术使用同步加速器X射线来绘制各种分辨率的多晶微结构（Rollett和Barmak 2015））。高能衍射显微镜（HEDM）是一类本身存在于两种主要变体中的技术。其中之一被称为近场HEDM（nf-HEDM），因为探测器相对靠近样品放置，使得衍射斑点的位置对样品中衍射光束的原点敏感。到衍射级（即，哪个晶面）。样品在光束前旋转，以从每个位置激发大量（至少20个）光束; 光束通常是平面的，以限制每个视图中的斑点数量。模拟退火算法筛选材料中的潜在方向，以匹配尽可能大的一部分点，并在常规网格中的每个点处指向方向（Li和Suter 2013））。HEDM（ff-HEDM）的远场变体从探测器获取数据，探测器的位置远离布拉格环明显的样本，这极大地促进了方向的指数化，但足够接近使得各个点与标准的偏差位置给出每个衍射颗粒的质心和（平均）弹性应变的信息（Bernier等，2011）。同样，样品在光束前面旋转，使得光点位置随着每个晶粒围绕旋转轴旋转而变化。重建以与nf-HEDM相似的非破坏性方式提供关于每个颗粒的信息，其具有关于弹性应变状态但不是3D方向图的附加信息（类似于EBSD（Schwartz等人，2000））。

最近，该领域的研究人员开始将HEDM的远场（ff-HEDM）和近场（nf-HEDM）变体结合起来。作为一个示例，ff-HEDM（Bernier等人2011）可以在3D体积上执行，其提供具有其质心和取向（以及弹性应变）的晶粒列表。然后可以使用该列表来“种子化”nf-HEDM重建，这节省了大量的计算时间，因为在定向空间中的搜索在其他方面是非常耗时的过程。然后，nf-HEDM重建提供了更精确的多晶微结构的空间图。这样的图提供了实例化全场微机械模拟的基础。特纳等人。（2017），用于从这样的组合的实验（Turner等数据2016）实例化全场有限元模拟。计算的弹性应变与测量值的比较显示出良好的一致性。然而，它们不能结合存在于未变形材料中的残余弹性应变。因此，他们必须使用初始值作为参考点来比较应变值的变化。之前的努力将残余应力作为本征应变（例如，Salvati等人2017），Pokharel和Lebensohn（2017年））证明有可能通过基于Eshelby的本征应变计算来近似颗粒尺度的初始（残余）应变，这很重要，因为退火多晶材料上的大多数弹性应变测量已经显示出可观的残余应变水平，例如，Oddershede等。（2010）。尽管如此，本征应变计算（Pokharel和Lebensohn 2017）只是将这种残余应变（应力）条件纳入多晶变形模拟的第一步，也是理解其起源的挑战。这表明在退火材料生产之前对材料加工进行建模可能有助于测试已提出的各种假设。

2.5残余应力（RS）
关于残余应力的大量文献清楚地表明它是焊接中的重要问题。主要原因是由熔池周围的加热引起的变形不是完全可逆的。在一定温度以上，几乎所有材料都松弛，使得在焊接的热影响区中的最高温度下，应变低（松弛）并且随后的冷却导致热收缩。收缩是一种本征应变，一旦材料冷却到应力松弛温度以下，即低于发生快速蠕变的点，就会产生应力。当然，残余应力的大小及其张力特性取决于焊缝周围材料的几何形状和程度，但这提供了基本的图像。在大多数情况下，没有使用深度穿透（带钥匙孔）。熔池的横截面为半圆形，应力的大部分变化与长宽比有关（Gratzke等。1992年）。如上所述，使用选择性熔化（激光或电子束）的增材制造技术基本上是微焊接工艺，专注于粉末的熔化和固结，而不是连接。因此，AM中残余应力的问题具有与焊接相同的物理基础。如Mach等人所做的重要贡献比比皆是。（ 2017）已经证明同步辐射可以有效地用于测量全弹性应变张量（可以从中得到应力）的空间变化。

对于增材制造的应用自然地通过以下事实产生：SLM在某种程度上是沉积50微米厚的连续凝固层的过程。与每层相关的收缩导致残余应力，其倾向于在部件上施加双轴收缩。如其他地方所述，AM机器制造商通常建议使用预热，这在电子束机器中通常比在SLM中高得多。怀特塞尔等人。（2016）描述了使用高能X射线使用实验室源进行残余应力测量。Park和Okasinski（2017）讨论了用同步辐射以空间分辨方式测量弹性应变的技术。基本思想包括使用狭缝来选择衍射光束来自样品的哪个部分。对于锥形狭缝，可以在输入光束穿过的路径内隔离150微米长的区域。Park和Okasinski（2017）报告使用螺旋形狭缝，尽管只允许每个布拉格环的一小部分到达探测器，但仍能提供更好的性能。然而，螺旋形狭缝旋转以捕获整个布拉格环。作者指出，螺旋狭缝对于晶体对称性低于立方的材料特别有用，因为每组锥形狭缝必须针对特定的布拉格角构造。然而，本文中描述的实验是在测试物品上，并且尚未报道添加制造材料的应用。

Mishurova等。（2017）描述了由拱形或桥形组成的典型测试件，其高度为8×10×20mm×深度×长度。该部件在SLM机器中由Ti-6Al-4V粉末印刷，并且在若干位置进行测量，这意味着尽管在光束位置之间最近的间距为0.5mm，但未探索到可获得的空间分辨率。测量在德国的同步加速器中进行，白色光束的能量范围为10-150keV。他们的主要结果是低能量密度，意味着更高的速度和更低的能量，导致显着更高的应力，最大弹性应变接近2.10 -3，相当于1 GPa以下的应力。毫不奇怪，在从基板移除部件之后残余应变状态发生变化，这反映了后者施加的约束的消除。

有关于残余应力建模的重要文献。在更直接的形式中，热历史通过移动热源的通过来模拟，通常简化为移动点源的热源。后一种简化方法的优势在于可以使用分析型罗森塔尔溶液（Rosenthal 1941），它可以根据熔池尺寸的实验数据进行直接校准（Promoppatum等，2018））。标准算法是适用的，并且正在使用各种商业支持的有限元代码。模拟热变形需要解决热机械问题，这反过来意味着必须以足够的精度知道材料的机械和热性能，直至熔点。Heigel等。（2015年）提到需要“基于测量的对流模型”以获得准确的结果。如上所述，由于热应变的发展取决于强度随温度的变化，另一个挑战是缺乏接近熔点的机械性能如蠕变强度，泊松比，模量等的详细数据。考虑到深度穿透（锁孔）条件的有力证据，例如Trapp等人，点热源的假设不太可能是准确的。（2017），这导致熔池中的纵横比大，并且有效地产生线热源。当然，有许多论文正在发表，它们对粉末床建模过程进行了不同程度的完整性建模。Lindgren等。（2016年例如，提供相对完整的这样的例子，其包括残余应力计算以及材料特性。

2.6动态X射线照相术（DXR）
当直接可视化时，所有过程都会获得新的含义。与许多3D打印机中使用的选择性激光或电子束熔化相关的高速，高度局部化的熔化和再冷却也不例外。已经使用俯视该过程的高速光学成像系统进行了许多有价值的实验（例如，Furumoto等人，2013）。然而，必须对尝试高温测定的温度场进行重要假设，并且通常需要过滤图像。赵等人提出的直接高速放射成像的出现。（2017年）承诺改造这个地区。样品被限制在具有真空能力的腔室中的透明碳板之间，该腔室具有用于X射线的窗口和用于520W激光器的光纤穿通装置; 激光系统与SLM机器中使用的类型相同（图3）。在APS的光束线32ID处使用2×2毫米的粉红色X射线束和高速成像系统，他们在一定范围的功率水平下成像局部点熔化，有粉末和无粉末。粉末颗粒的排出与直接测量速度一起可视化，因此说明了SLM机器内气流限制颗粒再沉积的重要性。熔池和蒸汽腔的形成是显而易见的。图4说明了事件的顺序（Zhao et al。2017）当激光撞击时，形成熔池，喷射粉末颗粒，最后在细长熔池冻结时留下空隙。这样的空隙是焊接中轨道末端缺陷的特征，并且它还提供了有时在SLM中观察到的系统空隙阵列的情景（Groeber等人 2017）。最近，增加了扫描系统（Parab等人， 2018年），开辟了研究与SLM实践相关的各种功率水平和速度的可能性。

3对建模和仿真的影响
建模和仿真中最重要的需求是改进零件变形的预测。这不是一项简单的任务：虽然热历史的预测相当简单，但计算实际位移并不可靠。温度场主要取决于热源和热性质，这些通常是众所周知的。然而，计算产生的扭曲更具挑战性，因为必须在熔点之前一直很好地定义额外的温度和应变率依赖性，例如屈服强度，泊松比和蠕变速率。这些性质难以测量，并且许多材料没有很好地表征。

现在可行的高速衍射实验表明，需要大量的正向建模工作来模拟X射线散射，并朝着更完整地理解动态微观结构演变的方向发展。衍射的正演模拟在许多不同的领域中进行，但是为了准确模拟任何特定实验而必须包括的实际物理学在很大程度上取决于具体情况。

此外，微观结构与热机械性能之间关系的多尺度建模，以及微结构演化的计算机模拟均受益于同步加速器实验研究。例如，诸如原子结构，晶格参数和局部有序的X射线衍射结果为微观结构特征的原子和中尺度建模提供输入。此外，层析成像结果为微观结构演化的中尺度模拟提供了丰富的验证数据。

4结论
至少在初步意义上，已经探索了同步加速器X射线辐射实验有助于提高我们对增材制造工艺的理解的大多数方式。鉴于增材制造中的高冷却速率，尤其是SLM，这一活动领域很可能会继续迅速发展，同步辐射将在阐明与增材制造相关的科学挑战中发挥重要作用。已经使用广角和小角X射线散射来测量微观结构演变，特别是扩散和沉淀过程。动态X射线照相术（DXR）已被用于直接对熔化和再冷冻过程进行成像，并对高速衍射进行了互补实验。尽管可以扫描的体积有限，但由于同步加速器X射线具有高空间分辨率以及可行的相对高吞吐量，计算机断层扫描（CT）已经做出了许多贡献。更高级的技术，如高能衍射显微镜看起来可能在不久的将来为AM材料的3D表征做出贡献。同步辐射X射线辐射实验为建立和验证微观结构演化的原子和中尺度计算机模拟提供了有价值的信息。更高级的技术，如高能衍射显微镜看起来可能在不久的将来为AM材料的3D表征做出贡献。同步辐射X射线辐射实验为建立和验证微观结构演化的原子和中尺度计算机模拟提供了有价值的信息。更高级的技术，如高能衍射显微镜看起来可能在不久的将来为AM材料的3D表征做出贡献。同步辐射X射线辐射实验为建立和验证微观结构演化的原子和中尺度计算机模拟提供了有价值的信息。
