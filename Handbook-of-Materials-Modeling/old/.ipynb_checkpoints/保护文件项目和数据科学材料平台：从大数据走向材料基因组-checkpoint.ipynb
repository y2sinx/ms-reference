抽象
描述了材料信息学中最古老的计划之一，即PAULING FILE项目。它包括无机结晶化合物的综合数据库，其原子结构，内在物理性​​质和相图。最重要的是，引入了功能强大的在线检索软件，称为MPDS，即数据科学材料平台。给出了存储，交换和分析大量材料数据的实用方法。重点是现代信息技术和软件工程。因此，从大型异构数据中，可以得出关于整套已知材料的整体结论。它们可以被视为系统性大规模预测的指南。

1简介
PAULING FILE项目是具有近25年历史的材料数据库，在同一框架下对无机结晶物质的晶体学数据，相图和物理性质进行分组。重点放在实验观察上。每个单独的晶体结构，相图或物理属性都源自特定的同行评审出版物。从1891年至今，涵盖了材料科学，工程，物理和无机化学领域的世界科学文献。数据科学材料平台是2016年创建的PAULING FILE项目的在线版本。它在两个在线界面中呈现所有PAULING FILE数据：图形用户界面和应用程序编程界面。前者适用于材料科学家; 后者适用于软件工程师和数据科学家。这些群体的研究兴趣的交叉点属于材料信息学的新学科的范围。

2材料大数据和保罗文件项目
2.1材料社区面临的现代挑战
与直觉相反，经验传统在基础科学，尤其是材料科学中广泛存在。化学信息学的先驱，Peter Murray-Rust（英国剑桥）称，材料科学是最保守的精确学科之一，与其他科学相比，最不透明，更开放的合作和众包，如生物化学，天文学，数学，计算机和环境科学（Murray-Rust 2013） - 不算付费墙。有趣的是，与化学和生物信息学相比，材料信息学领域不那么成熟和年轻（Gasteiger和Engel 2003，这里信息学假设信息交换））。在新颖的分布式计算项目中，任何人都可以自愿计算从个人计算机到特定原因的计算时间，只有很少一部分涉及材料科学。可以给出以下非详尽的解释。材料科学与工程最紧密地与行业紧密相关，例如航空航天，汽车，电子，军事工业等。材料研究和开发的相当一部分是私人资助的。与其他精确学科相比，材料科学的专业知识平均成本更高。这些成本可以与高能物理或天文学相提并论。然而，这些诀窍对技术进步的影响也更大，投资与创新之间的滞后相对温和（约15年或更短，奥巴马2011年））。此外，由于其广泛和极端的复杂性，材料科学也非常分割（“封建”）。因此，学术界和工业界（政府补助金，职员职位等）的竞争程度都很高。就复杂性而言，即使掌握现代固态物理学的基本形式主义也是一项非常重要的任务。多学科专家在材料科学中很少见。这些原因可能会给出一个想法，为什么材料科学界天生就非常保守。

此外，直到最近几年，共享基础研究成果的文化（例如，原始测量或模拟日志，与文章形成对比）完全缺席，有时甚至在材料科学中被禁忌。关于这些出版物，仍无法从期刊获得机器可读格式的科学信息，以便进一步重复利用和重新利用。然而，随着技术的进步，现代学者们正在沉入“数据的海洋”。今天，人们普遍抱怨研究人员出版的内容过多而且过快。为了估计这种增长，引用学术数据库的增长是错误的，因为没有数据库可以捕获所有内容。马克斯普朗克协会和瑞士联邦理工学院的文献计量分析师（Bornmann and Mutz 2015）估计全球科学产出每9年翻一番。此外，随着现代数据密集型方法（通常称为“材料设计”或“材料基因组”）的普及，有证据表明，通过实验或模拟生成的材料科学中的原始大数据量将继续增加成倍增长。谈到大数据，计算机科学界已经恰当地定义了这种现象，因为数据量由四个指标控制：数量，速度，种类和准确性。在材料科学中，体积指的是大尺寸，超过个人计算机设备，速度指的是利用实时数据采集（例如，来自动力学实验）。多样性涉及这样一个事实，即数据采用材料科学中的所有形式，从离散数值到材料行为和成像数据的定性描述。Veracity承认关于不确定性和大量“缺失”数据的实际现实（Rajan 2015）。如今是材料科学大数据的时代。然而，在这个大数据时代，材料科学仍然不足以鼓励和实现研究人员收集的大量数据的共享，分析和解释。材料科学界传统的科学信息交流手段非常不完善。在这种情况下，出现了自然的多样性，自然选择的原则引导了前沿研究的形式。在互联网和现代信息技术高度渗透的最近几年，情况开始缓慢变化。在这里，消化信息，得出正确结论的能力至关重要。这是数据科学工具的地方（例如，机器学习）需要与材料科学的基础相联系：理论，建模和实验。目的是使用于生成新信息的实验室，而不仅仅是用于检索已知或预期信息的存储库（Ghiringhelli等人2017）。

2.2保护文件背景
世界上有许多倡议试图克服材料科学中的上述挑战（Pizzi等人，2016年））。主要思想是系统地收集材料数据或以高吞吐量方式生成和处理它们。其中一项值得注意的举措是于1995年推出的PAULING FILE项目。其主要重点是对已发表的实验材料数据进行批判性评估。从历史上看，它是日本科学技术公司，瑞士材料阶段数据系统公司和东京大学RACE的合资企业。现在它完全由瑞士公司管理。从一开始就计划了三个步骤。第一个目标是创建和维护无机结晶物质的综合数据库，包括晶体学数据，衍射图，内在物理性​​质和相图。应仔细检查数据。术语“无机物质”定义为不含CH键的化合物。与数据库创建并行，第二个目标是开发适当的检索软件，以便在单用户界面中访问数据。从长远来看，作为第三个目标，应创建用于材料设计的新工具，以自动搜索数据库中的相关性。这被称为具有预定义的内在物理性​​质的新型无机材料的智能设计。试用版PAULING FILE二进制版于2002年作为桌面软件发布。现在，PAULING FILE数据的选定部分包含在多个印刷，离线和在线产品中。今天，PAULING FILE项目非常有名。已有数千种出版物在提及它。它的基础，数据库设计，2004年 ; 维拉斯等人。2008年 ; 徐等人。2011年 ; Kong等。2012年 ; Villars和Iwata 2013等）。最近实施的PAULING FILE检索软件和材料设计工具是一种名为Materials Platform for Data Science的在线产品。

现在必须给出最少的必要定义。一个数据库是服务于多个用户的需求相互关联的存储的数据的集合。在所谓的关系范例中，这些存储的数据被组织在表中。使用数据库而不是文件的动机包括为各种用户提供更高的可用性，集成数据以便更轻松地访问和更新复杂事务，以及减少数据冗余。一个数据库管理系统（DBMS）是用于操纵数据库的通用软件。DBMS支持逻辑视图（数据库模式），物理视图（访问方法，数据集群），数据定义和操作语言，以及实用程序，例如事务管理和并发控制，数据完整性，崩溃恢复和安全性（Teorey等人，2005）。例如，Oracle用作PAULING FILE的内部DBMS。PAULING FILE数据库具有以下结构。标准数据单元称为条目。所有条目细分为三种：晶体结构，物理性质和相图。它们相应地称为S-，P-和C-条目。条目具有持久标识符，类似于数字对象标识符（DOI），例如，S377634，P600028和C100027。PAULING FILE数据的另一个维度是不同的阶段。这三种条目通过它们所属的不同材料阶段相互关联。保护文件编辑在过去25年中做了大量工作，手工区分了超过150,000个无机材料阶段，出现在世界科学文献中。每个相具有（a）化学式和（b）改性的独特组合。这些是使用结构原型，Pearson符号和空间组定义的。每个阶段都有一个称为阶段ID的整数标识符。在接下来的三个部分中，将详细介绍每种条目类型（S-，P-和C）及其链接。

2.3物理性质
保护文件的P-条目包括实验和有限程度的模拟数据，用于固体结晶状态的无机化合物的广泛固有物理性质。所考虑的物理性质属于以下七个通用域之一：（a）电子和电学性质，（b）光学性质，（c）磁性，（d）机械性质，（e）相变，（f）超导性和（g）热和热力学性质。分类法由三个级别组成：提到的一般域，子域和特定的物理属性。例如，域“电子和电学性质”包含子域“电子能带结构”，其又包含“费米能量”性质等。目前，存在约100个子域和近2000个特定物理性质。 。该分类法由Fritz Hulliger（瑞士苏黎世瑞士联邦理工学院），Roman Gladyshevskii（乌克兰利沃夫Ivan Franko国立大学）和Karin Cenzual（瑞士日内瓦大学）编制。在一定程度上，它反映了上个世纪固态物理学的发展。

物理特性以四种不同的方式存储：数值，图形描述，属性类（例如铁磁体，压电等），以及源出版物中特定数据的存在的指示，例如，不同的光谱。最常见物理特性的符号已经标准化，主要基于CRC化学和物理手册（Lide 1997-1998）。数值存储在已发布的单位中并转换为标准SI单位。对于原子级别的某些属性，使用其他单位，例如eV或μB。将相对于确定量的物质（每kg，每摩尔）表达的性质转化为每原子 - 克。每个数值属性值都伴随有关于特定测量的实验条件的信息。通过指向参考表的链接提供了极大的灵活性。由此，可以选择新属性，并且可以控制它们的符号，单位和幅度范围。

所有数据均来自主要文献。每个P-entry对应一个特定的数据源，可以包含几个数值，图形描述和关键字。为了通过温度或压力诱导的结构相变来研究化合物，将存在两个P-条目，例如，一个用于室温改性，一个用于低温改性。默认情况下，假设铁电转变伴随着结构变化，并证明产生两个P-入口是合理的，而磁，电或超导跃迁则不然。具有一定均匀性范围的相的数据按代表性化学式分组。当与表示P-entry的组成不同时，特定测量的实际组成，在参数中指定。至于晶体结构部分，将有三个数据库条目用于两种三元化合物之间的连续固溶体：一种用于每种三元边界组合物，第三种用于包含四种化学元素的分组样品。还包括来自ab initio计算的一些模拟数据，特别是能带结构，但重点是实验测量的数据和直接从测量得到的值。

除了物理特性（以数值，图形描述或关键字的形式）和强制性项目（例如化学式）之外，还存储有关样品制备和实验条件的大量信息。以下数据库字段可以存在于物理性质P-条目中：（a）化合物，例如化学系统，公开的化学式（被调查的样品），代表性的标准化学式，改性; （b）书目数据，例如参考文献，作者，语言，标题等; （c）制备，例如原料和合成方法; （d）样品描述，例如形式，化学分析，温度，压力和组成的稳定性，弹性行为，密度，颜色，化学反应性; （e）晶体学数据，例如结构原型，空间群和细胞参数。

2.4结晶结构
目前，PAULING FILE包含超过350,000个晶体结构S-条目。保护文件中S条目的最低要求是一组完整的已发布单元格参数，分配给明确定义的组合物。每当公布的数据可用时，晶体学数据还包括原子坐标，位移参数和实验衍射线，并附有有关制备，实验条件，样品特征，相变和细胞参数对温度的依赖性的信息，压力和成分。为了给出实际结构的近似概念，为S-条目提出了一套完整的原子坐标和站点占用，其中可以分配原型。1987年）。相关时，进一步调整它们，以便可以直接比较同种型S条目的数据。衍生数据包括基于最大间隙法的各个原子位点的原子环境（Brunner和Schwarzenbach 1971 ; Daams等 1992））和Niggli减少的细胞。通过使用上一节中提到的程序包ESDD，检查S条目中S条目内的不一致性，并比较不同的S条目。对于5％的S条目，检测并校正公布的晶体学数据中的一个或多个印刷错误。关于剩余的短原子间距离，与名义成分的偏差等的警告在评论中添加。SI单位随处可用，晶体学术语遵循国际晶体学联合会的建议。

与物理属性类似，所有数据都是从主要文献中提取的。如果可用，作为CIF文件或其他格式存储的补充材料用作数据源。大约10％的已处理文件以原始版本（例如俄语）和翻译版本（英语）存在; 避免重复，并存储两个引用。通过ab initio计算模拟或通过其他方法优化的晶体学数据仅被认为通过实验观察证实。为特定论文中报告的所有完整改进创建了不同的S条目。对于没有公布原子坐标的单元参数，为每个化学系统和晶体结构准备S条目。例如，对于两种三元化合物之间的连续固溶体，将有三个S条目：一个用于每个三元边界组合，一个用于四元系统。后者可以包含描述细胞参数的组成依赖性的注释。对于可检索细胞参数的选择，优选在环境条件下确定的值。

3.2保护文件材料基础设施的开发
截至今天，材料信息学是一系列从计算机科学中获取并用于现代材料科学的食谱。从学术角度来看，主要的困难纯粹是技术性的 - 如何有效地处理材料大数据。这里讨论一些食谱。所有这些都是在开发和维护PAULING FILE在线检索软件--MPDS（数据科学材料平台）时考虑的。

解决复杂性的可能方法之一是用于软件工程的统一建模语言（UML），通用，开发，建模语言（参见例如Miles和Hamilton 2008）。从鸟瞰图来看，UML是一种以人类可理解的方式绘制概念的惯例。虽然它不能被计算机解释，但UML提供了有问题字段的正式描述，然后更容易用计算机编程语言进行编码。另一个UML优势是其在软件工程领域的标准化和高度普及（适用于协同工作）。UML中的概念必须与定义的关系类型相关。

通常，万维网基于互连的概念。实际上，在现代环境中，没有孤立的软件本身是有意义的，每个应用程序都与其他应用程序通信。为了提供快速有效的交互体验而不为每种情况开发新的访问接口，通常使用应用程序编程接口（API）。API调节任何类型软件之间的通信（无论是数据挖掘程序，仿真平台还是任何其他大数据消费者的复杂）。主要思想是将功能收集在一个地方，并通过在线API进行公开（封装）。在线媒体中的API通常遵循代表性状态转移或REST的原则（Fielding 2000）。REST为客户端 - 服务器软件架构提供了指导性约束，可以称为meta-API。它也用于MPDS API（参见图2），它以开发人员友好的机器可读方式呈现所有PAULING FILE数据，使用打开的格式，如CIF，JSON和MIF（见下文） 。重要的是，API不仅仅是客户端与服务器通信的方式。从更广泛的意义上讲，在线API是一种软件架构，宣告了所有通信的方式。这样，观众不受现有的面向人的图形用户界面的约束（无论其便利性如何）并且能够最大限度地使用服务提供商来实现其目标。

尽管如此，MPDS对其图形用户界面给予了很多关注。搜索输入字段允许一次无缝地键入基本搜索项的不同组合，以便正确识别它们并显示匹配的条目。负责处理这种无缝搜索的算法如下工作。首先，过滤停用词（“the”，“and”，“about”等）。然后，检查单词属于一学期类别：化学式，化学元素和晶格。这些类别不能包含多个单词。找到的单词被解释和排除。然后，检查剩余的单词属于材料类和物理属性类别。最开始检查最长单词组合的匹配，然后检查较短的组合。无法识别的单词被标记为“被忽略”。

从MPDS开始，所有子系统的开发性能都是基石。当服务器是单点故障时，这对于集中式在线软件尤为重要。通常，对于任何在线项目，始终建议关注操作速度。应该注意的是，为5个同时用户设计的软件与为50或500个同时用户设计的软件有很大不同。有两种方法来缩放性能：垂直和水平（Michael等人，2007年）。垂直缩放意味着随着负载的增加部署功能更强大的服务器，但软件架构保持不变。该解决方案快速有效，但有限。首先，服务器越强大，越昂贵，并且依赖性不是线性的。其次，每台服务器的计算容量有限。并且数据库性能通常比这个限制低得多。因此对于数据库经过一定的容差后，垂直缩放毫无意义。这种情况下的解决方案是水平缩放当软件架构以这样的方式改变时，在服务器集群中均匀地分配增加的负载，由廉价的商品硬件组成。这个想法是通过简单的快速复制将带有MPDS软件的新副本添加到服务器集群中。新副本加入服务器群集并开始接受传入的HTTP请求。此外，如果副本因任何原因失败，它可以自动替换为新副本。这样，服务器群集保持高可用性，即使是巨大的负载也不会影响性能，因为可以添加更便宜的副本。然而，虽然服务器价格仍然受到控制，但开发很复杂并且需要时间。

3.3材料数据的存储和交换
与任何其他科学数据类似，材料数据通常存储在文件和数据库中。这些文件通常呈现原始数据（模拟或测量日志，压缩档案等）。可重复使用的处理数据通常存储在数据库中。应该注意的是，尽管这种细分是非常有条件的，而且文件可以是数据库，反之亦然（例如，HDF5容器，SQLite等）。数据库模型是一种数据模型，它定义数据库的逻辑结构，并从根本上确定数据的存储，组织和操作方式。

MPDS使用了几种数据库模型，特别是关系和语义图模型。此外，面向文档的模型在准备数据时用作中间步骤。在面向文档的模型中，每个PAULING FILE条目都被视为文档。这些文档可能具有不同数量的属性，并且都存储在单个巨型数据库表中。所有文档属性都是可搜索的。该模型使用PostgreSQL DBMS实现。非常简单，该模型允许独特的灵活性（单表，任意数据查询）。这种灵活性的成本是非常低的访问速度。不适合生产，非常适合开发。关系模型稳健而成熟，是MPDS生产系统的核心。除了面向中间文档的模型之外，PostgreSQL DBMS还支持它。关系模型中的标准数据操作语言称为结构化查询语言（SQL），并基于关系代数。最终，面向文档的模型中的单个表被采用，精炼，并分成许多相互关联的简单表格。这个过程称为归一化（Teorey et al。2005）。这样可以实现更高的访问速度。然而，精炼的表结构与所选择的使用场景紧密相关，并且不提供额外的灵活性。

关系模型的缺点是数据必须遵守定义为数据库模式的严格规则。通常，这些规则并不意味着在应用程序业务逻辑之外的数据库级别存在其他术语。但是，专家系统和人工智能应用程序必须在开放的世界中行动，根据新收集的信息做出推论并确定新的事实。他们的数据库必须能够包含新术语并动态地遵循新逻辑，因此不适合传统的关系范式。通常，此类数据库采用所谓的语义Web方法（DuCharme 2013）。这种模型中的数据表示为图形。这样的图由“subject-property-object”形式的一组语句组成，称为三元组。使用语义图数据库模型，可以在知识图之上进行前所未有的灵活和富有表现力的查询。MPDS目前仅间接使用语义图，与关系模型互补; 然而，计划更广泛地采用这种模式。作为语义图模型的后端，使用了Virtuoso DBMS。

如前所述，材料科学中单一数据交换格式的发展是一项极其复杂的任务。如今，当特定的信息容器伴随着规则定义，即自动验证工具时，计算机科学提出了一种非常方便的范例。遵循这个想法有各种各样的例子：关系数据库的SQL和模式，语义图的A-Box和T-Box，机器可读数据传输和标记的XML和XSD，人类可读数据的JSON和模式在这方面，应该提到在材料科学中开发共同交换格式的两个成功成就：晶体信息文件（Hall等1991）和材料和物理信息文件（Michel和Meredig 2016））。它们都得到了MPDS的支持，部分得到了其他提到的材料信息基础设施的支持。

晶体学信息文件（CIF）由国际晶体学联合会（IUCR）于1990年代建立。CIF基于名为STAR（自定义文本存档和检索）的文本容器，其中例如作为X射线衍射或理论建模的结果获得的物理属性由标准标签标记。标准标签确定单位单元的参数，对称性，原子位置，相关科学出版物元数据等。这些标签在外部CIF字典中定义（参见XML文档的XSD模式），因此可以验证针对CIF字典的CIF文件，甚至可以推断出可用的新物理属性。不同之处在于CIF格式允许任意标签。根据IUCR，它们被CIF解析器忽略，但后来可以成为标准CIF字典的一部分。此外，CIF格式支持关系数据模型，因此可以通过其标识符来引用晶体结构中的特定原子。缺点是缺乏方便的多级层次结构支持，因此STAR容器承认XML。CIF格式用于MPDS的在线晶体结构可视化，并且不使用其他专有或自制格式。可视化只需要Web浏览器，不需要插件，小程序或其他软件。通常，结构渲染在GPU（即图形卡）上完成。但是，如果GPU过时或不可用，则在CPU（中央处理器）上完成渲染。在这种情况下，渲染质量降低。在线提供的代码总大小仅为150 Kb，加载代码后，无需进一步的Internet连接。MPDS上的CIF可视化基于开源技术。

JSON格式更简单，更灵活，更宽松。从历史上看，软件开发比狭窄用途的CIF更常见。由Citrine创业公司推出的材料和物理信息文件（MIF）的容器建立在JSON之上，充分利用了它的优势。JSON还提供了自己的模式方法，用于验证，文档和交互控制，即给定应用程序所需的JSON数据的合同，以及如何修改数据。所有类型的PAULING FILE数据的JSON模式（包括MIF规范）都是在2016年开发的，现在已公开发布。JSON格式（包括MIF）是上述材料基础架构中最广泛采用的格式。

3.4计算机辅助数据分析
毫无疑问，绘图和可视化对数据分析非常有用。MPDS的一个有趣的绘图功能是术语的语义图。通常，所有MPDS数据都是人类积累的材料科学结构知识的巨大语义图。不幸的是，没有人可以整体观察这个图。在线交互式可视化尝试以非常简化的形式仅显示与特定用户的输入相关的巨大MPDS语义图的微小部分。幸运的是，现代语义技术（DuCharme 2013）能够同时理解一个巨大的MPDS图。这是未来MPDS平台开发的计划方向。另一种可视化，即在MPDS上在线显示的动态渲染的相图（C-条目），是完全数字化的，以编程方式绘制的图。渲染引擎适用于所有现代Web浏览器，不需要插件，并且基于开源Web技术。相图中的相位与x  =  x（t），y  =  y（t），0 <= t  <= 1 的形式的参数方程相关联，其中x代表组成，y代表温度。

最近，MPDS推出了其机器学习预测的第一个版本。为了演示使用MPDS API的材料数据挖掘的一些实际使用场景，选择了相对简单但功能强大的预测机器学习算法 - 决策树回归。决策树是统计模型，其描述从关于某个项目（例如，晶体结构）的观察到关于项目的目标值的结论（例如，相应的物理属性）的数据。MPDS数据包含具有相应物理属性的晶体结构，因此在该数据集上训练模型是可行的。选择以下物理性质：等温体积模量，形成焓，恒压下的热容和熔融温度。通过对树木的替换和投票重复重新采样训练数据来构建多个决策树，从而产生共识预测。这种算法被称为“随机森林 “回归者。“随机森林”是一种统计估计器，它适合数据集的各种子样本上的许多分类决策树，并使用平均来提高预测准确性（Breiman 2001）。它目前使用的最先进的开源实现非常高效，并且需要几秒钟才能在平均台式PC上训练MPDS数据模型（McKinney 2010）。

3.5以数据为中心的观测
继Dmitri Mendeleev和Lothar Meyer之后，他们观察了19世纪60年代化学元素属性的周期性模式，人们可能会尝试在已知材料的整个范围内找到类似的模式，并向假设的周期性材料表迈出一步。基于整个PAULING FILE数据库的观察，制定了12个经验原则（Villars 1994 ; Villars等人2008 ; Villars和Iwata 2013）。这些原则可以称为自然的基石，所有这些原理都可以用现代电子结构理论来解释。它们定义了（a）化合物的形成，（b）结构原型中化学元素的排序，以及（c））元素周期表中化学元素的位置与结构原型中的占据位置之间的联系。

首先，化合物的形成：原子尺寸，电化学，价电子和内聚能量因子控制化合物的形成。例如，可以观察到所有化学元素组合中约30％在二元，三元和四元体系内不形成化合物。

其次，化学元素数量与原子环境类型的相关性：原子环境的最大多样性是在二元和三元无机化合物中实现的。四元，五元和其他高级化合物强烈偏好具有低配位数的原子环境类型。人们也可以将这种观察称为随着元素数量的增加而几何多样性的惊人减少。

第三，活性浓度范围：在一定的相当有限（活性）浓度范围内，二元，三元和四元无机系统存在系统性的二氢化物无机相。

第四，化学计量比条件：对于绝大多数的daltonide化合物，存在高度优选的化学计量比。

第五，化合物的简单性：绝大多数无机化合物平均每个晶胞仅有10个原子，因此在晶体结构中仅显示出三种或更少的原子环境类型。人们可能会注意到，大自然确实更喜欢简单。

第六，复合对称性：10％的空间群覆盖了近70％的无机化合物。最常见的11个空间群是12,62,63,139,166,191,194,216,221,225和227.如图所示，高对称性是优选的。

第七，原子环境安排：大约100种可能的原子环境类型中的18种是非常优选的，并且发现90％的所有PAULING FILE化合物。特别地，最常见类型的多面体是四面体，八面体，立方体，三顶三角棱柱，四顶三角棱柱，二十面体，立方八面体，双顶五角锥和反立方八面体。

第八，化学元素排序趋势：只有大约30个结构原型有超过1000个代表，1000个人口最多的原型及其代表覆盖了PAULING FILE的大部分晶体结构。

第九，结构原型与周期系统的相关性：绝大多数晶体结构在周期系统中化学元素的位置与其Wyckoff位置占据之间表现出非常严格的规律性。这是在1000个人口最多的原型的例子中得到证实的。

第十，结构与稳定性的联系：原子尺寸，电化学，价电子和原子序数因子决定了金属间化合物的晶体结构。同样，这适用于二进制，三进制和四进制系统。可以揭示例如前者与非前者系统，等化学计量结构稳定性图谱以及同一原型内的二元化合物之间的完全固溶度的清晰图案。

第十一，广义原子环境类型稳定性：使用周期数（来自Lothar Meyer的周期表），可以将不同的原子环境类型细分为不同的稳定域。发现周期数大于54的化学元素控制原子环境类型，与它们是作为中心原子还是配位原子无关。因此，在可能的和不可能的原子环境类型之间存在明显的分离。有趣的是，与二进制和三元相比，第四纪原子环境类型的多样性大大减少。

第十二，完全固溶体稳定性：原子尺寸，电化学和价电子因子控制固溶度。例如，在三元系统中，其中两个二元边界系统具有相同的结构原型，可以预测是否形成完整或有限的固溶体。类似地，对于给定的化学元素，可以预测有限的或扩展的固体溶解度。

可以看出，单相无机固体的所有固有物理性质与其晶体结构密切相关，强调了晶体结构分类的重要性。上述12条原则只能通过检查大量经过严格评估的实验确定的数据来发现。它们最终导致限制，这是开发实用且值得信赖的计算材料设计方法的要求。

3.6申请
已发布的实验无机固体与高通量DFT计算之间的可靠联系打开了数据库驱动，数据密集型研究和新材料发现的新视角。目前，中国和瑞士的两个高吞吐量计算计划开始采用带有PAULING FILE数据的MPDS API。

现今大多数的实际材料科学问题（清洁能源，储能，超导等）都与大量的复杂知识有关，即使不是天才人类，也不是人工智能代理人。随着神经科学和医学的进步，揭示了人类大脑的本质，新的强大的计算机科学技术应运而生，后来又在材料科学中得到应用。但是，在封闭（专有）环境中开发复杂工具极其困难甚至效率低下。这就是为什么开放性和自由交换思想的学术传统在开源策略中实现，展示了经济上有利的利他主义。

4总结
PaulING FILE是一项独特的，可能是最古老的组织材料数据的工作。从材料科学的基础开始，例如物理属性的分类，结构原型的概念，阶段，相图等，它涵盖了现代数据密集型科学，采用新颖的数据存储和分析技术并提供高质量的材料见解，在MPDS在线平台中实施。这一点非常重要，而在今天的大数据挑战之前，材料科学的高效快速知识交流受阻。还注意到其他努力及其对材料信息学的贡献。强调了培育当今材料信息学新兴生态系统的重要性。它基于JSON，MIF和CIF的标准，支持数据库，虚拟实验室，与在线REST API相连，这些API往往成为软件基础设施的主干。展示了高度抽象级别的PAULING FILE的概念UML建模。三种类型的PAULING FILE科学数据高度相互关联，因此，在MPDS在线平台内是不可分割的。丰富的数据结构假设存储方法的组合，以便利用它们的优势并同时减轻缺陷。探索大量不同的材料数据，可以提出无机物质的整体观点。再加上高通量的ab initio模拟，它可以为材料基因组的发现提供关键，为整套材料发挥周期表的作用。
