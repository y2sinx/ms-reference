抽象
优异的能量和功率密度，低毒性和延长的保质期促使锂离子电池作为电子和汽车工业中的能量存储装置的普及。然而，下一代锂离子电池将需要更高的能量密度，以满足对更长电池寿命的不断增长的需求。由于其极高的理论比容量（约为传统阳极材料的十倍）和低电化学还原电位（相对于H / H +为-3.04 V）参考电极），锂金属作为下一代锂离子电池的负极材料是极具吸引力的候选材料。然而，诸如枝晶生长的挑战已经阻止了锂金属电极的广泛商业使用，所述枝晶生长可能导致短路或由于生长的电隔离导致的容量损失。成功的商业化将需要稳定锂沉积。设计实现这一目标的策略需要了解控制电化学沉积和枝晶传播的基本机制，并且跨越多个长度尺度。在实验观察的基础上，已经开发了几种数学模型来评估各种物理现象（如电化学反应，扩散，迁移，机械应力和应变，和锂沉积过程中的表面张力。本章概述了锂沉积和枝晶生长的这些方法，总结了他们的发现，并讨论了枝晶建模的其余问题和未来方向。

1简介
为了发展可持续经济并尽量减少全球变暖和解决未来的能源问题，正在重点强调可再生能源的实施。能量存储装置对于实现这一现实至关重要。间歇性可再生能源需要高能量密度装置，用于存储在低需求时产生的多余能量，并在高峰消耗时提供储存的能量。特别需要适当的存储装置来为具有非化石能源的汽车提供动力。

科学界已经了解了电化学电池，其中能量可以以化学能的形式存储并作为电能提取（Blomgren 2017），持续近一个半世纪。然而，随着可充电铅酸电池的出现，它们在商用设备中的使用仅在大约50年前开始流行（Rahn 2013 ; Blomgren 2017）。在二十世纪的最后一个季度，镍镉和镍氢电池进入市场。这些产品比铅酸化学品具有更高的能量密度（Rahn 2013）它们较低的重量和毒性使它们成为手持电子设备的良好选择 - 第一代移动电话和笔记本电脑使用这些镍氢电池。在上个世纪末，索尼通过基于锂离子化学的第一种电池的商业化（Blomgren 2017）彻底改变了能量存储行业。它不仅含有显着更高的能量和功率密度，而且还表现出比铅酸和镍金属氢化物基化学品更高的效率（Rahn 2013）。更长的保质期，更高的多次循环容量保持率以及可忽略的记忆效应使锂离子化学成为电子行业中非常优越的选择。然而，在汽车工业中，商业上可获得的锂离子电池在能量和功率密度方面远远落后于汽油发动机（Gallagher等人，2014）。对于汽车工业的成功电气化，需要对现今的锂离子电池进行重大改进（Gallagher等人，2014）。

1.1锂离子电池技术的简史和未来发展方向
索尼公司于1991年首次将锂离子电池商业化，使用碳基阳极，锂钴氧化物阴极和有机液体电解质溶剂（Blomgren 2017）。阴极和阳极材料都通过插入机制储存锂，并且离子在充电和放电过程中在两个电极之间穿梭。碳被用作阳极材料，因为它具有低的电化学势（相对于Li + / Li）和储存锂而不形成枝晶的能力。在第一个商业上成功的版本中，锂离子化学能够证明能量密度为200 Wh / l（Blomgren 2017）。这足以为电子设备供电，但不足以运行电动汽车。

为了降低与钴基阴极材料相关的成本和毒性，已经研究了几种其他过渡金属作为阴极材料。例如，富镍阴极因其高容量而被广泛使用，并且还研究了富锰阴极的高稳定性。目前，镍，锰和钴的某些组合（Ni x Mn y Co （1-x-y））被用作锂离子电池阴极，以结合它们的高容量，增强的电化学稳定性和大的电子传导性的固有特性，分别（Myung et al.2017）。这些先进的阴极材料能够显示400 Wh / l的能量密度，足以满足短程电动汽车的需求（Blomgren 2017 ; Myung等2017）。

为了克服与当今电动车辆相关的范围焦虑，必须提高现有锂离子电池的能量密度。它可以通过几种方式实现，其中一些方法如下：
（一个）
增加细胞电位的操作窗口。这可以通过将电池充电到更高的电压并从阴极提取一些过量的锂来实现（Myung等人，2017）。这具有通过从晶格位置除去氧来破坏阴极材料的晶体结构的缺点。这种不可逆转的变化只会在第一个周期产生高容量。

 
（b）中
减小阳极电势也可导致运行的（Xu等较大的电势窗2014）。只有锂金属本身表现出比石墨更低的潜力。因此，从常规石墨到锂金属的阳极材料的转变对于增加当今电动车辆的能量密度是至关重要的。

 
（C）
增加电极中使用的阴极和阳极活性材料的比容量。由于其较高的比容量，已经彻底研究了富镍NMC和NCA型阴极材料（Nitta等人2015 ; Myung等人2017）。从阳极的角度来看，通过用锂或硅/锡（Si / Sn）基活性材料代替石墨可以提高比容量（Nitta和Yushin 2014）。

 
（d）
采用“超越锂离子”化学物质，如锂 - 硫 - 和/或锂 - 空气电池（Gallagher等，2014）。这些下一代技术使用锂金属作为阳极材料和一些非常规阴极材料。例如，锂 - 硫电池使用相变硫作为阴极，在放电过程中理论上可以消耗16个电子。锂氧气/空气电池显示阴极的理论重量几乎为零，使其具有与汽油几乎相当的过高能量密度。

 
（e）中
用固态陶瓷或聚合物基电解质代替液体电解质（Kerman等，2017）。由于与非液体电解质相关的更高安全性，它们可能允许去除液体电解质所需的某些额外的安全装置。这个额外的空间可用于插入额外的电池并增加电池组的能量密度。固态电解质还可以帮助加速锂金属阳极的实施。

 
1.2锂金属作为负极
大多数上述提高能量密度的技术都涉及锂金属作为阳极，因为它具有相对于氢参比电极-3.04 V的极低电化学电位，理论比容量为3870 mAh / g，这几乎是一个订单比传统的石墨基阳极（375 mAh / g）更高（Xu et al。，2014 ; Lin et al。，2017）。值得注意的是，锂金属阳极在上世纪下半叶被积极研究用于锂离子电池，但由于几个问题没有看到商业上的成功（Blomgren 2017 ; Cheng等2017）：
（一个）
由于与阳极表面反应和形成固体电解质界面而导致电解质损失，这导致显着的容量衰减

 
（b）中
形成不参与电化学反应的死锂，如果不使用过量的锂，则可能导致容量褪色

 
（C）
树突状突起的生长，其可穿透分离器并使细胞短路

 
（d）
由于内部短路和随后的电池爆炸导致的热失控

 
锂沉积过程中树枝状突起的成核和传播与所有这些问题相关。已经进行了大量研究以理解和预防这些现象。例如，证明在电解质中使用少量氢氟酸作为添加剂导致在金属阳极顶部形成机械刚性的氟化锂层，这可以稳定金属 - 电解质界面几十个周期（Takehara 1997）。后来假设在电极-SEI界面处的不均匀锂沉积和随后的固体电解质中间相层的破裂导致锂枝晶的成核（Aurbach 2000））。与隔板一起添加坚硬的保护层可以防止树枝状突起的生长。这种保护层的一些实例是硫和氮共掺杂石墨烯（Shin等人2015）和沉积在锂顶部的氧化铝（Al 2 O 3）（Peng等人，2016）。覆盖该锂表面与聚合物网络或碳纳米球可以有助于最小化分别不均匀锂析出或SEI形成，（Zheng等人2014 ; Peng等人2016）。为了防止锂枝晶的生长，还有一些研究努力改变电解质的传输性质（电导率，扩散率和迁移数）。例如，在聚合物电解质中添加增塑剂以改善室温附近的电导率以稳定锂沉积（Khurana等人，2014）。在液体电解质中还添加了纳米颗粒以固定阴离子并增加锂迁移数，这有助于在阳极附近保持较高的阴离子浓度（Lu等人，2012））。然而，据作者所知，尚未开发出确凿的产品或策略，可以确保锂的稳定沉积达数千次循环，且容量衰减最小。

在早期的努力中，下一代聚合物或固态陶瓷电解质并不容易获得，但现在使得实际的锂金属阳极看起来更容易实现。然而，实验室规模的初步实验研究仍显示通过这些非液体电解质形成锂枝晶（Harry等人2016 ; Kerman等人2017）。更好地设计这些电解质可能能够稳定锂沉积过程。

此外，已经观察到不同的枝晶形态，根据电解质的类型，操作条件和应用的过电位而变化（Harry等人2016 ; Cheng等人2017）：
（一个）
在低电流密度下形成苔藓沉积物。这通常与枝晶底部的锂沉积有关。如果锂沉积过程受电荷转移控制，预计会形成苔藓样沉积物（Bai等人，2016）。

 
（b）中
在沉积期间以较高速率发展的针状突起通常与突起尖端处的沉积相关联。在传质有限的锂沉积过程中，观察到针状枝晶形成（Bai等人，2016）。

 
（C）
还观察到树状树枝状生长，并且通常归因于表面不均匀性的存在。

 
（d）
在机械应力影响下沉积过程中也报道了球状树突状突起（Harry等人，2016）。

在所有不同的可能微观结构中，针状枝晶具有最大的穿透隔膜，与阴极接触并使细胞短路的潜力（Li等人，2014 ; Bai等人，2016）。因此，即使枝晶生长是不可避免的，重要的是控制在沉积过程中产生的锂微结构，并且可能仅在其仅形成苔藓沉积物的条件下操作电池。为了制定控制锂沉积过程的策略，重要的是要了解整体锂树枝状核化和生长机制。

 
1.3锂枝晶核化和生长的数学模型
在过去的几十年中，对锂的电化学沉积和随后的枝晶生长的广泛研究已经能够建立各种操作条件和枝晶微观结构之间的相关性（Cheng等人2017 ; Lin等人2017）。基于对电化学还原过程的热力学和动力学理解，可以开发能够捕获在金属锂沉积期间观察到的实验特征的数学模型（Chen等人，2015 ; Wood等人，2016）。虽然成核和生长机制可能根据电极和电解质的性质而变化，但锂金属的沉积机制一般可分为两部分：
（一个）
锂的成核

 
（b）中
树突状突起的生长

 
在本书的章节中，将讨论用于预测锂离子成核的不同数学模型。2。在Sects中将讨论用于捕获与液体和非液体（聚合物和陶瓷）电解质接触的树枝状突起的生长的不同数学模型。分别为3和4。一些未解决的问题可以帮助更好地理解锂阳极顶部的金属沉积过程，将在Sect。5。最后，关于树突的成核和传播建模的未来展望将在Sect。6。

2枝晶成核模型
经典成核理论（CNT）已广泛用于捕获假定无缺陷的表面顶部的锂成核过程（Paunovic和Schlesinger 2006）。已经开发了一种替代的瞬时或渐进成核理论来描述表面缺陷为锂枝晶提供成核位点的情况（Paunovic和Schlesinger 2006 ; Stark等人2013）。

请注意，这些理论是为液体电解质系统开发的。在固体电解质存在下，成核应受到使附近电解质或集电器变形所需的额外机械功的影响。在假设集电器弹性变形的情况下，Motoyama等人。发现沉积的过电位几乎与集电器的杨氏模量线性增加（Motoyama等人，2015）。

2.1经典成核理论
在CNT下，成核仅在其降低系统的总吉布斯自由能时才进行（Mullin 2001）。对总自由能的贡献包括电解质置换，锂沉积，以及表面和电解质与新沉积的锂之间的新界面的形成。

基于这些考虑，伊利和加西亚确定（伊利和加西亚2013）高于该半球形锂存款变得热力学稳定的和低于该细胞核应自发溶解（Schmickler和桑托斯临界半径2010）。然而，沉积过程还以更大的过电位（Mullin 2001 ; Paunovic和Schlesinger 2006）的形式消耗额外的能量，以克服与离子锂（在溶液中）转化为原子锂（在基板上）相关的动力学障碍，沿着与弯曲表面的情况下（门罗和Newman表面张力2003 ;伊利和加西亚2013）。使用改进的Butler-Volmer方程结合表面曲率（Monroe和Newman 2003）对反应电流的影响，Ely和García确定了动能临界半径，在此处非负极过电位和表面曲率的影响是平衡的。这个临界半径成反比的超电势（伊利和加西亚2013），并且通常比临界热力学半径大。稳定增长预期用于锂原子核比动能临界半径大，虽然那些比热力学临界半径大可以与相邻的核聚结并变得稳定（伊利和加西亚2013）。实验证实，锂核的尺寸随着过电位的增加而降低（Sano等，2014 ; Pei等，2017）。

2.2锂的瞬时和渐进成核
当存在表面缺陷时，锂最初沉积在这些缺陷部位并且可能由于金属的进一步沉积或通过与其他附近的核的聚结而生长（Paunovic和Schlesinger 2006 ; Stark等人2013）。假设缺陷位点上的核形成速率在可用位点的数量上是线性的，快速动力学（“瞬时成核”）极限表明几乎所有预先存在的活性核在金属沉积开始的同时开始生长（Paunovic和Schlesinger 2006） ; Schmickler和Santos 2010 ; Stark等人2013）。另一方面，慢动力学（“渐进成核”）限制对应于细胞核数密度随时间的逐渐，近似线性增加（Stark等人，2013）。这些数字密度，以及关于核形状的假设，可用于估计（Paunovic和Schlesinger 2006 ; Schmickler和Santos 2010 ; Stark等人2013）电流密度和核大小在核被隔离的早期时间，以及之后的邻近核已经合并。

3枝晶生长模型：液体电解质
使用液体电解质溶液在电化学电池中的现有锂沉积物的树枝状生长已经使用各种方法建模，解决了从原子水平到电化学电池的现象。本节将从最小的长度尺度开始，然后通过连续更大的长度尺度来研究这些技术。

3.1原子尺度建模
Ozhabes，Gunceler和Arias（Ozhabes等人2015年）构建了与电解质溶液接触的结晶表面的密度泛函理论（DFT）模型，其由连续近似和计算的表面能和用于一系列SEI材料的锂表面扩散能垒代表。发现具有小阴离子，特别是卤化物的化合物的空间排列允许较低的能量势垒，并且对于大多数化合物，较大的能量势垒与较大的表面能相关（并且因此树枝状物增加表面积，因此具有更大的热力学稳定性）。此外，他们报告说，细胞短路的时间越长，扩散障碍越大，预计会阻止原子迅速扩散，防止枝晶生长。

3.2分子尺度建模
分子尺度模型适用于模拟沉积少量材料的情况。它们已被用于探索通过使用脉冲电流充电降低枝晶生长速率的可能性。

Mayers和同事创建了（Mayers等人，2012年）3D半蒙特卡罗（MC）模拟在半球形电极尖端周围的半球形SEI区域中的枝晶生长。允许在模拟域内扩散的阳离子以与自由能考虑因素确定的概率在与电极尖端或先前沉积的金属原子碰撞时沉积。较低的反应概率允许阳离子平均通过电极表面上生长的涂层更深地扩散，增加沉积密度。类似地，短充电脉冲与长休息时间相结合促进了更密集的枝晶生长。

Aryanfar和同事构建了（Aryanfar等人，2014）类似的2D MC树枝状生长模型，但是通过扩散和迁移的组合确定了溶液中锂离子的传输。在平面电极上的枝晶生长的脉冲电流模拟表明，适当长度的脉冲并且以适当的时间间隔分开可以降低枝晶生长速率。

尽管未建模锂析出，Li和同事构造（Li等人。2017年）的电解质溶液的三维分子动力学（MD）模拟锂（LiTFSI溶解在碳酸丙烯酯），观察不同物种间的相互作用。他们发现脉冲电场的应用影响了阴离子对锂离子的协调作用，并且通过降低配位来改善锂离子的扩散性。与在恒定电流条件下循环的那些相比，用脉冲电流循环的实验电池显示出更稳定的行为和更均匀的沉积。

3.3连续体建模
连续体尺度的模拟通常涉及具有光滑表面的更大，生长更快的树枝状晶体。这种尺度的模型通常被表述为描述感兴趣的局部现象的偏微分方程，尽管Yamaki等人。提出（Yamaki等，1998）基于大量实验观察的定性锂枝晶生长机制，包括以下步骤：由于表面不规则性和保护性表面膜的不均匀离子导电性，锂沉积不均匀。由此产生的机械应力驱动表面内的锂原子的传输，其通过膜中的断裂而形成晶须。最后，当晶须覆盖变得足够致密时，额外的锂优先沉积在晶须上的缺陷和不规则处。

刘和鲁模拟（Liu and Lu 2017）涂有生长的SEI层的二维枝晶的生长。SEI厚度通过电化学反应和下面的表面应变而改变，并反映在用于确定反应速率的表面过电势的电阻贡献中。电极反应另外受与表面能和局部曲率相关的流体静压应力项的影响。由表面曲率驱动的枝晶尖端处的相对高的反应速率最初导致局部SEI由于下面的锂区域的膨胀而变薄。具有较低电阻的变薄的SEI区域反过来有利于在枝晶尖端处的进一步反应。在枝晶的基础上，压应力增加了SEI厚度，这减缓了局部锂沉积并相对于周围材料产生了凹陷。相同的机制导致在其他平坦电极表面中SEI缺陷位置处的枝晶形成。

Chazalviel观察到（Chazalviel 1990），在稳态对称单元的一维模型中应用稀溶液理论和泊松方程，但不坚持电中性。）大部分细胞的准中性条件，但阴极附近的电位急剧下降，阳离子浓度低，阴离子浓度很低。与相应的实验不同，施加的电位不影响电流密度，因为电流几乎完全由扩散确定。这种差异归因于枝晶生长，其基于动态模拟，其速率预计与阴极附近的阴离子迁移速度相匹配。这个概念通过模拟2D控制体积进一步探索，具有周期性边界条件，随着薄的，直的金属枝晶的尖端以稳定的速度生长而移动。在临界间隔距离以下，枝晶之间的溶液由于枝晶生长而耗尽了离子，它以与阴离子速度匹配的预期速率进行。在临界间隔距离以上，有理由认为生长速率比阴离子速度快，但树枝状晶体之间离子的存在会导致分支生长，从而减少朝向临界距离的分离距离。

3.4相场建模
此前调查的连续体枝晶生长的动态模型涉及精确表面位置的计算。然而，对于不规则界面，模拟通常在计算上是昂贵的（Provatas和Elder 2010 ; Acharya 2016）。相场建模可以是一种较便宜的替代方案（Chen 2002）。相位场模型不是跟踪接口，而是计算“顺序参数”以表示相位标识，通常一个相位被赋值为0而另一个相位在两相系统中被赋值为1。具有非整数阶参数值的区域可以被解释为具有有限厚度的界面区域。为模拟电化学系统，相场模型通常基于根据场参数编写的现象学自由能函数（Ferguson和Bazant 2012）），例如，浓度，电位。对于界面移动问题，Cahn-Hillard方程（对于表示守恒量的变量）和Allen-Cahn方程（对于非保守变量）用于表示场参数值与估计的总自由能之间的关系。通过降低系统的总自由能随时间推移来模拟界面运动。

相场建模在电化学系统中的应用是一个正在进行的研究领域。Guyer等人。（2004a）是最早通过Allen-Cahn动力学模拟电化学界面演化的尝试之一，其中他们研究了界面上的电荷分离，并与经典的Gouy-Chapman-Stern模型进行了比较，以证明其有效性。在随后的论文中，Guyer等人。（2004b）进一步探讨了相场模型对电沉积过程的动力学行为。作为典型的固 - 液界面问题，锂枝晶生长也已经使用相场法广泛建模。同样，Cogswell（2015通过调整Marcus微观电荷转移理论来研究枝晶生长模式，以获得与法拉第反应动力学，分形生长维度和锌电极上的曲率半径的定量一致性。Cogswell还指出，小的交换电流密度可以减少枝晶的生长。为了研究锂离子电池中电极 - 电解质界面的演变，Liang等人。（2012）提出了一种相场模型，其中他们通过假设界面速度与自由能衰减成比例来简化Allen-Cahn方程。按照同样的方法，梁等人。（梁和陈2014年然后研究了锂沉积过程中锂沉积物的形态，发现锂沉积物的生长具有很强的方向性趋势，这可以通过大浓度梯度的存在大大加速。加西亚等人。（Ely等人2014 ; Jana等人2015）开发了一个相场模型来描述锂电沉积在石墨电极表面上的生长动力学和分布，并发现在孤立的枝晶顶端的电沉积速率更高而不是周围的电极。陈等人。（2015年）还提出了一种新的相场模型来研究电沉积过程中表面的枝晶生长，其中他们利用每种物质的电化学势制定了用于锂沉积的Butler-Volmer动力学。Bazant等人也开发了一种非常相似的通用相场建模方法。（Ferguson和Bazant 2012 ; Bazant 2013 ; Smith和Bazant 2017），其中电化学过程（包括广义Butler-Volmer动力学）以及广义化学动力学被建模以提供磷酸铁锂相变行为的定量描述。

相场方法可用于轻松地模拟复杂的界面形状和拓扑变化。它特别适用于研究电化学系统中相扩散，形态变化或扩散场的分裂和合并（Chen 2002 ; Acharya 2016））。它可以将各种物理效应结合到控制方程中，这些方程可以通过简单的数值方法求解，而无需复杂的界面处理。但是，应该认识到相场法也有缺点。尽管相场模型提供了与其他建模方法更深层次的基本热力学连接，但将这些技术应用于电化学系统往往涉及非常坚硬的高阶偏微分方程，这些方程需要较小的数值时间步长，这使得模拟实际时间间隔变得不切实际（ Provatas和Elder 2010）。现在正在开发能够解决相场模型中固有的薄界面同时捕获实际时间尺度上的微结构演变的多尺度数值方法（Provatas和Elder 2010），这有可能克服电化学系统建模的一些当前挑战。

4机械应力对枝晶生长的影响：非液体电解质
液体电解质溶液不对树枝状晶体的形成和传播提供机械抗性。这些可导致形成电隔离的死锂或通过内部短路电池而导致灾难性故障。

已经提出使用非液体电解质通过促进锂金属电极上的均匀沉积来减少枝晶生长（Monroe和Newman，2005）并消除与易燃液体电解质泄漏相关的火灾危险（Cheng等人，2017）。从微结构的角度来看，非液体电解质可分为两类：（a）结构均匀的和（b）结构不均匀的电解质。大多数基于聚（环氧乙烷）（PEO）或聚（丙烯腈）（PAN）的电解质属于第一类，而固态电解质，例如基于陶瓷的Li 7 La 3 Zr 2 O 12（LLZO）和李10 GeP 2小号12（LGPS），它具有晶粒内部/晶界的微观结构，属于第二类。具有纳米级微观结构特征的嵌段共聚物属于第一类，因为这些异质性不影响枝晶生长过程（Inceoglu等人，2014）。然而，在陶瓷固态电解质中发现的微米级晶粒内部/晶界微观结构异质性可通过电极 - 电解质界面处的电流聚焦影响锂沉积过程（Kerman等人，2017）。因此，结构均匀和非均相电解质中的枝晶生长机制可能显着不同并且需要单独的建模方法。

4.1结构均匀的电解质
非液体电解质以压缩应力的形式为树枝状突起的生长提供机械抗性。此外，这种压缩可以改变它们的形状，从针状（在液体电解质下）到球状（通常在均相聚合物电解质中观察到）。

压缩应力的大小取决于电解质的弹性模量。将电解质模量增加一个数量级可以显着减缓树突状突起的生长速度（Harry等人，2016）。图1显示，在90°C时弹性模量约为10 MPa的硬质嵌段共聚物电解质（SEO，由聚苯乙烯加强的PEO组成）降低了锂突起相对于PEO基电解质的生长速率，其具有接近的弹性模量值高温下1 MPa。将树枝状生长的模型中的机械应力结合到非液体电解质中显然是重要的。

之前在本小节中讨论过的所有分析都使用了小变形力学。哈利等人。收集了生长到固体电解质区域的锂枝晶的静态X射线显微断层图像的时间序列，并使用大变形力学计算了电解质区域中的机械应力（Harry等人，2016）。在突出尖端附近发现了压应力，而在基部发现了拉应力。如图3e所示，压缩力降低了反应电流密度的大小，而拉伸应力使其增加。因此，聚合物电解质应该抑制峰值处的生长并增强基部的生长（Harry等人，2016），与球状树突的形成一致（如图2所示）。1b）。

4.1.2锂金属电极的弹塑性应力演变
在弹性变形的假设下，含有硬电解质的细胞可能经历非常高的应力（在GPa范围内）（Monroe和Newman，2005 ; Barai等，2017a）。然而，锂金属或固体电解质不能维持这种程度的应力。延性材料塑性变形，脆性电解质分别经历超过其屈服和极限强度的断裂。因此，对锂金属和电解质中观察到的弹塑性变形和破裂进行建模以正确估计作用于电极 - 电解质界面的应力是非常重要的。

其他研究小组也研究了锂金属弹性塑性变形对树枝状突起生长的影响。巴特勒-Volmer方程通过配合表面的曲率，并且包括塑性变形的模型修改（Ferrese和Newman 2014 ;亚娜和加西亚2017）。枝晶生长过程被分为两个类别（亚娜和加西亚2017）：（ⅰ）尖端受控生长由浓度过电压限制和（ii）碱驱动的增长管辖发生由于锂的塑性变形。得出的结论是大流体静压应力和最小的偏应力是理想的用于抑制枝晶生长（亚娜和加西亚2017）。包括硬质陶瓷隔板（弹性模量在几十吉帕斯卡范围内）的模拟显示金属锂的严重塑性变形和成功抑制树枝状突起（Ferrese等人2012 ; Ferrese和Newman 2014）。发现锂金属的塑性变形在抑制树枝状突起的生长方面比弹性变形或压力对反应动力学的影响更为重要（Ferrese和Newman，2014）。

这里描述的所有三个研究都包含了小应变和线性应变 - 位移关系，这些研究结合了锂金属的塑性变形。然而，显着的枝晶生长将导致大的变形，违反通常的小位移假设。因此，有必要开发消除这种约束的模拟。

4.2结构不均匀的电解质
还有其他固体电解质显示出微观结构的异质性。这些非均相固体电解质的锂离子传输，可塑性和断裂的机制可以显着变化（Stephan和Nahm 2006 ; Kerman等人2017），导致具有不同生长机制和形态的树突突起。具有不均匀微结构的电解质可大致分为两个子类别：（a）显示出明显的晶粒内部/晶界微观结构的固态电解质和（b）其中聚合物基质用陶瓷填料增强的复合电解质。在这些结构上不均匀的电解质中几乎没有关于枝晶生长的计算研究，这些电解质是用于锂离子电池的相对新的候选物。

4.2.1单离子导电固态电解质
在均相电解质（液体和聚合物）中，阴离子的扩散有助于锂离子穿过电解质（Stephan和Nahm 2006）。然而，在陶瓷电解质中，电流仅由锂离子携带，因为阴离子作为稳定晶体结构的一部分被固定（Kamaya等人，2011）。单离子传导电解质的一些实例是LLZO和LGPS，它们分别是氧化物基和硫化物基陶瓷。然而，对于化学稳定性，它们通常掺杂有其他元素，例如镓或铝（Rettenwander等人，2016）。陶瓷固态电解质具有极高的弹性模量值，使其适合于防止树枝状突起的生长; LGPS和LLZO的弹性模量值分别约为20GPa和150GPa（Kerman等人，2017）。然而，所有固态电解质在足够的应力下以脆性方式破裂。

在均匀电解质的假设下，机械应力对枝晶生长的影响较早（Monroe和Newman，2004）。修改该模型以研究陶瓷电解质中枝晶生长的影响（Ahmad和Viswanathan 2017）。固态电解质中锂离子的部分摩尔体积显着减小会影响枝晶生长过程。结论是，对于比锂更软的电解质，树枝状突起不应生长（Ahmad和Viswanathan 2017）。由于晶体结构的各向异性，机械刚度张量通常是各向异性的。进一步得出结论，各向异性行为不会影响整体稳定性，而只是简单地改变稳定性极限（Ahmed和Viswanathan 2017）。

基于平衡的计算模型表明，相对于锂在电解质另一侧的沉积，破坏通过固态电解质的路径的细长树枝状突起的传播在能量上是有利的（Porz等人，2017），其也得到了支持。通过实验观察。脆性断裂现象和低断裂能量被认为是导致陶瓷单离子导体中锂枝晶容易生长的主要因素。而且，固态电解质缺乏柔韧性导致电极 - 电解质界面处的分层，这可导致高过电位。

4.2.2复合聚合物 - 陶瓷电解质
标准聚合物电解质（例如PEO）既不具有足够高的导电率也不具有弹性模量值，用于在室温下用锂金属阳极进行持续操作。然而，它们的无定形区域比其结晶区域更容易传导锂离子，因此将操作温度提高到熔化温度以上改善了锂离子传导。或者，有时会引入陶瓷填料（如微米级或纳米级Al 2 O 3或SiO 2）来生产室温下具有更高导电率和弹性模量的聚合物 - 陶瓷复合材料，这会抑制聚合物链的结晶（Stephan）和Nahm 2006）。

一些纳米尺寸的填料颗粒，例如由SiO 2组成的那些，形成三维网络，其中阴离子被固定（Schaefer等人，2013），增加了电解质的转移数。Tikekar及其同事构建了一个包含电解质区域的电池的稳态模型，其机械应力被忽略，但具有固定阴离子的特定部分（Tikekar等人，2014））。在高电流密度下沉积期间，他们观察到阴极附近的陡峭电位变化和离子耗尽，这被认为是驱动枝晶生长。增加固定阴离子的比例降低了该区域中的电场强度，特别是在高电流密度下，这有望促进稳定的电极表面。

5个问题
尽管在过去几十年中已经进行了大量的建模工作以了解锂枝晶的成核和生长背后的机制，但目前的理解还远远不能令人满意。没有开发出能够完全稳定锂沉积过程的材料或策略的事实反映了这种显着的知识差距。科学家们已经能够开发出可以解释锂沉积过程中观察到的电压和电流分布的连续模型（Barton和Bockris 1962 ; Diggle等人1969 ; Motoyama等人2015 ; Wood等人2016 ; Pei等人。2017年）。基于相场和蒙特卡罗技术的其他计算模型也能够预测在不同操作条件下枝晶形态的演变（Chen等人，2015）。不幸的是，这些没有导致锂在金属电极上的稳定沉积。重要的是要问正确的问题以取得合理的技术进步。这里提供了一些科学家应该关注的将锂金属阳极技术提升到更高水平的问题。

重要的是要认识到锂金属阳极的成功商业实施需要稳定沉积数千次充电 - 放电循环。在单次充电步骤中，树突可能不会长得足够大以使电池短路。通常，树突状突起穿过分离器并使细胞短路需要多次充电 - 放电循环（Stone等人，2012 ; Sharafi等人，2016）。因此，单次充放电循环的稳定沉积可能无法保证无枝晶操作。然而，大多数计算研究试图在单一沉积过程中模拟枝晶的生长（Monroe和Newman，2003 ; Ely等，2014））。即使可以稳定锂在平坦表面上的沉积，锂在放电期间的不均匀溶解也可能导致在随后的充电步骤中树枝状晶体生长的位点的发展（Wood等人，2016）。重点不仅应集中在锂的首次沉积，还应集中在后期的溶解和电镀步骤上。

一旦锂金属与有机电解质接触，固 - 电解质 - 界面（SEI）就在锂 - 电解质界面处发展。SEI层组成尚不确切（Cheng等人，2016）。与金属电极相邻的具有可变锂离子扩散性和导电性的不同区域的存在可导致锂-SEI界面处的不均匀沉积。已经提出争论，SEI层的破裂允许从这些区域开始增强锂和树枝状突起的沉积（Aurbach 2000））。另一方面，还有人认为SEI层内存在机械硬化的氟化锂可使沉积过程稳定数十个循环。因此，金属 - 电解质中间相对锂沉积的影响尚未完全阐明。SEI层如何改变锂溶解过程也不清楚。存在关于锂离子和溶剂分子如何在与锂电极反应之前迁移和扩散通过SEI层的问题。应开发适当的计算模型以协调SEI层对锂枝晶生长机制的影响。

从实验的角度来看，已经观察到在小电流密度下锂沉积期间形成苔藓样结构，而在大电流操作期间针状树枝状晶体发展（Bai等人，2016）。通过计算方法获得的树枝状微结构并不总是与实验观察结果良好相关。例如，相场建模技术在低过电位时产生线性树枝状结构，在较高过电位时产生树状结构（Chen等人，2015）。这似乎与实验观察无关。此外，其他计算方法已经观察到即使在阴极过电位下也会形成死锂（Jana等人，2015年），这在现实情况下是不太可能的。因此，未来的模型应侧重于预测与实验观察更密切相关的情景。尚未彻底研究的另一个重要特征是发生沉积的锂表面的晶体取向。应进行适当的基于第一原理的分析，以确定不同锂晶体取向的表面能。应适当地阐明在不同类型的锂晶体表面上沉积期间树枝状晶体形成的倾向。

有限的室温导电性一直是妨碍商业设备中非液体电解质广泛实施的主要问题（Stephan和Nahm 2006）。最近的发展产生了陶瓷固态电解质，其室温电导率甚至大于液体电导率（Kato等人，2016）。然而，没有计算模型可以分析陶瓷电解质中锂枝晶的生长。一些唯象模型已被开发，以阐明陶瓷电解质内的锂枝晶生长的倾向较高（的Porz等人2017年）。还应彻底研究这些树枝状突起在阳极过电位下的溶解。关于具有陶瓷电解质的锂阳极的循环性能仍然存在重大问题，例如，晶界和表面缺陷对整个锂沉积 - 溶解过程的影响（Kerman等人，2017）。还应通过计算模型进一步研究聚合物基固体电解质稳定锂电化学沉积的能力。如果理解成核和生长的物理现象，成功的树突防止机制可以来自计算模型。

6展望
鉴于锂金属阳极的优点，与其低电化学势和高重量容量相关，下一代锂离子电池应该旨在使用这些阳极材料来增加能量密度。锂金属的最大问题以枝晶生长，死锂的形成和固体电解质中间相层的电解质的连续分解的形式出现。计算方法应该在设计解决这些问题的策略方面发挥主导作用。由于计算模型的成功取决于所采用理论的详细能力，因此对物理化学现象的良好理解对于开发准确的理论框架至关重要。考虑到锂 - 电解质界面可能发生多种物理现象，如质量传递，电荷平衡，传热和机械平衡，应开发多物理场计算模型，以成功预测锂枝晶生长过程。此外，在树枝状突起的生长期间观察到的物理现象跨越多个长度尺度。例如，锂的成核发生在纳米级。阴极和阳极通常在电池中间隔几微米放置。为了引起短路，枝晶同样必须生长几微米才能到达阴极。为了捕获在多个长度尺度上发生的所有现象，应该开发多尺度计算模型。应该开发机械平衡，多物理场计算模型，以成功预测锂枝晶生长过程。此外，在树枝状突起的生长期间观察到的物理现象跨越多个长度尺度。例如，锂的成核发生在纳米级。阴极和阳极通常在电池中间隔几微米放置。为了引起短路，枝晶同样必须生长几微米才能到达阴极。为了捕获在多个长度尺度上发生的所有现象，应该开发多尺度计算模型。应该开发机械平衡，多物理场计算模型，以成功预测锂枝晶生长过程。此外，在树枝状突起的生长期间观察到的物理现象跨越多个长度尺度。例如，锂的成核发生在纳米级。阴极和阳极通常在电池中间隔几微米放置。为了引起短路，枝晶同样必须生长几微米才能到达阴极。为了捕获在多个长度尺度上发生的所有现象，应该开发多尺度计算模型。树枝状突起生长过程中观察到的物理现象跨越多个长度尺度。例如，锂的成核发生在纳米级。阴极和阳极通常在电池中间隔几微米放置。为了引起短路，枝晶同样必须生长几微米才能到达阴极。为了捕获在多个长度尺度上发生的所有现象，应该开发多尺度计算模型。树枝状突起生长过程中观察到的物理现象跨越多个长度尺度。例如，锂的成核发生在纳米级。阴极和阳极通常在电池中间隔几微米放置。为了引起短路，枝晶同样必须生长几微米才能到达阴极。为了捕获在多个长度尺度上发生的所有现象，应该开发多尺度计算模型。

DFT模型已被广泛开发并用于分别理解锂和电解质的行为以及它们的界面。但是，这些模型只能在原子尺度上预测局部特征。使用DFT和ab-initio分子动力学（AIMD）模拟技术进行预测时，形成锂枝晶核所需的原子数也过大。因此，有必要使用经典分子动力学（CMD）或基于蒙特卡罗（MC）的计算程序来理解在纳米尺度上发生的物理现象，其中使用AIMD程序测量性质。

另一方面，存在发展良好的计算技术，用于对在连续体（或微米尺寸）长度尺度下观察到的各种物理现象进行建模。通常这些连续体模型具有需要通过实验或较低长度尺度模拟确定的参数。

大多数散装材料在连续体水平上看似均匀。然而，在微米或亚微米长度尺度上也可能存在不均匀性，也称为中尺度，其在连续体水平中不可见。基于DFT / AIMD的较低长度尺度模拟也不受这些不均匀性的影响，因为它们具有局部性质并且提供完美晶体的参数。实际上，这些微观结构的异质性导致更高的应力集中或电流聚焦，这导致树突的开始和随后的细胞失效。通过对中尺度现象的建模，成功地桥接适用于不同长度尺度的这些计算技术对于开发精确模型至关重要。在中尺度上观察到的特征的适当预测可以将计算模型的重要性提升到下一个层次。具有高计算能力的更快的超级计算机还可以通过对在中尺度处观察到的物理现象的建模来帮助弥合原子和连续长度尺度之间的差距。
