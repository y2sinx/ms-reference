
抽象
近年来,第一性原理建模技术取得了长足的进步.这使研究人员能够估计材料的各种特性,并从微观角度提供对物理过程的宝贵见解,而这是无法通过实验直接评估的.随着计算能力的不断提高,预计第一原理方法将在材料设计中扮演更重要的角色.本章旨在作为面向电池的计算手册,供那些可能对第一原理计算不熟悉的普通读者使用.具体而言,本章将从热力学和动力学两个方面介绍在电池相关研究中广泛使用的公认的从头建模方法.将要讨论的热力学方法包括平衡电压和电压曲线的计算,电子结构和稳定性分析.动力学方法将涵盖离子扩散研究的常用方法:过渡态理论和微动弹性带方法以及从头算分子动力学.接下来是结论和展望.

1引言
全球变暖加上能源需求已使可再生能源技术(太阳能,地热,潮汐,风能等)的发展成为全球关注的焦点.但是,大多数可再生能源技术都具有高度的可变性和供需之间的不匹配.因此,可靠且具有成本效益的电能存储是可再生能源以及无污染运输手段的关键推动因素(Lewis 2007).在众多储能技术中(例如,锌锰电池,镍镉电池,镍氢电池,铅酸电池,碱离子电池,燃料电池,氧化还原液流电池等),可充电碱由于其高的比能量,高的工作电压,宽的工作温度和长的使用寿命,离子电池已成为电网和车辆应用中最有吸引力的候选者之一,并成为移动消费应用中的主导者(Tarascon和Armand,2001年). .

任何电池中包含的能量都是电压乘以充电容量的积分.为了在碱离子电池中获得高能量和高功率密度以延长循环寿命,电极应具有高的比容量(每单位质量或体积存储的电荷),高的工作电压,合理的电子和离子电导率以及良好的相和电化学稳定性.碱性离子电池中的一对电极的合适电解质选择是另一个重要方面.为了防止电解质的还原和氧化,阳极的费米能级的能量应低于电解质的最低未占用分子轨道(LUMO),而阴极的费米能级的能量应高于最高的费米能级.电解质(Roy和Srivastava)的占据分子轨道(HOMO)2015).在固态电解质的情况下,LUMO(HOMO)被称为导带最小值(价带最大值).另外,期望电解质,特别是对于固态电解质,具有良好的界面稳定性,比电极更严格的易离子传导性和优异的电子绝缘性.

由于近年来计算能力的重大进步,计算机模拟已成为用于了解和设计材料的实验的重要补充工具包.第一原理密度泛函理论(DFT)(Hohenberg和Kohn 1964),尤其是Kohn和Sham(1965)提出的框架通过求解具有最小逼近的单粒子Kohn-Sham方程来评估材料的各种属性,在这种情况下特别成功.第一性原理计算使研究人员能够从原子水平评估材料的关键特性,从而为无法通过实验直接评估的基本物理过程提供宝贵的见解.此外,以计算为指导的新型材料发现已在能源应用方面取得了巨大进展,包括新型碱离子电池材料(Deng等人2016 ; Urban等人2016).).不仅通过计算机功能的增强,而且通过自动化高通量第一原理筛选的发展,这些筛选都可以从成千上万的初始候选人中快速识别出最有前途的候选人.

在以下各节中,我们将简要介绍用于阐明电池材料热力学和动力学的公认的第一性原理建模技术.

2.3电子结构
碱离子电池中不同成分的电子能级的相对能量决定了存储能量的限制.图3a示意了平衡电压与电极,电解质的电子结构之间的关系.在开路条件下,阳极和阴极之间费米能级的能量差会建立电子电化学势.假设没有电极/电解质反应,该电子电位差等于阳极和阴极之间的碱化学电位差,以开路电压(V OC)进行测量).一旦施加了充电电流,电子便进一步从阴极的价带经电路泵出到阳极的导带,分别导致阴极和阳极的费米能级降档和升档.随着其他离子(例如TM离子)的氧化变化,碱金属离子从阴极提取到阳极,以平衡电荷.在理想情况下,整个过程在放电过程中是完全可逆的.氧化还原电位(V 氧化还原在充电/放电过程中测得的电导率受电解质"窗口"的限制.如果阳极和阴极的电位超出电解质的LUMO和HOMO,则电解质会在阳极上还原并在阴极上氧化,从而形成固体电解质中间相(SEI).理想的SEI是离子导体和电子绝缘体,可防止电解质进一步分解,从而增加稳定的电压窗口.

2.4稳定性分析
2.4.1 0 K相位稳定性
前面小节中提到的伪二进制稳定性图本质上适用于电极材料.在一般情况下,任何给定化合物(例如,固体电解质)的热力学稳定性可以通过评估其分解能来分析,该分解能参考该组成空间中稳定化合物的线性组合.这可以通过使用该组成空间中所有已知化合物的DFT能量构造凸包来实现.例如,在橄榄石LiFePO 4的情况下,相关的稳定性分析要求所有现有的Li x Fe y P z O u化合物(x,y,z和u是非负整数),包括在Li-Fe-PO四元组成空间中的自身.在这种情况下的稳定性度量称为船体上方的能量(E 船体)(Sun等人,2016年; Ong等人,2008年),该标准通过测量化合物在凸体船体之上的能量来衡量化合物的稳定性.它始终是非负的,对于稳定的化合物,它等于零.E 壳越高,化合物趋向于越不稳定.随着大开源材料的出现数据库如材料项目(Jain等人.2013),一种常见的方法是利用这些数据库中预先计算的DFT能量来获得主要关注化合物以外的化合物,这将显着减少所涉及的计算工作,尤其是对于高维组成空间.在高通量第一原理筛选研究中,E 壳通常被用作相稳定性和可合成性的有效描述子,以快速排除在环境温度下不太可能稳定的结构(高E 壳).

2.4.2表面稳定性
实际的表面配置很少与理想的晶格平面相吻合,后者是晶体结构的几何抽象.实际上,稳定的表面可以包含从环境中吸附的带电物质,电荷密度的重排以及阶梯状或扭结的表面轮廓.这使得确定表面热力学非常困难.因此,第一性原理计算可能是确定刻面特定表面能的有用工具.计算表面能的标准模型是二维表面平板,其体积化学计量取向为暴露感兴趣的小平面,并在平板顶部添加了足够大的真空层,以确保平板与其周期之间的相互作用最小沿表面法线的图像.因此,表面板的k点网格生成为k 1  ×  k 2  ×1,可确保没有能量从表面向外散布.

生成表面平板的标准程序包括基础转换,原子坐标缩放,终止选择和极性检查.通常通过常规单位晶胞使用Miller索引符号定义表面.但是,第一性原理计算要求基础平面与构造的平板表面重合.必须对常规单位晶格的基础进行转换,以使新基础的(001)平面平行于目标Miller平面.第三个向量应与底面正交,且长度较短.然后基于新的基础重新定义原子坐标.对于给定的表面方向,由于不同的原子构型,通常会存在许多终止层,可以通过将基数放置在沿着第三矢量的不同位置进行选择.请注意,在重复单元中具有垂直偶极矩的带电表面(根据Tasker的表面分类,类型为III)会在整体中产生极化电场,从而导致无限的周期性电池能量(Tasker1979).例如,NiO离子晶体的(111)面具有交替的Ni 2+和O 2-离子层,形成垂直于表面的永久偶极矩.通过将一个表面上一半的带电物种移动到另一侧以消除2D偶极子来进行原子重建是一种稳定III型表面的常用方法(所谓的Tasker III至Tasker II重建)(Bruno2013).

2.4.3电解质/电极界面稳定性
对电解质/电极界面的理解已经成为对电化学能量存储系统的性能改善最重要的近期焦点.对于带有液态有机电解质的碱离子电池,由于电解质分解而形成的SEI对循环性能和安全性至关重要.对于所有带有硫化物型电解质的固态碱离子电池,通常在阴极/电解质界面周围观察到速率确定过程.预测电极/电解质界面在大环境条件下的稳定性是电池研究的重大挑战之一.使用基于DFT的方法的第一原理计算已经发展到调查两种晶体固体(春山等人的最稳定界面匹配.2014).按照传统方式,研究由两种化合物形成的界面的过程包括三个主要步骤:(i)确定每种材料的最低能量表面,如表面稳定性计算中所述;(ii)使用相当大的超级电池(真空层约为10Å)修补两个表面平板,以确保尽可能小的晶格参数失配;(iii)列举界面超级电池的所有可能配置,包括一个表面板相对于另一个表面板的滑动,并进行DFT松弛.然后可以在不同的热力学平衡条件下进一步研究最稳定界面中的碱金属离子位点和碱金属离子转移性质.

由于使用了超级单元,因此显式接口计算通常需要非常高的计算成本.为了研究电解质/电极的界面稳定性,关键步骤是确定在界面处可能形成的反应产物.在固体电解质/电极界面的情况下,有两种基于两种不同热力学近似值的有效分析方法(Zhu等人2015a ; Richards等人2016 ; Tang等人2018).
快速离子扩散极限:在这种近似中,假设固体电解质中的扩散离子是主要的移动物质.在这种情况下,可以将接口建模为相对于移动物种的开放系统.相关的热力学势是大势.以锂离子的固体电解质作为一个具体的例子中,锂巨势表示为φ  =  ë  -  μ 李 Ñ 栗,其中ê,Ñ 栗,和μ 栗分别是内部能量,开放系统中的Li原子数和Li化学势.当金属Li被作为阳极,锂固体电解质/负极界面被建模为在高固体电解质μ 李听,说:μ 李 0,与Li的固体电解质/阴极界面在低温建模为固体电解质μ 利听,说:μ 李 0  -  V,其中V是电池电压,μ 李 0作为金属Li的化学势.

3种动力学方法
有两种广泛使用的第一性方法可以评估电池材料中碱离子扩散的动力学:基于过渡态理论的方法,例如微动弹性带方法和从头算分子动力学模拟.接下来,我们将介绍离子扩散研究的每种方法的关键概念及其局限性.

3.1.1 NEB方法的局限性
在碱离子电池的研究中,NEB方法已主要应用于块状系统(原始电极和固体电解质).在非晶体系统的情况下,例如电解质/电极界面,由于较低的对称性需要考虑大量的不同的离子迁移路径,因此不可避免地需要更高的计算工作量.与从头算分子动力学模拟相比,NEB方法还有两个局限性(见第3.2节).

首先,必须将迁移路径的起点和终点指定为NEB计算的输入.对于具有复杂扩散特性的碱性电池材料,例如,当离子运动高度协调时(在超离子导体电解质中很常见),相关迁移事件和局部环境的识别变得非常困难.在这种情况下,从头算分子动力学(请参见第3.2节)将成为更合适的技术.因此,NEB计算主要是在稀载流子极限条件下进行的,例如,在原本完全碱化的结构中出现单个碱金属离子空位迁移,或在原本没有碱金属离子的骨架结构中出现单个碱金属离子迁移.

其次,NEB计算实质上输出了0 K下的迁移势垒,即,在计算中未考虑温度影响.此外,为了估计与迁移路径相关的准确跳变率,并因此估计扩散率,可将等式中的前置因子确定为.必须明确计算10.但是,通常假定其值为10 11到10 13 s -1之间的某个值.

3.2从头算分子动力学模拟
分子动力学(MD)通过积分牛顿运动方程来模拟有限温度下的原子运动.这种方法的关键输入是原子间相互作用的描述,而主要输出是原子运动的轨迹.根据如何获得原子间相互作用,MD模拟可分为两大类:经典MD和从头算MD(AIMD)模拟.在经典的MD模拟中,原子间的相互作用用解析函数表达,参数使用实验数据或第一性原理数据拟合,而在AIMD模拟中,原子间的相互作用是通过近似求解薛定er方程直接获得的.

尽管经典的MD模拟在计算上比从头计算便宜,但它确实存在严重的缺点,这些缺点是由于使用了经验潜力而引起的,例如,跨不同化学物质的可传递性差,以及难以处理具有简单功能形式的复杂原子间相互作用.另一方面,AIMD模拟通常可以应用于广泛的化学空间,并且原子间相互作用的准确性基本上受从头算的基本限制.

如今,AIMD模型有两种主要变体:Born-Oppenheimer(BO)变体和Car-Parrinello(CP)变体(Car和Parrinello 1985; Hutter 2012).对于电池相关的研究,更普遍采用前者.这些AIMD技术背后的理论超出了本章的范围,并且被许多优秀的教科书和评论广泛涵盖(Hutter 2012; Marx and Parrinello 1996).AIMD方法的主要缺点是与传统的MD以及NEB方法相比,其计算成本明显更高(第3.1节)).这对AIMD模拟中可访问的像元大小和时间尺度构成了限制.但是,这些可以通过计算机能力的提高和所研究问题的性质来缓解.

AIMD模拟的关键输出是离子运动的轨迹,从中可以评估材料的扩散特性,例如离子扩散率和电导率.AIMD技术在碱离子电池材料的研究中取得了巨大的成功,特别是在理解和发现新型碱离子超离子导体方面.在下文中,我们将首先概述完善的分析方法以及如何将其应用于了解电池材料中的离子扩散.我们还将讨论有关参数选择,AIMD在电解质/电极界面中的应用以及该技术的局限性的实际考虑.我们应该强调,这些分析方法与MD模拟的类型(即经典和从头算)无关,

3.2.4 AIMD仿真中的实际考虑
由于AIMD技术的计算成本很高,因此在选择相关的仿真参数时需要特别注意,以使其在计算成本和实际结果之间达到良好的平衡.通常,这些参数的选择主要取决于(i)研究中的特定系统,例如快导体或慢导体,以及(ii)感兴趣的问题的类型,例如对具有相似化学/拓扑结构的系统的比较评估,或准确的扩散率计算.对于任何感兴趣的AIMD仿真,关键参数包括仿真像元大小,总仿真时间,时间步长,仿真温度,整体等等.有关参数选择的全面摘要可以在感兴趣的读者的文献中找到(Shin和Saal 2018).

3.2.5电解质/电极界面建模
AIMD仿真还可用于显式建模电解质/电极界面,这在充放电过程中对电池性能起着至关重要的作用.为了获得更实际的结果,要求此类系统的模拟超级单元尺寸要比散装材料的超级单元大得多,尤其是沿着垂直于界面的晶格方向.原则上,可以通过AIMD仿真来估计界面处的扩散特性.但是,由于所需的超级单元大且仿真时间长,因此需要大量的计算工作.AIMD模拟通常用于模拟碱化/脱碱过程中的结构演变(Leung和Budzien,2010年))以及电解质(固体或液体)与电极材料之间的化学反应,可以从中识别出可能的界面反应产物(Tang等人,2018)).在那些仿真中,可以利用具有更短时间尺度的更小的超级单元尺寸来加速AIMD仿真并以定性的方式获得现实的结果.近年来,已经开发了旨在从AIMD仿真中有效确定界面产品的分析工具.其中包括径向分布函数(RDF)分析.在该方法中,将处于晶相的所有可能产物的RDF与界面的RDF相比较,以确定可能的反应产物.我们应该注意,所有这些现有技术都基于各种近似值,它们是进一步改进所必需的.

3.2.6 AIMD仿真的局限性
与NEB方法类似,AIMD仿真主要应用于批量扩散性能的体系,特别是碱离子超离子导体.对于碱离子超离子导体,通常可以在可访问的AIMD时间范围内估计关键量,例如离子扩散率和电导率,该时间范围通常长达数百皮秒.对于诸如碱过渡金属氧化物阴极的块状电极材料,过渡金属的存在通常需要自旋极化计算.而且,电极材料中的离子传导通常比超离子导体中慢得多,这意味着可能需要更长的模拟时间范围才能正确收敛离子传导性.因此,使用AIMD模拟对这些材料进行扩散研究通常需要高得多的计算成本,3.1).尽管如此,仍然可以针对其他目的对电极材料执行AIMD仿真,例如,以估计碱离子的扩散途径,位置占用率等.

作为AIMD模拟的最后说明,材料在室温下的离子扩散率(和活化能)基本上是通过线性外推法从高温下获得的离子外推法获得的.这意味着环境温度和高温之间的扩散机制必须保持不变.对于不满足这种条件的材料,应改用NEB方法.

4结论与展望
本章从热力学和动力学两个方面概述了与电池相关的研究中使用的成熟的第一性原理(电极,电解质及其界面).其中包括碱离子电池的许多关键特性:平衡电压,相稳定性,电子结构和离子扩散特性.第一性原理计算需要与实验设计和数据进行闭环交互.模拟可以洞悉新材料的开发,并有助于解释实验观察结果.实验结果反过来证明了理论模型和提出的机制的有效性.尽管近年来第一原理方法取得了显著进步,

首先,应注意,此处介绍的大多数第一原理方法,例如电压计算和NEB方法,已主要应用于完美的结晶固体.但是,从实验的角度来看,合成材料中总是存在固有的缺陷,例如位错,堆垛层错和微应变.这些固有缺陷会对电池材料的电化学性能产生重大影响.原则上,可以通过超级电池的构造来模拟缺陷(例如,掺杂剂),并且超级电池的尺寸主要取决于缺陷浓度.鉴于在引入缺陷时通常存在不止一种可能的配置,因此需要执行有效的结构枚举,随后是针对每个可能的不同配置的DFT计算.因此,最终导致能量最低的掺杂结构.该过程通常在计算上非常昂贵,尤其是在稀缺陷浓度方面.

固体电解质/电极界面是非晶第一性原理建模的另一个重要示例,这在当前阶段极具挑战性.为了获得真实的结果,所需的超级单元必须足够大,尤其是沿界面法线的长度,该长度旨在最大程度地减少周期图像之间的人为交互.原则上,AIMD方法适用于有限温度下的界面扩散研究.然而,所需的长仿真时间尺度和长度尺度表明,这种仿真在计算上几乎是不可行的.

对于阴极材料,应该注意的是,本章主要介绍了插层电极材料,其中通过在稳定的主体骨架中进行可逆插层,可以进行能量存储.尽管最近已经发现并证实了阴离子氧化还原,但是这种嵌入型电极材料的容量主要限于每个过渡金属一个电子.因此提出电极材料中的转化和合金反应以获得更高的容量.在转换和合金型电极中,更多的TM氧化态是可行的,这允许更多的电子在电化学循环中可逆地变化.与插层反应类似,转化/合金反应的热力学可以通过计算所有涉及化合物的总能量从第一性原理计算中进行研究.确实存在额外的复杂性,因为转化/合金反应会产生具有不同化学组成的多个相或区域,其中界面和纳米尺寸效应(即,要求进行非晶体建模)通常主导着材料的热力学性质.

解决上述局限性和挑战的方法是多种多样的.首先,计算能力将继续快速增长,这可以进行更长的仿真,更大的系统尺寸和化学复杂性.其次,旨在跨越多个时间和长度尺度的多尺度方法有望在未来得到进一步改进.这使我们能够以更高的计算精度探究更现实的模型.第三,自动化软件工具的不断开发也将促进电池材料的研究,并大大减少研究人员监视仿真的时间.第四,通过标准化的第一性原理高通量计算来建立开放获取的材料科学数据库也可以促进第一性原理建模.更重要的是,
