## 抽象
我们简要概述了表面催化建模的最新进展.在多尺度建模方法之后建立的对所需精度水平的不同看法导致出现了高通量计算筛选方法.现在,越来越多地使用机器学习技术来挖掘以这种方式创建的大量数据.我们将讨论描述从电子到反应器的催化系统的各种令人兴奋的方法论中的现状和挑战.除了对理解和预测催化活性的传统关注外,我们认为动态催化剂重组,伴随的热量管理以及催化剂寿命的方法是未来十年的重要主题.

## 引言
当《材料建模手册》(Handbook of Material Modeling)的第一版于2005年问世时(Yip 2005),许多章节都将表面催化作为重要的应用领域.但是,与我们现在所处的状态相比,十年后的今天,这项研究更加分散,着眼于各自方法在相应长度和时间范围内的方法.多尺度建模尚处于初期阶段,但随后成为催化建模的主要主题.在自下而上的方法中,电子结构理论家试图将第一性原理技术的预测质量能力扩展到更长的长度和时间范围,从头开始进行热力学和微观动力学模型以捕获反应性环境的影响或相互作用中产生的整体效应.催化循环中的基本过程 在自上而下的方法中,化学工程师的目的是将更详细的化学方法引入反应堆模型中有关宏观传热和传质的传统连续描述中.时至今日,两种方法之间的差距仍未消除,但已大大缩小,形成了一个跨学科的社区,该社区被用来应对采用多种方法的表面催化的多物理挑战.

这种新范式最重要的含义可能是对准确性和所需细节水平的差异化看法.长期以来,始终如一的化学准确性一直被认为是第一性原理计算中预测质量的前提.相反,更粗糙的模型经常是由于人们认为微观细节(及其描述中伴随的错误)将以某种方式平均化而引起的.在多尺度建模方法中建立的敏感性分析越来越多地仔细研究了这些观点,发现实际上通常是一些微观参数严格地控制了目标宏观功能(如活性或选择性)(Campbell 2017 ;Döpking等人2018 ; Medford等人2014); Meskine等.2009年 ; 萨顿等.2016).这为高通量计算筛选方法奠定了基础,该方法专注于这些微观参数(称为描述符),针对各种材料进行计算,从而可以在广阔的材料空间中寻找有希望的催化剂(Andersen等人,2016年,2017a ; Andersson等2006 ; Besenbacher等1998 ; Greeley和Mavrikakis 2004 ; Greeley等2006 ; Linic等2004 ; Loffreda等2009 ; 2014 ; Toulhoat和Raybaud 2003).现在,越来越多地使用机器学习技术来挖掘以这种方式创建的大量数据.这提供了对材料特性与(催化)功能之间相关性的进一步洞察力,并将预测的第一原理转化为改进系统的计算机设计(Jinnouchi和Asahi 2017 ; Li等人2017 ; Ma等人2015 ; Ulissi等人2017a,b).

详细的多尺度建模,高通量筛选和数据科学的新兴方法三角将当代的计算催化研究转化为可能性和令人振奋的发展的强大动力.其中许多是由于迫切需要转向替代能源并通过可持续能源技术来实现其转换而驱动的.除其他外,这极大地促进了电催化建模.应对与电气化的固液界面相关的挑战,将可能继续成为下一个十年的主题.另一个这样的主题将是目前在原子反应方法中可行的结构模型与复杂反应器几何结构中的实际高表面积催化剂之间可行的复杂性差距,并结合对催化剂降解的描述.在下文中,我们将简要调查其中的一些发展和挑战,然后在本节的以下各章中详细讨论.从简短的概述中可以清楚地看到,催化建模的下一个十年将至少与过去一样快.在这方面,令人难以置信的是,想象在未来的第三版《表面催化》中表面催化部分的内容是什么.材料建模手册.

## 表面催化建模方法
第一性原理的电子结构理论以其对计算光谱学的支持和对反应性表面化学的描述固有地专注于催化,因此一直是催化研究中任何建模的自然基础.密度泛函理论(DFT)和它更加复杂和强大的电脑包,比如VASP(Kresse和Furthmüller 1996年a,b)等是目前延长系统的事实标准.对于金属催化剂和在缺乏对系统的任何改进甚至概念性途径,低梯级广义梯度近似(GGA)函仍普遍,可能会是相当长的一段(伯克2012 ; Cohen等人2012 ; Feibelman 2010).因此,关键的高能参数(例如反应壁垒)中随之而来的不确定性将继续成为关键问题,尤其是当考虑到计算出的反应壁垒(及其误差)以指数形式输入动力学速率常数时.对于像大多数氧化物这样的间隙体系,在更多应用研究中,对GGA功能的简单+ U校正仍然是最简单的技术水平(Himmetoglu等,2014; Kulik 2015).在混合,筛选的混合或图解功能中精确交换的混合为这些系统的系统改进提供了一条替代途径,该方法在表面催化研究中的使用也在增加(Goerigk和Grimme 2011; Pacchioni 2008; Ren等人2012).; Schimka等.2010).通过范德华(vdW)校正方案或vdW功能来有效解决色散相互作用的可能性对于所有类型的系统都具有无与伦比的价值,并且无疑也为处理更大的反应中间体打开了大门(Klimeš和Michaelides 2012; Liu)等人,2014; Maurer等人,2016),例如在费-托催化的背景下.固态嵌入方法将量子力学计算集中到一个活动中心周围的有限核心区域,也开始使基于波函数的相关电子结构理论计算得到更广泛的应用,尤其是提供了高度准确的参考值(Boese和Sauer2013 ; Kubas等.2016 ; Libisch等.2014年;Richter等.2013 ; Stodt等.2013).

除了这种并发的多尺度建模技术之外,尤其是现在在表面催化研究中大量采用的分层多尺度建模方法,与十年前的情况相比,已经取得了质的进步.在自下而上的方法中,通过第一原理技术计算的数量进入了更粗粒度的模型,从而可以解决更长的时间尺度和更大的长度尺度.最普遍的方法是从头算热力学,其中使用第一原理计算得出的总能量或振动频率来确定热力学势,如吉布斯自由能(Reuter 2016).以化学储集层的形式表示周围的气体或液体环境,向其或从其提供或获取物质或电荷,这种方法可以(例如)洞悉(近似)反应环境中催化剂表面的表面结构和组成(在非均相催化中表示为表面相图,在电催化中表示为Pourbaix图)和反应条件下的纳米粒子形状(Garcia-Mota等人,2015; Ouyang等人,2013; Wang等人,2013),或者它提供了首先对假定反应路径上的热力学壁垒进行评估,例如,本节中Li和Schneider的章节对此进行了讨论.

对于表面电催化,对这类车型的突破来了,当Nørskov和同事意识到,(否则计算很难确定)溶剂化物种的化学势可以与在标准条件下的气相化学电位(Nørskov等人.2004年,2005年).然后,这种计算性氢电极(CHE)或其他物种的等效参考物可以快速获得对超电势的估计以及引起它们的限速步骤.在原始配方中,未在电子结构水平上处理对表面高能学的溶剂化作用,这尤其导致了pH效应描述的不足.因此溶剂化作用的有效治疗是目前研究的一个重要课题,具有隐溶剂化模型(Andreussi和Marzari 2014 ;马修等人.2014 ;林格等人2016 ; Sinstein等人2017年),静态的冰层(Chan和Nørskov 2015年; Rossmeisl等人;2008年;Skúlason等人.2010年)或完全显式的动力学模拟(Chen等,2013; Mattioli等,2013; Quaranta等,2017; Stercher等,2016),代表了活动的三个主要方面,在Bagger的章节中有更详细的讨论等,由Groß和Oberhofer撰写.

反应壁垒的显式计算是考虑动力学效应的第一步(Sabbe等,2012).在多相催化中,对应计算主要进行了使用过渡态搜索算法等被轻推弹性带(Henkelman等人2000).尽管名义上几乎所有主要计算机程序包都具有"即插即用"功能,但实际上,使用它的每个学生在实践中仍难以解决,收敛性差且计算成本高.对于固液界面的反应,情况甚至更糟,因为原则上甚至需要更严格的动力学或自由能模拟(Chen等,2013;Groß等,2014; Li等,2014).; Mattioli等.2013 ; Pinto等.2014年;Schnur和毛利率2011 ; Quaranta等.2017 ; Stecher等.2016).目前,在这种电催化环境中还出现了进一步的复杂性,因为用于扩展系统的标准第一原理超级电池方法是恒定电荷而不是恒定电势(Bonnet等人,2012; Chan和Nørskov,2015; Filhol和Bocquet,2007; Otani和Sugino ,2006;Bill 等. Rossmeisl等人,2008年,2013年; Surendrala等人,2018年),如Groß在其章中详细讨论的那样.因此,对允许对反应壁垒进行可靠(定量或至少半定量但具有可控制的误差线)计算的数值有效方法的研究是一个热门话题,在未来几年中其重要性可能还会增加.在这方面,在过去十年中,由Nørskov和他的同事在所谓的比例关系的更一般的范围内率先提出的Brønsted-Evans-Polanyi(BEP)关系的量化方法已经在表面催化研究中真正改变了游戏规则(Greeley 2016 ; Nørskov等人.2008年,2014).通过BEP关系,可以从要求不高的热化学计算中得出近似壁垒,而其他比例关系则可以从其构成元素的数据中得出用于更复杂的反应中间体的相应的能级参数,或者可以推断出不同催化剂材料上的能级; 有关更多详细信息,请参见Raman和Vojvodic的章节.

这种(近似)第一原理反应能学在基本过程中的可用性不断提高,极大地推动了微动力学模拟的发展,该仿真评估了这些基本过程在反应网络中的相互作用,并产生了固有的催化活性(以及越来越多的选择性).所施加的反应条件功能(Sabbe等人2012 ;路透2016 ; Sutton和Vlachos 2015).尽管速率方程形式的平均场微动力学模型仍然主导着该领域(Nørskov等人2014 ; van Santen and Neurock 2006),请参见.通过Vlachos和Wittreich章,即占反应中间体在催化剂表面的空间分布更为准确动力学蒙特卡罗(KMC)模拟正在迅速赶上(Stamatakis和Vlachos 2012 ;路透2013,2016).尤其是在利用基于比例关系的大型能量数据库时,两种类型的微动力学模拟都已成功并且越来越多地用于处理更复杂的反应网络,例如,合成燃料.因此,kMC模拟通常仍受到具有非常快速的过程(在金属催化剂上的典型扩散)特征的网络的挑战.这已开始在新兴的加速kMC方法中得到解决(Andersen等人2017b; 查特吉和选民2010 ; Dybeck等.2017).

在微观动力学仿真领域中,一个特别有趣的发展是认识到,通过使用比例缩放和BEP关系,可以显着降低问题的维度.虽然原则上任何给定的微动力学模型都取决于所有基本过程的所有速率常数(也称为反应障碍),但是当利用速率常数的各个组相互缩放时,这些依赖性的复杂性会大大降低.当进一步(部分地)对反应机理和限速步骤进行假设时,可以将微观动力学模型的整体依赖性降至最低(Nørskov等人,2014年)).然后通常可以将催化活性(或选择性)讨论为仅一种或两种基本过程的速率常数的函数,并通过BEP和比例关系来讨论仅取决于一种或两种基本反应中间体(如O或O或)的吸附能的函数. C.催化研究的这一链条极大地增进了我们对过去十年趋势的了解,如今,这种催化研究已被广泛用于通过简单地计算相应的描述符(例如多种催化剂材料的O或C结合能)来"筛选"更好的催化剂.这些能量是否落入一个范围,得到了根据简化microkinetic模型最佳活性(Andersen等.2016,2017A ; Besenbacher等人1998年; Greeley和Mavrikakis 2004;Greeley等.2006年 ; Linic等.2004 ; Loffreda等.2009年 ; 2014年;Toulhoat和Raybaud,2003年).在这种情况下,令人兴奋的新发展是与数据科学的融合,使用机器学习技术来识别更好的描述符或通过回归技术扩展和完善能量数据库(Jinnouchi和Asahi,2017年 ; Li等人,2017年 ; Ma等人,2015年); Quaranta等人,2017; Stecher等人,2016; Ulissi等人,2017a,b).

为了还考虑给定反应器中的宏观热量和质量传输效应,将微动力学模型集成到连续流体动力学模拟中.这更是建立了平均场率方程模型(贾纳德汉南和Deutschmann 2012 ;马斯特里和Cuoci 2013),但最近也被扩展到基于KMC-microkinetic模型(马特拉和路透社2009年,2010 ;马泰拉等人.2014).这种耦合模拟尤其有助于分析来自越来越多的现场和操作人员的数据测量,通常受热量和质量传输的限制,然后无法使用微动力学模型提供的正常内在活性与实验数据相关(Sabbe等人2012 ; Reuter 2016).迄今为止,基于第一原理的微动力学模型(均场或kMC)在反应器级的流体动力学模拟中的集成仅限于简单的模型催化剂,实际上基本上仅限于单晶.取而代之的是通过Hettel,Wörner和Deutschmann在本章中讨论的有效动力学模型来描述更复杂的高表面积实际催化剂中的反应化学.对于这种结构化催化剂,显式或有效的孔隙率处理,湍流的出现或复杂的反应器方法(如流化床)反而成为流体动力学研究本身的难题.

## 个公开挑战
由于上面简要强调了令人印象深刻的方法学进展,因此建模和仿真为从活性位点到反应器的各个层面的表面催化研究做出了重要贡献.实际上,到目前为止,建模和仿真已牢固地集成为几乎所有项目中必不可少的部分,即使在工业环境中也是如此.尽管取得了这些进步,但是实际催化系统的复杂性仍然对建模提出了严峻挑战.合适的模型系统的开发仍然包含基本特征,但复杂度大大降低,这仍然是未来几年的核心问题和任务之一(Sauer和Freund,2015年)).在活性位点的水平上,流行的方法通常集中于晶体纳米粒子各个面上的奇异的高对称位点.这并不能充分地解决在特定化合物的材料,它提供多个这样的位点的复杂性-与现有往往很少洞察这些位点的其中占主导地位的催化活性(Reuter等人2017).此外,点缺陷或扩展缺陷通常可以起关键作用,在半导体材料中,点缺陷或扩展缺陷可以另外以多种电荷状态存在.Operando越来越多的研究报道了活性颗粒的强烈形态变化,这严重质疑简单地基于切割标称本体催化剂材料晶格的结构建立任何类型的活性位点模型的普遍方法(Reuter 2016).有操作的催化剂更动态的性质也越来越多的证据,表明持续的结构性变化,甚至让人联想到近液态(牛顿2008 ;路透社2016 ;施勒格尔2015年).目前,此类动力学无法通过流行的微动力学模型来捕获,特别是不能被通常仍基于晶格的kMC模型捕获(Reuter 2013).因此,没有在相应的建模中"看到"这样的动力学,并不能证明它的存在,而仅仅是我们目前方法论能力的一个缺点.

第二个核心问题是(特别是基于第一原理的学术性)计算催化研究仍主要集中在催化活性作为一种好的催化剂的单一度量.这忽略了实际上催化剂是复杂系统的一部分,如果不考虑整个系统就无法评估催化剂.催化剂需要包含廉价的材料,需要廉价地制造,需要耐受进料中的污染物,需要持续很长时间,并且需要避免产物的精心分离.另外,它必须允许廉价的热量管理.尽管一些计算研究开始考虑材料成本,例如,在使用成本作为附加描述符的高通量筛选中寻找最佳解决方案(Andersson等人,2006年); Nørskov等.2009),这还没有解决用这种材料制造催化剂的实际成本.回想一下,例如钌仍然比铁占主导地位的原因是工业氨合成催化剂是用廉价铁制造铁催化剂的高成本(Ertl 1980).迄今为止,无论是在详细的微动力学模型中,还是在计算筛选中未考虑催化剂和污染物的优化相互作用,计算研究都还没有真正旨在以可预测的质量水平评估物流中痕量杂质的影响.因此,杂质"中毒"并不总是有害的.有时可能会选择性地中毒发生不良反应的表面部位.在部分氧化反应中,人们经常试图使发生完全氧化的表面部位中毒.无论哪种情况,绝对效果都是巨大的.在氨制造中,催化反应器约占工厂成本的1%.主要费用是产生不含CO或硫污染物的清洁氢气.该方法的根本改进将是找到对进料中的杂质免疫的催化剂.即使反应速率小于现有催化剂的反应速率,在实践中仍将使用这种催化剂.关键是高比率是一个很好的标准,但在许多情况下并不是主要因素.关于热管理和催化剂降解的重要性,可以做出类似的例子和论据.在过去的十年中,计算催化建模的主要特征是根据潜在的材料特性建立对催化活性的全面趋势理解.与这种趋势理解相比,例如,我们目前对微观散热通道的见解(Rittmeyer等.即使反应速率小于现有催化剂的反应速率,在实践中仍将使用这种催化剂.关键是高比率是一个很好的标准,但在许多情况下并不是主要因素.关于热管理和催化剂降解的重要性,可以做出类似的例子和论据.在过去的十年中,计算催化建模的主要特征是根据潜在的材料特性建立对催化活性的全面趋势理解.与这种趋势理解相比,例如,我们目前对微观散热通道的见解(Rittmeyer等.即使反应速率小于现有催化剂的反应速率,在实践中仍将使用这种催化剂.关键是高比率是一个很好的标准,但在许多情况下并不是主要因素.关于热管理和催化剂降解的重要性,可以做出类似的例子和论据.在过去的十年中,计算催化建模的主要特征是根据潜在的材料特性建立对催化活性的全面趋势理解.与这种趋势理解相比,例如,我们目前对微观散热通道的见解(Rittmeyer等.关于热管理和催化剂降解的重要性,可以做出类似的例子和论据.在过去的十年中,计算催化建模的主要特征是根据潜在的材料特性建立对催化活性的全面趋势理解.与这种趋势理解相比,例如,我们目前对微观散热通道的见解(Rittmeyer等.关于热管理和催化剂降解的重要性,可以做出类似的例子和论据.在过去的十年中,计算催化建模的主要特征是根据潜在的材料特性建立对催化活性的全面趋势理解.与这种趋势理解相比,例如,我们目前对微观散热通道的见解(Rittmeyer等.2018年),基本上不存在它们在烧结或其他时效过程中的作用.在这方面,催化剂寿命,催化剂对污染物的耐受性以及可能的催化剂动态状态无疑是未来十年的主题.
