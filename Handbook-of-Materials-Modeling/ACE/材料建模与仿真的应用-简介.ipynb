**摘要**:  
材料研究的最终目标是改善其性能和控制过程,例如化学降解或机械故障,以及设计新材料和加工路线.《材料建模手册》的前版和现版旨在阐明建模和虚拟实验如何为这项工作做出贡献,并监督其多年来的发展.特别是,《材料建模手册:应用–当前和新兴材料》(ACE))充分介绍了材料科学和技术领域的最新成就.它涵盖了对系统的研究,例如催化剂,磁性材料或复杂氧化物,传统上一直以建模和模拟为重点,尤其是过去15年左右的一些"发现",例如2D材料或新型光伏或生物启发性的分层系统.可以认为以前的研究是对2005年版《材料建模手册》(HMM)中报告的研究的更新.因此,相比之下,ACE自那时以来,向读者提供了计算材料科学如何发展的观点.另一方面,有关新兴材料的文章表明,计算的作用对于帮助解决(通常是前所未有的)相关科学和技术问题的实验起到了多么关键的作用.在这里,我们简要介绍ACE的15个部分,并强调其中讨论的主题和主要发现的重要性.几章通过报告成就和失败来简要评估所采用的方法.我们将强调从这些结果中学到的东西,并指出ACE与专用于该方法的配套手册(MTM)之间的联系.

## 简介和与MTM的连接
技术的最新进展很大程度上与使用前所未有的复杂性的新材料有关,其理解似乎要求使用高级建模和仿真.的确,在过去的十年左右的时间里,计算材料科学的应用领域已显着扩大,其协议已从使用单一方法更改为使用多种方法以更全面地理解,并且它们的目标已逐渐朝着合理的要求发展.设计.但是,仍有许多挑战尚未解决,从而使虚拟实验的范例成为真实实验的重要伴侣,从而有助于技术和基础理解的进一步发展.

该材料手册:应用程序-当前和新兴材料(ACE)叙述了各种措施,成功和失败的这一努力至今.尽管我们不假装所介绍的场景是完整的,但我们相信它可以使材料科学的各个分支得到广泛的了解,而计算在其中发挥着重要作用.

## ACE部分的简要指南
在这里,我们将简要介绍ACE中涉及的主题,并重点介绍每个主题的相关性以及在各种应用中似乎很常见的方法问题.对于更广泛和具体的讨论,我们请读者阅读各节编辑的概述.特别是,它们提供了对建模和仿真当前状态的全球展望,并阐明了短期和长期挑战.

在分别关注每个部分之前,我们注意到,在每个部分中,都需要特别注意超越对通过应用某些模型和计算得出的现象的理解,并将其转化为改进或创新的建议的需要. .实际上,在过去的10到15年中,这种对计算方法结果进行实际使用的驱动力已经得到了发展,这肯定是由于当前技术中对"更好"的材料的需求日益增长和广泛需求所致.同时,计算科学家越来越意识到真实材料的复杂性.一方面,这意味着需要获得对其行为的全局理解,例如,在确定热力学稳定性时化学和机械性能之间的相互作用,或在实际条件下在给定化合物在运行条件下的性能方面,以及需要学习和利用各种理论和计算工具方面.也可以在各部分中确定有关后者的局限性的共同问题,例如,在基于DFT的计算中缺乏标准近似的准确性,以及需要能够桥接不同长度范围的系统描述的算法,因此例如,要弄清楚原子间过程对中尺度行为的影响.而且,正如我们将在下面看到的那样,机器学习技术已经开始在几个领域中用于寻找新材料.

与MTM类似,ACE的开头是一系列关于我们希望读者特别感兴趣的主题的文章("全会主题",编辑:Andreoni和Yip).如本节概述所示,其中某些章可以轻松地与以下主题节中的多个相连,从而提供了更多信息和新观点.其他一些则涉及其他非常有趣的主题,这些主题不符合表1的特定分类  .

第一个主题部分是"光伏:第一原理建模"(编辑:Bernardi和Grossman),重点是从头开始研究"现代"光伏的方法,即有机-无机卤化物钙钛矿(原型CH 3 NH 3 PbI 3)和超薄6族金属硫属化物,例如WS 2.在过去的十年中,这两类材料已经成为非晶硅的极有希望的替代品(参见例如社论2019)).为了理解它们的特性并帮助设计性能更好的器件,人们面临着许多基本问题,而这些问题不能用当前的第一原理方法全部解决.这些问题在节编辑器的概述中指出.但是,正如Filip,Volonakis和Giustino在本章中针对钙钛矿的案例中充分讨论的那样("混合卤化物钙钛矿:基本理论和材料设计")目前的从头算方法已经成功地预测了(不仅是复制的)几种光电特性,并且从早期开始就一直为实验研究提供宝贵的支持,同时也寻求铅化合物的更稳定和无毒的替代品.例如,我们将读者引向有关无毒双钙钛矿的合成和利用的相关专利,该专利已通过上述计算确定(Snaith等人2018).

下一节名为"磁性,磁性材料和自旋电子学的建模应用"(编辑:Sanvito)与MTM的配套部分(编辑:Sanvito)紧密相连,描述了研究电磁波的最新理论和计算方法.磁性和自旋电子学的不同方面.此处介绍的应用涉及宏观磁化动力学的描述以及原子级电流与自旋之间相互作用的描述.此外,还包括一章,内容涉及使用组合的机器学习策略和来自电子结构计算的高通量数据来搜索新型磁性材料.(有关一般信息,请参见MTM中的 "材料信息学"部分.)

"纳米尺度的低维材料"一节(编辑:Chelikowsky)记录了二维材料,纳米团簇,纳米线和石墨烯纳米带的结构和电子性能计算的成功应用(主要基于DFT).例如,这些计算阐明了有限的尺寸和量子限制之间的相互作用以及它们对掺杂的影响,这对于材料功能化至关重要.它们还使人们能够了解随着系统规模的增大而发生的物理性质的演变,从而预测了本体的某些特征(如能隙)是如何建立的.

"热传输"部分(编辑:Donadio)提供了广泛的观点,说明了为解决现实系统中与热传输有关的多个问题并获得可靠的建模工具和仿真而正在进行的工作.一方面,有几章详细介绍了理论算法和计算方案的最新进展,并指出了方法开发所面临的挑战.另一方面,当前方法在从聚合物到金属合金再到纳米结构材料的各种系统中的应用,阐明了仿真应在不久的将来解决的问题的复杂性.

"用于能源和信息设备的氧化物建模的挑战和机遇"部分(编辑:Franchini和Yildiz)向读者介绍了广阔的研究领域,即电化学材料(主要是复杂的金属氧化物)的研究领域.对(复杂)金属氧化物的极大兴趣源于其作为能量转换和存储设备以及逻辑和存储系统一部分的用途广泛.本节中的文章解决了对于提到的设备的性能至关重要的基本问题.特别地,主要关注点缺陷,表面和界面的性质和作用,它们对离子或电子传输的影响,以及当前用于提供定量信息的计算的局限性.

我们注意到,其中一章着重于钙钛矿氧化物及其在燃料电池中作为电催化剂的用途.特别是,它指出了缺乏催化剂表面知识,而建模和模拟可能与之相关.实际上,电催化是下一节中更广泛讨论的主题之一,该主题专门致力于表面催化.总而言之,我们指出,除了使用本节中重点讨论的传统计算方法外,机器学习策略也已开始用于优化电磁氧化物的特性(例如,见Min等人  2018)).

关于"表面催化"的部分(编辑:路透社和Metiu)也涉及广泛的研究领域.足以认识到催化反应在我们的生活和工业过程中无处不在,并且数十年来一直在为寻找"最佳"催化剂而付出巨大的努力.这里收集的文章集介绍了相关的子主题,并特别说明了在过去的10到15年中如何应用当前的计算方法-从头算到微动力学模型,再到计算流体动力学再到多尺度方法.尤其是,我们清楚地显示了我们在上文提到的计算机模拟设计的建模和仿真范式的变化.实际上,除了特定的化学反应机理以及可能的活性位点和点缺陷的性质外,最近解决的重大问题涉及影响催化剂性能和寿命的众多因素.例如,当前的研究集中在催化剂与其环境的相互作用上,例如异质界面,作为催化活性的促进剂或抑制剂引入反应器的其他化学物质以及溶剂(在电催化的情况下)或浓缩物操作条件下催化剂的动态重组

关于催化过程的大量可用数据要求在催化器设计中使用人工智能程序.除了章节编辑器概述中对机器学习技术的新兴应用提供的一些参考文献外,我们还引导读者进行有趣的讨论,以探讨催化领域机器学习的最新进展以及一些严峻的挑战(Li等.2018).我们还建议进一步阅读有关逆设计催化方法的最新综述(Freeze等人2019).

接下来的两个部分说明了如何将各种长度尺度的最新模型应用于具有复杂结构的"日常"材料,以研究其基本特性.了解机械性能与结构特征之间的联系无疑是这项研究的主要目标.我们注意到,在分子或纳米尺度上的动力学模拟(即模型链的第一步,导致这些系统的更全面表示)依赖于完善的粗粒度方案或简单的原子间势.对于方法论问题和广泛的观点,我们请读者参考MTM,尤其是"软物质/聚合物模拟"部分(编辑:Kremer).

"结构材料的多尺度建模:化学和机械性能"部分(编辑:Buehler和Martin-Martinez)主要关注分层的天然材料,例如丝绸,木材和竹子,以及基础设施复合材料,例如沥青和混凝土.它还包括一章有关二氧化硅气凝胶的章节,即具有异常结构组织的纳米结构."复杂软材料的结构发展和力学建模"部分(编辑:Del Gado和Pellenq)专门介绍了另一类引人入胜的"软材料",尤其是有关软凝胶,水泥和粘土的建模和模拟的报告.我们注意到,以上两个部分也可以作为介绍其中所处理材料的丰富现象学和实验研究成果的简介.

机械行为也是下一节"材料的纳米力学:结构和变形"(编辑:Zhu)的重点.纳米力学涵盖了各种各样的纳米级材料,例如纳米线,纳米管,纳米膜,纳米马达和一般的低维材料.相对于松散材料,尺寸减小对材料的机械性能具有强烈而独特的作用,并且通常导致强度和韧性的显着提高.还出现了导致变形和断裂的不寻常机制.在选定的章节中给出了一些示例.其中讨论的大多数计算研究都基于经典分子动力学.这些模拟提供了对例如 主要缺陷的性质和作用,或导致纳米孪晶材料变形或2D材料的塑性行为的机理.还强调了这种方法的局限性,特别是在仿真的长度和时间尺度方面.为了更全面地了解"纳米力学",我们建议您进一步阅读,例如,《晶体力学建模的最新进展》(编辑:Cai和Ghosh)中的章节.MTM以及最近的评论(Taloni等人2018)描述了不同尺寸尺度下材料的失效行为.我们还指出了与ACE先前有关"纳米级低尺寸材料"(编辑:Chelikowsky)的内容有关的内容,涉及二维结构纳米材料的不同方面,主要是电子特性.

"玻璃科学与技术:预测建模"部分(编辑:Mauro和Onbasli)显示了本构模型是如何普遍用于实际工程目的的,特别是针对玻璃行业的.这种选择肯定是由于功能性眼镜在建筑,运输,汽车或光伏应用等各个领域的普遍存在.确实,建模已被证明对于优化和设计各种眼镜功能特性至关重要.例如,要求制造眼镜以确保独特的稳定属性,例如高光学透明度,高机械强度和高声传输损耗.此外,正如各节编辑提供的广泛概述所指出的那样,

接下来的两个部分,"核材料"(编辑:Marian和Devanathan)和"辐射损伤"(编辑:Short和Nordlund)在很大程度上是相关的,因为尽管它们截然不同,但它们处理的问题相似,需要使用相同的计算方法.前者报道了对核裂变燃料(如二氧化铀)和结构材料(如锆合金)的基本性能的研究,以及中子辐照对核反应堆性能和寿命的有害影响.后者考虑了中子和离子辐射效应以及除核以外的材料.这两部分都对从原子级到连续体级的各种建模策略和模拟进行了全面的描述.但是,可以说,尽管理论上有进步,超过50年的建模,仿真和实验,仍然存在着巨大的挑战.对于进一步的具体讨论,读者应查阅专用于该主题的主要参考著作(Cacuci2010 ; 孔宁斯2012 ; IAEA2018).

我们还指出,本手册未考虑电离辐射对生物物质(例如DNA或蛋白质)引起的损害.

"疾病的多尺度建模"部分(编辑:Dao和Karnidakis)介绍了新兴的人类疾病实验和计算模型领域.它是生物物理和生物医学研究领域,主要是在过去十年中受到三个方向的发展,先进的多尺度建模能力,新的力学生物学实验技术以及在目标专业领域中的建模科学家,实验者和临床医生之间的跨学科合作的推动.本章讨论了在中尺度(基于粒子)和连续体水平上建模的各种计算方法,以及对人类红细胞(与疟疾有关)和镰状细胞性贫血的具体研究.

该ACE该手册集以题为"计算晶体结构:预测和材料发现"的部分(编辑:Wang和Wolverton)作为结尾,提供了另一个有趣的材料开发主题.对于给定化学组成的材料,提出了多种方法用于预测晶体结构的应用.基于对势能面和能量最小化过程的探索,基本算法的开发可以追溯到20多年前.但是,仅在最近十年中才采用了几种新的,更准确的程序,以提高预测复杂化合物的低能结构和发现无法预料的多晶型物的能力.尤其是,我们强调遗传算法和粒子群优化技术的扩展和适应.特别是,我们指出了一种原始方法的问世,该方法结合了从X射线衍射分析(Rietveld精修)和从头算(DFT)计算获得的结果.

## 结论
ACE提供了材料科学几个领域的概述,其中在过去10到15年中广泛应用了不同复杂程度的计算.现在确定了计算材料科学的作用.

但是,很明显,基于物理的建模和仿真需要成为具有指定特性的新的复杂材料或用于材料过程控制和合成的新程序的设计中的重要组成部分,还需要进一步的发展.为了实现这些宏伟目标,学术界与工业界和社会之间必须进行深入交流.

尽管我们很可能忽略了读者感兴趣的某些主题,但我们相信他/她会发现许多非常有趣和及时的文章,尤其是对可预见的未来计算方法应用的指导性观点.确实,在比较了ACE阐明的对更强大,影响深远的算法的需求以及MTM中讨论的正在进行的以及预期的方法开发之后,我们可以通过重复莱布尼兹著名的句子得出结论:"现在怀有未来. "
