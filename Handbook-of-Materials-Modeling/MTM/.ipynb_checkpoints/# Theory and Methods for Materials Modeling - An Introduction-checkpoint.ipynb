# 材料建模与模拟的属性(Attributes)

**材料设计的原理植根于分子结构与物理性质的相关性**. 根据这些相关性,可以构建能够预测微观结构演变的模型. 这种模型使研究人员能够研究材料关键行为的潜在机制,并系统地得出改进的设计.

在材料建模社区的兴起和发展过程中,对这一原则的认识是显而易见的,因为可以遵循美国政府机构和资助组织的一系列报告以及参考出版物(PITAC 2005; SBES 2006; NRC 2008;2008). WTEC 2009 ; MGIGC 2011 ; DOE-BES 2012 ; Konings 2012),以及评论(Yip and Short 2013)). 值得一提的是,在这些讨论中,信息技术对全球社会的影响,以及中尺度在建立多尺度材料模型中的特殊挑战( It is worthwhile to note the perceived global societal impact of information technology in these discussions and the particular challenges of the mesoscale in building multiscale materials models.). 前一个问题引起了材料创新(materials innovation)领域中新兴的人工智能领域,而后一个问题使人们意识到了介观层面的建模和模拟的重要性(DOE-BES 2012). 这是"规模专制"表现的结果(SBES 2006).

材料创新的好处可以用纳米科学和纳米技术的最新进展来说明,纳米科学和纳米技术是可以与计算药物设计相比的全球性事业(worldwide enterprise). 能够描述能够在原子尺度上描述系统的方法(包括量子机制)和能够描述中尺度现象的技术将有望确保材料创新和开发的不同阶段,从设计到合成到测试到性能和寿命评估,可以全部进行模拟,分析和优化.

在广阔的社会影响领域,人们可以期望材料建模和模拟具有巨大的社会效益潜力,因为它们处于计算科学与材料科学与技术的十字路口.

为了使各个社区保持联系,我们一开始就宣布,除非另有说明,否则"材料建模和模拟"和"计算材料科学"一词应视为同义词.

我们可以简短地回顾一下材料建模和模拟的三个属性,这些属性将在开发和应用程序阶段持续存在(SBES 2006).它们共同构成了MTM和ACE的"知识产权胶"的本质.

出色的频宽(bandwidth:一定时间内传送信息量的量度) –材料建模和模拟的概念基础涵盖了物理科学的所有学科,甚至还包括其他学科.它不区分属于物理,化学,工程等领域. 材料建模事业的普遍性在人们考虑了建模模型的实际涉及时就变得显而易见.如果我们将建模定义为概念的"物理化",而将模拟定义为该概念的数字实现,则很明显,建模和模拟是所有涉及通过模拟(或实验)提出假设和检验的科学学科的基础),然后通过理解和后续操作(设计)重复此过程.

经验主义的消除 –多尺度建模和模拟的一个优点是,两者的结果在概念上和操作上都是可量化的. 这里的"可量化性"是指可以以一种可控的方式询问特定的操作并确定其结果,实际上类似于研究"因果关系"的能力. 因此,可以一次探究一个复杂的现象. 这很重要,因为它具有广泛的含义,可以用基于物理的,可量化的描述系统地替换经验假设.

现象的可视化 –模拟的数字输出通常以表示模型特征的自由度的数据形式出现. 此类数据的可用性不仅有助于直接进行动画和可视化,而且还有助于进行数据分析以及随后对感兴趣的功能特性进行可视化. 例如,在显微镜下,人们可以获得结构信息,但通常没有能量学信息. 通过模拟,可以同时拥有两者. 对于需要关联空间和能量信息的机理和反应途径的研究也是如此.

MTM中介绍的理论和计算方法在很大程度上代表了(硬和软)材料的物理,化学和力学研究的最新水平. 请注意,它们不包括任何应用程序所需的附加程序,即,将特定的复杂实际系统简化为可以用可用的计算手段处理的虚拟模型,选择理论和计算方法并评估计算的准确性和局限性,以及对结果的分析,旨在了解并提供物理洞察力以及与实验的联系. 为什么这些程序如此重要?让我们不要忘记,我们模拟的是虚拟系统,而我们从计算中学到的是该模型的物理特性. 因此,我们在假装模型代表实际系统时必须谨慎. 对于某个过程的动力学仿真,情况也是如此,因为例如虚拟实验的持续时间可能不允许检测相关机制. 另请参阅我们的讨论"原子模拟:简介"

## 中尺度的建模和仿真挑战
在评估材料建模和仿真的优点时反复提到的一个问题是,人们如何利用对分子机制的理解来预测宏观上材料的功能行为. 本着多尺度材料建模的精神,可以将"中尺度科学"定义为将微观模型和模拟与宏观行为联系起来的过程.中尺度的科学挑战引起了社会的特别关注(DOE-BES 2012). 为了说明材料现象中长期存在的问题,对它们进行了说明,例如过冷液体粘度的温度变化,水泥浆凝固过程中剪切模量的时间演变以及载荷对应力中裂纹扩展的影响. 玻璃的腐蚀开裂(Yip and Short 2013).在每种情况下,在分子模拟中处理介观时间尺度都存在一个共同的挑战. 专门讨论了已提出的各种问题(编辑:Sarrao和Stan).

## 通过模拟超越材料特性
如上所述,结构-属性关系的搜索是材料科学的核心,多年来一直指导着材料设计计算工具的发展. 当今的巨大挑战是利用先进的计算方法来模拟"真实"材料所经历的物理和化学过程:它们的形成(见生长,成核),它们的转化以及最终的降解. 实际上,原则上,计算机实验可以提供有关给定材料的独特信息,因为它们可以在实际实验无法访问的条件或范围内探索其行为. 正如Galli在"全体会议主题"部分所指出的那样,最雄心勃勃的目标之一就是将第一原理模拟作为综合协议的一部分.另一方面.

如所证明的,例如,过去30年中基于密度泛函理论的分子动力学的成就,成功将取决于巧妙的理论和计算算法的协同作用,以及计算机科学和代码开发的有效程序的辅助.

## 人工智能:材料建模的新范式
在过去的几年中,材料建模社区对包含人工智能,机器学习和深度学习的概念和工具的兴趣激增. 如上所述,本手册包含了该领域的一些贡献和许多参考资料. 可能还需要一些评论来指出这种新的研究方法的重要性. 我们指出,到目前为止,基于数据的人工智能策略被认为构成了计算材料科学的第四范式. 然而,这种新的"范式"同时诞生于许多不同的实验室,但仍处于起步阶段,并且正在努力建立准确的透明和可共享的数据. 毫无疑问,这一艰巨的努力将继续提供独特的机会,将计算机科学和计算材料科学的社区联系起来,使两者都受益. 一方面,材料科学的发展和影响将继续受到传统方法,理论,实验和计算以及它们之间相互作用程度的限制. 另一方面,人工智能工具的应用势必会带来新的且也是未曾料到的挑战,例如,将要合成的虚拟材料(肯定是大量)的合成. 实验,计算以及它们之间的交互程度.
