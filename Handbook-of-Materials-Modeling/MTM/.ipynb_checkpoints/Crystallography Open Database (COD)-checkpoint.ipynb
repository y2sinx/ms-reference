抽象
晶体学开放数据库（COD，http：//crystallography.net/）截至撰写最大的开放式矿物，金属有机，有机金属和小有机晶体结构时，不包括单独存放的生物分子。蛋白质数据库（http://wwpdb.org/））。与其他现有的化学晶体结构数据库不同，COD是完全开放的 - 所有结构都可以不受限制地下载，使用和重新传播，以及从中获得的结果。目前，COD包含> 385,000条记录并且不断增长，包括在同行评审的学术出版物和个别研究人员捐赠中发表的大多数结构。本文介绍如何在COD中组织数据以及如何为各种目的查询，下载和处理数据库。

1简介
X射线晶体学是确定凝聚态物质内部结构的极其有效的方法。X射线（伦琴的发现后不久，1896年）以及它们对晶体样品衍射的第一记录（Friedrich等人1912A，b），通过这种技术来确定结构的数目开始增长。使用第一原理解释X射线散射（Bragg and Bragg 1913 ; Bragg 1913允许以统一的方式确定各种固体材料的结构模型，从简单的无机物到非常大的生物分子。随着越来越多的晶体结构出现，很明显，它们描述中的数字（如晶体晶胞参数，原子坐标）由于普通散射理论的可用性而变得均匀，自身具有很大的价值并且收集了很多他们系统地开始了。所述第一集合为以纸件形式（赫尔曼和埃瓦尔德1931 ; IUCr 2017c ; 1955年来自最初的出版物（例如，IUCr于1948年开始的Acta Crystallographica期刊（Clews and Cochran 1948））中，数字数据伴随着专门用于该领域的期刊的晶体学出版物。

电子计算机的可用性和功率的增长使得晶体学家能够将它们用于结构确定，并促使人们认识到晶体结构数据也可以自动处理（Brown and McMahon 2002）。第一个专门的晶体学数据库CSD由CCDC于1965年建立（Groom和Allen，2014），用于收集小有机分子的结构，并采用计算机辅助的信息存储和检索方法（Allen等，1979）。有关无机晶体（Kaduk 2002），合金（White等人，2002年）和粉末衍射数据（Kabekkodu等人，2002 年）的数据）历史上被保存在单独的档案中。今天，我们拥有一系列数据库，其范围，规模和许可模式各不相同，涵盖了晶体学数据的各个方面（表1）。

2 COD简史
COD项目最初是一项社区倡议，当时SDPD（粉末衍射结构测定）的结晶学家讨论了晶体学数据传播的可能模式。2003年，计算机变得便宜，互联网连接广泛可用，免费/自由开源软件（F / LOSS）无处不在。Armel Le Bail提出了一个问题，即是否有可能通过加入社区工作，为每个人建立一个完全开放和免费的结构数据库。回答这个问题，Michael Berndt（1964-2003）列出了社区资源创建和策划所必需和充分的三个条件：“一个在数据库和软件设计方面具有一定经验的小型科学家团队，以协调项目; 作者（即 科学界=你）为项目提供数据库条目/ ... /; 自由软件（a）用于维护数据库，（b）用于数据评估和计算衍生数据。“考虑到这一计划，COD项目开始并成为自上而下，重资金的可行替代方案数据库项目。从2003年到2007年，COD数据库主副本由法国勒芒大学的Armel Le Bail维护。2007年，其收集的50,000份记录被移植到立陶宛维尔纽斯的生物技术研究所，COD的软件开发，并继续进行数据库维护。当生物技术研究所于2011年与维尔纽斯大学合并时，

尽管维护人员有多次转移，但COD由国际COD顾问委员会（AB）管理，该委员会在COD网站上列出并通过邮件列表进行操作。COD AB建立COD数据管理策略并设置COD数据的包含标准。以这种方式，保持了数据库质量的连续性。

在2007年以来的10年期间，COD不断增长，2017年达到了385,000个记录（图1）。这可以通过引入新的数据存储网站（图2）实现，该网站允许手动和自动上传数据到COD，并且在开发自动数据收集和沉积软件之后自动将可用结构存入COD。此自动化又高度通过引入晶体交汇处框架（CIF）（Hall等人的便利1991 ; IUCr 2017b）。CIF框架最初用于促进结晶纸的出版，并通过提供自动晶体学数据处理方法减少数据中的输入错误（Brown and McMahon 2002）。在发布过程中引入电子数据处理显着减少了数据发布中的输入错误，这是实现可靠数据重用的重要一步。不仅如此：以标准化的机器可读形式提供晶体结构描述作为科学出版物的补充材料极大地促进了该数据的重复使用。因此，COD数据采集子系统可以自动摄取所有必要的值并制定结构描述记录，使用IUCr出版物公开的信息以及公开提供必要信息的其他一些出版商的期刊。相同的CIF框架使COD能够以广泛接受的标准化形式呈现其所有数据集合，

但是，COD进行的数据收集程序并不完全是直截了当的。实际上，所有结构，即使在出版物旁边以标准CIF格式表示，也缺乏必要的元数据，例如出版物参考书目; 有时计算的项目（如单元格卷或空间组名称）缺失或以非标准化形式显示。这些信息由COD数据处理管道自动插入（Gražulis等人，2009）。此外，补充文件的一个不可忽略的部分，虽然它确实包含类似于CIF的形式的必要数据，但并不严格遵循CIF语法。由于此类案例的数量太大而无法手动纠正，因此实施了纠错CIF解析器（Merkys等人，2016年））。当研究人员使用Web沉积界面将数据存入COD时，遵循相同的程序（Gražulis等人，2012）。通过这种方式，COD确保进入其集合的所有结构描述在语法上都是正确的，即符合IUCr（2017a）定义的语法。

通过这种设置，COD可以进一步发展，为研究人员和所有相关方提供对晶体结构数据的开放访问，并不断发展以应对新千年的挑战。计算环境变化迅速，每天都有新的技术，语言，格式和协议，计算机体系结构变化得足够快，因此任何合理的科学档案都必须比许多代计算机软件和硬件更长久。COD设计的基本原则和COD的成功运行十多年，暗示COD创始人选择的方法是合理的，COD将成功地发展到未来。

3 COD的范围和内容
COD收集机器可读的无机化合物，矿物，小有机分子，金属有机和有机金属化合物的晶体结构描述。蛋白质，核酸及其复合物，糖蛋白等通常被排除在COD之外，因为它们被系统地收集在开放存取数据库中，蛋白质数据库（PDB）（Berman等人，2012）。COD中的大多数“小分子”结构使用独立原子参数（使用全矩阵最小二乘精修）和球形原子模型的假设来精炼。例如，这使得COD适用于对分子几何结构产生限制并改善较大分子或分子组装（Long et al.2017a，b）。但是，我们必须注意，这个假设并不一定适用于所有COD条目。对于较大的条目，或者当存在无序时，可以通过作者对热位移参数进行约束。对于使用粉末衍射技术求解的结构，也可以使用对粘合长度和角度的约束。最后，通过混合方法解决COD中的一些结构，使用粉末衍射进行Rietveld精修并使用DFT进一步细化原子参数; 一些结构完全基于DFT计算报告。显然，从受约束的结构确定键长和角度参数将导致循环推理，因为在结构细化过程中已经使用了相同的约束。从而，

4 COD数据语义和选择
为了便于结构选择，COD维护一组标志，描述用于结构确定的实验和细化技术。在COD SQL表中，“数据”表的“方法”列描述了可以是“单晶”，“粉末衍射”或“理论”的实验技术。如果此列的值为NULL，则该方法很可能是单晶衍射。不幸的是，在许多结构中，最流行的方法“单晶”并未明确提及，因此这种假设是一种猜测; 但结构通过“粉末衍射”或“粉末衍射”解决“理论”方法通常标记得更准确，数量更少，因此猜测应该是合理安全的。标记为“理论”的结构实际上通过DFT计算解决，而不使用任何结构特定的实验数据。这些结构当然更适合于不同的数据库，TCOD，专用于理论结构，实际上也很可能存放在那里。他们最终进入了COD，因为它们是作为一些论文的补充材料提供的，并没有被标记为理论，只有后来的数据管理才显示出他们的决定方法。几个重要的理论结构，例如，来自DFT方法的误差估计研究（Lejaeghere等，2014），在TCOD完全运行之前存放到COD，但被认为足够重要，因此必须在数据库中永久存储这些数据记录。由于COD策略不是要删除任何记录，因此一旦分配的COD ID保持稳定，COD的策略就是用适当的标记标记其条目，但不删除它们。

此外，COD数据库表包含几个描述实验技术的字段，取自IUCr Core CIF字典。''data''表的''辐射''，''radType''和''radSymbol''列分别直接来自CIF数据项_diffrn_radiation_probe，_diffrn_radiation_type和_diffrn_radiation_xray_symbol。这些数据项可以区分从X射线，中子和电子衍射数据获得的结构（''辐射''列可以有值''X射线''，''中子'或''电子'''对于各自的辐射类型）。同样，与“单晶”值一样，最流行的辐射类型“X射线”通常没有标记，因此表示为NULL值。我们可以期待作者在提交由不太常见的方法制作的结构时更加专注，但是某些谨慎当然是合适的。

从COD中选择记录时，必须记住某些簿记数据项。某些结构存放在COD中，故意比最佳解释更糟糕; 这通常在出版物中进行，以证明作者提供的数据的主要解释是正确的或者确实是最好的。COD策略是包含这样的结构（以便可以轻松验证论文声明），但将其标记为“次优”。在COD CIF中，此类结构标有_cod_suboptimal_structure yes和_cod_related_optimal_struct数据项，以及COD '数据'表，它在''最佳''中具有非NULL值列指向相关的最佳结构。除非寻求对次优和最佳结构进行明确比较，否则只应选择具有NULL “最佳”值的结构。

另一个问题是包含已知问题的结构。同样，COD政策不是要删除这些结构，一旦它们被包含在COD中，而是适当地标记它们。该标志记录在COD数据库'data'表''status''列中。此列的可能值包括“警告”，“错误”和“已撤消”。该“”警告“”级表示该结构可能毕竟正确的，但也有奇怪的特点，不寻常的描述，或者是错误的元数据。在'错误'标记结构要么被后来发表的观察结果证明是错误的，作者的更正，要么包含严重的数据一致性问题，妨碍对结构的正确解释。在所有情况下，_cod_error_description都提供了人类可读的问题描述。最后，“状态”列中的“缩回”表示结构已缩回，不应在任何情况下使用。撤回的原因可能有所不同，但通常这个标志表明直到完全科学欺诈的非常严重的问题，就像在一次IUCr调查中发现的那样（Harrison等人，2009年）; 在这种情况下，原始出版物也会被撤回。

最后要注意的是COD中存在重复的条目。不幸的是，由于在COD的早期阶段不太严格的准入程序，或者由于编程或数据编码错误，有时相同的结构不止一次地存在于COD中。再次，当检测到这种情况时，COD中的条目都不会被删除; 相反，一个条目（通常是最完整的条目）被声明为描述此结构的“主”条目，而其他条目使用_cod_duplicate_entry数据项标记为“重复” 。如果主条目缺少重复项中存在的某些信息，则此信息将合并到主条目中并作为新修订提交。重复条目由非NULL''duplicateof ''标记在列“数据”表。因此，要仅选择那些未标记为重复项的条目，需要选择将“duplicateof”列设置为NULL的条目。

必须注意的是，在COD中仅标记技术副本，即，仅来自相同原始描述和来自相同出版物的结构。在不同出版物中报告的相同化合物的两种结构不被认为是重复的并且存储为不同的COD记录。即使将相同的数据文件作为补充材料发布到两个不同的出版物，它也存放在两个不同的COD标识符下。这里的基本原理是COD记录报告了某处报告的晶体结构解决方案的实例，所有这些情况都必须在数据库中表示。进一步减少多个记录是COD用户的责任，事实上，

5.2使用RESTful接口
COD提供RESTful接口，允许人们根据特定条件以及晶体结构文件本身检索有关COD条目的信息。REST（REpresentational State Transfer）是一种基于网络的程序的架构风格，在Roy Fielding的博士论文中有所描述（2000年）。与COD相关的此体系结构的主要思想是使用客户端 - 服务器设计（COD服务器为多个客户端提供服务），以尽可能使COD服务器无状态（因此对COD服务器的相同请求应产生相同的结果如果重复多次），使用基于HTTP协议的标准连接和稳定的Web URI，并使用标准格式（CIF，HTML）来交换信息。基于REST的概念的接口，即所谓的RESTful接口，具有不需要专门的客户端程序的优点，因为查询可以由能够解析URI的任何软件执行，包括但不限于大多数因特网。浏览器。

COD RESTful搜索查询URI遵循HTTP GET查询格式，以http://www.crystallography.net/cod/result为基础URI。例如，返回与包含Li和O原子的结构相关联的COD ID列表的查询将于2017年发布，其形式为：http：//www.crystallography.net/cod/result ?el1 = Li＆el2 = O年= 2017年＆FORMAT = LST

如上所述，不需要专门的软件，但是它可以简化查询字符串的构造。清单2给出了重写使用cURL程序的相同请求的示例。

5.3查询SQL数据库
SQL（结构查询语言）可以说是查询关系数据库最强大的方法，并提供比COD网页甚至COD RESTful接口更多的功能。Crystallography Open Database提供对其数据表的只读访问，以便SQL查询可以由用户或第三方软件执行。涵盖SQL语言及其用法超出了本章的范围，但是存在大量教科书和SQL的在线参考，以及SQL的几个F / LOSS实现的优秀文档（MySQL就是其中之一）。在本文中，我们仅提供一些示例，演示如何使用SQL查询来查询开箱即用的COD。

COD SQL表是从COD CIF集合自动构造的。表由Subversion存储库的post-commit钩子更新; 因此，SQL表应始终与CIF集合同步。在COD中，数据流总是从CIF到SQL数据库; 因此，必须首先在主存储库中记录和版本化表中的所有更改。因此，MySQL基本上充当COD的快速搜索缓存，利用索引表和查询优化器。COD MySQL '数据'表还包含''svnrevision''记录从中生成每一行的Subversion修订版的列。除此之外，所有COD MySQL表都以文本形式每晚转储，并提交给与CIF集合相同的Subversion存储库。如果出现科学计算再现性的必要性，这些档案提供了重现前一段时间运行的查询的方法。

6 COD应用
尽管COD没有一些旧的晶体学数据库那么大，但由于其开放性，它有许多应用。COD擅长的一个直接可能性就是教学。使用COD可以为学生提供一些真实的数据搜索和晶体学应用，说明各种化合物的结构，并提供对现代化学研究领域的见解（Gražulis等人，2015）。COD的优点是其极快的发布周期（数据库每天更新），允许学生将任意部分甚至整个数据库下载到他们的计算机的许可许可证，以及在因特网上可以从外部访问或从外部访问的可用性教室。

另一个被广泛接受的COD应用是在粉末衍射法和搜索匹配程序的帮助下用于材料识别。最大的衍射仪供应商（其中包括Bruker，PANalytical，Rigaku）已经为其软件调整了COD集合，并将其与其设备一起发货，在COD网站或他们自己的页面上定期更新。由于COD是开放式数据库，因此这些更新对最终用户是免费的。COD目前已经积累了足够的矿物结构，因此它可以用于SOLSA项目（http://solsa-mining.eu），其中使用数据库以及其他信息源作为材料识别和数据的工具传播。

在生物信息学和药物设计中，COD被用作约束文库的开放数据来源（Long et al.2017a，b）。它也被用于DataWarrior（Sander等人，2015）作为化学信息来源之一，并在OpenMolecules网站（http://www.openmolecules.org/）中使用。软件测试受益于大量COD数据的收集，需要检查不同的案例，并且需要将数据存储在回归测试中。最后，COD用于基础研究以回答关于物质的不同问题（参见，例如，最近关于MOF的工作（First and Floudas 2013），氢存储（Breternitz和Gregory 2015），或2D材料的表征（Mounet等人。2018年）。

7结论
COD十多年的历史证明，使用开放式许可模式可以建立一个持久，高质量的科学数据库。在目前的状态下，COD可用于一系列学术和工业应用。最重要的是，这个开放的数据库为每个人提供了在自己的小分子晶体学领域获取知识的机会。与此同时，可以做很多明显的改进。显然，COD需要更全面的数据收集。应该开展更多的社区组织工作，让更多人参与数据更正，收集和确保COD记录的质量。应该与其他互联网数据资源建立更多链接，将COD更紧密地集成到链接开放数据云中。
