[面向数据科学的 Pauling文件项目和材料平台: 从大数据到材料基因组(Materials Genome)](https://link.springer.com/referenceworkentry/10.1007/978-3-319-42913-7_62-2):  


**摘要**:  
描述了材料信息学中最古老的举措之一,即 Pauling文件项目(PAULING FILE).它包括有关无机结晶化合物,其原子结构,固有物理性质和相图的综合数据库.最重要的是,引入了功能强大的**在线检索软件,称为MPDS,即数据科学材料平台**.给出了存储,交换和分析大量材料数据的实用方法.重点放在现代信息技术和软件工程上.结果,从大量的异构数据中,得出了关于整套已知材料的整体结论.它们可以被视为系统的大规模预测的指南.

# 引言

PAULING FILE项目是具有将近25年历史的材料数据库,在同一框架下将结晶数据,相图和无机结晶物质的物理性质分组.它的重点放在实验观察上.每个单独的晶体结构,相图或物理性质均来自经过同行评审的特定出版物.从1891年到现在,涵盖了材料科学,工程,物理和无机化学领域的全球科学文献.数据科学材料平台是PAULING FILE项目的在线版本,创建于2016年.它通过两个在线界面显示所有PAULING FILE数据:图形用户界面和应用程序编程界面.前者是供材料科学家使用的.后者适用于软件工程师和数据科学家.这些小组的研究兴趣的交集属于材料信息学的新颖学科的范围.

# 材料大数据和PAULING FILE项目

## 材料界的现代挑战
违反直觉的是,经验传统在基础科学尤其是材料科学中广泛存在.化学信息学的先驱Peter Murray-Rust(英国剑桥)将材料科学称为最保守的精确学科之一,与其他科学(如生物化学,天文学)相比,它是最不透明且对协作和众包开放的开放性,数学,计算机和环境科学(Murray-Rust 2013)–不包括付费专区.有趣的是,与化学信息学和生物信息学相比,材料信息学的领域还不成熟,而且还比较年轻(Gasteiger和Engel,2003年,此处信息学假定了信息交换).在新颖的分布式计算项目中,任何人都可以自愿将计算时间从他们的个人计算机用于特定的原因,其中只有很少一部分与材料科学有关.可以给出以下不详尽的解释.材料科学和工程学与航空航天,汽车,电子,军事工业等行业紧密相关.材料研发的相当一部分是私人资助的.与其他精确学科相比,材料科学知识的平均成本更高.这些成本可以与高能物理或天文学相提并论.但是,此类专有技术对技术进步的影响也更大,投资与创新之间的滞后相对温和(大约15年或更短时间,奥巴马,2011年)).此外,由于其广泛而极端的复杂性,材料科学也非常细分("封建").因此,无论是在学术界还是产业界,竞争水平都很高(政府补助,职员职位等).在复杂性方面,即使掌握现代固态物理学的基本形式主义也提出了一项艰巨的任务.通常,多学科专家在材料科学领域很少见.这些原因可以说明一个问题,为什么材料科学界固有地非常保守.

此外,直到最近几年,完全没有共享基础研究成果的文化(例如,与文章相比,原始测量或仿真日志),甚至在材料科学中甚至被禁忌.关于出版物,仍然没有办法从期刊获取机器可读格式的科学信息,以供进一步重用和重新利用.但是,随着技术的进步,现代学者正在沉入"数据海洋".如今,人们普遍抱怨研究人员发布的内容太多或太多.要估算这种增长,引用学术数据库的增长是错误的,因为没有数据库可以捕获所有内容.马克斯-普朗克学会和瑞士联邦理工学院的文献计量分析师(Bornmann and Mutz 2015)估计,全球科学产出每9年大约增加一倍.此外,随着通常称为"材料设计"或"材料基因组"的现代数据密集型方法的普及,有证据表明,通过实验或模拟生成的材料科学中的原始大数据量将继续增加.呈指数增长.在谈到大数据时,计算机科学界已经适当地将这种现象定义为由四个指标控制的数据量:数量,速度,多样性和准确性.在材料科学中,体积是指超出个人计算机设施的大体积,速度是指利用实时数据采集(例如,来自动力学实验).各种涉及该数据采取各种形式在材料科学,从离散的数值到的材料特性和成像数据的定性描述的事实.Veracity承认有关不确定性和大量"缺失"数据的实际现实(Rajan,2015年)).如今是材料科学中大数据的时代.然而,在大数据时代,材料科学仍不足以鼓励并支持共享,分析和解释研究人员正在收集的大量数据.在材料科学界,传统的科学信息交流手段是不完善的.在这种情况下,自然多样性出现了,自然选择的原则指导了前沿研究的形式.在过去的几年中,随着Internet和现代信息技术的高度普及,情况开始缓慢变化.在这里,消化信息,得出正确结论的能力至关重要.这是数据科学工具(例如,机器学习)需要与材料科学的基础联系起来:理论,建模和实验.目的是使实验室用于产生新的信息,而不仅仅是库用于检索已知或预期信息(Ghiringhelli等人2017).

## PAULING文件背景
有许多在世界努力克服在材料科学(上述挑战的举措皮兹等人.2016).主要思想是系统地收集材料数据或以高通量方式生成和处理它们.1995年启动的PAULING FILE项目是著名的计划之一.它的主要重点是对已发表的实验材料数据进行严格的评估.从历史上看,它是日本科学技术公司,瑞士材料相数据系统公司和东京大学RACE的合资企业.现在,它完全由瑞士公司管理.从一开始就计划了三个步骤.**第一个目标是创建和维护一个全面的无机晶体物质数据库,其中包括晶体学数据,衍射图,固有物理性质和相图**.应格外小心地检查数据.术语"无机物质"定义为不含CH键的化合物.与数据库创建并行,**第二个目标是开发合适的检索软件,以使数据可以在单用户界面中访问**.从长远来看,**作为第三个目标,应该创建用于材料设计的新工具,以自动在数据库中搜索相关性**.这被称为具有预定义的固有物理特性的新型无机材料的智能设计.试用版PAULING FILE Binaries Edition于2002年作为桌面软件发布.现在,PAULING FILE数据的选定部分已包含在几种印刷,脱机和在线产品中.今天,PAULING FILE项目已广为人知.已经有成千上万的出版物引用它.它的基础,数据库设计,2004 ; Villars等.2008 ; 徐等.2011年;Kong等.2012 ; Villars和Iwata2013等).PAULING FILE检索软件和材料设计工具的最新实现是一个称为 **"材料科学数据平台"**的在线产品.

现在必须给出最低限度的定义.一个数据库是服务于多个用户的需求相互关联的存储的数据的集合.在所谓的关系范式中,这些存储的数据被组织在表中.使用数据库而不是文件的动机包括:使更多的用户具有更大的可用性;集成数据以便更轻松地访问和更新复杂事务;以及减少数据冗余.一个数据库管理系统(DBMS)是用于操纵数据库的一个广义的软件.DBMS支持逻辑视图(数据库模式),物理视图(访问方法,数据集群),数据定义和操作语言以及实用程序,例如事务管理和并发控制,数据完整性,崩溃恢复和安全性(Teorey等,2005).例如,Oracle用作PAULING FILE的内部DBMS.PAULING FILE数据库具有以下结构.**数据的标准单位称为条目**. **所有条目都细分为三种:晶体结构,物理性质和相图**.它们分别称为S,P和C项.条目具有类似于数字对象标识符(DOI)的持久标识符,例如S377634,P600028和C100027.PAULING FILE数据的另一个维度是不同的阶段.这三种条目通过它们所属的不同材料阶段相互链接.在过去的25年中,PAULING FILE编辑人员进行了巨大的工作,以手动区分超过15万种无机材料相,这些相出现在世界科学文献中.每个相具有(a)化学式和(b)修饰的独特组合.这些是使用结构原型,Pearson符号和空间组定义的.每个阶段都有一个称为阶段id的整数标识符.在接下来的三个部分中,将详细介绍每种条目类型(S-,P-和C)及其相互链接.

## 物理性质
PAULING FILE的P项包括在固态,结晶状态下无机化合物的多种固有物理性质的实验数据,并在一定程度上包括了模拟数据.所考虑的物理性质属于以下七个一般领域之一:(a)电子和电气性质,(b)光学性质,(c)磁性性质,(d)机械性质,(e)相变,(f)超导性和(g)热和热力学性质.分类法包括三个级别:提到的一般域,子域和特定的物理属性.例如,域"电子和电气属性"包含子域"电子能带结构",而子域又包含"费米能"属性,等等.当前大约有100个子域和近2000个特定物理属性.该分类法由Fritz Hulliger(位于瑞士苏黎世的瑞士联邦理工学院),Roman Gladyshevskii(位于乌克兰利沃夫的伊凡-弗兰科国立大学)和Karin Cenzual(位于瑞士的日内瓦大学)共同编写.在某种程度上,它反映了上个世纪固态物理学的发展.

物理属性以四种不同的方式存储:数值,图形说明,属性类别(例如铁磁体,压电等),以及源出版物中特定数据的存在指示,例如不同的光谱.最常见的物理性质的符号已标准化,主要基于《CRC化学与物理手册》(Lide 1997)–1998年).数值以发布的单位存储,并转换为标准SI单位.对于原子级的某些特性,可以使用其他单位,例如eV或μB.相对于确定数量的物质(每千克,每摩尔)表示的特性被转换为每原子克.每个数值属性值都附带有关特定测量的实验条件的信息.通过参考表的链接提供了极大的灵活性.因此,可以选择新的属性,并且可以控制其符号,单位和幅度范围.

所有数据均取自原始文献.每个P条目都对应于一个特定的数据源,并且可以包含多个数值,图形描述和关键字.为了通过温度或压力诱导的结构相变研究化合物,将有两个P项,例如,一个用于室温修饰,一个用于低温修饰.默认情况下,假定铁电跃迁会伴随结构变化,并且将证明创建两个P项是合理的,而磁,电或超导跃迁则不会.具有一定同质性范围的相的数据归为代表性的化学式.当特定测量的实际组成不同于表示P项的组成时,在参数中指定.对于晶体结构部分,将存在三个数据库条目,用于两个三元化合物之间的连续固溶体:一个用于每个三元边界成分,另一个用于包含四个化学元素的分组样品.还包括一些从头算的模拟数据,特别是能带结构,但重点是实验测量的数据和直接从测量得出的值.

除了物理特性(以数值,图形描述或关键字的形式)和诸如化学式之类的必填项外,还存储了大量有关样品制备和实验条件的信息.物理属性P条目中可能存在以下数据库字段:(a)化合物,例如化学系统,公开的化学式(研究的样品),代表性的标准化化学式,修饰形式;(b)书目数据,例如参考文献,作者,语言,标题等;(c)制备,例如起始原料和合成方法;(d)样品描述,例如形式,化学分析,相对于温度,压力和组成的稳定性,弹性行为,密度,颜色,化学反应性;(e)晶体学数据,例如结构原型,空间群和晶胞参数.

## 晶体结构
目前,PAULING FILE包含超过350,000个晶体结构S条目.在PAULING文件中对S条目的最低要求是将已发布的单元格参数的完整集合分配给定义明确的化合物.只要有公开的数据,晶体学数据还将包括原子坐标,位移参数和实验衍射线,并附带有关制备,实验条件,样品特性,相变以及晶胞参数对温度的依赖性的信息,压力和成分.为了给出实际结构的大致概念,对于可分配原型的S入口,提出了一套完整的原子坐标和位置占用量.1987).相关时,可以对其进行进一步调整,以便可以直接比较同型S条目的数据.导出的数据包括基于最大间隙法的各个原子位点的原子环境(Brunner和Schwarzenbach1971; Daams等人1992).),以及减少了Niggli的细胞.使用上一部分提到的程序包ESDD,通过比较不同的S条目来检查S条目中是否存在不一致.对于5%的S条目,检测到并纠正了已发布晶体学数据中的一个或多个错印.备注中还包含有关剩余原子间距离短,与标称成分的偏差等的警告.SI单位随处可见,其晶体学术语遵循国际晶体学联合会的建议.

## 相图和不同的相
PAULING FILE的相图部分包含用于二元系统的温度组成相图,以及用于三元,四元和其他高阶系统的水平和垂直部分以及液相线或固相线投影.实验确定和计算的图均被处理.优先考虑原始文献.还有一些著名的汇编中的图,例如二进制相图的纲要(Massalski等人,1990年)和有关三元相图的系列丛书(Petzow和Effenberg 1988年)–1995年),已包括在内.所有图表均已转换为at.%和°C,并以单一比例重绘,因此可以轻松比较同一化学系统的不同报告.单相场以蓝色着色,三相场以黄色着色.图上标识的阶段是根据PAULING FILE约定命名的,但原始名称也被存储.每个相图都链接到一个C条目,该条目通常包含以下数据库字段:(a)分类,例如化学系统和图的类型;(b)调查,例如实验或计算技术,APDIC标准化;(c)书目数据,例如参考文献,作者,隶属关系,语言和标题;(d)原始图的详细信息,例如出版物中的图号,边框,比例,尺寸等;(e)重画图表细节,例如浓度范围,温度和浓度换算;(f)图表中存在的相的列表,标准化的相名称,原始出版物中使用的名称,编辑者分配的结构原型,原始出版物中给出的结构信息以及指向代表性晶体结构S-entry的链接.对于二元系统,还存储相存在的上限和/或下限的温度和反应类型.

物理性质,晶体结构和相图条目通过不同的相概念关联.在数据库级别,所有三种不同类型的数据(P,S和C条目)都链接到不同的阶段表.为了准备该表,已经评估了每个化学系统.例如,报告的TiO 2晶体的三个主要不同相(金红石,锐钛矿和板钛矿)相对于温度或压力是分开的.然后报道的TiO 2的物理性质或晶体结构仅与相应的不同阶段关联.最后,每个C条目(相位图)由一组特定的已知不同相位组成.因此,实现了数据的唯一互连.归因于相(化合物类别,矿物名称等)的一定数量的特征存储在不同的相表中.每个不同的相都有一个唯一的名称,该名称包含一个代表性的化学式,必要时可以加上诸如" ht"," rt"," cub"等的规格.存在以下几种特殊情况.第一,区分了在相同结构原型中结晶但在相图中被两相区域分隔的相.对于温度或压力引起的同构相变,也是如此,其中报告了单元参数的不连续性.第二,在某些情况下,具有不同排序次序的结构已被单独考虑,而在另一些情况下,则没有进行明确考虑,这取决于明确分配一个或另一个修改的可能性.考虑到例如分裂原子位置的结构改进通常在父原型下分组.第三,在较新的文献中指出是不正确的结构建议已与最新报告相符,被归为一个阶段.即,例如,在这种情况下,报告六边形单元格的S条目可以在正交相下分组.第四,此处使用的结构原型的定义表明,连续的固溶体可能会从一个原型平稳过渡到另一个原型.然后将考虑一个或另一个原型的优化组合在一起.第五,报告的忽略晶体结构的物理特性,原则上是指环境条件,被赋予" rt"修饰,或者,如果不知道温度依赖性,则被赋予最常观察到的修饰.第六,默认情况下,假定顺电-铁电相变伴随有结构转变,并且在转变温度之上和之下都考虑了不同的相.相反,假定磁排序不会在很大程度上改变核结构,因此不会导致新的相.仍然存在很少探索的化学系统,因此文献中的报告是矛盾的.在这里,阶段分配变得非常困难,并且不同阶段的列表有时会包含比实际存在更多的阶段.应当注意的是,在分配相位标识符时存在一定量的主观性.然而,这种方法代表了很大的优势.

## 应用
由于存储在数十个表和数百个不同的数据库字段中的大量信息,PAULING FILE提供了几乎无限的检索可能性.当然,它可以用于基于化学系统或文献数据的各种平凡搜索,但它还有很多其他用途.转换为标准单位有助于在特定数值范围内搜索属性,并且不同相的分配起着至关重要的作用,从而可以对数据库三部分中存储的数据进行组合搜索:晶体结构,相图,和物理性质.数百个互连的数据库字段可用于创建不同的产品.PAULING FILE数据包含在各种印刷产品以及离线和在线软件(例如桌面目录,仿真环境,材料研究工具包等.其中一些产品仅包含结构数据,其他产品仅包含相图和晶体学数据,而其他产品则仅包含三组数据.根据生产者的喜好,某些产品仅包含已发布的晶胞参数,其他产品仅包含标准化的晶胞参数,而另一些产品则包含已发布和标准化的晶体学数据.一些产品仅限于PAULING FILE数据,而其他产品也包含来自其他来源的数据.其他的仅是标准化的晶胞参数,还有其他的既是公开的又是标准化的晶体学数据.

数据科学材料平台(MPDS)是最新的在线基础结构,显示了PAULING FILE数据的所有三个部分.它包含近70,000个相图,超过350,000个晶体结构和近700,000个物理特性条目.这些数据的来源大约有300,000种材料科学,化学和物理学方面的科学出版物.可以以开发人员友好的格式(通过所谓的MPDS应用程序编程接口)以大规模方式远程请求大约80%的数据,以准备用于任何外部现代数据密集型应用程序.接下来的部分概述了PAULING FILE的实现,其技术细节和使用场景,以及该领域的常规工作.

## PAULING FILE物料基础结构的开发
迄今为止,材料信息学是计算机科学和现代材料科学所采用的食谱的集合.从学术角度来看,主要困难是纯技术性的-如何有效地处理材料大数据.这里讨论一些食谱.在开发和维护PAULING FILE在线检索软件– MPDS(数据科学材料平台)时,考虑了所有这些因素.

解决复杂性的一种可能方法是用于软件工程的**统一建模语言(UML)**,通用,开发性建模语言(例如,参见Miles和Hamilton 2008).从鸟瞰的角度来看,UML是一种以人类可理解的方式绘制概念的约定.尽管UML无法用计算机解释,但UML提供了有关问题领域的正式描述,因此可以更轻松地用计算机编程语言进行编码.UML的另一个优点是它的标准化和在软件工程领域的流行(适合于协作工作).UML中的概念必须与定义的关系类型相关.

通常,万维网基于互连的思想.实际上,在现代环境中,没有孤立的软件本身就没有意义,并且每个应用程序都可以与其他应用程序进行通信.**为了提供快速有效的交互体验,并且不针对每种情况开发新的访问接口,通常使用应用程序编程接口(API).API规范了任何一种软件之间的通信(无论是复杂的数据挖掘程序,模拟平台还是任何其他大数据使用者).主要思想是将功能收集在一个地方,并通过在线API进行公开(封装).在线媒体中的API通常遵循代表性状态转移或REST的原则(Fielding 2000).REST提出了对客户端-服务器软件体系结构的指导约束,可以称为meta-API.它也用于MPDS API(请参见图2),该API  以开发人员友好的,机器可读的方式使用打开的格式(例如CIF,JSON和MIF)显示所有PAULING FILE数据(请参见下文). .重要的是,API不仅是客户端与服务器通信的方式.从广义上讲,在线API是一种软件体系结构,它声明了所有通信方式.这样,观众就不必受限于现有的以人为本的图形用户界面(无论其便利性如何),并且能够最大程度地使用服务提供商来实现其目标.**

但是,MPDS的图形用户界面引起了很多关注.搜索输入字段允许一次无缝地键入基本搜索词的不同组合,以便正确识别它们并显示匹配的条目.负责处理这种无缝搜索的算法如下工作.首先,过滤停用词(" the"," and"和" about"等).然后,检查单词属于一类类别:化学式,化学元素和晶格.这些类别不能包含多个单词.找到的单词将被解释和排除.然后,检查剩余的单词,使其属于材料类别和物理属性类别.最长的单词组合的匹配项会在开始时进行检查,然后检查较短的组合.无法识别的单词被标记为"忽略".

从MPDS的开发之初起,所有子系统的性能便是基石.当服务器是单点故障时,这对于集中式在线软件尤为重要.通常,对于任何在线项目,始终建议着重于操作速度.应当注意,为5个同时用户设计的软件与为50或500个同时用户设计的软件有很大不同.有两种缩放性能的方法:垂直和水平(Michael等,2007).垂直缩放意味着将部署功能更强大的服务器,并增加负载,但是软件架构保持不变.尽管存在局限性,但是该解决方案是快速有效的.首先,服务器功能越强大,价格越昂贵,并且依赖性不是线性的.其次,每台服务器的计算能力有限.而且数据库性能通常会饱和,大大低于此限制.因此,对于具有一定容差的数据库,垂直缩放毫无意义.这种情况下的解决方案是水平缩放,当以这种方式更改软件体系结构时,可以在由廉价商品硬件组成的服务器群集之间平均分配增加的负载.这个想法是通过简单的快速复制将带有MPDS软件的新副本添加到服务器群集中.新副本将加入服务器群集,并开始接受传入的HTTP请求.此外,如果某个副本由于某种原因而失败,则可以自动将其替换为新副本.这样,服务器群集将保持高可用性,甚至巨大的负载也不会影响性能,因为可以添加更便宜的副本.但是,尽管服务器价格仍在控制范围内,但开发过程很复杂并且需要时间.

## 物料数据的存储和交换
类似于任何其他科学数据,材料数据通常存储在文件和数据库中.这些文件通常显示原始数据(模拟或测量日志,压缩档案等).可重用的已处理数据通常存储在数据库中.应该注意的是,尽管这种细分是有条件的,但是文件可以是数据库,反之亦然(例如,HDF5容器,SQLite等).数据库模型是一种数据模型,它定义数据库的逻辑结构并从根本上确定可以哪种方式存储,组织和处理数据.

MPDS使用几种数据库模型,特别是关系和语义图模型.此外,在准备数据时,将面向文档的模型用作中间步骤.在面向文档的模型中,每个PAULING FILE条目都被视为一个文档.这些文档可能具有不同数量的属性,并且都存储在单个巨型数据库表中.所有文档属性都是可搜索的.该模型是使用PostgreSQL DBMS实现的.非常简单,该模型具有独特的灵活性(单表,任意数据查询).这种灵活性的代价是访问速度非常低.不适合生产,非常适合开发.关系模型健壮且成熟,是MPDS生产系统的核心.除了中间的面向文档的模型之外,PostgreSQL DBMS还支持它.关系模型中的标准数据操作语言称为结构化查询语言(SQL),并基于关系代数.最终,从面向文档的模型中提取单个表格,并分成许多相互关联的简单表格.这个过程称为归一化(Teorey et al.2005).这样可以实现更高的访问速度.但是,精炼的表结构与所选的使用场景紧密相关,并且不提供额外的灵活性.

关系模型的缺点是数据必须遵守严格的规则,即数据库模式.通常,这些规则并不意味着在应用程序业务逻辑之外的数据库级别还存在其他术语.但是,专家系统和人工智能应用程序必须在开放的世界中行动,以新收集的信息为基础进行推理并确定新事实.他们的数据库必须能够包含新术语并动态地遵循新逻辑,因此不适合传统的关系范式.通常,此类数据库采用所谓的语义Web方法(DuCharme 2013).这种模型中的数据表示为图形.此类图由一组形式为"主体属性对象"的语句组成,称为三元组.使用语义图数据库模型,在知识图之上进行前所未有的灵活表达的查询成为可能.MPDS当前仅间接使用语义图,与关系模型互补.然而,计划更广泛地采用这种模式.作为语义图模型的后端,采用了Virtuoso DBMS.

如上所述,材料科学中单一数据交换格式的开发是一项极其复杂的任务.如今,当特定的信息容器附带规则定义(即自动验证工具)时,计算机科学提出了一种非常方便的范例.遵循此想法的示例有很多:关系数据库的SQL和模式,语义图的A-Box和T-Box,机器可读数据传输和标记的XML和XSD,人类可读数据的JSON和模式在这方面,应该提到在开发材料科学通用交换格式方面取得的两个成功成就:晶体学信息文件(Hall等人,1991年)和材料与物理信息文件(Michel和Meredig,2016年).).它们都在MPDS上得到支持,部分在其他提到的材料信息学基础设施上得到支持.

晶体学信息文件(CIF)由国际晶体学联盟(IUCR)于1990年代建立.CIF基于称为STAR(自定义文本存档和检索)的文本容器,在该容器中,例如由于X射线衍射或理论建模而获得的物理属性由标准标签标记.标准标签决定了晶胞的参数,对称性,原子位置,相关的科学出版物元数据等.这些标签是在外部CIF词典中定义的(请参阅XML文档的XSD模式),因此可以验证针对CIF词典的CIF文件,甚至可以从可用的属性中推断出新的物理属性.区别在于CIF格式允许使用任意标签.根据IUCR,它们被CIF解析器忽略,但稍后可以成为标准CIF词典的一部分.此外,CIF格式支持关系数据模型,因此可以通过其标识符引用晶体结构中的特定原子.缺点是缺少方便的多层次层次结构支持,因此STAR容器在此允许使用XML.CIF格式用于MPDS的在线晶体结构可视化,并且不使用其他专有或自制格式.只需使用Web浏览器即可进行可视化,并且不需要插件,小程序或其他软件.通常,结构渲染是在GPU(即图形卡)上完成的.但是,如果GPU过时或不可用,则在CPU(中央处理器)上进行渲染.在这种情况下,渲染质量降低.在线提供的用于渲染的代码总大小仅为150 Kb,并且在加载代码后,无需进一步的Internet连接.MPDS上的CIF可视化基于开源技术.

JSON格式更简单,更灵活且更宽松.从历史上看,它比狭窄用途的CIF在软件开发中更为常见.由Citrine初创公司引入的材料和物理信息文件(MIF)容器基于JSON构建,并具有所有优势.JSON还提供了自己的模式方法,该方法用于验证,文档编制和交互控制,即,给定应用程序所需的JSON数据协定以及如何修改该数据.所有类型的PAULING FILE数据(包括MIF规范)的JSON模式于2016年开发,现已公开提供.JSON格式(包括MIF)是上述材料基础结构中使用最广泛的格式.

## 计算机辅助数据分析
毫无疑问,绘图和可视化对于数据分析非常有帮助.MPDS有趣的绘图功能之一是术语的语义图.通常,所有MPDS数据都是人类积累的材料科学中结构化知识的巨型语义图.不幸的是,没有人可以观察到整个图.在线交互式可视化试图以非常简化的形式仅显示与特定用户输入有关的巨型MPDS语义图中的一小部分.幸运的是,现代语义技术(DuCharme 2013)能够一次理解一个巨大的MPDS图.这是未来MPDS平台开发的计划方向.另一种可视化类型是在MPDS上在线显示的动态渲染的相位图(C项),是完全数字化的以编程方式绘制的图.呈现引擎可在所有现代Web浏览器中使用,不需要插件,并且基于开源Web技术.相图中的相位与参数方程式相关联,形式为x  =  x(t),y  =  y(t),0 <= t  <= 1,其中x代表成分,y代表温度.

最近,MPDS已经发布了其机器学习预测的第一个版本.为了演示使用MPDS API进行物料数据挖掘的一些实际使用情况,选择了一种相对简单但功能强大的预测性机器学习算法,即决策树回归.决策树是一种统计模型,描述了从对某个项目(例如晶体结构)的观察到有关该项目目标值(例如相应的物理属性)的结论的数据.MPDS数据包含具有相应物理特性的晶体结构,因此在此数据集上训练模型是可行的.选择以下物理性质:等温体积模量,形成焓,恒压下的热容量和熔融温度.通过反复对训练数据进行重采样(对树进行替换和投票)来构建多个决策树,从而产生共识预测.该算法被称为"随机森林 "回归器."随机森林"是一种统计估计量,适合数据集各个子样本上的许多分类决策树,并使用求平均值来提高预测准确性(Breiman 2001).它目前使用的最先进的开放源代码实施非常高效,并且花费数秒钟从普通台式机上的MPDS数据中训练模型(McKinney 2010).

## 以数据为中心的观察
继德米特里-门捷列夫(Dmitri Mendeleev)和洛萨-迈耶(Lothar Meyer)在1860年代观察到化学元素性质的周期性模式之后,人们可能会尝试在整个已知材料范围内找到相似的模式,并朝着假设的材料周期表迈出一步.基于对PAULING FILE数据库的观察,制定了12条经验原则(Villars 1994; Villars等人2008; Villars and Iwata 2013).这些原理可以称为自然的基石,并且可以使用现代电子结构理论来解释所有这些原理.他们定义了(a)化合物的形成,(b)结构原型中化学元素的顺序,以及(c)化学元素在元素周期表中的位置与其在结构原型中所占位置之间的联系.

首先,化合物的形成:原子大小,电化学,价电子和内聚能因子决定着化合物的形成.例如,可以观察到所有化学元素组合中约30%的化合物在二元,三元和四元系统中不形成任何化合物.

第二,化学元素数量与原子环境类型的相关性:在二元和三元无机化合物中实现了原子环境的最大多样性.四元,五元和其他高阶化合物强烈偏爱配位数低的原子环境类型.人们也可以称这种观察为随着元素数量的增加而几何多样性的惊人减少.

第三,活性浓度范围:在一定的相当有限的(活性)浓度范围内,二元,三元和四元无机体系存在系统发生的道尔顿无机相.

第四,化学计量比条件:绝大多数道尔顿化合物具有高度优选的化学计量比.

第五,化合物简单:绝大多数无机化合物平均每个晶胞只有十个原子,因此在晶体结构中仅显示三种或更少的原子环境类型.在这里可能有人指出,自然界确实更喜欢简单.

第六,化合物对称性:10%的空间基团覆盖了近70%的无机化合物.最常见的11个空间组是12,62,63,139,166,191,194,216,221,225和227.可以看出,高度对称是优选的.

第七,原子环境安排:在大约100种可能的原子环境类型中,有18种是高度优选的,并且在所有PAULING FILE化合物中发现90%.特别是,最常见的多面体类型是四面体,八面体,立方体,三峰三角棱镜,四峰三角棱镜,二十面体,立方八面体,双峰五角锥和反立方八面体.

第八,化学元素的排序趋势:只有大约30个结构原型具有1000多个代表,而1000个人口最多的原型及其代表涵盖了PAULING FILE的大多数晶体结构.

第九,结构原型与周期系统的相关性:绝大多数晶体结构在周期系统中化学元素的位置与其Wyckoff位置占据之间显示出非常严格的规律性.一个例子是1000个人口最多的原型.

第十,结构与稳定性的联系:原子大小,电化学,价电子和原子序数因素决定了金属间化合物的晶体结构.同样,这适用于二元,三元和四元系统.一个人可能会发现清晰的图案,例如以前的系统与非以前的系统,等化学计量的结构稳定性图以及同一原型中二元化合物之间的完全固溶性.

第十一,广义原子环境类型的稳定性:使用周期数(来自Lothar Meyer的周期表),可以将不同的原子环境类型细分为不同的稳定性域.已发现,周期数大于54的化学元素控制原子环境的类型,而与它们是作为中心原子还是配位原子无关.因此,可能的和不可能的原子环境类型之间存在明显的区分.有趣的是,与二元和三元相比,四元的原子环境类型的多样性大大降低了.

第十二,完全固溶体的稳定性:原子大小,电化学和价电子因素控制固溶度.例如,在三元系统中,两个二元边界系统具有相同的结构原型,一个可以预测是形成完整的还是有限的固溶体.类似地,对于给定的化学元素,可以预测有限或扩展的固溶度.

可以看出,单相无机固体的所有固有物理性质都与它的晶体结构紧密相关,从而强调了晶体结构分类的重要性.上面概述的12条原则只能通过检查大量严格评估的实验确定数据才能发现.它们最终导致了限制,这是开发实用和可信赖的计算材料设计方法的要求.

## 应用
公开发表的实验无机固体与高通量DFT计算之间可信赖的联系打开了数据库驱动的,数据密集型研究和新材料发现的新视角.目前,在中国和瑞士,有两个这样的高通量计算计划开始将MPDS API与PAULING FILE数据一起使用.

如今,大多数实用材料科学问题(清洁能源,储能,超导性等)都与大量要被吸收的复杂知识有关,如果不是由有天赋的人,则不是由人工智能代理吸收的.随着神经科学和医学的进步,揭示了人类大脑的本质,新的强大计算机科学技术应运而生,后来在材料科学中得到应用.但是,在封闭(专有)环境中开发复杂工具非常困难,甚至效率很低.这就是为什么开放性和思想自由交流的学术传统在开源策略中得以实现的原因,它展示了经济上有利的利他主义.

# 小结
审查了PAULING FILE,这是组织材料数据的唯一且可能是最早的工作.从材料科学的基础开始,例如物理性质的分类学,结构原型的概念,相,相图等,它跨越了现代数据密集型科学,采用了新颖的数据存储和分析技术并提供了在MPDS在线平台中实施的高质量材料见解.这对于在当今大数据挑战之前阻碍材料科学中的高效,快速的知识交流非常重要.还指出了其他努力及其对材料信息学的贡献.强调了培育当今材料信息学新兴生态系统的重要性.它基于JSON,MIF和CIF的标准,支持数据库,虚拟实验室,并与在线REST API相连,这些REST API往往成为软件基础结构的主干.展示了高度抽象级别的PAULING FILE的概念性UML建模.因此,三种类型的PAULING FILE科学数据是高度互连的,这在MPDS在线平台内是不可分割的.丰富的数据结构假定组合了存储方法,以便同时利用它们的优势和减少缺陷.探索大量不同的材料数据,可能会提供有关无机物质的整体视图.结合高通量的从头计算,它可以为发现材料基因组提供关键,在整个材料周期表中发挥重要作用.
