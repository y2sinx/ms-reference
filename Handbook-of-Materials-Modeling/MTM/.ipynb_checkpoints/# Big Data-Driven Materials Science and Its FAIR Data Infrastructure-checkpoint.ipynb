[大数据-驱动的材料科学以及FAIR数据基础设施(Data Infrastructure)](https://link.springer.com/referenceworkentry/10.1007/978-3-319-42913-7_104-1)

**摘要**:  
本章介绍材料研究的第四种范式(the forth paradigm of materials research)--大数据驱动的材料科学. 描述了其概念和最新技术,并讨论了其挑战和机遇. 为了推动这一领域的发展,开放数据和无所不包的共享,高效的数据基础架构以及社区中使用的丰富的计算机代码生态系统至关重要. 为了塑造第四范式并为改进和新颖的材料的开发或发现做出贡献,数据必须满足(现在所谓的)FAIR要求--可查找,可访问,可互操作和可重用/可重用(FAIR: Findable, Accessible, Interoperable, and Repurposable/Reusable). 这为处理大数据集的人工智能方法的发展奠定了基础,以发现无法通过单个计算甚至无法直接从高通量研究获得的趋势和模式. 回顾和展示了最新进展,并以前瞻性的观点总结了本章,以应对尚未解决的重要挑战.

# 引言

材料科学正在进入一个时代,在这个时代中,来自实验和计算的数据的增长已超出了既定方法可以处理的水平. 所谓的4V挑战--涉及体积(数据量),多样性(数据形式和含义的异质性),速度(数据可能更改或新数据到达的速率)和准确性(数据质量的不确定性)--显然变得突出.(`The so-called 4 V challenge – concerning Volume (the amount of data), Variety (the heterogeneity of form and meaning of data), Velocity (the rate at which data may change or new data arrive), and Veracity (the uncertainty of data quality) – is clearly becoming eminent.`) 然而,最重要的是,材料科学的大数据为获得科学知识和见识提供了新颖,非凡和广阔的机会. 这些机会需要新的研究概念和思路. 尽管本章着重于计算材料科学,但我们强调此处描述的内容也适用于实验数据.

当今计算材料科学中的通用方法是将结果发布为重点研究报告,仅报告与各自主题直接相关的少量数据. 因此,即使从大量的计算中(花费数千或数百万个CPU核心小时),与社区共享的结果也很少. 大多数数据,特别是在不被认为具有直接相关性的情况下,都被保密甚至被丢弃. 几年以来,计算材料科学与工程领域正在经历"科学文化的变化",并开始了该领域数据的广泛共享. 全部分享数据,即计算的完整输入和输出文件,意味着不必一次又一次地重复计算,并且该领域将可以使用大数据,而大数据可以以全新的研究方式使用,例如人工智能方法(Draxl和Scheffler 2019及其参考文献). 正如将在下一节中详细说明的那样,人们可以在大数据中找到结构和模式,获得新的见解,而这是研究小数据集无法获得的,因此,据称甚至不准确的数据也可以获得价值. 日常生活中大数据分析影响的一个流行示例是跟踪移动电话的移动,该移动电话提供有关流量和拥堵的即时信息. 另一个示例是Google搜索的本地信息流感症状和药物,可反映烟道波的传播. 后一个示例还说明了数据分析的危险,因为"Google烟道趋势"在开始的几年中运行良好,但随后在2012年,某些方面变得不可靠. 原因可能尤其是由公共和社交媒体的报道以及Google搜索技术的变化引起的焦虑刺激的搜索,例如,根据其他搜索对象推荐搜索(Lazer等人,2014年)). 该示例还说明,大数据不应被视为替代品,而应作为对传统分析的补充. 具体而言,我们注意到即使数据量可能很大,但在关联数据时独立信息也可能很小,或者出于感兴趣的目的,数据甚至可能不相关或具有误导性. 如果没有适当考虑这些方面,那么统计分析将是有缺陷的.

考虑到现在所谓的FAIR数据的各个方面,NOMAD卓越中心(NOMAD; Draxl和Scheffler 2019)克服了计算材料科学的"孤岛思想",并开发和实施了广泛的数据共享概念. 原则(NOMAD 2014,Wilkinson等,2016):数据对于任何感兴趣的人都是可以找到的;它们以易于访问的方式存储;其表示如下公认标准(Ghiringhelli等人2016,2017A); 并且所有规范都是开放的,因此数据可以互操作. 所有这些使数据可以用于可能与原始目的不同的研究问题. 因此数据是可重用的. 我们注意到,NOMAD CoE使用了可重复使用的术语,而在FAIR概念中则将其称为可重复使用的. 前者明确指出,数据不仅可以重复使用,甚至可以用于其他目的. 显然,FAIR数据也可用于人工智能工具.

本章的结构如下. 在Sect.2中,我们简要概述了材料科学的**四种研究范式的历史**,特别着重于第四种"大数据驱动的材料科学". 然后,第3节强调了广泛的数据共享对于科学和科学发展的重要性. 工程.第4节通过一些具体示例介绍了材料科学数据的人工智能概念. 最后,在Sect.5,我们对大数据驱动材料科学的发展和开放性问题进行了广泛的展望.

# 材料科学的四个研究范式
图1概述了材料科学方法和概念的历史演变. 我们记得实验研究可以追溯到石器时代,冶金的基本技术是在公元前六千年后期开始的铜和青铜时代发展起来的. 火势的控制促使实验取得重大突破. 到16世纪末,科学家开始通过方程式描述物理关系. 早期著名的名字是第谷-布拉赫(Tycho Brahe)(1546–1601),托马斯-哈里欧(Tomas Harriot)(约1560–1621),伽利略-伽利莱(Galileo Galilei)(1564–1642),约翰内斯-开普勒(Johannes Kepler)(1571–1630),艾萨克-牛顿(Isaac Newton)(1643–1727). 以及Gottfried Wilhelm Leibniz(1646–1716). 后两者还发展了数学微分和导数的概念. 因此,解析方程成为理论物理学的中心工具. 图1左数第二 ,这种新的思维方式-2.范式-用薛定方程表示. 不用说,第一个范式,经验科学和实验科学并没有过时,但是理论物理学代表了一种重要的补充方法论.

自50年代以来电子结构理论是由JC斯莱特推进了材料(例如,铺瓦工1937,1953,1965,1967 ;斯莱特和Johnson 1972),蒙特卡罗方法由Metropolis等人提出.(1953年)和桤Wainwright的(1958年,1962年,1970年),和拉赫曼(1964年)中引入分子动力学. Hohenberg和Kohn(1964)以及Kohn和Sham(1965)奠定了密度泛函理论(DFT)的基础(参见Li and Burke的章节2018))在1960年代中期. 所有这些发展一方面使基于计算机的研究和热力学和统计力学分析,另一方面使固体和液体的量子力学性能得以分析. 他们定义了计算材料科学的开端,即当今的第三范式. 因此,引入了"计算机实验",即模拟,其结果经常像实验研究那样被处理和分析. 最初是独立开发的,现在电子结构理论,统计力学和热力学领域正在共同发展(Reuter等,2005年及其参考文献). 同样,与DFT并行,开发了基于Green函数的多体技术(Hedin 1965),现在已与DFT协同插入,形成了包括激励在内的电子结构理论的前沿.

如今,大数据和人工智能彻底改变了我们生活的许多领域,材料科学也不例外(Gray 2007 ; Agrawal and Choudhary 2016). 吉姆-格雷(Jim Gray)可能首先讨论了第四种范式,明确指出大数据揭示了研究小数据集时无法看到的相关性和依赖性. 让我们通过注意材料科学(以及其他科学)的复杂性来概括"大数据"概念:表征或识别感兴趣的特性和功能的潜在但最初未知的描述性参数的数量可能非常大. 因此,机制的多样性和复杂性也代表了材料科学中的大数据问题. 与第二个范式的另一个重要区别是,我们接受许多材料属性(即大数据中的模式和相关性)不能用封闭的数学公式来描述,因为它们受几种方法控制,理论概念和多层次,复杂的过程交织在一起. 结果,这种模式代表并促进了知识,但不一定反映理解.

我们对开发本领域的第四种研究范式的愿景是什么?图2提供了一个示意图:不同化学和结构材料的空间实际上是无限的. 但是,当要求提供高性能的热障涂层时,只有几种合适的材料. 因此,就功能材料而言,由于大多数现有材料或未来合成材料都不相关,因此材料空间稀疏. 因此,相关数据是统计上例外的少数,如果忽略这种情况,那么对所有数据赋予相同重要性的统计分析很可能会出现问题. 寻找区域或图案对应于具有优越的功能性能的材料,需要的适当的描述符的识别,记为d1和d2在图2. 显然,通常,维数可能会高于仅两个. 此时,图2只是一个草图,对于大多数属性来说,合适的描述符是未知的,并且图案保持隐藏. 科学的挑战是找到合适的描述符. 让我们强调一下,由于实际相关的数据还不够大(在大多数情况下对材料科学感兴趣),因此使用尽可能多的领域知识不是至关重要的. 压缩感知,亚组发现和其他人工智能方法能够识别这些描述符和模式,我们将在Sect3中解决这些方法.

# 广泛的数据共享:为什么以及如何进行?

数据是本世纪的重要原材料. 多年来,我们的社区一直在通过CPU密集型计算来生成物料数据. 通常,结果存储在PC,工作站或本地计算机上,尽管信息内容可能很重要,但大多数数据并未使用,甚至经常被丢弃. 现在,该领域正在逐渐将其科学文化转变为开放科学和开放数据,并且这样做的原因很多. 数据的开放访问意味着任何人都可以使用数据,而不仅仅是开发或运行高级计算机代码的专家. 如果数据是公开可用的(并得到了很好的描述),则将有更多的人使用该数据,例如,计算机科学家,应用数学家,分析凝聚态科学家等. 当人们使用数据时可能会使用当前计算材料科学界甚至不知道的工具,我们会对人们将如何处理数据感到惊讶.

让我们举一个例子,互联网平台Kaggle的"众包"数据分析竞赛(Kaggle/NOMAD2018 2018). 这项比赛使用(AlxGayInz)2O3化合物(x+y+z=1为数据集)寻找新颖的透明和半导电材料. 这项挑战的目的是确定用于预测与光电应用相关的两个关键物理特性的最佳机器学习模型:电子带隙能和晶体形成能. 为2400个系统提供了这些目标特性,这些系统涵盖7个不同的晶体空间组和10到80个原子范围的晶胞. 六百套附加系统为比赛提供了测试设备.

比赛于2017年12月启动,两个月后完成时,来自世界各地的研究人员或研究团队采用了不同的方法提交了数量惊人的883个解决方案. 有趣的是,在最近的出版物(Sutton et al.2019)中总结的三种顶级方法采用了完全不同的描述符和回归模型. 例如,获胜的解决方案采用晶体图表示法,通过对各种长度(称为n克)的独特原子位点的连续序列进行计数,并将晶体结构转换为特征,并将其与核岭回归(KRR)相结合.这种n克方法对于材料科学来说是新的.

为了实现大数据驱动材料科学的预期成功,该领域显然需要合适的数据基础架构,以促进高效的收集,以元数据表示的数据描述以及数据共享. 对于计算材料科学领域,这是由NOMAD卓越中心(NOMAD)开发的,该中心还促进了全面的数据共享. 与其他主要关键数据库的协同关系,特别是AFLOW,Materials Project和OQMD,使其成为该领域最大的存储库. 上载的数据由持久性标识符(PID)标记,用户可以请求DOI(数字对象标识符)以使数据可引用. 数据下载不需要任何注册,请参考知识共享署名3.0许可.

那么,FAIR背后是什么?对计算材料科学意味着什么?

F代表Findable. 许多研究组织现在都要求公开研究数据并将其保存至少10年. 从实用的角度来看,避免工作量加倍,从而节省人力和计算资源以及能源也很有用. 由于个别研究人员在从工作站到计算集群再到高性能计算(HPC)中心的各种平台上创建数据,因此,在离开研究小组后的某个时间,通常无法找到学生或博士后的数据. 除了组织方面的问题外,问题还可能与HPC中心中的数据自动擦除,本地计算机上缺少权限,数据保护等有关. 显然,使数据可查找需要适当的数据基础结构,包括文档,元数据,搜索引擎和硬件.NOMAD,NOMAD2014).

A表示Accessible. 材料科学中的可访问性具有不同的方面. 首先,我们不应忘记允许快速访问数据的适当硬件. 其次,我们需要提供应用程序编程接口(API). 为了使数据完全可访问需要一个重要的附加步骤,即数据的形式描述,即也考虑元数据相互关系的元数据. 这将公平地连接到I.

I代表 Interoperable(互操作). 在这里,我们首先要考虑计算材料数据的极端异质性. 广泛的社区正在使用大约40种不同的主要计算机代码(考虑材料的电子结构,分子动力学和量子化学软件包),这些方法在方法和实现的各个方面都不同. 因此,有必要使它们的结果具有可比性,这不仅是在需要将它们转换为通用格式和通用单位的意义上,也是一个重大挑战. 我们还记得,在不同的(子)社区中,一个量的名称可能不同,或者一个相同的表达式在一个或另一个区域中可能具有不同的含义. 因此,需要"词典"在它们之间进行翻译. 显然,如果将所有内容限制为一个计算机程序包,无需翻译或转换. 但是,那么将丢失大量可用数据,甚至更重要的是社区. 从这个意义上讲,NOMAD概念是通用的,实际上与基本上集中于一种计算机代码的所有其他数据存储库正交(请参阅Draxl和Scheffler2019).

尽管如此,我们仍然需要问一问是否可以一种有意义的方式处理所有可用数据. 除了格式和单位以外,还有更严格的限制可能会妨碍这项重要的工作. 这些涉及用于不同计算的计算参数,因此也涉及数值精度. 我们在这里回想起,为了进行广泛共享,在提供代码和代码版本时,所有打开的数据都可以视为有效,因为输出文件以可再现的方式对应于提供的输入文件. 尽管如此,已经出于不同目的创建了数据,这些数据可能需要在基础集大小等方面进行不同程度的收敛(例如,确定原子几何形状可能需要比电子能带结构中的细节更严格的参数(另请参见图的讨论.3和 4)). 不仅如此,我们甚至可能会问,针对"安全"设置的一个相同问题的解决方案的不同代码是否给出相同的结果. 对于后者,我们指出了由Stefaan Cottenier(Lejaeghere等人,2016年)领导的社区工作,其中使用许多不同的从头算电子结构包计算了71种元素固体的状态方程. 事实证明,经过几年的时间,对代码,基集,尤其是伪电位的改进,所有代码都能得出基本相同的答案. 实际上,在 https://molmod.ugent.be/deltacodesdft 的列表中,必须选择一个参考,显然全电子代码是自然的选择,而全电子封装WIEN2k,激振和FHI- ams的精度实际上是相同的. 这样的研究非常有帮助,并为建立标杆文化奠定了基础. 标杆文化早在数十年前就已经在量子化学中为分子所确立. Lejaeghere等人的研究.(2016年)只是开始. 显然,其他属性(如能垒,带隙,光谱等)和系统(如点缺陷,表面,界面以及无机/有机杂化材料等)的容忍度要比简单的散装固体和会使差异更加明显.因此,正在沿着这些方向进行更多的研究.

尽管只能使用代表完全收敛的参数集进行上述比较(Lejaeghere等人,2016),但日常生活中进行的计算通常离此理想情况还很远,实际上通常是足够的. 显然,这种情况引发了一个问题,即如何比较和操作已在不同设置下执行的计算,例如,在基础集,布里渊区域积分的网格等方面. 在下面,表明实际上是可行的.

让我们展示的是完全融合的结果复杂的材料可以通过从计算的误差学习来估算简单的材料(Carbogno等人.2019). 这项研究采用了四种不同的代码:两种非常不同的全电子代码和两种投影仪增强波/平面波代码. 这些令人兴奋(Gulans等人,2014年)和FHI-aims(Blum等人,2009年),GPAW(Enkovaara等人,2010年)和VASP(Kresse和Furthmüller,1996年). 出版物和/或各种数据存储库中不同计算差异的来源是不完整的基集,对k的近似处理空间集成,伪电位的使用等. 由于不完整的基集确实是最严重的问题,图3显示了固定几何形状的总能量如何随Lejaeghere等人采用的71种元素固体的基集质量而变化.(2016年).图3中的红色符号表示显示最大误差的材料,黑色符号表示所有材料的平均值. 对于每种材料,相对于使用Lejaeghere等人使用的设置甚至更精确的设置获得的完全收敛值,定义了误差ΔE.(2016年). 在所有情况下,误差都会从小基数组的1eV数量级下降到高质量基数组的meV精度,逐渐降低. 不是所有的计算机代码显示了小基组的大小单调的行为其实反映了不同的方法和实施的具体特点(见Carbogno等人2019详细讨论). 基于这些元素固体的误差,作者使用一个简单的分析模型,即通过线性组合以相同设置获得的各成分(元素固体)的各个误差,估算了多组分系统中基集不完备性引起的误差:

$\overline{{\Delta E}_{\mathrm{tot}}}=\frac{1}{N}\ \sum_N{N}_i{\Delta E}_{\mathrm{tot}}^i \tag{1}$

在这里,Ni是存在于该化合物中的物种i的原子数,${\Delta E}_{\mathrm{tot}}^i$ 是该物种在相应元素固体中的误差. 该模型适用于63种二元固体,这些二元固体被选择为覆盖化学空间(原子序数为71的每个元素1个,不含稀有气体). 通过将估计的误差与这些二进制文件的相应DFT计算的误差进行比较来验证该方法,如图4所示. 这项工作的作者发现所有四种代码的预测错误与实际错误之间有明确的关联(具体取决于特定代码的特殊性). 为了进行更深入的分析,包括相对能量和三元合金,以及对所采用方法/基组的作用进行广泛讨论,我们请读者参考Carbogno等人的文章.(2019). 从本质上讲,可以得出这样的**结论:即使在未收敛的基础集合中计算总能量时,也可以推断出完全收敛情况的能量**. 这项研究为`满足I的重要条件的来自不同来源的数据奠定了基础`.

# 材料科学数据的人工智能概念

我们将"人工智能(AI)"用作"从经验中学习"的计算方法的总称. 如上所述,对于材料科学而言,致动机构的复杂性很大,但是相关数据的数量通常在下侧. 张-凌(2018)最近通过模型的自由度和预测精度解决了这个问题. 在类似的意义上,我们认为,对数据进行适当的分析至少需要考虑驱动感兴趣相关性的因果关系的某些方面,即,为了获得可信赖的描述,需要在学习过程中包括领域知识,可解释性,甚至可能更深入地了解数据中的结构或模式背后的原因. 具体而言,我们注意到,从实际上无限数量的可能材料中,对于特定目的而言,可能只有10或100种可能相关. 标准的机器学习工具解决了人群或大多数人群的描述问题,并优化了总体均方根误差等,并引入了正则化方法以避免过度拟合和/或获得流畅的描述. 结果,很可能隐藏了统计上例外的数据. 但是,只有这些可能携带我们感兴趣的信息. 所采用的领域知识越少,学习过程所需的数据量就越大,可能会发生数据拟合但没有预测甚至插值的情况. 可靠. AI是一个广泛而跨学科的领域,机器学习(从数据中学习)和压缩感测(源自信号压缩;主要是为了识别低维描述符)是重要的子领域.

让我们添加另一个有关材料科学大数据的警告. 可能使用的材料数量实际上是无限的,我们希望找出性能或功能比今天使用的材料更好的新材料. 显然,材料科学中的可用数据量越来越大,尽管迄今为止合成的无机材料很少(约250,000种),我们通常只知道原子结构,而几乎不了解其电子,弹性或其他特性. 但是,获取越来越多的数据并不意味着所有数据都与所有感兴趣的属性相关. 材料科学显示出高度的多样性,即它受大量不同特性和机制的支配,经验似乎表明,最后,对某些功能组有用的材料数量很少. 例如,如果我们要求具有高导热性和耐刮擦性的高度透明材料,可能没有什么比金刚石更好,甚至没有金刚石. 另一个例子是Singh等人最近的研究.(2019)研究了68,860种用于光催化还原CO2的候选材料. 事实证明,其中只有52个可能相关. 通常,很可能在少数"高性能材料"中,一种机制是活动的(或不活动的),而该机制与大多数材料无关. 因此,可用数据量可能很大,但相关数据的数量数据,即包含所需信息的数据很小. 简而言之,在材料科学和工程学中,我们经常寻找"干草堆中的针",即,从某种角度来说,对于极少数统计上异常的系统. 用单个全局模型拟合所有数据(即干草)可以平均消除少数族裔(即针头)的特长. 因此,我们需要立即识别大量数据中相关的,可能很小的,统计上例外的子组的方法.

$\operatorname{Min}\ \left\{\ {\sum}_{i=1}^N\left({\hat{P}}_i-P\!\left({\boldsymbol{d}}_i\right)\right)^2-\lambda \left|\left|m\right|\right|\Big\}\right\} \tag{4}$

## 子组发现
如上所述,处理整个数据集的定量描述的全局模型可能难以解释. 对于材料研究中的许多要求,识别和描述子组的局部模型将是有利的. 为了便于说明(请参见图 6),全局最优回归模型可以预测数据之间的负相关关系(图  6). 但是,子组发现分析可能会发现存在两个统计上例外的数据组(在图的右半部分由蓝色和红色表示). 因此,数据集中的关系不具有负斜率(全局模型),而是具有正斜率(两个子组). 作为物理实例,元素周期表中的过渡金属是一个亚组,而the系元素,镧系元素和卤素是其他亚组. 因此,亚组的识别有助于了解材料族之间的相似性和差异.

![fig6](https://media.springernature.com/original/springer-static/image/chp%3A10.1007%2F978-3-319-42913-7_104-1/MediaObjects/116042_0_En_104-1_Fig6_HTML.png)
图6: 左边的数据点和一个全局回归机器学习模型预测了材料特性y 1和y 2之间的负相关关系. 正确,亚组发现确定了统计上标记为红色和蓝色的特殊区域,这两个区域的机器学习呈现出正斜率. 亚组发现旨在通过布尔选择器函数,描述这样的亚群(σ 1和σ 2的基本选择的)定义为连词(逻辑AND,表示为∧)(该C我)

亚群发现(SGD)的概念是在1990年代初引入的,当时大型数据库的出现推动了探索性和描述性分析工具的发展,作为对全球建模的可解释补充(Duivesteijn等人,2016;Klösgen,1996; Atzmueller,2015;Azmueller,2015). Herrera等,2011; Siebes 1995; Wrobel 1997; Friedman and Fisher 1999).

简而言之,子组的标识建立在三个组件上:
1. 使用描述语言来标识给定数据池中的子种群.这些是布尔表达式,例如"原子A的电离势减去原子B的电离势小于X ",其中X是可以迭代选择的数字.这些表达式称为选择器.

2. 效用函数的定义,这些函数正式化了子种群的趣味性(质量).这可能包括以下要求:"材料的带隙在1.1和1.5 eV之间"和"内聚能大于3 eV".

3. 搜索算法的设计,用于找到描述感兴趣的子种群的选择器(例如,Goldsmith等人,2017年).

# 展望

计算材料科学在Moruzzi等人的早期工作中取得了令人瞩目的成就.(1978)关于金属的各种特性,以及Cohen和他的同事(Yin和Cohen 1982)关于硅和锗的内聚力和相变的观点(请参见本手册中M. Cohen的章节,特别是图  4 Cohen 2018).). 从那时起,已经开发出许多用于求解Kohn-Sham方程的计算机代码,最初涉及近似值,例如伪势(去除核电子,产生平滑势)或引入接触或略微重叠的原子中心球体,在该球体中势化了球形.在1980年代,原始的伪电势方法取得了重大进展(请参见Vanderbilt和同事的工作:Garrity等人,2014年及其中的参考文献),并且已经开发了不需要电子形状近似的全电子代码. 电位(例如Wimmer等人1981 ; Blaha等人1990 ; Gulans等人2014 ; Blum等人2009).Lejaeghere等人的工作.(2016年)对元素散装固体的现代电子结构规范的精度进行了很好的概述,还演示了如何使社区参与. 显然,这种工作很重要,需要扩展到更复杂的结构和组成,缺陷,表面和界面. 在这个方向上的工作正在进行中,如可用于先进的电子结构的方法的研究,如GW方法的(van Setten等人2015). 此外,该领域迫切需要各种数值逼近和交换相关电位的基准,以便不仅要解决数值精度问题,还要解决准确性问题. MSE(材料科学与工程)项目(Zhang et al.2019)是朝着这个方向迈出的有希望的一步. 没有所有这些,数据驱动科学的能力将受到限制.

目前,计算材料科学被第三种研究范式所主导(参见图  1),但近年来AI方法的进步非常显着,第四种范式正发挥着越来越大的作用. 尽管如此,目前在AI可以做的事情上,宣传比现实主义要多. 其中大部分与以下事实有关,即相关或独立数据的数量通常并不大. 当人们对大趋势感兴趣并且需要适应和预测"一大群人的行为"时,机器学习技术可以提供很多帮助(例如,参见Kaggle竞赛中用于预测透明导体特性的方法) .2019). 但是,材料科学和工程的明智需求通常是不同的:我们通常不是在寻找人群行为,而是在寻找在某些功能或特性(通常甚至是多个特性的组合)方面具有非凡性能的材料. 在大量可能的材料(已经存在的材料以及将来将要合成的材料)中,通常只有少数几种统计学上非常合适的合适材料. 但是,在许多情况下,我们不知道如何识别这种可能有趣的功能材料的"类别". 我们如何区分哪些数据/材料相关,哪些不相关? 从成千上万的不相关数据中了解不到0.01%的相关材料显然是有问题的,而且标准方法,它优化了正规化的均方根误差,甚至强调了无关数据的重要性,同时超越了特殊情况.如果可以将数据分为"多数类"和"少数类",则已经开发出了解决该问题的方法(Chawla等人.2002年及其参考文献).但是,这些类别通常是未知的,因此迫切需要对材料科学领域的亚组发现概念进行改进.

目前缺少什么?让我们列出一些问题:
材料性能预测与稳定性分析以及合成路线的预测紧密结合.

高通量研究亚稳材料和这些亚稳状态的寿命.

实际条件下的材料(T,p和反应性环境):稳定性和性能.这非常涉及具有鲁棒,受错误控制的链接的多尺度建模,这些链接具有各种模拟方法之间的不确定性知识.过去经常提到这一点,但至今仍未完全实现.

根据特定属性(结构,弹性和电子属性等)的数值近似值(基集,伪势等),计算的误差估计.

超出标准DFT的计算,例如,用于固体计算的耦合聚类方法(可能还准备用于量子计算机).

完整说明对数据异质性的科学结果,即改善和补充当前的元数据定义.虽然显著已经取得了进展朝向计算数据的变换从许多计算机代码和对应的元数据的发展(Ghringhelli等人2016,2017A ; 元信息在NOMAD),其优点将只充分变得明显时,同样会一直还获得了实验数据.后一个挑战甚至比第一个更大.实验研究中使用的样本材料对应于计算的输入文件;实验条件(T,p,环境)和实验设备中的计算机代码.一个尚未完全解决的挑战是样品材料的完整表征.实际上,这通常是不可能的.显然,与元数据的定义紧密相关的是样品制备和实验运行中工作流程的描述.

该领域正在开发第四范式的方法.连接范例1,2,3和4的学习曲线显然越来越陡峭. 因此,即使到目前为止,第四个范例还没有很好地开发,下一个范例也可能接近.下一个范例是什么?考虑到"未来已经来临–分布不均"(Gibson,1999),今天可能已经暗示了这一点.我们猜测这可能是虚拟现实,可以直接与新计算进行即时连接,也可以是理论(预测和分析)与实验的结合.我们面前还有激动人心的时刻.
