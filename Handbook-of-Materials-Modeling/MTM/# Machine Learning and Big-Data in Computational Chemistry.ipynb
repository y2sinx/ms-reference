# 简介
几十年来,化学一直是一个大数据企业. 在该术语推广之前很久,化学家们一直在努力使用化学数据集,其数量,种类和速度超过了他们可用工具的能力. 首先,化学家们一直在努力解决化学空间的巨大问题. 自计算机问世以来,化学家一直是数字技术的创造者和早期采用者,用于化学数据的存储,处理和检索. 同样,互联网的出现引发了加速化学科学知识创造和共享的新时代.

在诸如消费者分析,广告,运输,医疗保健或金融等领域,大数据和机器学习方法的最明显收益来自于尽可能多的有利特征的子区域:数据生成成本低廉,数字化起源或容易数字化,高度组织化,一致化,极其丰富. 虽然丰富的化学数据往往是昂贵的,生成,异构,手工制作和稀疏. 化学数据输入和管理通常是手动完成的; 由于异构的来源和报告标准以及化学报告的技术复杂性,这些任务需要大量的专业知识. 近年来,随着诸如显微镜或光谱仪的化学仪器产生数字化数据,可以由单个装置生成非常大的数据集. 本地大数据,但在尝试合并来自不同组或设备的结果时仍然面临类似的障碍.

计算化学方法具有一系列特征,使得它们非常适合这种数据驱动的范例,并且被证明是实现化学中大数据愿景的完美测试基础. 计算化学软件的输出从头到尾进行数字化. 即使它依赖于自定义格式的输入和输出文件,自动创建和处理这些文件也是一项简单的任务. 当使用相同的计算机代码时,计算结果可以复制到浮点精度,并且即使在不同的程序中也是如此,只要它们使用相同的基础算法. 因为计算本质上是一种商品,所以结果可以通过成本线性缩放生成任意数量 - 或者甚至是次线性的,这要归功于机器学习方法.最后.

在本章中,我们将分析计算化学如何包含大数据以加速发现急需的新分子,以及这些新生成的数据集如何允许应用和开发机器学习技术,使其更加强大.

# 化学数据库
化学中大数据方法的最初发展历来侧重于巩固,策划和索引文献和经验数据. 解决这些任务是计算机在化学中的首批应用之一,并且仍然是开发新的化学计算工具的关键驱动因素.

化学文摘社(CAS)创建于1907年,用于保存化学文献和报告物质的完整记录,遵循欧洲先前的摘要期刊的理念,如Chemisches Zentralblatt和19世纪的编辑,如Beilsteins Handbuch der organischen Chemie,Gmelins Handbuch der anorganischen Chemie,或默克索引. 它在第一年包含超过12,000个索引,在1937年创下了百万分之一的记录. CAS历史上依靠成千上万的志愿者为每篇论文创建摘要和索引信息 - 而不是今天在亚马逊的机械土耳其人那么远 - 并逐渐将它们分开六十年代末和九十年代. 此外,到20世纪50年代,CAS雇用了数百名员工,强调了大数据对化学的巨大价值以及将相同数据整齐地划分为可用形式的巨大成本.

20世纪60年代后期,化学应用数字计算机的应用通过促进化学数据库的存储,进入和非常重要的查询,开始了化学大数据的数字化. 基于CAS计算机的化学品登记系统,就像之前的Beilstein登记号码一样,为化学品分配独特的个人任意数字. 它于1964年在内部首次亮相,并逐步更新和扩展,为越来越现代的查询和检索工具奠定了基础,使用命令行和图形界面从本地或在线记录中检索数据. 在七十年代中期,记录显示每年约有三百万条记录并且每年增加数十万条,登记册在2007年呈指数增长,达到第3000万种化学物质,2009年增长了5000万,2015年增长了1亿.

1965年,剑桥大学的Kennard小组开始收集已发表的小分子晶体学数据. 这项工作已发展成为所谓的剑桥结构数据库,该数据库现在拥有近100万个来自X射线和中子衍射分析的精选项目(Groom等人,2016年). 同样在20世纪60年代后期,布鲁克海文国家实验室和得克萨斯A&M大学的共同努力启动了蛋白质数据库,这是一个大型生物分子(如蛋白质和核酸)的三维结构数据库,现在拥有超过120,000个条目(Berman等人2000 ; Bernstein等人1978 ; Meyer 1997).

实验晶信息的其他储存库(布鲁诺等人2017年)是晶体开放式数据库(COD)(Gražulis等人2009,2012),其存储用户提交的数据; 无机晶体结构数据库,由FIZ Karlsruhe和NIST合作生产,用于存储无机固体的晶体结构(Belsky等,2002); 和存储粉末衍射图谱的国际衍射数据中心(ICDD)(Faber等,2002).

互联网的兴起引发了其他大数据存储库的推出,以及现有化学和材料在线可用性的扩展. 研究数据库登记处汇总了包括化学和材料科学在内的许多科学学科的多个储存库和数据集的列表(Pampel等,2013).

美国国家标准与技术研究所自1996年以来一直主持NIST化学WebBook(Linstrom和Mallard 2001),这是一本主要是光谱和热力学数据的汇编,最初是在手册和表格中编写的. 此外,近200个标准参考数据集包括化学家感兴趣的其他数据. NIST 101(III 1999)特别包含约1800个气相原子和小分子的计算数据量子化学数据和用于比较实验和计算理想气体热化学性质的工具.

由于专利和专利申请是开放的,它们是一个大数据,信息丰富的资产. 然而,专利文献倾向于仅基于文本或基于图像,并且将它们数字化为可用于数据驱动方法的格式是一个长期挑战. 欧洲,美国,日本专利局和世界知识产权组织可以在线访问和查询,但没有系统地注释. 美国专利商标局通过其复杂的工作单位计划,在专利申请中提供了分子的数字表示,从而简化了数据处理管道.

在开放数据领域,NextMoveSoftware发布了近百万种化学反应,通过自动文本挖掘提取专利中报告的相关实验部分,涵盖1976年至2013年期间. SCRIPDB包含策划合成,化学品和反应来自从CWU文件收集的专利文献来自授予的美国专利(Heifets和Jurisica 2012).

SureChem是另一个分子数据库,于2012年开放给SureChemBL. 它由IBM的战略IP洞察平台使用,最初从大约470万个专利(1976-2000,仅文本)中提取了200多万个化学结构,随后扩展到包括截至2010年底发布的专利,并且化学结构还得到了来自美国CWU和图像(2001-2010)(Papadatos等,2016). UniChem试图统一并交叉引用这些多个数据库(Chambers等,2013).

ChemSpider是一个化学品数据库,拥有来自480多个数据源的6000多万个条目. 它最初是一家私营企业,于2009年被皇家化学学会(RSC)收购. 它包括一个众包组件,但只有有限的下载量(Pence和Williams 2010).

由于具有生物活性的可能的小有机分子的大尺寸,生物靶标的大量和多样性,许多生物测定和测量的高维度以及医疗应用,化学信息学,生物化学和药物的高价值发现应用程序是化学中大数据应用的关键目标. 药物和化学品的毒理学信息也是有趣的生物相互作用,记录在几个公开的数据库中.

国家生物技术信息中心(NCBI)是美国国立卫生研究院(NIH)的美国国家医学图书馆(NLM)的一部分,它拥有生物技术领域的多个数据集和信息学工具. 对化学特别感兴趣的是PubChem,这是一个分子数据库及其针对生物分析的活动. 该项目于2004年启动,现已汇集了540多种来源,超过9亿种化合物,其中近250万种具有2.33亿种生物活性结果. 作为联邦政府资助的服务,
PubChem被认为与CAS(Kaiser 2005)等营利性存储库发生冲突.

ChemBL是一个开放式数据库,包含针对药物样生物活性化合物的结合,功能和ADMET(吸收,分布,代谢,排泄)测量,对超过100万种化合物和11,500种蛋白质靶标进行了近1500万次生物活性测量(Gaulton等)人2012).DrugBank是一种网络可访问的化学信息学数据库和服务,结合了药物分子的结构和生物学目标数据.该数据库包含超过9000种小分子药物,3000 FDA批准的药物,以及数据近17000药物靶协会(威沙特等人.2006年,2018).化学实体生物学兴趣数据库也遵循类似的重点(Degtyarenko等人,2008).

PDBbind数据库匹配来自文献的已发表的亲和常数,其配体 - 蛋白质系统的3D结构存储在PDB中,经历定期更新(Liu等人,2015 ; Wang等人,2004).BindingDB(Chen等人2001 ; Gilson等人2016)和AffinDB(Block等人2006)也旨在完成类似的任务.比较毒物基因组学数据库是一个基于志愿者的基因组资源,致力于毒理学相关基因和蛋白质及其与化学物质和毒素的相互作用(Davis等人2017 ; Mattingly等人2003年)).MACiE含有酶反应机制,侧重于酶催化机制的进化和化学机理的分类(Holliday等,2005).

这些服务是化学领域的大数据先驱,突出了早期对化学如何成为高维,稀疏数据领域的理解,大数据方法可以创造巨大价值,以及收集和管理化学数据的高成本.对于机器学习应用程序,这些存储库提供极其有价值的标记数据.

# 大计算化学
计算化学领域很快就跟随了第一台数字计算机的部署,并且发展迅速,与算法和硬件的发展相匹配,并且也有自己的贡献. 通常情况下,执行的化学计算将推动可用的硬件运行:使用大量内存或存储在超级计算机中的多个核心上并行运行.

从这个意义上说,从它们产生的输出数据的数量和速度来判断,以及硬件要求的压力,计算化学计算一直是大数据,基本上与可用资源允许的一样大,并且是大的驱动因素 - 数据技术.

在本章中,我们将重点关注计算化学的应用,这些计算化学结合了体积和速度,并且由于具有很大的粒度,因此也具有很高的多样性. 这些是(i)分布式计算的情况,其中小的计算有效载荷被分配到大型小型计算机网格,并且在计算之后,结果被合并和处理,以及(ii)高吞吐量虚拟筛选,其中数千个候选分子或材料以自动化方式单独计算,并进行数据驱动的最高性能候选材料搜索.

## 计算化学中的自动化和数据库
已经开发了许多工具来自动化计算化学数据的创建,提交,转移,处理,解析,存储和查询. 最基本的网络示例之一是 Basis Set Exchange,最初在环境分子科学实验室组装,其中大多数元素周期表的无数精选基础集可以多种格式下载(Feller 1996 ; Schuchardt et人.2007).

存在近六个平台,其具有类似的理念,以实现以下部分或全部:材料科学自动化和固态电子结构计算; 数据处理和这些计算的分析; 以及用于虚拟发现和机器学习目的的这些计算输出的集中式,Web可访问的存储库. 多种解决方案提供了一些重叠的功能,通常在材料领域,并且最近已经过审查(Lin 2015).

**电子结构项目/The Electronic Structure Project**)(Klintenberg等,2002 ; Ortiz等,2009)利用ICSD的结构数据来筛选新型无机材料.

**计算材料库/The Computational Materials Repository**(兰迪斯等人.2012)提出了一种用于计算机驱动的材料设计集成软件解决方案. 它是量子材料信息学项目生态系统的一部分,该项目还包括原子模拟环境,这是一个用于原子模拟的python库(Hjorth Larsen等人2017).

**AFLOW**是高通量材料发现的自动框架(Calderon等人2015 ; Curtarolo等人2012a),匹配库Aflowlib.org主持并向公众提供这些计算结果(Curtarolo等人2012b) . 它现在包含1,748,704种物质化合物 - 计算出的物质超过173,121,696.

**pymatgen/The Python Materials Genomics** Python的材料基因组学(pymatgen)是用于固态DFT计算的分析材料(Ong等.一个开源Python库2013),其还具有匹配材料中的应用程序编程接口(Ong等.2015年)进行交互材料项目是一个大型材料计算数据库(Jain等,2013).该生态系统中的其他计算工具包括用于错误处理的Custodian,用于工作流管理的Fireworks等工具. 材料项目工具和数据已在100多篇已发表的论文中使用,近200篇由其创作者使用.

**开放式量子材料数据库/The Open Quantum Materials Database**(OQMD)是一个完全开放的项目,可以对来自ICSD的化合物以及假设化合物进行400,000多次DFT能量计算,可能发现有效但尚未合成的化学物质(Kirklin等人,2015).

**AiiDA**(Pizzi等人,2016)是一个灵活且可扩展的信息学基础设施,用于模拟,数据和工作流程,重点关注材料的平面波DFT计算和对数据来源的重视(Merkys等人2017).

**新材料发现存储库/The Novel Materials Discovery repository** 的建立是为了以与管道无关的方式托管,组织和共享材料数据(Goldsmith 2016),并通过各种计算机代码托管超过4400万开放访问用户提交的总能量计算.

**ioChem-BD平台/The ioChem-BD Platform** 提供了类似的解决方案,重点是分子数据(Álvarez-Moreno等人,2015). 几个工具包解决了缺陷的模拟和作用,例如MAST(Mayeshiba等人2017),PyDII(Ding等人2015)和其他(Goyal等人2017). PyChemia是一个用于自动化原子模拟的python库,侧重于一些DFT代码和数据挖掘功能的材料和接口.

计算化学结果的多个数据集纯粹旨在产生不同的训练数据. **QM9**含有B3LYP/6-31G(2df,p)结果,用于由CHONF组成的134k稳定小有机分子,包括谐波频率(Ramakrishnan等,2014). **PubChemQC** 最近尝试为机器学习方法创建训练数据,该方法基于 B3LYP/6-31G* 水平的密度泛函理论(DFT)和超过2个的10个最低激发态计算了300万个分子的基态电子结构在TD-DFT / B3LYP/6-31+G* 理论水平上的百万分子(Nakata和Shimazaki 2017). **ANI-1**包含2000万个构象的能量和DFT水平特性,超过50,000个沿正常模式扭曲的小有机分子(Smith等.2017). 也可获得从头算分子动力学:在理论的PBE水平上,500 K时的5,000帧,对于C 7 O 2 H 10的 113种结构异构体和8种小有机分子的数十万帧(Chmiela等人 2017 ;Schütt等人. 2017年).

大量用于数据处理的内部工具已发展成可用于计算输出的数据分析的开放库: 基于python的**cclib**,用于解析和解释计算化学包的结果(O'boyle等人,2008); **ESTEST**,量子化学计算产出的比较,验证和共享的自由框架(Yuan和Gygi 2010); **ORBKIT**,也是基于python的库,用于对来自多个代码的量子化学波函数输出进行后处理; (Hermann等人,2016年); 和**PyGlobal**,对于DFT计算面向电子表格的输出后处理工具(Nath等.2016). 其他工具也可用于输入生成和流水线计算,例如**JACOB**(Waller等人2013)用于企业应用的计算化学框架,**PyADF**用于使用ADF包编写多尺度量子化学(Jacob等人2011).

这种充足的景观表明,增加量子化学计算的速度和体积是非常有趣的. 然而,这些增加伴随着各种应用的折衷. 由于增加的复杂性,其中大多数很少被视为在创建它们的团体或财团之外采用(Thygesen和Jacobsen 2016).

## 高吞吐量虚拟筛选
随着计算方法变得更加准确并且计算硬件更加经济实惠,在合成之前自动预筛选化合物的可能性变得更有希望. 各个团队已经使用HTVS发现了许多无机材料. 这些在本书的其他章节中进行了深入的处理. 在这里,重点将放在有机光电子领域的有机分子材料上,用于光能互换(光伏和发光二极管)和电能储存(Pyzer-Knapp等,2015).

### 光电子学
清除能源项目(CEP)包括志愿者,分配计算工作和HTVS愿景的概念. 自2006年左右开始以及整个两个阶段开始,该项目在 IBM World Community Grid上运行,志愿者将计算机时间捐赠给虚拟屏幕p型和后来的n型有机光伏低聚物.

CEP解决了化学化合物自动化虚拟测试的许多挑战:候选分子的程序化生成,量子化学计算的自动化,数据存储和分析.

通过活性位点的共价连接组合20个片段的池,并且还通过形成与这些标记的反应性原子相邻的稠合环,处理的CEP分子产生上升至四聚体并产生超过200万个候选物,其被彻底筛选. 进行的量子化学计算包括多个DFT功能,相当于迄今为止最大的计算化学项目. 使用Scharber模型评估候选四聚体的光伏效率,假设富勒烯作为电子受体以及原始和经验校准的供体能量水平. 大数据集的统计分析产生,得到了最高和最低有前途的片段和片段的组合的相关性(Hachmann等人2011,2014; Olivares-Amaya等.2011).

由于理论数据集较大,并且OPV领域存在大量独立实验结果,CEP已经证明是材料发现新方法的试验场. 使用来自文献的一组实验结果(Lopez等人2016)改进理论气相结果和捕获体积效应的基于概率核的校准方案改善了DFT计算的预测性能(Pyzer-Knapp等人.2016).

后来的子项目专注于从100多种常见有机部分的组合中虚拟筛选超过50,000个非富勒烯电子受体. 对选出的先导化合物也进行了时间密度泛函理论计算.二酮吡咯并吡咯和醌型噻吩衍生物显示出良好的前景,并被提议进行额外的研究(Lopez等人,2017).

在较不庞大的规模上,其他工作也涉及对体相异质结太阳能电池组件的筛选,例如超过780种不同共聚物供体材料的组合带隙设计策略(Shin等人,2014).

一些研究已经解决了其他类太阳能电池的分子虚拟发现,例如基于TiO 2的染料敏化太阳能电池,优于普通染料(Martsinovich和Troisi 2011)以及卟啉(Ørnsø等人,2014).

### 电解质和储能
能源存储是材料科学和工程领域最活跃的领域之一,因为对轻型,高能量密度应用(如移动电话或运输和静态,低成本,网格级存储)的需求强劲.

液流电池是大型静电电池,其中液体电解质储存在罐中并在充电或放电时在电化学电池中循环. 尽管与诸如锂离子的固态电池相比它们显示出稍差的能量密度,但是由于较低的成本和使用地球丰富且廉价的电解质材料,液流电池可能是电网规模电存储的更好的解决方案. 功率的独立缩放(取决于诸如电化学反应的动力学和电极表面积的影响)和能量(与罐的总尺寸,溶液中电解质的浓度和电离的程度有关).电解液为流动电池解决方案提供更灵活的工程设计. 使用水作为溶剂进一步降低了安装成本并提高了操作安全性. 因为它们可以来自石油和可持续生物物质,有机分子的巨大设计灵活性,以及它们有前途的性能,液体电池的有机电解质被认为是竞争对手(Kowalski等.2016年 ; Leung等.2017年 ; 魏等人.2017).

高通量模拟,特别是与实验后续相结合,已成为发现有机液流电池电解质材料的有力工具. 使用蒽醌二磺酸盐和溴/氢溴酸的酸性水溶液的高效,无金属,低成本流动电池设计的报告利用了数千个潜在醌分子的DFT计算,以优化negolyte及其水溶性的还原潜力(Huskinson)等人,2014). 允许通过苯并的域来识别有用的设计规则和结构-性能关系,naphto-理论预测,和蒽醌轴承13个的任何官能团的(的仔细分析铒等人2015年). 生物激发的苯醌醌衍生物的其他大量计算也确定了用于液流电池的潜在有用的电解质(Pineda Flores等人,2015).

电解质基因组项目中使用的材料项目骨干上的一组从喹喔啉和蒽醌,thiane,噻吩衍生大多近5000分子进行高通量的计算,和联吡啶衍生物(Qu等人.2015年). 特别注意错误处理和氧化还原电位的估计,离子对解离和复盐形成. 还报告了近1400种喹喔啉的非水流电池估算氧化还原电位,溶剂化能量和结构变化的详细分析(Cheng等人2015),以及4000多个具有一个或两个官能团的五元或六元环附件(Pelzer等人,2017年).

还报告了一些规模较小的虚拟搜索. 从在PubChem数据库可用实验报道分子开始,和下选择手动至315部署神经网络之后,提高观察到喹喔啉衍生物(Park等人的实验性能2015,2016).

用于电化学应用的有机分子的化学稳定性影响电池的保质期和长期可循环性,是虚拟发现应用的常见问题之一. 为此,已经使用DFT分析了锂离子电池的有机溶剂的第一次和第二次还原和氧化(Borodin等人,2015). 类似地,生物启发的咯嗪电解质已经通过实验证明,通过在虚拟搜索中包括稳定性标准,除了氧化还原和溶解度要求之外,在碱性含水液流电池中提供优异的稳定性(Lin等人,2016).

已经报道了较少关注自动化和极高通量的方法,侧重于诸如对32种有机分子的低Li +结合亲和力(Park等人2011),15种镁电解质的抗衡离子(Qu等人2017),氧化还原- 可转换的聚合物基膜(Ward等人,2017),或调整有机分子的还原潜力,以优化小氧化还原活性分子与导电聚合物的组合(Araujo等人,2017).

HTVS还应用于与其他电化学应用相关的材料,如双层电容器(Schütter等人2016)或其他绿色化学应用,如可切换亲水性溶剂(R. Vanderveen等人,2015).

# 应用机器学习加速发现
与许多其他领域的情况一样,化学机器学习应用并不是一种新型事业(Pierce and Hohne 1986). 与许多其他应用中的情况一样,深度学习算法与较大数据集和专用计算硬件的组合已经导致机器学习的许多有效应用用于化学.

## 定量结构 - 属性关系
机器学习到计算化学应用的大多数应用涉及在三维原子排列或化学图上建立快速统计代理到昂贵的计算. 在某种意义上,这是建立在分子结构或3D几何上训练的定量结构 - 性质关系,通常使用分子力学在运行中生成.

同一出版物中的其他章节对最近对该领域的许多贡献进行了全面回顾,因此仅提供简要概述. 我们鼓励读者在同一本书中进一步探索.

使用神经网络模型在潜在能量表面上进行插值的想法可以追溯到20世纪90年代(Blank等人1995). 将机器学习应用于计算化学的早期当代作品是Behler的那些,其中神经网络用于分子动力学模拟和其他PES探索技术(Behler 2011b,a ; Behler等人2007 ; Behler和Parrinello 2007). 神经网络的这些和许多其他应用,主要是批量和异质系统,最近已经被审查(Behler 2017).

机器学习技术应用于计算化学的另一个应用领域,更重视分子,而不是大量材料最初利用核方法,重点是绘制3D坐标(Rupp等人2012 ;Schütt等人2017)或电子密度(Brockherde等人,2017年 ; Snyder等人,2012年). 这些方法在本出版物的其他章节中有详细介绍.

基于分子结构的预测中最重要的问题之一是选择代表分子的特征. 化学家倾向于将分子视为无向图,节点中的原子和边缘中的键,以及立体化学的一些额外考虑因素. 在量子化学中,分子是原子的三维排列,其对应于势能超曲面中的局部最小值. 几十年来,化学信息学应用(主要是制药和药物发现)已经为两种图形的机器学习创建了描述性的丰富功能:药效团指纹,拓扑指纹以及3D构象.

最近机器学习对化学分子的应用主要集中在量子化学特性的QSPR上,并将重点放在更多信息丰富的特征上. 例如,库仑矩阵是依赖于置换的对称不变修正距离矩阵. 当代高度强大的神经网络方法也利用3D信息(Gilmer等人2017 ; Lubbers等人2017 ;Schütt等人2017年). 在三维结构未知的分子的情况下,这通常是在进行预测时的情况,这些特征是根据在理论的分子力学水平上即时计算的猜测几何计算的. 分子力学方法通过潜在的力场进行了大量的参数化,通常是在量子化学计算上,但最终还是用于参数化的实验数据. 因此,应用于这些特征的机器学习方法受到严格监督,其输入基本上定量地嵌入了大量的先前化学知识和手动调整.

由于图形同构,图形学习提出了一系列非常独特的挑战. 图形结构上的机器学习方法比化学更受关注,例如在运输,数据库,电信等任何类型的网络中.而扩展连通性圆形指纹地址图同构,编码图形,散列,混淆化学信息的过程可以导致不同的化学子结构激活相同的特征. 神经指纹是ECFP的一种可区分的深度学习扩展,最近被提出作为一种更灵活和可学习的替代方案(Duvenaud等人,2015). 这些化学拓扑指纹和衍生图形卷积的连续对应物(Kearnes等,2016已证明其优于QSPR的一些传统化学信息学描述符,特别是对于较大的数据集,并开辟了分子筛选的新途径.

## 化学空间的搜索和优化
机器学习应用于计算化学/化学信息学界的一个有趣的子集正在由所有可能的分子或相关子集(例如具有芳环的小有机分子或类似子集)跨越的离散的高维空间中移动.出于兴趣.

化学空间非常大,估计数量级或数量级从10 23到10 60不等. 据报道,基于规则的计算小分子化合物和彻底检测非实用但正式有效甚至可能稳定的分子的基于规则的努力如下:GDB-11列出了2640万个C,N,O最多11个原子的小有机分子.和F(Fink和Reymond 2007); GDB-13列举了978亿个具有多达13个C,N,O,S和Cl原子的分子(Blum和Reymond 2009); 和GDB-17含有1664亿个分子,最多17个C,N,O,S和卤素原子(Ruddigkeit等人,2012)). 从这些集合推断出,药物化学空间的大小已推断为10 33(Polishchuk等人,2013年).

除了空间的大小之外,分子是离散的图形,并且存在关于允许的连接类型和程度的规则. 更进一步,正式有效的分子在感兴趣的温度,压力和时间尺度下仍可能是化学不稳定的. 因此,进行局部甚至更具挑战性的全球优化对材料和药物设计非常感兴趣,也是一个巨大的挑战. 化学空间项目利用GDB数据库开发可视化和探索工具,重点是候选药物分子(Reymond 2015).

定制分子库已被证明是一种非常有效的化学空间导航方式,因为人为驱动的设计允许嵌入难以捕获的规则和化学要求,而无需强有力的监督,特别是关于分子的合成可接近性. 这种成功反映在上一节中实例的实验适用性的简易性中.

该领域的多种机器学习方法表明,可以利用来自专利文献(Lowe 2012)或商业数据库的大量现有化学反应数据集来自动构建用于有机合成反应和反合成分析软件的反应性预测工具. 最近的例子包括在对规则生成的例子进行训练之后预测反应物指纹的化学反应结果(Wei等人,2016),并通过对自生成候选者列表进行排序来选择主要产品(Coley等人,2017a ; Jin等人,2017).遵循机器翻译方法的序列到序列模型(Fooshee等,2018; Schwaller等.2017)或相似性搜索(Coley等人,2017b)已被证明能够以自动,数据驱动的方式自动预测有机合成反应. 从更大,更好的策划数据集中获得了更有希望的结果(Segler等人2017 ; Segler和Waller 2017a,b).

一种众所周知的化学优化方法是遗传算法,其中来自手工制备的化学转化列表的突变并随机应用于起始分子,并且所得化合物(如果显示出所需性质的改善)保持用于进一步进化. 因为它们依赖于手工挑选的突变和超参数,并且通常还依赖于目标属性之间的手动调整权衡,这些方法涉及一定程度的化学直觉. 通过复合可以单独但不能组合使用的突变,产生的分子往往非常复杂. 这些类型的遗传方法已被用于导航化学空间(van Deursen和Reymond 2007 ; Virshup等人2013)并优化多类分子,如有机发光二极管(Rupakheti等2016 ; Shu和Levine 2015),有机光伏(Kanal和Hutchison 2017),金刚石(Teunissen等2017),可见发色团高超极化率(Elward和Rinderspacher 2015),或具有高电偶极子的小分子(Rupakheti等,2015).

## 生成模型
生成模型是机器学习模型,旨在产生捕获训练群体的内在统计特性的自然表观数据. 它们可以以无人监督的方式进行训练,因此不需要标记数据. 鉴于化学空间的大小,以及已知的大量化学物质(见第2节),已知的化合物数量约为1亿,最近已经评估了基于现有分子的无监督深度学习模型.

化学中生成模型的最基本的例子之一是使用递归神经网络(RNN)来预测分子的SMILES表示的下一个特征(Voss 2015). 通过向网络提供n个字符来预测第n + 1个,已经证明RNN在生成有效的 - 如果是荒谬的 - 文本(Karpathy 2015)方面非常强大. 可以存储单个字符串的简单而全面的分子字符串表示:简化的分子输入线路输入系统(SMILES)(Weininger 1988)). SMILES使用一系列规则和规范化程序包含完整的分子连接. 事实证明,为文本显示的相同原理通过基于字符串的SMILES表示生成分子非常有效. 另外的工作已经针对产生用于产生捕获训练数据的统计行为的引线的分子(Bjerrum 2017 ; Ertl等人2017 ; Gupta等人2017). 进一步的步骤涉及分子库的进化精制以使该组偏向给定的性质(Olivecrona等人2017).

自动编码器是深度学习模型,经过训练可以再现高维输入,使其受到低维信息瓶颈的影响. 分子设计中一些最棘手的问题来自分子的离散性和搜索域的极大尺寸. 因此,这种低维,实值,可解码的嵌入可能允许将基于梯度的优化算法应用于分子优化. 进一步的改进是使用变分自动编码器,其中随机噪声被添加到编码步骤,这导致潜在空间中更连续的覆盖以及更少的与有效解码点不对应的死区.

如上所述,存在使用化学图作为机器学习系统的输入的可用策略. 然而,有效编写化学图的深度学习工具仍然是一个悬而未决的问题. 神经网络架构已被证明可以有效地编写文本和图像,因此,第一份报告(Gómez-Bombarelli等人,2018年)用于化学发现的变分自动编码器(VAE)使用基于串的化学表示. 这项工作展示了VAE如何从连续的实值阵列表示中精确地重建分子. 更进一步,它探索了潜在空间中的变换,特别是关于感兴趣的性质的分子优化,如何进行. 该应用在联合训练的VAE加预测系统的情况下特别有效,其中深度学习系统同时被训练为无监督发生器和监督预测器. 在这种情况下,潜在空间在拓扑上受化学性质的影响(图2).

具有联合属性预测的变分自动编码器的图表:编码器神经网络接收分子的串重复并将其转换为可通过解码的神经网络解码回字符串的实值数组.通过与属性预测网络联合训练,可以导航潜在空间以进行属性优化

最初的VAE模型存在几个缺陷,通常与其写出分子的能力有关,特别是字符串输出是概率性的. 由于使用了字符串表示,并且需要内部一致性甚至算术(打开和关闭环和分支),VAE模型产生许多无效分子,不止一种意义上. 一方面,语法错误导致字符串作为SMILES无效,并且不对应于实际化学图(打开但不关闭的循环,打开但不关闭的括号,表示未完成的分支等),a类似的失败是写出非化学有效图的完整图(通常与化合价和八位组合规则有关:化合价大于5的碳原子,化合价3的氧原子等).

许多工作正在迅速扩展这一领域,解决这些性能问题,并探索化学中深层生成模型的进一步途径(Blaschke等人2017 ; Wang 2017). 这些包括使用SMILES的一系列语法规则作为解码器的输出(Kusner等人2017),主动学习输出的有效性(Janz等人2017),执行约束贝叶斯优化以避免探索死区潜在空间(Griffiths和Hernández-Lobato 2017),在编码潜点附近进行局部优化(Mueller等人,2017年)),结合额外的RNN,通过强化学习产生更高质量的产出(Jaques等,2016).

已经报道了进一步改进和应用于产生DNA(Killoran等人2017)和蛋白质序列(Sinai等人2017),以及具有向所需特性施加偏差的能力的对抗性方法(Guimaraes等人2017 ; Sanchez) -Lengeling等人,2017).

# 结论
机器学习和数据驱动的应用程序风靡计算化学. 由于采用了深度学习方法,几十年来被认为是圣杯的任务似乎比以往任何时候都更接近:符合或超越实验准确性的可负担的计算预测,可以击败人类的计算机化反向合成等.

特别是化学和计算化学,长期以来一直在努力研究高容量,高速度和高品种的数据. 然而,痛点不同. 在计算化学中,采集速度越来越不成为瓶颈. 在实验上,它们对速度和再现性的限制更多,而且它们的成本要高得多. 另一方面,计算化学已经显示出难以解决的变化:随着参数化增加了可转移性和对预测能力下降的信任.

它的数字化和自动化程度越高,实验化学物质就越多地与计算化学同化,因此最容易利用这些极具前景的数据驱动工具. 为了产生更广泛的影响,如果计算化学确实是一个沙箱和最终改变实验化学的想法的加速器,则需要更多地关注异构的非结构化数据. 无监督的机器学习和转移学习是这项任务的有前途的工具.
