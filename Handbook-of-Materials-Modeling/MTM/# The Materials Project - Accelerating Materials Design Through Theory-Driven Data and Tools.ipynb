# 材料项目的历史和概述

材料科学家和工程师始终依靠材料属性数据来提供信息,指导和解释研究与开发. 传统上,此类数据几乎完全来自实验研究. 在过去的10-15年中,使用可扩展的物理基本方程式(例如Schrödinger方程)的计算机模拟,已经可以快速生成可靠的材料数据. 这种范式转变是由理论上的进步相结合引起的,最著名的是密度泛函理论(DFT)的发展,算法上的改进以及低成本的计算.

材料项目(MP,简称"该项目")成立于2011年,是一项协作工作,旨在利用理论和计算方面的不断进步来加快新材料的研究和设计. 该项目基于一个全面的材料预测特性数据库,该数据库是对超级计算资源执行数百万次DFT模拟的结果. 在撰写本文时,该数据库包括 > 69,000 种具有晶体结构和总能量的无机材料, > 57,000种具有电子能带结构的材料, > 48,000种具有电子传输特性的材料(图1)(Ricci等人2017), > 30,000种XANES k边缘光谱(Dozier等人,2017年),> 15,000有转换电池特性,> 6000有弹性张量(de Jong等.2015a), > 3,000个具有插入式电池特性, > 1,000个带有压电张量(de Jong等人2015b), > 1,000个带有介电张量(Petousis et al.2017年)和 > 1000个元素表面能(Tran等人2016年). 该数据库正在不断扩展,提供更多的材料和更多的属性(有关当前迭代中列出的属性的示例,请参见图 2).

# 基础理论形式主义和材料设计"应用程序"的开发

## 理论方法

材料项目的核心数据集由从无机化合物库中的密度泛函理论(DFT)计算获得的结果组成. DFT非常适合创建材料属性数据库,因为它具有较少的参数,需要针对不同的材料系统进行调整,并且因为中小型(约300个原子或更少)的系统的计算成本是可控制的. DFT方法已在很大程度上实现了标准化,因此具有略微不同的参数(例如,伪电势)的各种软件实现产生的结果非常相似(Lejaeghere等人,2016年).

尽管如此,为高通量计算选择一组可靠的参数仍然不是一件容易的事. 重要的是要强调,目前还没有完善的DFT功能,因为它们都是近似于定义材料现象的完整物理学组. 例如,高度相关的系统仍然具有挑战性. 在单个研究中,通常甚至可以使用专门的方法处理复杂的系统. 但是,当构建包含许多化合物的大型数据库时,这种专门的处理实际上很难实现,并且还会导致各种化合物之间的结果不一致且经常不兼容. 另外,必须更加谨慎地在计算成本与预期信息收益之间取得平衡. 如果单个研究的计算在数值上过度收敛并且使用的计算能力比必要的多50%,则可能不会对单个研究产生明显的影响,但是这种情况将严重降低高吞吐量的数据库项目,例如消耗数千万CPU的MP每年的计算时间. 因此,MP必须做出实际的折中,以在保持完全自动化和计算效率的同时,保持专业化,精确计算的准确性,并保持程序与其他计算的清晰度和一致性.

`材料项目用于实现这种平衡的一种方法是将材料分为两类,并应用不同的DFT功能为每一类建模`. 第一类化合物是`过渡金属氧化物和硫化物`.对于这些化合物,标准的DFT功能(例如局部密度近似(LDA)(Kohn和Sham 1965)和广义梯度近似(GGA)(Perdew等人,1996))不准确,由于这些化合物的自相互作用误差更明显,缺少导数间断而引起的轨道占用(Zhou等2004年; Cococcioni和德Gironcoli 2005年). 一种计算有效的方法来处理这些化合物是使用 GGA + U 框架,其中将类似Hubbard的校正应用于局部d轨道. 如前所述(Wang等,2006),将特定的U校正值拟合到形成能数据中. 重要的是要注意,这些相同的U值对于准确表示其他特性(例如电子能带结构)可能不是最佳的. `第二类化合物包括所有其他系统,并用标准的GGA-PBE功能性化合物处理`(Perdew等,1996).

与仅对整个数据库使用单个功能(例如GGA)相比,通过允许对不同的化合物使用两种不同的功能进行处理,可以提高所得数据库的准确性. 然而,由于这些结果不直接兼容,因此必须另外设计一种方案来混合从不同方法获得的结果(例如,总能量). 在材料项目中,通过对实验生成焓数据进行基准测试,可以在不同功能的结果之间进行这些调整(Jain等,2011b). 图5通过仅使用GGA,仅使用GGA + U以及混合使用GGA和GGA + U的 Fe-PO相图来描述这种情况的一个实例总能量. 只有使用两种不同功能的图表版本(应用了混合调整)才能重现该系统中所有已知的稳定相位.

图5: 使用从(a)仅GGA,(b)仅GGA+U和(c)混合GGA和GGA+U功能的总能量,计算构建的Fe-PO三元相图. 只有混合相图才能在相位图中将所有已知的相复制为稳定状态. (转载图经Jain等人(2011b)许可.美国物理学会2011年版权所有)

材料项目采取的另一项实用措施涉及分子系统.尽管可以在相同的密度泛函理论框架(例如具有平面波基集的PBE-GGA)中对分子系统和固体进行建模,但是包含分子和固体的计算出的反应能通常会显示出很高的误差,因为自相互作用误差之间存在显着差异以局部(例如分子或高度相关的系统)和非局部(例如金属)电子为特征的系统(Grindy等,2013; Perdew等,1998)). 同样,纯GGA官能团无法很好地描述气体,二维材料和液体中存在的分子间相互作用,对于构建避免这种系统性错误的,由DFT衍生的综合热力学框架提出了进一步的挑战.

`MP不直接计算液体/气体的能量,而是根据实验反应焓(例如金属的氧化)来调节室温下为液态或气态的几种元素的能量`(Wang等,2006). 以下所有化合物均已调节能量,以更好地重现固相的反应能: O2,N2,Cl2,F2和H2.

最后,`我们提到MP还根据化合物的类型调整某些数值参数`. 例如,与半导体和绝缘体相比,MP在计算金属(由初始的松散k点网格计算确定)时使用更密集的k点网格. 此外,随着时间的推移,材料项目使用的数值公差越来越精确. 可通过Materials Project网站获得用于每个计算的参数,有关参数设置的最新描述,请参见https://www.materialsproject.org/docs/calculations.

## 用于数据探索的"应用程序"
MP数据集的大部分价值来自在原始数据之上执行的辅助分析. 这些辅助分析通常结合多个数据点,并且可以采用材料科学中常用的图表(例如相稳定性图或Pourbaix图表),专用材料设计工具(例如评估MP化合物作为电池电极)或简单地采用作为附加信息(例如,报告可能与目标材料形成连贯晶格的潜在基材). 此类工具对于帮助用户从数据集中获取尽可能多的价值至关重要.

材料项目开发了执行许多此类二次分析的方法,并将其作为开源软件实现(通过pymatgen(Ong等人,2013)软件包)和作为Web应用程序("apps")发布. 应用程序为这些功能强大且通常很复杂的分析例程提供了直观,用户友好的界面.在下面的示例中,我们描述了基础方法以及用于生成和操作相图的随附应用程序.

### 相图应用
相图在材料科学中有多种应用. 由实验生成的传统相图不仅显示出稳定的相,而且还描述了溶解度极限和温度依赖性. 相比之下,由于MP当前仅在零温度和零压力下对材料进行建模,而没有对溶解度极限进行建模,因此所得到的相图可能更准确地称为相稳定性图(我们在这里可以互换使用术语). 然而,这种相稳定性图显示了给定化学体系中的稳定相以及各种组成比下的相关相平衡. 这种相图的一个主要应用是用作对新的假设材料的"真实性检查". 如果所提出的材料的能量足够低,能够(或几乎)在相稳定性图上显示,则该材料将具有足够的稳定性,足以在实验室中合成(Sun等人,2016年). 相稳定性图还可用于识别可能与目标相竞争的可能分解产物.

生成这样的计算相图需要了解化学系统内所有可能材料的形成能. 例如,计算三元相图需要了解该系统中所有相关的一元,二元和三元相的形成能. 对于典型的三元系统,计算所有已知相的能量将需要数十次计算. 但是,由于MP数据库已经包含大多数已知无机化合物的预先计算出的能量,因此现在可以避免运行所有这些模拟,并使用MP数据集直接创建合理完整的相图.

从数学上讲,可以使用凸包构造来确定相图上的稳定点集,这是一种根据n自由度确定最小值的方法(Barber等,1996). 通过计算各种DFT能量的总能量的凸包,可以找到全局稳定的结构以及连接稳定相的各种联系线. 凸包构造可用于构造任意数量组件的相图.

许多已知化合物不是热力学稳定的,即它们没有出现在相稳定性图上(Sun等人,2016). 然后需要一个额外的指标来区分这些化合物的亚稳定性. 凸包的构造提供了稳定性的包络. 凸壳上的化合物是稳定的,而能量在壳上的化合物是亚稳态的. 因此,在组合物中到船体的能量距离是可量化的度量,并且与该化合物的亚稳性直接相关. 合成时通常希望在船体上方具有较低的能量,因为与已知的稳定相相比,这意味着形成目标化合物所需的能量损失更少. 材料项目中许多已知的亚稳化合物都在船体15原子电子伏/原子以内,但根据化学反应,可以延伸到船体上方60毫电子伏/原子(Sun等人,2016)). 尽管此分析着重于已知化合物的亚稳性,但仍需要进行量化亚稳极限的工作.

到目前为止,我们已经描述了封闭系统的形式主义,即那些元素的化学计量比是固定的,但是同样的形式主义也可以等同地应用于开放系统,在开放系统中,一个或多个元素保持在固定的化学势而不是保持固定金额. 例如,可以在空气中进行实验,该空气本质上是在特定化学势下无限存储诸如氧和氮等大气元素的功能. 在氩气流动下进行的同一实验仍将是一个开放的系统,但其中的那些元素的化学势大大降低. 因此,在开放的特定元素的环境中,相关的控制变量是该元素的化学势(μ 我),而不是其构成价值. 然后将化学势作为外部变量来获得大势相图.

MP的用户不必熟悉计算相图构造的所有方法论细节(Ong等,2008)即可生成和使用它们. MP网站允许用户简单地键入(或单击可视化的周期表)他们感兴趣的系统元素. 这将生成一个相图,该相图将以图形方式显示该相图以及稳定和亚稳态的列表. /不稳定的材料. 图6显示了具有-5.288 eV氧电势的Li-Fe-PO大电势相图的MP相图应用的屏幕截图. 请注意,由于氧的组成是由电势规定的,因此在相图中不存在自由度,因此会将四元相图折叠为三元相图(带有用于控制氧化学势的滑块).

类似地,其他应用程序也向广大受众提供了强大的基础方法. 例如,类似于大电势相图,Pourbaix图是对电势-pH空间的整体稳定性的投影,以对电化学稳定性进行建模. 先前已经开发出了一种方法,该方法通过利用实验测得的水离子自由能和材料项目中可用的固相计算出的DFT能量来计算(Persson等,2012).). 该方法已集成到"Pourbaix应用程序"中,该应用程序允许用户简单地选择感兴趣的化学系统,元素比率和离子浓度,以便生成熟悉的Pourbaix图,该图利用MP数据集并且可以在视觉上和交互方式进行操作由用户探索. 此外,单个材料相对于最稳定的分解产物的稳定性可以通过叠加在Pourbaix图上的热图生成,从而为用户提供了一种在含水条件下估算亚稳定性的工具(Singh等人2017).

# 计算基础架构和软件工具
开发和维护诸如材料项目之类的计算数据库需要对计算和软件基础结构的大量关注. 在撰写本文时,"材料项目"是超过一百万次单独计算的结果,这些计算代表了超过一亿小时的中央处理器(CPU)小时的计算时间投入. 设置,执行,分析和管理所有这些计算并非易事. 在本文中,我们在撰写本文时描述了材料项目的基础结构. 但是,我们注意到计算的经济性以及软件库的最佳选择可能会很快发生变化.

## 计算资源
材料项目(MP)已在美国国家能源研究科学计算中心(NERSC)和其他地方使用了高性能计算(HPC)资源,迄今为止已消耗了超过1亿个CPU小时.多任务计算工作流(Raicu等,2008)正在越来越多地使用HPC环境,因为这些资源通常具有潜在的大量总计算时间,良好的硬件规格(例如,中到高内存)以及足够的存储空间. 但是,HPC环境对运行高通量计算提出了一些挑战,因为这些环境最初是为满足少数大型,高度并行应用程序的需要而设计的,这些应用程序运行可预测的时间并执行所有对磁盘的输入/输出. 相比之下,高通量计算通常非常众多,并且可实现的并行性受到限制,并且需要不可预测的,通常很长的总运行时间. 此外,它们通常更适合通过外部服务进行管理,而不是仅通过磁盘上的平面文件进行管理. 为了克服这些挑战,2015年解决了与在HPC资源上运行高吞吐量作业相关的许多计算难题.

## 选择数据库软件
高通量计算工作流程的许多部分都需要有效地存储,检索和搜索信息,包括:
* 管理高通量计算的状态
* 存储原始计算结果,以及
* 可搜索的一组已处理数据,用于数据分发和分析

材料项目选择使用非-SQL(NoSQL)"文档存储"(Cattell 2011),MongoDB作为执行这些任务的主要数据库技术(还保留原始输出文件). 我们注意到,这代表了先前在高通量计算材料科学中使用的其他基于SQL的数据管理策略的转变(Jain等人2011a). 之所以做出此决定,主要是因为MongoDB可以适应材料项目所需的数据异质性和快速的数据模型开发速度. 例如,与典型的SQL关系数据库管理系统(RDBMS)(例如MySQL和PostgreSQL)不同,MongoDB在项目开始时不需要在所有数据类型之间进行预先设计的标准化模式. 随着我们向项目中添加新的计算类型,存储的数据类型也在不断发展. 通过选择MongoDB,MP可以在应用程序代码中进行少量更改即可快速适应这些更改,而无需重构复杂的关系模式.

在面向文档的数据存储中,MongoDB以其简单但功能强大的查询语言,易于管理以及在读取大量工作负载(大多数常用数据(所谓的工作集)可以放入内存)中的良好性能而著称. 对于MP来说,它在链接异构数据(数据库"联接")和繁重的工作负载方面的相对弱点是一个合理的权衡. MongoDB的生产力优势是查询语言和本机数据模型都是JavaScript Object Notation(JSON)(Bray 2017),这是现代Web应用程序的标准数据格式,可以轻松地表示和操作为Python中的本机数据类型. 编程语言(范Rossum的等.2007),以编写其他软件库. 因此,熟悉Python的用户,尤其是其"dict"对象的用户可以迅速适应使用MongoDB理解和开发数据模型. 我们的经验是,与我们以前使用RDBMS的情况相比,这些方面使我们团队的更多成员可以在数据库开发方面进行协作,在RDBMS中,只有一个或两个团队成员对系统足够熟悉,可以进行更改. Gunter等人描述了有关在科学的HPC生态系统中部署这种类型的集中式数据存储时遇到的经验和挑战的更多详细信息.(2012).

根据要研究的属性,DFT"计算"实际上可能涉及一系列单独的计算,这些数据需要数据传递以及计算之间的几何形状或输入设置的修改. 获得所需输出属性所需的一组计算,以及这些计算之间的依赖关系和数据传递要求,定义了一个"工作流". Materials Project开发了两个Python编程语言的软件库来管理此类工作流. 第一个库称为" FireWorks"(Jain等人,2015年),是通用的工作流程库.FireWorks不包含任何材料科学或DFT特定代码. 它的范围是为用户提供一个框架,以定义任意计算序列,将其存储在数据库中,在各种类型的计算资源上执行它们以及管理整个系统中数以百万计的工作流的状态. 因此,FireWorks与各种各样的科学计算工作流程兼容(尽管它最适合高通量应用程序),并且在材料科学领域之外经常使用. 第二个用于创建工作流的库"atomate"(Mathew等人,2017年)),其中包含在FireWorks中实现的特定材料科学工作流程,并使用pymatgen作为基础库.可 以认为雾化包提供了材料科学工作流程实施的库(例如,电子带结构,弹性特性以及压电,介电和铁电特性的标准工作流程). Atomate用户可以为材料和所需的工作流程类型指定输入几何,原子化将提供该工作流程的基于FireWorks的实现,可以在超级计算中心执行. 此外,atomize利用pymatgen库自动解析计算输出并创建可以由用户查询的材料属性数据库. FireWorks和atomate中的各种功能允许您针对特定情况自定义行为,从低级问题(例如与各种排队系统进行交互)到高级问题(例如使用多个DFT功能选择运行相同的工作流). 计算工作流还可以根据从早期计算中获得的结果,自动调整其过程以用于以后的计算.

在执行计算时,可能会遇到与计算收敛有关的各种错误. 该材料项目已发展到模拟软件类型作业的包装(例如,VASP Kresse和哈夫纳1994年; Kresse和Furthmüller 1996年.或QChem Kong等2000)被称为"托管人"(Ong等.2014自动监控输出)计算文件并根据一组规则自动修复错误(通过停止作业,更改输入文件并重新启动作业). 保管人软件还可以用于自动执行线性计算序列(例如,收敛协议,可以收紧数字参数,直到输出没有变化为止).

一旦执行了计算,结果将被解析并存储在各种数据库集合中. pymatgen将原始数据解析为雾化工作流程的一部分. 我们注意到pymatgen可以解析输出文件(解析为结构化数据或具有可调用函数的Python对象),还可以执行高级数据分析,例如相图创建或绘图. 另外,我们使用称为"构建器"的代码来收集,重组原始数据并将其后处理为单独的数据库集合,与原始数据集合相比,这些数据库更易于分析. 例如,建造者可以将单一材料的所有计算结果汇总在一起,以建立该化合物的单一摘要报告("材料"文档). 建造者还可能从多种化合物中收集信息,进行分析,并将结果存储在数据库中. 为此类过程提供服务,我们开发并使用轻量级库来自动化, 简化并最终简化创建MongoDB数据库的过程. 对于需要使用MongoDB执行提取-转换-加载(ETL)操作的任何项目,我们的通用"生成器"代码都可能有用. 例如,它们可以并行运行,而无需作者明确编码并行性.这使数据的CPU密集型转换可以在包括大多数现代硬件的多核计算机上更快地运行(正在与消息传递接口(MPI)标准集成,以实现跨超级计算资源的并行化). 此外,我们代码中用于增量构建的功能允许源MongoDB集合的连续构建仅对自上次构建以来添加的记录进行操作,这可以节省大量的计算时间. 总体而言,构建器框架允许针对不同类型的查询和使用模式量身定制的多个数据库集合的高效生成和可靠更新.

材料项目非常重视用户体验,用户界面设计,并确保有效地分发数据,以便广泛的用户能够将数据应用于研究,开发和教育(Jain等人,2016b). 为此,我们建立了一个交互式Web门户(https://www.materialsproject.org),重点是科学家作为最终用户.该Web门户使用成熟的Django Web框架构建(Dja 2015),因为其将前端视图与后端业务逻辑完全分开. ## 相图和化合物稳定性
在旨在基于实验证据增进我们的理解或针对给定应用合成新材料的研究中(Bayliss等人,2014; Krishnamoorthy等人,2015; Martinolich和Neilson,2014),生成相平衡数据是最常用的方法之一MP功能. 例如,Bayliss等.对钠掺杂的硅酸锶锶材料进行了一项研究,该材料据称具有非常高的氧化物离子传导性(Bayliss等人,2014). 通过结合实验(中子粉末衍射,两点交流阻抗谱,飞行时间二次离子质谱)和DFT计算,他们可以表明电导率比以前报道的要低,并且氧空位形成的高能成本是根本原因. 来自MP的数据用于交叉检查DFT-PBEsol衍生的SrSiO3分解的主要极限相的研究结果.

Shi和他的同事(Shi et al.2017)采用了高通量DFT筛选方法来稳定成分ABX 2的铜铁矿和相关分层相,其中A和B是元素周期表中的任何元素, X是硫属元素(O, S,Se和Te). 从最初的15624种化合物中,发现285个在距凸包50 meV /原子以内. 尽管这些结构的大多数包含在"材料项目"数据库中,但作者强调指出,这些稳定系统中不存在79个. 这强调了晶体结构数据库(如MP)在化合物完整性方面仍具有可观的增长潜力,并强调了基于用户的化合物提交的作用(通过MPComplete服务,请参见第6.1节).)可以扩展此类数据库.

一个类似的例子是Krishnamoorthy等人的工作.(2015年),他使用了一种基于DFT的高通量筛选技术来鉴定可用于光收集的无铅碘化锗钙钛矿. 研究人员计算了360种AMX 3成分的PBE带隙,发现了9个有趣的候选对象. MP相平衡数据被用来通过要求材料在热力学上稳定以防止分解为更简单的二元相来减少清单. 从计算筛选中剩下三种材料,RbSnBr 3,CsSnBr 3和CsGeI 3,其中后者已成功合成和表征. 我们将感兴趣的读者推荐给先前的评论文章Jain等.(2016b),了解使用MP计算的相图进行实验研究的更多示例.

## 晶体结构分析,2D材料和机器学习
材料项目中可用的大量数据可以用作开发用于处理晶体学数据的新算法的试验台. Ashton及其同事(Ashton等人,2017年)就是这种情况,他们开发了一种拓扑缩放算法来识别给定晶体结构的尺寸. 他们使用该算法在MP数据库中搜索可能是潜在2D材料的材料. 确定了826种稳定的分层材料,其中基于计算的剥落能量,预测其中680种是可行的2D材料候选对象.

类似地,Cheon及其同事(Cheon等人,2017年)提出了一种算法,该算法可以识别三维晶体结构的弱键合子成分的维数. 他们将此算法应用于50,000种以上的MP材料,并识别出1,173个二维分层材料以及487个弱键合的一维分子链,表示已识别材料的数量增加了一个数量级. 此外,建议其中的325种材料是压电单层.

有趣的是,通过特别利用典型DFT计算中的一个弱点,即色散力没有得到很好的解释,因此层状材料的晶格参数通常非常不准确,Choudhary和他的同事(Choudhary等人,2017年)能够识别出两个尺寸候选材料. 为此,作者要求实验和(主要是)MP数据库的晶格常数之间的偏差≥5%. 为了验证他们的方法,作者使用了基于剥落能量的公认标准,发现他们的预测中有88.9%符合该标准.

构成材料科学中设计规则基础的许多结构－属性关系都是基于与本地协调环境有关的信息. 因此,非常需要具有能够有效地识别基本的局部结构基序的工具,例如四面体,八面体,bcc,fcc和hcp环境. Zimmermann等.提供了基于局部结构顺序参数的这些图案分类标准,这些参数用于在整个材料项目数据库中自动识别这些图案. 此外,这些工具也可能引领通往替代结构匹配途径的道路(Zimmermann et al.2017).

材料项目中的大量数据也为开发新的机器学习(ML)技术提供了机会,新的机器学习(ML)技术可用于对材料属性进行建模并更好地理解结构与属性之间的关系. 这样的一个例子出现在Faber等人的著作中.(2015)旨在发展适应于机器学习模型的周期性系统的表示. 在这项研究中,材料项目的4000个结构用于基于三种不同的晶体结构表示方案并使用核岭回归来评估预测的形成能中的泛化误差,这表明旨在模拟无限大库仑和的正弦矩阵方法是效率和准确性都很高.

同样,de Jong等.(2016年)展示了一种机器学习方法来预测k元化合物的弹性模量,该方法在一组高度多样化的化学方法中有效. 更具体地说,本研究对MP弹性张量数据集使用了梯度提升机局部多项式回归(GBM-Locfit)来确定一组相关的描述符,并得出弹性模量预测. 最终,利用该模型来估算整个MP材料库的维氏硬度,从而可以快速搜索超硬材料.

在大多数情况下,为了便于预测更复杂的属性(例如地层能量或弹性张量),使用了易于检索或计算的数据(例如空间组,组成和密度). Shandiz和Gauvin追求逆向路线(Shandiz和Gauvin 2016):作者对MP数据库中的339种潜在锂离子硅酸盐阴极材料进行了分类研究(一般成分:Li-Si-(Mn,Fe,Co)-O). 特别是,他们根据从输入晶体结构和DFT输出获得的特征,测试了是否可以预测晶体系统(单斜晶系,斜方晶系或三斜晶系):单位晶胞体积,带隙,位点数在晶胞中,地层能量和凸壳上方的能量.这些特征的成对相关图表明,任何特征与晶体系统之间均没有可利用的直接相关性.基于决策树的方法(随机森林和极端随机的树)显示出高达75%的预测准确性,k-最近邻居分类.

## 筛选应用材料
从材料项目的数据中得出的最一致的材料筛选策略也许是按照适合给定应用空间的连续更严格的标准对材料进行筛选的策略. 在这种方法中,对于大多数应用而言,常见的过滤器通常通过ΔE 船体来提高稳定性,这可以提供一种化合物在实验上是否可行的指标. 如图9所示,依次的过滤器依次减少了要考虑的物料数量,直到物料达到可处理的数量为止,以便通过更复杂的计算进行跟踪或进行实验查询.

这是Sendek及其同事采取的方法(Sendek等人2017)谁在寻找可以用作锂离子电池固态电解质的新候选材料. 作者从材料项目中筛选了12,831种含锂化合物,以过滤具有高结构和化学稳定性,低电子电导率和低成本的化合物,从而消除了92.2%的初始材料. 随后,对离子电导率分类模型进行了训练,该模型针对40种晶体结构和文献中的相关测量进行了训练,将有趣的候选对象列表减少到仅21种材料. 在后面的步骤中,考虑多描述符模型而不是单描述符函数对于实现预测能力至关重要. 其余21种材料中有许多尚未进行实验研究,

为了筛选结构敏感特性和对材料进行分类而鉴定结构相似的化合物也已开始作为筛选策略和设计范式出现. Dagdelen和同事(Dagdelen等人.2017年)证明这样的筛选程序用于预测新拉胀材料(具有负的泊松比化合物). 作者通过材料项目的REST API系统地筛选了整个MP数据库,并将每种结构与α-  方石英SiO 2进行了比较,是以前已知的唯一具有负均质泊松比的无机结晶材料之一. 通过将pymatgen的结构匹配算法(可以在用户定义的公差范围内匹配结构,而不管晶体设置,超级晶胞大小或成分如何)与更常规的筛选策略结合,从超过65,000个结构中收集了30种可能的候选物. 然后计算每个候选人的全弹性张量,然后推导其泊松比. 在这30个结构中,发现3个具有均匀的膨胀性,另外9个显示出接近零的均匀泊松比,并正在进行实验确认.

Lau等人介绍了"材料计划" 中进行计算机筛选的一个例子,该实验导致了实验证实的材料发现.(2017年).这些作者通过连续标准过滤("漏斗"方法)在MP数据库中搜索了有希望的化学回路空气分离(CLAS)材料. 具体而言,应用的搜索过滤器包括(i)限制二元和三元化合物;(ii)识别可发生氧化反应的化合物(在此步骤中,使用相图app);以及(iii)限制温度和氧分压氧化反应进行到合理极限的范围. 该方法产生了5,501种试验化合物和20,861种相关的氧化还原反应. 由于还原焓和重量O 2容量(图10)没有揭示任何可利用的趋势,因此,作者不得不采用一种更具启发性的方法来减少候选人名单. 首先,他们要求反应的复杂性和反应中存在的总相数必须最小,从而产生292种材料. 其次,他们通过排除具有昂贵和有毒物质的化合物以及还原后涉及非氧化物的反应,将其进一步减少至108个. 他们从其余的材料中选择了ABO 3钙钛矿,因为它们在氧化学计量方面具有灵活性,而无需进行大的结构变化,并且通常容易合成钙钛矿. 他们随后合成并表征了SrFeO 3− δ由于其热力学和出色的循环稳定性以及在整个工作温度下的抗碳化能力,其已成为有前途的CLAS候选物(Lau et al.2017). 通过材料项目发现化合物的更多示例可以在先前的评论中找到(Hautier等,2012; Jain等,2016c).

## 超越模拟:社区贡献的材料数据
MP已成为材料科学界的全球资源,有40,000多名用户依靠该门户网站作为可信赖的来源来加速他们的研究.这为扩大MP的任务范围提供了机会,也包括协助研究人员向更大范围的材料科学家群体传播他们自己的数据集(无论是计算数据还是实验数据).因此,MP不仅可以用作集中生成的计算数据的集线器,而且还可以托管和分发由全球研究小组生成的各种数据集.这也将使MP的用户更全面地了解化合物,因为他们将获得来自各种技术的计算和实验信息.

## MPCite:出版物中的引用材料数据
科学和技术信息的能源办公室,美国能源部(OSTI)(Elliot等人.2016)提供了E-链接服务,从而使研究人员提交有关OSTI产品(在XML元数据记录的形式)和检索持久数字对象标识符(DOI)在万维网上进行标识. DOI最常用于参考和查找期刊论文,因为它们提供了唯一的URL链接到期刊的在线登录页面,其中包含有关出版物的更多信息. 我们的开源软件MPCite(Huck 2016a,b)允许对所有MP化合物的DOI进行连续的请求,验证和分发. MPCite还可以用于将DOI分配给非核心数据库条目,例如通过MPContrib提供的理论和实验数据或用户生成的分析或结构数据.

材料项目还将把重点放在帮助用户将基础数据资产转换为有关结构－属性关系的新见解上. 特别是,新功能将允许用户使用可视界面制定复杂的查询,并执行交互式数据分析和实时过滤. 用户将能够在机器学习算法以及传统理论计算的指导下快速迭代材料设计探索. 材料设计平台的愿景的四个组成部分是查询,过程,可视化和模型/计算(请参见图12). 接下来,我们详细讨论这些组件.

**询问**  
如今,材料项目为其基础数据库提供了基于Web的可视化搜索界面,该界面最适合识别与一系列约束条件匹配的材料集. 但是,许多用户需要更复杂的数据管道,在其中可以直观地添加或删除过滤器,并在分析中的多个点检查结果,或者合并来自独立查询流的结果. 对于那些能够编写计算机程序以通过MAPI提取"材料项目"数据的功能来说,这种功能已经是可行的,但对于其他人则仍然很难. 将来,将开发允许用户获取数据并与数据进行交互的新技术,从而使人们能够使用可视查询界面精确调用所需的数据.

**处理**  
用户通过查询工具编译了感兴趣的数据集后,Materials Project将使用户轻松地以有助于可视化,解释和模型构建的方式向数据添加描述符/特征. 我们设想了一个系统,用户可以通过该系统显示任何结果集(例如,100种感兴趣的材料),并且通过单击按钮可以快速生成描述符库,例如平均电负性,局部环境类型或多面体连接类型数据集中的每种材料.用户将能够使用这些描述符通过常规数据分析(例如,可视化,统计报告)以及数据挖掘和机器学习方法来探索潜在的结构-属性关系.

**可视化**  
新的软件库和Web框架(例如Pashly的Dash和Crossfilter)使在网络上生成可以交互式浏览和操纵的高质量图表变得前所未有的容易. 这样的库可以使用户在MP网站本身上执行大部分数据探索.作为一个简单的示例,用户可以决定在两个用户选择的感兴趣属性之间生成标准的XY散点图,该散点图还允许将鼠标悬停在特定点上以显示该材料的详细信息.一个更复杂的示例将包括用于交互式过滤数据集的工具,每次修改或添加过滤器都会显示实时直方图,以显示数据集中剩余材料的各种材料特性的分布.

**模型/计算**  
在准备和探索数据之后,下一步就是对数据采取行动.一种可能的动作是产生一个描述材料特性之间各种关系的模型. 例如,可以尝试建立一种模型,该模型将结构描述符(例如局部环境)和成分描述符(例如氧化还原活性物质)与计算出的输出(例如电池电压)相关联. 机器学习模型是一种有趣的前进方式,因为一旦经过训练和验证,它们就可以用于获取关于材料在如何进行模拟之前几乎如何表现的即时反馈. 因此,这样的模型可以作为替代品,用于更复杂,更费时的物理模拟,以进行定性估计和排序.

有了这些元素,单个交互式Web会话将允许用户对数据集执行复杂的查询,自动生成可用于形成结构－属性关系的描述符,以可视方式浏览(并在必要时进一步优化)数据集,生成描述数据的模型,并使用这些模型来驱动进一步的计算. 实际上,今天的材料项目中存在许多这些元素.例如,对于尚未计算弹性模量的材料,用户可以基于机器学习模型立即获得估计值(de Jong等人,2016),并支持基于密度泛函理论的完整计算.将来,这种类型的数据挖掘和传统理论模型的混合使用将变得更加普遍,并且对于用户而言越来越自然.

## 结论
从头开始进行仿真一直是理解和设计材料的强大工具. 随着高通量计算的进步,现在可以创建模拟结果库,该库可以以远远超过过去的速度生成有关材料的信息. 此外,软件框架和网络技术的进步已使这些结果以无障碍的方式传播给全球成千上万的研究人员. 材料项目致力于利用这些进步来构建有价值的材料数据资源以及可改变材料设计方式的软件工具. 此外,该材料项目旨在通过开源软件的开发,并通过从实验组的反馈进行计算材料学一个更具协作性的过程.

这是理论激动人心的时刻–前所未有的材料数据或计算潜力对材料设计产生更大的影响. 实验学家和理论家都能够利用材料计划以弥合传统知识鸿沟的方式进行科学研究和工业研究. 这些用例可能是理论技术和大型材料数据库将越来越有影响力并有助于创建新的材料设计范例的未来的早期迹象.
