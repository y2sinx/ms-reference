# 概述

** 摘要 **

物质电子结构的理论分析, 为描述观察到的各种现象, 提供了理解和定量方法. 这些现象的清单类似于精简版物质物理学教科书的目录，并且自然地分为基态和激发态电子性质. 本章的目的是对电子结构提供一个介绍, 而无需借助数学公式; 目的是阐名电子在决定物质性质中所起的作用, 并且概述电子结构理论所面临的挑战.

物质的性质自然分为两类, 分别是基态性质和激发态性质. 在材料的物理性质上, 这种区别很明显, 并且也决定了理解和发展整个电子结构领域的理论框架. 从本质上讲, 在大多数关于凝聚态物理学的教科书中, 基态和激发态电子特性列表是相同的[84,86,88]():  
* 基态: 内聚能(cohesive energy), 平衡晶体结构(equilibrium crystal structure), 结构相变(phase transitions between
structures), 弹性常数(elastic constants), 电荷密度(charge density), 磁性顺序(magnetic order), 静电介质(static dielectric)和磁性磁化率( magnetic susceptibilities), 核振动和运动(在绝热近似中,adiabatic), 以及许多其他性质. 
* 激发态: 涉及具体热的(specific heat)金属低能激发, Pauli自旋极化率(Pauli spin susceptibility), 电子输送(transport)等; 确定绝缘体中绝缘带隙(insulating gaps)的高能激发, 光学性质, 添加或去除电子的光谱, 以及许多其他性质.  

这种划分的原因是材料是由核和电子结合在一起的. 由于电子的典型能量尺度远大于与更大质量核的自由度相关的那些东西(的尺度), 所以, 电子的最低能量基态决定了原子核的结构和低能运动. 大阵列物质形式--从已知最硬物质金刚石C到软润滑剂石墨C, 再到由许多复杂的晶体和分子(由周期表中的元素形成)--主要是电子基态的表现. 在大多数材料中, 在晶格振动中原子核的运动, 原子核的时间尺度比典型电子的时间尺度要大得多, 因此当原子核移动的时候, 电子可以被认为处在其瞬时基态(instantaneous ground state).  这是众所周知的绝热近似或 **Born-Oppenheimer近似**[89,90]()(见附录C). 绝热近似可以将体系中的原子核与电子奋力出来分别讨论, 即电子运动与原子核的运动是互相独立的.

由于电子的基态是电子结构的重要组成部分, 因此当前理论努力的很大一部分致力于寻找精确, 稳健的基态处理方法. 为了建立理论所需的基本特征, 我们需要了解材料中涉及的典型能量. 为了能够做出准确的理论预测, 我们需要有非常准确的方法来区分物质不同相之间很小的能量差异. 到目前为止, 对`固体第一原理定量计算`(for "first principles" quantitative calculations of solids)最普遍的方法, 是密度泛函理论[91-93](), 这是本书的核心主题. 此外, 目前已知最准确的多体方法是量子蒙特卡罗方法[81,94,95], 被明确设计为计算基态或热平衡的性质.

另一方面, 对于由核组成的给定结构, 电子激发是物质"电子性质"的本质--包括导电性, 光学性质, 电子的热激发, 在半导体中外在电子引起的现象等等. 这些性质由激发能谱和激发态的性质决定. 有两种主要的激发类型: 单电子的加或减(addition or
subtraction of single electron**s**), 和保持电子数量恒定的激发. 由于激发可以被严格地视为对基态的微扰, 因此, 微扰理论方法往往是理论理解和计算这些特性的关键[96](). 

电子激发也会与核运动耦合, 这会产生诸如电声相互作用(电子-声子相互作用)此类的效应. 这会导致电子态的扩展并潜在地在金属中造成很大的影响, 因为普通金属在任意低能处总是具有激发能量, 因此会与低能核激发混合. 耦合可以导致相变和物质的定性新状态, 例如超导. 在这里, 我们将考虑可以理解和计算电声相互作用的理论(例如在第19章中), 但我们不会明确地处理相变或超导本身的结果.

## 电子基态: 键合和特征结构 

电子结构理论面临的挑战是提供通用方法, 可以准确的描述自然界中的真实体系. 它们不得局限于任何特定类型的键合(应该适用于所有类型的键合), 否则他们就不是成功. 并且我们希望理论提供理解力, 我们寻求的方法不仅可以提供数值分析(例如结合能)还可以使用简单的图像(方法)对一般性问题进行可能性分析, 用以描述控制物质结构稳定的主要机制.

## 体积或压力是最基本的物理量

最基本的物理量有: 能量, 压力, 体积模量以及能量的高阶导数:   
 
 \begin{equation}
 \begin{split}
 & E = E(\Omega) \equiv E_{total}(\Omega),\\  
 & P = - \frac{\text{d}E}{\text{d}\Omega},\\  
 & B = -\Omega \frac{\text{d}P}{\text{d}\Omega} = \Omega \frac{\text{d}^2E}{\Omega^2} 
\end{split}  \tag{2.1}
\end{equation}  

其中E是晶胞的能量, 所有的量都有固定数量的粒子数.  
第一个测试是检测理论对平衡体积($\Omega^0$)的预测, 此时能量最小, 压力为0, B是零压晶体结构的体模量. $\Omega^0$ 和 B可以在很大的精度内测量, 并且很容易外推到 T=0 的情形, 所以该测试是对理论强有力的测试.   
对于不同的体积计算其能量, 就可以拟合出一个解析公式(analytic form), 比如 Murnaghan状态方程就是这么得来的[108].   
可选的, P可以从 Virial定理或它的推广形式计算(Sec 3.3), 而 B 可以从相应函数得来(Ch19).    
1960s和1970s期间的计算力(computational power)和算法(主要使用原子轨道基组和APQ方法), 使第一批可靠的自洽计算成为可能($E-\Omega$). 比如KCl[109,110], alkali metals, [111,112],and Cu[113].   
$Janak, Moruzzi, and Williams$[106,114]的工作成为转折点, 因为他们将  Kohn-Sham DFT 的功效(定性?说明)变为可实际计算固体的性质. 他们使用 `Koringa–Kohn–Rostocker (KKR) method (Sec. 16.3)`计算了所有过渡金属的平衡体积和体模量(使用局域近似, LDA?), 如图2.3所示.

## 弹性: 应力-应变关系

材料中应力和应变的古老主题也被纳入了电子结构之中.  
这意味着应力-应变关系的起源可以追溯到**量子力学中关于应力的基本定义**, 以及实用方程式已经衍生出来, 现在是电子结构理论中的常规工具[104,129].  
应力张量 $\sigma_{\alpha\beta}$ 的基本定义是(2.2)对各向异性应变的推广:  
$$\displaystyle \sigma_{\alpha\beta} = - \frac{1}{\Omega} \frac{\partial E_{total}}{\partial u_{\alpha\beta}}\tag{2.3}$$  
其中 $u_{\alpha\beta}$ 对称应力张量(G.2). 同样, 该理论表达式是维里定理的推广(从压力到各向异性应力)[104,129].

## 磁性与电子-电子相互作用

**磁性体系**是其中基态具有**破坏对称性**的体系, 具有电子自旋和(或)轨道角动量. 
在铁磁体系中存在净磁矩, 在反铁磁体系中存在随空间变化的磁矩, 其通过对称性平均为零.
磁性基态的存在本质上(intrinsically)是一个由电电子相互作用引起的多体效应.
在量子力学出现之前, 人们已经认识到磁性材料的存在是物理学中的关键问题之一, 并且可以证明在经典物理学中, 体系的能量不可能受到外部磁场的影响.[66-68].
该解决方案在量子力学的早期被认识到,因为单个电子具有n/2(half-integral)自旋,并且互作用电子在基态中可以具有净自旋和轨道角动量.
在开壳原子中(没有形成8电子稳定结构), 这在Hund的规则中得到了总结,即基态具有最大的总自旋和最大轨道动量, 与壳中可能的态电子占据情况一致.

在凝聚态物质中, 关键问题是将**磁矩**排列成长程有序的**磁态**.  
在许多情况下, 这是一个类似原子的自旋的排序问题, 即问题分为两部分: 由电电相互作用形成的原子磁矩, 以及这些局部磁矩排序, 这可以通过海森堡或伊辛(Heisenberg or Ising)模型来描述[131]. 另一方面, 即使电子状态与原子的电子状态有很大的不同, 许多材料也是磁性的, 这种情况被称为"band magnetism"或"巡回电子磁"(itinerant electron magnetism)[132].

磁学的定性图像出现在能带图中, 其中, 电子的交换关联效应被有效的塞曼场 $H_{Zeeman}$ 取代, 用哈密顿量 $ m(\mathbf{r}) V_m(\mathbf{r})$ 中的一个附加项表示, 其中 m 是自旋磁化(spin magnetization), $ m = n^\uparrow - n^\downarrow $, $V_m = \mu H_{Zeeman}$. 与2.2节中的能量-体积曲线类似, 寻找固定场 $V_m$ (**V表示势能, 磁场中的势能**?)的能量是非常方便的, 现在的问题变为求能量和磁化率的最小值. 基本公式如下:   

\begin{equation}
\begin{split}
E &= E(V_m) \equiv E_{total}(V_m),\\   
m(\mathbf{r}) &= - \frac{\text{d}E}{\text{d}V_m(\mathbf{r})},\\  
\chi (\mathbf{r}, \mathbf{r}') & = -\frac{\text{d}m(\mathbf{r})}{\text{d}V_m(\mathbf{r}')} = \frac{\text{d}^2E}{ \text{d} V_m(\mathbf{r}) \text{d} V_m(\mathbf{r}')}.
\end{split}\tag{2.4}
\end{equation}

如果电子无相互作用, 在磁化强度为零时(m为0, 能量的一阶导数为零), 能量的曲率 $\chi$ 为正(能量的二阶导数为正), 则**能量**(是不是遗漏了?)此时有最小值(该处需要知道如何求二元函数的极小值,才能理解这句话的含义), 对应于充满自旋成对排列的能带. 然而, 交换能倾向于使电子自旋平行排列, 因此 $V_m(\mathbf{r})$ 本身依赖于 $V_m(\mathbf{r}')$, 并且可以导致一个最大值(在 $V_m(\mathbf{r})$时的时候), 以及 一个最小值(非零磁化时); 铁磁(均值$bar{m}$是非零的话), 否则反铁磁. 通常情况小, $V_m(\mathbf{r})$ 和 $V_m(\mathbf{r}')$ 必须保持自我一致。

磁性的平均场处理(treatment)是一个对很多问题都适用的原型(在电子结构理论当中). 磁性极化率是相应函数的一个例子(App.D), 自洽可以导致平均场理论的表达式(D.11); 磁化率的表达式最早由斯托纳(Stoner)得到[131,133]:   

$\displaystyle \chi = \frac{N(0)}{1-I N(0)} \tag{2.5}$    

此处, N(0) 是费米能处的态密度, 有效场被扩展为磁化强度的线性阶 $V_m = V^ext_m + I m$ (对于有效交互). 2.5的分母表示独立粒子磁化率的重正化 $\chi^0 = N(0)$, 磁性的不稳定性被斯托纳参数 $I N(0)$ 等于1时的散度所预示. 图2.7由莫鲁齐等人整理(汇编), $I N(0)$来自于两次独立的DFT计算(一次计算I,一次计算N(0)). 显然, 这一理论是相当成功的, 因为只有铁磁金属, Fe,Co和Ni的Stoner参数超过了1, 而对已知的增强顺磁情况(如Pd)近乎为1.

现代计算可以使用"冻结"自旋结构或响应函数, 以与声子相同的形式, 精确地处理DFT中的自旋极化率和激发. 基于Berry相[135,136]的前一种方法的优雅公式将在Sec.19.2中描述. Berry相位和响应函数方法应用于铁元素的例子如图19.3和19.5所示.

## 声子与位移相变(displacive phase transitions)

振动谱可以提供材料的很多信息, 而振动谱很容易通过实验检测, 例如红外吸收, 光散射, 非弹性中子散射等.固体对电场等的响应也是如此. 这种性质最终成为电子结构的一部分, 因为电子决定材料能量的变化(如果原子移位或外加电场先引起电子变化,在引起能量变化, 或者说能量的变化可以在电子的变化上表现出来). 只要频率较低，电子就可以被认为保持基态，基态随原子核的位移而演化。总能可以视为原子核位置的函数, $E(\{\mathbf{R}_I\})$, 而与原子核速度无关. 这是绝热近似或者 波恩-奥本海默 近似, 是有关晶格振动的一个很好的近似.(将电子运动和核运动分开了). 下式与2.2节的类似. $C_{IJ}$力常数矩阵.

\begin{equation}
\begin{split}
E &= E(\{ \mathbf{R}_I \}) \equiv E_{total}(\{ \mathbf{R}_I \}),\\  
\mathbf{F}_I &= - \frac{\text{d}E}{\text{d} \mathbf{R}_I},\\  
C_{IJ} &= - \frac{\text{d} \mathbf{F}_I }{\text{d} \mathbf{R}_J } = \frac{\text{d}^2 E}{\text{d} \mathbf{R}_I \text{d} \mathbf{R}_J  } 
\end{split}  \tag{2.6}
\end{equation}  

在定量计算上有两个特征性的方法:  
* 总能直接计算为原子位置的函数, 被称为"frozen phonon"方法(冻结声子法)
* 计算能量的任意阶导数, 这被称为"response function" 或者 "Green's function"方法(响应函数法或格林函数法)

## 热学性质: 固液气相图

作为表面的大量理论研究的说明,离子半导体,例如III-V和II-VI晶体,对电子结构计算提出了挑战性问题.
阴离子和阳离子终止的可能性具有不同的化学计量:不仅必须比较具有相同原子数的各种可能重建的总能量,而且还必须比较每种类型的不同原子数的结构,即不同的化学计量.
作为说明的第一步,可以通过简单的电子计数规则[175,176]预测不同类型结构的原子的化学计量,即通过填充所有阴离子悬空键在表面上的电荷补偿和阳离子悬空键.

要最小化的数量不是自由能E-TS,而是巨势[179]:  

\begin{equation}
\Omega = E - TS - \sum_I \mu_I N_I \tag{2.7}
\end{equation}

如何将其恰当地纳入理论? 幸运的是,二元AB化合物[179]有简化:
1. 晶体$E_AB$的能量接近其在T= 0时的值(如果需要,可以对有限T进行校正).
2. 假设表面在体相内平衡,则只有一个自由化学势,可以认为是 $\mu A$,因为$\mu A + \mu B = \mu A_{AB} \approx E_{AB}$ .
3. 在平衡状态下,$\mu A$和 $\mu B$的范围是有限的,因为每个都不能超过浓缩纯元素的能量,$\mu_A \leq E_A, \mu_B \leq E_B$

这足以使理论预测表面的重建,假设平衡作为实际实验条件的函数.

## 电子激发: 带和带隙

电子激发可分为两种类型:
与基态具有相同数量的电子(N)的激发态,
单粒子激发, 其中一个电子减去`N→N-1`或加`N→N+1`.
前者的激发决定了比热,线性响应,光学性质等,
而后者则通过隧道效应和光发射或反向光发射进行实验探测[85].

添加和去除电子最重要的物理量是基本间隙,
这是加入和减去电子的能量之间的最小差异.
最低间隙不是限于独立粒子近似的近似概念.
它在一般的多体系统中定义为添加电子和去除电子之间的能量差异:
如果基态具有N个电子,则基本间隙为:

$\displaystyle E^{min}_{gap} = min \bigg\{ \big[ E(N + 1) − E(N) \big] − \big[ E(N) − E(N − 1) \big] \bigg\} \tag{2.8}$

金属是间隙消失且最低能量电子态离域的系统.
另一方面,如果基本间隙不为零或者状态是局部的(由于无序),则系统是绝缘体.

## 电子激发: 热容, 传导, 和光学谱

电子数不变的激发可被视为**电子-空穴激子**,其中额外的电子通过"空穴"(去除电子留下的)相互作用. 

$$ E^{min}_{ex} < E_{gap}^{min}, \tag{2.9} $$

最低能量小于或等于 $E_{gap}^{min}$, 因为电子-空穴相互作用是有吸引力的.
因此, 保存电子数量的测量(例如比热和光谱)可用于建立界限并确定材料是金属或绝缘体. 

最普遍的激发测量(保存电子数)是热容量(因为它包含所有可能的激发)(没有偏差).
低温下的热容量T是可测量的量(最终将具有能隙的系统(用于电子激发)与没有间隙的系统分开); 在完美的水晶中,这也是金属和绝缘体之间的区别.
普通金属的特征在于具体的热量$ \propto$T,这是Landau Fermi液体理论的基本证据[96,225,226].
这导致了"准粒子"的概念,即低能量激发,其作用类似于弱相互作用的电子,即使实际上存在强烈的相互作用.
这些图片通常提供比热,电导率和泡利顺磁性的优秀描述.
独立粒子的图像在Sec.4.7讨论, 单粒子DOS的例子如 图2.31和16.13所示.
如果在低T时比热量呈指数小,则这是间隙和低能量激发的明确证据.

费米表面是倒易空间中的表面(在多体体系中定义), 作为点的轨迹(其中准粒子寿命无限长并且准粒子能量等于费米能量μ).
在独立粒子近似中,态在所有能量下具有无限寿命,并且费米表面(分离填充状态和空状态)是由ε(k)=μ定义的表面.
在低温下极低能量激发的许多晶体中已经非常详细地绘制了表面(例如,使用电子磁化的小周期性变化作为施加的磁场的函数)(这在1930年首次观察到)由de Haas和van Alfen提出,并成为20世纪50年代的关键实验工具.
许多文献[84,86,88]描述了理论分析.
图2.31中所示的计算的MgB2费米表面是说明费米表面的非常不同部分的实例,其在热和电性质和超导性中起不同作用.

## 持续的挑战:电子相关

(由于相互作用引起的)相关性和(由于动能引起的)离域之间的竞争导致了(凝聚态物质中)电子理论中最具挑战性的问题. 
相关性负责金属-绝缘体转变,Kondo效应,重费米子系统,高温超导体和许多其他现象(参见,例如,[216]). 
低维度导致相关性的更大影响和量子霍尔效应等新现象.  
在一个方面,费米液体被Luttinger-Tomanaga液体代替,其中激发是"完全子"(holons) 和 "闪子"(spinons).  

在某些情况下,独立粒子方法足以通过有效的平均场相互作用包括相关效应;实际上,这是密度泛函理论在具有弱相关或中等相关系统的定量描述中成功的关键. 
术语"强相关"通常用于表示那些系统: 其中,平均场近似崩溃(无法适用)(break down), 无法描述重要的物理现象.  
这些问题是整篇文章的适当主题,本节仅仅是一个指针,提醒读者当前近似形式的密度泛函理论和其他独立粒子方法未解决的重要问题.  

表现出强相关效应的材料体系位于是定域化和局域化体系之间的边界.
它们通常涉及后3d和后4d过渡金属; 异常的4f稀土Ce,Sm,Eu和Tb; 和5f act系元素.
这些状态介于高度局域化的4f稀土和高离域的**带似**(形成能带)金属之间(包括前3d和前4d元素以及非过渡金属).
3d过渡金属氧化物处于金属-绝缘体转变发生的中间范围,伴随着高温超导性,这被认为涉及电子相关性,即使理解仍然是难以捉摸的[216].
`异常稀土元素是添加或去除能量接近费米能量的元素,表明4f占据不稳定,导致"混合价"和"重-费米子"行为.`
例如, CeCu2Si2的比热系数比"电子"的频带计算[245]预期的大约1,000倍.

随着电子结构理论变得更强大和更具预测性,记住多体电子 - 电子相互作用的可能后果的"大图"更为重要.
不仅要正确地说明对真实材料进行定量描述所必需的相互作用的影响,
但是,对相关性的富有想象力的探索可以导致令人兴奋的新现象,这些现象在质量上与通常的平均场理论的预测不同.
